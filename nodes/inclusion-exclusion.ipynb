# Principle of Inclusion-Exclusion (PIE)
The Principle of Inclusion-Exclusion is a counting technique to find the size of a union of sets.

### The Formula
For $n$ sets, the probability of the union is:
$$P(\cup_{i=1}^n A_i) = \sum P(A_i) - \sum P(A_i \cap A_j) + \dots + (-1)^{n-1} P(\cap A_i)$$

This is essential for problems where outcomes are not mutually exclusive.
