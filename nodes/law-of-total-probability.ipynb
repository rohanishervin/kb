# Law of Total Probability

## 1\. Theorem Statement

Let $A_1, A_2, \dots, A_n$ be a **partition** of the [sample space](sample-space.ipynb) $S$. (This means the [events](event.ipynb) $A_i$ are mutually disjoint, and their union covers the entire space $S$). Assume $P(A_i) > 0$ for all $i$.

For any event $B$, the probability of $B$ can be found by calculating the weighted average of the [conditional probabilities](conditional-probability.ipynb) of $B$ given each partition:
$$P(B) = \sum_{i=1}^n P(B|A_i)P(A_i)$$

## 2\. Proof

**Proof:**

Because the collection $\{A_1, \dots, A_n\}$ forms a partition of $S$, any event $B$ can be perfectly sliced into smaller, disjoint pieces corresponding to its intersection with each $A_i$:
$$B = (B \cap A_1) \cup (B \cap A_2) \cup \dots \cup (B \cap A_n)$$

Since these intersections are disjoint, the third [axiom of probability](probability.ipynb) (Countable Additivity) allows us to sum their probabilities:
$$P(B) = P(B \cap A_1) + P(B \cap A_2) + \dots + P(B \cap A_n)$$

Finally, we apply the multiplication rule ($P(X \cap Y) = P(X|Y)P(Y)$) to every term:
$$P(B) = P(B|A_1)P(A_1) + P(B|A_2)P(A_2) + \dots + P(B|A_n)P(A_n)$$
$\square$
