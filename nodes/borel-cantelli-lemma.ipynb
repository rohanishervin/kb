# The Borel-Cantelli Lemma

The **Borel-Cantelli Lemma** is a fundamental result in measure theory and probability, providing a condition under which a certain event will almost certainly occur only finitely many times. 

## 1. Statement of the Lemma

**Theorem (The Borel-Cantelli Lemma):** 
Let $(X, \mathcal{M}, \mu)$ be a [measure space](measure.ipynb) and let $\{E_k\}_{k=1}^\infty$ be a countable collection of [measurable sets](). If the sum of their measures is finite, that is:

$$
\sum_{k=1}^\infty \mu(E_k) < \infty
$$

Then [almost all](continuity-of-measure.ipynb#almost-everywhere-terminology) $x \in X$ belong to at most a finite number of the sets $E_k$. 

In other words, the measure of the set of points that belong to infinitely many of the $E_k$'s is zero.

<!-- CELL BREAK -->

::: {.callout-note}
## Intuition: Throwing Darts at Shrinking Targets
To understand why this holds, imagine throwing darts at a wall. Each set $E_k$ is a target on the wall, and its measure $\mu(E_k)$ represents its area. The condition $\sum_{k=1}^\infty \mu(E_k) < \infty$ means that the total area of the targets shrinks very rapidly as $k$ increases. 

Because the total area of all targets combined is strictly finite, the targets must get infinitesimally small. If you look at the "tail" of the sequence of targets (ignoring the first $n$ targets), the remaining area approaches zero as $n$ goes to infinity. Therefore, the chance (or measure) of a point landing in infinitely many of these rapidly shrinking targets is precisely zero. You might hit a few of the earlier, larger targets, but you will almost certainly only hit a finite number of them.
:::

## 2. Proof

To prove this, we must first construct the set of points that belong to infinitely many $E_k$'s and then show its measure is zero.

::: {.callout-tip collapse="false"}
## Proof of The Borel-Cantelli Lemma

**Step 1: Construct the target set.**
Let $E^*$ be the set of all points in $X$ that belong to an infinite number of the sets $E_k$. We can express $E^*$ mathematically using the limit superior of the sequence of sets:

$$
E^* = \bigcap_{n=1}^\infty \bigcup_{k=n}^\infty E_k
$$

This expression means that for any integer $n$, a point $x \in E^*$ must be in at least one $E_k$ where $k \ge n$.

**Step 2: Define a decreasing sequence of sets.**
Let $F_n = \bigcup_{k=n}^\infty E_k$. 
Notice that this sequence is nested and decreasing, meaning $F_1 \supset F_2 \supset F_3 \supset \dots$. The set $E^*$ can now be written simply as $E^* = \bigcap_{n=1}^\infty F_n$.

**Step 3: Apply the properties of the measure $\mu$.**
By the [countable subadditivity](measure.ipynb#countable-monotonicity-subadditivity) (or countable monotonicity) of the measure $\mu$, we can bound the measure of $F_n$:

$$
\mu(F_n) = \mu\left(\bigcup_{k=n}^\infty E_k\right) \le \sum_{k=n}^\infty \mu(E_k)
$$


**Step 4: Use the continuity of measure from above.**
Because the total sum $\sum_{k=1}^\infty \mu(E_k)$ is finite, we know that $\mu(F_1) \le \sum_{k=1}^\infty \mu(E_k) < \infty$. Since $F_n$ is a decreasing sequence of sets and $\mu(F_1) < \infty$, we can apply the [continuity of measure](continuity-of-measure.ipynb) from above:

$$
\mu(E^*) = \mu\left(\bigcap_{n=1}^\infty F_n\right) = \lim_{n \to \infty} \mu(F_n)
$$


**Step 5: Evaluate the limit.**
We substitute our inequality from Step 3 into the limit:

$$
\mu(E^*) = \lim_{n \to \infty} \mu(F_n) \le \lim_{n \to \infty} \sum_{k=n}^\infty \mu(E_k)
$$


Since the infinite series $\sum_{k=1}^\infty \mu(E_k)$ converges to a finite value, the tail of the series must converge to zero. That is:

$$
\lim_{n \to \infty} \sum_{k=n}^\infty \mu(E_k) = 0
$$


Because measure cannot be negative, we conclude that $\mu(E^*) = 0$. Thus, almost all $x \in X$ belong to at most a finite number of the $E_k$'s. $\blacksquare$
:::

::: {.callout-warning}
## Notation Convention: Probability vs. Measure Theory
In strict probability theory, the Borel-Cantelli lemma is often stated in terms of a probability space $(\Omega, \mathcal{F}, \mathbb{P})$ rather than a general measure space $(X, \mathcal{M}, \mu)$. Furthermore, the set $\bigcap_{n=1}^\infty \bigcup_{k=n}^\infty E_k$ is commonly denoted as $\limsup_{n \to \infty} E_n$ or simply "$E_n \text{ infinitely often}$" (i.o.). Ensure you recognize these equivalent forms depending on the context of the literature.
:::
