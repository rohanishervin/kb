# Experiment

## 1\. Definition

In probability theory, a **random experiment** (or trial) is a physical or conceptual procedure that satisfies three specific criteria:

1.  **Repeatability:** The procedure can be repeated, at least conceptually, under identical conditions.
2.  **Defined Outcomes:** The set of all possible results is known perfectly in advance.
3.  **Uncertainty:** The specific, individual result of any single execution of the procedure cannot be predicted with certainty beforehand.

## 2\. Breaking the "Circular Definition"

It is common to feel stuck when defining an experiment because an experiment requires a probability space, but a probability space is meant to model an experiment.

The distinction lies between the **physical reality** and the **mathematical model**:

  * The **Experiment** is the actual real-world or conceptual process (e.g., "flipping a coin").
  * The **Probability Space** $(\Omega, \mathcal{F}, P)$ is the rigorous *mathematical construct* we build to model that process.

We perform the experiment in reality, which yields an [outcome](outcome.ipynb). We map that outcome to our sample space $\Omega$ to do math with it.
