# Bose-Einstein Problem

How many ways are there to choose $k$ items from a set of $n$ objects with replacement, if order does not matter (we only care about the frequency of each object chosen, not the order)?

# Solution
## Tools needed:
1. [multiplication rule](multiplication-rule.ipynb)
2. [sampling with replacement](sampling-with-replacement.ipynb)
3. [sampling without replacement](sampling-without-replacement.ipynb)
4. [binomial coefficient formula](binomial-coefficient-formula.ipynb) 

Imagine that every time we sample an object, we put a dot on it. The problem can be expressed as: after $k$ samples, how many dots are on each object? To solve this, we can use the "stars and bars" argument.

To be valid, a sequence must start and end with a `|`, and have exactly $n - 1$ `|`'s and exactly $k$ $*$'s in between the starting and ending `|`'s. Conversely, any such sequence is a valid encoding for some configuration of particles in boxes.

Imagine that we have written down the starting and ending `|`'s, which represent the outer walls, and there are $n + k - 1$ fill-in-the-blank slots between these outer walls. We need only choose where to place the $k$ $*$'s (since then, the positions of the $n + k - 1$ interior `|`'s are completely determined).

So, the number of possibilities is given by 
$$ \binom{n+k-1}{k} $$
This counting method is sometimes called the "stars and bars" argument.
