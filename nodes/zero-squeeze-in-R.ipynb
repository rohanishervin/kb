# Zero Squeeze in $\mathbb{R}$

It is also correct for any ordered field.

We are using [ordered field](ordered-field.ipynb) and its propeties implicitly here. 

Let $x$ be a real number. If $|x| < \epsilon$ for each $\epsilon > 0$, then $x = 0$.

**proof-1:**
Assume, for contradiction, that $x \neq 0$. by [absolute value](absolute-value.ipynb) properties we can say:
$|x|>0$. so to find the desired $\epsilon$ we can wait until we get $x$. then we set $\epsilon = |x|$ and it's a contradiction because: $0<|x|<\epsilon=|x|$. 

**proof-2:** 
Assume, for contradiction, that $x \neq 0$.

If $x \neq 0$, then there exists a positive number $\delta$ such that $|x| \geq \delta$.

According to the given condition, $|x| < \epsilon$ for each $\epsilon > 0$. We can choose $ \epsilon = \frac{\delta}{2} $, which is a positive number.

By the given condition:
$|x| < \frac{\delta}{2}$

However, we previously established that $|x| \geq \delta$.
