# Naive Definition of Probability

Let $A$ be an event for an experiment with a finite sample space $S$. The naive probability of $A$, denoted as $P_{\text{naive}}(A)$, is given by the ratio of the number of outcomes favorable 
to $A$ to the total number of outcomes in $S$:

$$P_{\text{naive}}(A) = \frac{|A|}{|S|}$$

where $|A|$ denotes the size (number of elements) of event $A$.
