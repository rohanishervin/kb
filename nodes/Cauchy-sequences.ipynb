# Cauchy [Sequences](sequence.ipynb)


## Theorem: 
A monotone sequence converges if and only if it is bounded.

A sequence $\{x_n\}$ is a Cauchy sequence if:

$$ \forall \epsilon > 0, \forall m,n \in \mathbb{N}, \exists N\in \mathbb{N},:  \implies |x_m-x_n| < \epsilon $$

## Proposition: 
if a sequence is convergent  $\Leftrightarrow$  it's Cauchy.
