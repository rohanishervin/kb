# Continuity and Compactness

A major theme in topology is understanding which properties of a space are preserved when transformed by a continuous function. Just as continuous functions preserve sequential convergence, they also perfectly preserve compactness.

## 1. The Continuous Image of a Compact Set

**Theorem:** If $f: M \to N$ is continuous and $A$ is a [compact](compactness.ipynb) subset of $M$, then the image $f(A)$ is a compact subset of $N$. Simply put, the continuous image of a compact set is always compact.

**Proof:**
Suppose $(b_n)$ is an arbitrary sequence in the image set $f(A)$. 
By definition of the image, for each $n \in \mathbb{N}$, we can choose a corresponding pre-image point $a_n \in A$ such that $f(a_n) = b_n$.
Because $A$ is a compact set, the sequence $(a_n)$ must possess a subsequence $(a_{n_k})$ that converges to some limit point $p \in A$.
Because the function $f$ is continuous, it preserves limits. Therefore, applying $f$ to this convergent subsequence yields:
$$b_{n_k} = f(a_{n_k}) \to f(p) \in f(A) \quad \text{as } k \to \infty$$
Thus, we have shown that every sequence $(b_n)$ in $f(A)$ has a subsequence that converges to a limit strictly inside $f(A)$, which is the exact definition of sequential compactness. $\square$

## 2. The Extreme Value Theorem (Min/Max Theorem)

From the previous theorem follows a natural and incredibly powerful generalization of the classic Min/Max theorem from single-variable calculus.

**Theorem (Extreme Value Theorem):** A continuous real-valued function defined on a compact set is bounded; furthermore, it definitely assumes its maximum and minimum values.

**Proof:**
Let $f: M \to \mathbb{R}$ be a continuous function, and let $A$ be a compact subset of $M$. 
By our previous theorem, the image $f(A)$ must be a compact subset of $\mathbb{R}$. 
From our foundational study of Euclidean space, we know that every compact subset of $\mathbb{R}$ is strictly closed and bounded. 
Because the set is bounded, the completeness of the real numbers guarantees that it has a greatest lower bound (infimum), $v$, and a least upper bound (supremum), $V$. 
Crucially, because $f(A)$ is a *closed* set, it must actually contain its boundaries. Therefore, the values $v$ and $V$ belong to $f(A)$.
This means there exist specific points $p, P \in A$ such that for all $a \in A$, we have:
$$v = f(p) \leq f(a) \leq f(P) = V$$
Thus, the function achieves an absolute minimum at $p$ and an absolute maximum at $P$. $\square$

## 3. Counterexamples: Why Compactness is Mandatory

If the domain is not compact, the Extreme Value Theorem completely fails. A domain might be bounded but not closed, or closed but not bounded. In either case, a continuous function can "escape" to infinity or fail to reach its boundaries.

* **Bounded but NOT Closed (Open Interval):** 
  Consider $f(x) = \frac{1}{x}$ on the open interval $(0, 1)$. 
  The domain is strictly bounded, and $f$ is perfectly continuous everywhere inside this domain. However, as $x \to 0$, the function shoots off to $+\infty$. The image is $(1, \infty)$, which is not bounded and possesses no maximum.
* **Closed but NOT Bounded (Infinite Domain):**
  Consider $f(x) = e^x$ on the closed set $[0, \infty)$.
  The domain is closed (it contains its only boundary point, 0). However, the domain is infinitely wide. The image is $[1, \infty)$. While it has a minimum ($1$ at $x=0$), it has no maximum.
* **Failing to reach the supremum:**
  Consider $f(x) = x$ on the open interval $(0, 1)$.
  The domain is bounded but open. The image is $(0, 1)$. The supremum of the image is $1$, but there is no point $x \in (0, 1)$ where $f(x) = 1$. It has an upper bound, but it never *assumes* the maximum. 

Only true **Compactness** (both closed and bounded in $\mathbb{R}^n$) guarantees that the image is trapped and reaches its extremes.
