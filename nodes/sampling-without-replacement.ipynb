# Sampling w/o replacement

Consider $n$ objects and making $k$ choices from them, one at a time without replacement (i.e., choosing a certain object precludes it from being chosen again). Then there are $n(n − 1) · · · (n − k + 1)$ possible outcomes (based on [multiplication rule](multiplication-rule.ipynb)) for $1 ≤ k ≤ n$, and $0$ possibilities for $k > n$ (where order matters). By convention, $n(n − 1) · · · (n − k + 1) = n$ for $k = 1$.
