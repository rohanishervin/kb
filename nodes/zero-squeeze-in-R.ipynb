# The Zero Squeeze Theorem in $\mathbb{R}$

## 1\. Theorem Statement

The "Zero Squeeze" theorem is a foundational concept in real analysis. It formalizes the idea that if a number is infinitesimally small (smaller than any positive value you can possibly choose), that number must be exactly zero.

**Theorem:** Let $x$ be a real number ($x \in \mathbb{R}$). If $|x| < \epsilon$ for every $\epsilon > 0$, then $x = 0$.

*(Note: While stated here for the real numbers, this property holds true for any [ordered field](ordered-field.ipynb), as the proofs rely implicitly on ordered field axioms and the properties of the [absolute value](absolute-value.ipynb)).*

::: {.callout-note}

## Intuition: The Foundation of Equality in Calculus

This seemingly simple theorem is the logical engine behind limits and equality in calculus. In analysis, to prove that two numbers $a$ and $b$ are perfectly equal ($a = b$), we rarely prove it directly. Instead, we prove that the distance between them is smaller than any arbitrary error margin: $|a - b| < \epsilon$. By the Zero Squeeze theorem, this implies $|a - b| = 0$, thus $a = b$.
:::

## 2\. Proof 1: Direct Contradiction

This is the most direct and elegant way to prove the theorem, using the number itself as the arbitrary bound.

**Proof:**
Assume, for the sake of contradiction, that $x \neq 0$.

By the properties of the absolute value, if $x \neq 0$, then its absolute value must be strictly positive:
$$|x| > 0$$

The premise of our theorem states that $|x| < \epsilon$ holds true for *every* possible positive number $\epsilon$. Since we have established that $|x|$ is a positive number, we are perfectly allowed to choose $\epsilon = |x|$.

Substituting this specific $\epsilon$ into the premise yields:
$$|x| < |x|$$

This is a strict logical contradiction, as no number can be strictly less than itself. Therefore, our initial assumption must be false, and $x = 0$. $\square$

-----

## 3\. Proof 2: Halving the Bound

This alternative proof relies on the density of the ordered field—the idea that you can always find a smaller positive number by dividing by two.

**Proof:**
Assume, for the sake of contradiction, that $x \neq 0$.

Since $x \neq 0$, its absolute value is strictly positive. Let there exist a strictly positive real number $\delta$ such that:
$$|x| = \delta \quad (\text{where } \delta > 0)$$

The premise of the theorem states that $|x| < \epsilon$ for *every* $\epsilon > 0$.
Since $\delta > 0$, we can construct a new positive number by dividing it by two. Let us choose $\epsilon = \frac{\delta}{2}$.

Because $\epsilon > 0$, the premise must hold for this value, meaning:
$$|x| < \frac{\delta}{2}$$

However, we already established that $|x| = \delta$. Substituting this in yields:
$$\delta < \frac{\delta}{2}$$

Since $\delta$ is a strictly positive number, dividing it by two must yield a smaller number, making this inequality an absolute mathematical impossibility. Thus, our assumption that $x \neq 0$ must be false. $\square$
