# Limit Inferior and Limit Superior

Let $\{ x_n \}$ be a [sequence](sequence.ipynb) of [real numbers](real-numbers.ipynb).

1. if $\{ x_n \}$ is bounded below then the limit inferior is defined as follows:
   $$ \liminf_{n \to \infty} x_n = \lim_{n \to \infty}(\inf{\{x_k:k\geq n\}}) $$

2. if $\{ x_n \}$ is bounded above then the limit superior is defined as follows:
   $$ \limsup_{n \to \infty} x_n = \lim_{n \to \infty}(\sup{\{x_k:k\geq n\}}) $$

If sequence $\{ x_n \}$:
   $$ \limsup_{n \to \infty} x_n = \liminf_{n \to \infty} x_n = L \implies  \lim_{n \to \infty} x_n = L  $$
