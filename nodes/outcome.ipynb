# Outcome

## 1\. Formal Definition

An **outcome** (often denoted by the lowercase Greek letter $\omega$) is a single, specific, and mathematically indivisible result of a random [experiment](experiment.ipynb).

The set of all possible outcomes forms the [sample-space](sample-space.ipynb), denoted $S$ or $\Omega$. Therefore, an outcome is simply an element of the sample space:
$$\omega \in S$$

## 2\. Mutually Exclusive and Exhaustive

For a mathematical model of an experiment to be valid, its outcomes must be defined such that they are:

1.  **Mutually Exclusive:** Exactly one outcome occurs per trial. Two different basic outcomes cannot happen simultaneously.
2.  **Exhaustive:** Every possible result of the experiment maps to at least one defined outcome.

## 3\. Outcome vs. Event

It is crucial to distinguish between an outcome and an [event](event.ipynb).

  * An **outcome** $\omega$ is a single *element* of the sample space ($\omega \in S$).
  * An **event** $A$ is a *subset* of the sample space ($A \subseteq S$).

Even if an event contains only one outcome (a "simple event", e.g., $A = \{\omega\}$), it is still formally a set. Probability is measured on *events*, not naked outcomes.
