#  Metric Spaces and Subspaces


## 1\. Definition of a Metric Space

A **metric space** is a pair $(M, d)$, where $M$ is a [set](pre-set-theory.ipynb) and $d$ is a metric (a distance [function](def-functions-mappings.ipynb)) $d: M \times M \to [0, \infty)$ that satisfies three properties for all $x, y, z \in M$:

1.  **Positive Definiteness:** $d(x, y) \geq 0$, and $d(x, y) = 0 \iff x = y$
2.  **Symmetry:** $d(x, y) = d(y, x)$
3.  **Triangle Inequality:** $d(x, y) + d(y, z) \geq d(x, z)$

### The Reverse Triangle Inequality

A direct consequence of the triangle inequality that is extremely useful for bounding distances is the Reverse Triangle Inequality:
$$|d(x, y) - d(x, z)| \leq d(y, z)$$
**Proof:**
By the standard triangle inequality, $d(x, y) \leq d(x, z) + d(z, y)$.
Rearranging terms gives: $d(x, y) - d(x, z) \leq d(y, z)$.
Similarly, starting with $d(x, z) \leq d(x, y) + d(y, z)$ gives $d(x, z) - d(x, y) \leq d(y, z)$, which means $-(d(x, y) - d(x, z)) \leq d(y, z)$.
Combining these yields the absolute value inequality. $\square$

## 2\. Examples of Metrics in $\mathbb{R}^n$

While the Euclidean metric is the most common, different metrics fundamentally change the "shape" of the space. Let $x = (x_1, \dots, x_n)$ and $y = (y_1, \dots, y_n)$.

  * **Euclidean Metric ($L^2$):** $d_2(x, y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}$
  * **Taxicab Metric ($L^1$):** $d_1(x, y) = \sum_{i=1}^n |x_i - y_i|$ (Movement is restricted to grid lines).
  * **Supremum/Max Metric ($L^\infty$):** $d_\infty(x, y) = \max_{1 \leq i \leq n} |x_i - y_i|$ (Distance is solely determined by the coordinate with the largest difference).

## 3\. Metric Subspaces

If $(M, d)$ is a metric space and $S \subset M$ is a subset, we can consider $S$ as a metric space in its own right by letting it "inherit" the metric $d$ from $M$.

## 4\. The Discrete Metric

A trivial but highly useful metric that can be applied to *any* arbitrary set $M$ is the **discrete metric**:
$$d(x, y) = \begin{cases} 1 & \text{if } x \neq y \\ 0 & \text{if } x = y \end{cases}$$
*(Intuition: In a discrete space, every distinct point is exactly a distance of 1 away from every other point. It is impossible to get "infinitely close" to another point.)*
