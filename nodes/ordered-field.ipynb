# Ordered $(F, +, \cdot, \textless)$

An ordered [field](field.ipynb) is an [ordered set](ordered-set.ipynb) with the following additional properties:


1. If $x > 0$ and $y > 0$, then $x + y > 0$.
2. If $x > 0$ and $y > 0$, then $xy > 0$.
3. If $x < y$, then $y - x > 0$.

If $a>0$ then $-a<0$.

**Proof**:
if $-a > 0$, based on 5th property of a [field](field.ipynb) we have:
$$-a + a = 0$$
on the otherhand based on first property of ordered field:
$$-a>0, a>0 \implies -a+a > 0$$
which is a contradiction.

if $-a = 0$ again:
$-a+a = 0$ and $-a+a = 0+a = a$ is a contradiction.

$$ -a>0, -a>0 \implies (-a)(-a)=a^2>0$$
