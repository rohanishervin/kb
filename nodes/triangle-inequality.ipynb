# Triangle Inequality and Its Reverse

## 1\. Theorem Statements

For all real numbers $a, b \in \mathbb{R}$, the following fundamental inequalities hold:

1.  **The Triangle Inequality:** $$|a + b| \leq |a| + |b|$$

2.  **The Reverse Triangle Inequality:** $$||a| - |b|| \leq |a - b|$$

-----

## 2\. Proof of the Triangle Inequality

**Proof:**

By using [absolute value properties](absolute-value.ipynb), we know that any real number is bounded by its own absolute value. Therefore, we can write:

$$-|a| \leq a \leq |a|$$
$$-|b| \leq b \leq |b|$$

By using the rules for [adding inequalities](adding-inequalities.ipynb), we can sum these two compound inequalities straight down the columns:

$$(-|a| - |b|) \leq a + b \leq (|a| + |b|)$$

We can factor out the negative sign on the left side to make the symmetry clear:

$$-(|a| + |b|) \leq a + b \leq (|a| + |b|)$$

Recall the core absolute value property: if we have a real number $c \geq 0$ such that $-c \leq x \leq c$, this is strictly equivalent to writing $|x| \leq c$.

Let $x = (a + b)$ and let our positive boundary $c = (|a| + |b|)$. Applying the property directly yields:

$$|a + b| \leq |a| + |b|$$
$\square$

-----

## 3\. Proof of the Reverse Triangle Inequality

The reverse inequality bounds the *difference* of absolute values. It is proved by applying the standard Triangle Inequality (proved above) using a clever algebraic substitution.

**Proof:**

Consider the absolute value of $a$. We can add and subtract $b$ inside the absolute value without changing its value:

$$|a| = |a - b + b|$$

Now, treat $(a - b)$ as the first term and $b$ as the second term. Apply the standard Triangle Inequality ($|x + y| \leq |x| + |y|$):

$$|a - b + b| \leq |a - b| + |b|$$

Substitute $|a|$ back into the left side:

$$|a| \leq |a - b| + |b|$$

Subtract $|b|$ from both sides:

$$|a| - |b| \leq |a - b| \quad \text{--- (Equation 1)}$$

Next, we do the exact same process starting with $|b|$, adding and subtracting $a$:

$$|b| = |b - a + a| \leq |b - a| + |a|$$

Since $|b - a| = |a - b|$ (due to the symmetry of absolute values), we can write:

$$|b| \leq |a - b| + |a|$$

Subtract $|a|$ from both sides:

$$|b| - |a| \leq |a - b|$$

Multiply by $-1$ (which flips the inequality sign):

$$|a| - |b| \geq -|a - b|$$

Which can be written as:

$$-|a - b| \leq |a| - |b| \quad \text{--- (Equation 2)}$$

Combining **Equation 1** and **Equation 2** creates a compound inequality:

$$-|a - b| \leq |a| - |b| \leq |a - b|$$

Using the same absolute value property we used in the first proof ($-c \leq x \leq c \implies |x| \leq c$), we conclude:

$$||a| - |b|| \leq |a - b|$$
$\square$
