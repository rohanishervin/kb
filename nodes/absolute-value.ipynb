# Absolute Value

we can define absolute value function on any arbitrary [ordered field](ordered-field.ipynb).

$$
|x| =
\begin{cases}
    x & \text{if } x \geq 0 \\
    -x & \text{if } x < 0
\end{cases}
$$

## Properties:
1. $$-|a|\leq a \leq |a|$$
2. $$ |ab| = |a||b| $$
3. $$ |-a|=|a| $$
4. if $c>0$. then $|a|<c$ if and only if $-c<a<c$
