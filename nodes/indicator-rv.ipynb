# Indicator Random Variables

## 1. Definition

An [**indicator random variable**](random-variables.ipynb) $I_A$ is a binary variable that "indicates" whether a specific mathematical event $A$ has occurred. 

* $I_A = 1$ if event $A$ occurs.
* $I_A = 0$ if event $A$ does not occur.

Because $I_A$ is fundamentally just a [Bernoulli random variable](bernoulli-and-binomial-distributions.ipynb) where "success" is defined as event $A$ happening, we can easily find its [expected value](bernoulli-and-binomial-distributions.ipynb):
$$E[I_A] = 1 \cdot P(A) + 0 \cdot P(A^c) = P(A)$$

::: {.callout-note title="Remark: Probability to Expecation Bridge"}
This simple equality is incredibly profound: **it establishes a direct bridge between the probability of an event and the expected value of a random variable.** 
:::

## 2. Solving Complex Problems with Indicators

Indicators are standardly used to calculate the expected value of a complicated counting variable $X$. If $X$ counts the number of times a certain condition is met, we can break $X$ down into a sum of simple indicators: $X = I_1 + I_2 + \dots + I_n$. 

By applying the **Linearity of Expectation**, we can find $E[X]$ without ever knowing the overall probability distribution of $X$, *even if the indicators are highly dependent on one another*.

### Application 1: [The Matching Problem](de-montmorts-matching-problem.ipynb#problem-statement)
**Scenario:** A well-shuffled deck of $n$ numbered cards is dealt one by one. What is the expected number of cards that perfectly match their position in the deck (e.g., card #3 is the 3rd card dealt)?

**Solution:** Let $X$ be the total number of matches. Calculating the exact Probability Mass Function (PMF) of $X$ is notoriously difficult because a match at position 1 slightly changes the probability of a match at position 2 (they are dependent). 

However, let $I_j$ be the indicator that the $j$-th card is a match. 
$$X = I_1 + I_2 + \dots + I_n$$

By symmetry, the unconditional probability that any specific card matches its position is simply $\frac{1}{n}$. Thus, $E[I_j] = P(\text{match at } j) = \frac{1}{n}$. 

Applying linearity:
$$E[X] = E[I_1] + E[I_2] + \dots + E[I_n] = n \cdot \left(\frac{1}{n}\right) = 1$$

The expected number of matches is always exactly 1, regardless of whether the deck has 10 cards or 10,000 cards!

### Application 2: [Birthday Matches](birthday-problem.ipynb)
**Scenario:** In a room of $n$ people, what is the expected number of "birthday matches" (pairs of people who share the exact same birthday)? Assume 365 days in a year and equal likelihood.

**Solution:**
Let $Y$ be the number of birthday matches. We can define an indicator $I_{j}$ for every possible *pair* of people in the room. There are $\binom{n}{2}$ total pairs.
$$Y = I_1 + I_2 + \dots + I_{\binom{n}{2}}$$

For any specific pair of people, the probability they share a birthday is $\frac{1}{365}$. Therefore, $E[I_j] = \frac{1}{365}$.

Applying linearity over all possible pairs:
$$E[Y] = \sum_{j=1}^{\binom{n}{2}} E[I_j] = \frac{\binom{n}{2}}{365}$$
