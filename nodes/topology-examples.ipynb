# Topology Examples

## Example 1: $(a, b)$ is an open set in $\mathbb{R}$

**Proof:**
Let $S = (a, b)$ and choose an arbitrary point $p \in S$. By definition, $a < p < b$.

By [definition](metric-space-topology.ipynb), we must find a radius $r > 0$ such that the neighborhood $M_r p \subset S$. 
Choose $r = \min(p - a, b - p)$. Since $p$ is strictly between $a$ and $b$, $r$ is strictly positive.

For any point $q \in M_r p$, we have $|q - p| < r$, which means:
$$p - r < q < p + r$$

Substituting our bounds for $r$:
$$a = p - (p - a) \le p - r < q < p + r \le p + (b - p) = b$$

Thus, $a < q < b$, meaning $q \in S$. Therefore, the entire neighborhood $M_r p$ is contained in $S$, proving $(a, b)$ is an open set. $\square$


## Example 2: Clopen Sets

**Definition:** A set is **clopen** if it is both open and closed. 

In any metric space $M$, the empty set $\emptyset$ and the whole space $M$ are always clopen.


## Example 3: Clopen subsets of $\mathbb{R}$

In the metric space $\mathbb{R}$ with the standard Euclidean metric, the **only** clopen sets are $\emptyset$ and $\mathbb{R}$. There are no other subsets of $\mathbb{R}$ that are both open and closed. This is a fundamental property related to the **connectivity** of the real line.


::: {.callout-note}
## Intuition: The Relativity of Topology
Openness and closedness are **not intrinsic properties** of a set. They are relative to the background metric space $M$. A set $S$ might be open when viewed in one space, but not open when viewed as a subset of a larger space. 
:::


## Example 4: Clopen sets in $\mathbb{Q}$

If we change the background metric space to $M = \mathbb{Q}$, the "gaps" left by irrational numbers allow for more clopen sets:

1. The set $\mathbb{Q}$ is both open and closed in $\mathbb{Q}$ (by the general property of $M$).
2. The set $S = \{x \in \mathbb{Q} : x^2 < 2\}$ is **clopen** in $\mathbb{Q}$.
   * **It is open:** For any $q \in S$, since $q^2 < 2$, there is a rational neighborhood $(q-r, q+r) \cap \mathbb{Q}$ that stays within the bounds of $\sqrt{2}$.
   * **It is closed:** In $\mathbb{R}$, the limit points are $\pm\sqrt{2}$, which are irrational. In the world of $\mathbb{Q}$, these points do not exist. Therefore, every sequence in $S$ that converges *to a rational number* must converge to a point already in $S$.


## Example 5: Infinite Intersection of Open Sets

While [axiom of topology](metric-space-topology.ipynb) states that the intersection of any *finite* number of open sets is open, this property does not hold for *infinite* intersections.

**Example:**
Consider the metric space $\mathbb{R}$ with the standard metric. Define an infinite sequence of open intervals:
$$U_n = \left(-\frac{1}{n}, \frac{1}{n}\right) \quad \text{for } n = 1, 2, 3, \dots$$

Every $U_n$ is an open set (as proven in Example 1). Now, let $S$ be the intersection of all these sets:
$$S = \bigcap_{n=1}^{\infty} U_n = \bigcap_{n=1}^{\infty} \left(-\frac{1}{n}, \frac{1}{n}\right)$$

By the Archimedean property, the only real number that is strictly between $-1/n$ and $1/n$ for *every* positive integer $n$ is $0$. Therefore, the infinite intersection collapses to a single point:
$$S = \{0\}$$

**Proof that $S$ is not open:**
For $S = \{0\}$ to be open, there must exist some radius $r > 0$ such that the entire neighborhood $M_r(0) = (-r, r)$ is contained within $S$. 
However, for any strictly positive $r$, the interval $(-r, r)$ contains infinitely many non-zero numbers, meaning $(-r, r) \not\subset \{0\}$. 
Thus, $\{0\}$ is not an open set. (It is, in fact, a closed set). $\square$


## Example 6: Every open ball is an open set

**Problem Statement:**
Prove that in any metric space $(M, d)$, the open ball (or $r$-neighborhood) $M_r p$ is an open set.

**Proof:**
Let $S = M_r p = \{x \in M \mid d(x, p) < r\}$. 
To prove $S$ is open, we must show that for any arbitrary point $q \in S$, there exists some radius $s > 0$ such that the neighborhood $M_s q$ is entirely contained within $S$.

Since $q \in S$, we know that the distance $d(p, q) < r$. 
Let us define a new radius:
$$s = r - d(p, q)$$
Because $d(p, q) < r$, it is strictly true that $s > 0$.

Now, we must verify that $M_s q \subset S$. Let $x$ be any point in $M_s q$. 
By definition, $d(x, q) < s$. 

To determine if $x \in S$, we check its distance to the center $p$ using the **triangle inequality**:
$$d(x, p) \leq d(x, q) + d(q, p)$$

Substitute our known strict bound $d(x, q) < s$:
$$d(x, p) < s + d(q, p)$$

Now, substitute our definition of $s$:
$$d(x, p) < (r - d(p, q)) + d(q, p)$$
$$d(x, p) < r$$

Since $d(x, p) < r$, the point $x$ belongs to $S$. Because this holds for any arbitrary point $x \in M_s q$, we conclude that $M_s q \subset S$. 
Therefore, $M_r p$ is an open set. $\square$
