# The Banach Contraction Principle

The Banach Contraction Principle (often called the Banach Fixed Point Theorem) is one of the most famous and useful fixed-point theorems in analysis. It guarantees not only that a fixed point exists for a contracting map, but also that it is unique and can be found constructively via iteration.

## 1\. Statement of the Theorem

**Theorem (Banach Contraction Principle):**
Suppose that $(M, d)$ is a [complete metric space](completeness.ipynb) and $f: M \to M$ is a [contraction](contractions.ipynb). Then $f$ has a unique fixed point $p \in M$. 

Furthermore, for any initial choice $x_0 \in M$, the sequence of iterates defined by $x_n = f^n(x_0)$ converges to $p$ as $n \to \infty$.

*(Note: Here, $f^n$ denotes function composition $f \circ f \circ \dots \circ f$ repeated $n$ times, not the $n$-th power of a real number.)*

## 2\. Dynamical Proof

::: {.callout-tip collapse="true"}
## Proof of the Banach Contraction Principle

**Step 1: Constructing a sequence**
Choose an arbitrary starting point $x_0 \in M$. We define a sequence $(x_n)$ by repeatedly applying the function $f$:
$$
x_n = f^n(x_0)
$$
This means $x_1 = f(x_0)$, $x_2 = f(x_1)$, and generally $x_n = f(x_{n-1})$.

**Step 2: Bounding the distance between successive terms**
Because $f$ is a contraction with some constant $k < 1$, we can bound the distance between adjacent terms in our sequence:
$$
\begin{aligned}
d(x_n, x_{n+1}) &= d(f(x_{n-1}), f(x_n)) \\
&\le k d(x_{n-1}, x_n) \\
&\le k^2 d(x_{n-2}, x_{n-1}) \\
&\le \dots \\
&\le k^n d(x_0, x_1)
\end{aligned}
$$

**Step 3: Showing the sequence is Cauchy**
We must show that $(x_n)$ is a [Cauchy sequence](Cauchy-sequences.ipynb). For any integers $n, m$ with $m > n$, we can use the Triangle Inequality to estimate the distance between $x_n$ and $x_m$:
$$
\begin{aligned}
d(x_n, x_m) &\le d(x_n, x_{n+1}) + d(x_{n+1}, x_{n+2}) + \dots + d(x_{m-1}, x_m) \\
&\le k^n d(x_0, x_1) + k^{n+1} d(x_0, x_1) + \dots + k^{m-1} d(x_0, x_1) \\
&= k^n (1 + k + k^2 + \dots + k^{m-n-1}) d(x_0, x_1)
\end{aligned}
$$
Since $k < 1$, the finite geometric series is strictly bounded by the infinite geometric series $\sum_{l=0}^\infty k^l = \frac{1}{1-k}$. Thus:
$$
d(x_n, x_m) \le \frac{k^n}{1-k} d(x_0, x_1)
$$
Let $\epsilon > 0$ be given. Because $k < 1$, the term $k^n \to 0$ as $n \to \infty$. We can choose $N$ large enough so that for all $n \ge N$:
$$
\frac{k^N}{1-k} d(x_0, x_1) < \epsilon
$$
Therefore, for all $m > n \ge N$, we have $d(x_n, x_m) < \epsilon$, which proves the sequence $(x_n)$ is Cauchy.

**Step 4: Existence of the fixed point**
Since $M$ is a complete metric space, every Cauchy sequence converges to a limit in $M$. Let $p = \lim_{n \to \infty} x_n$. 

Because a contraction is trivially continuous, we can pass the limit inside the function:
$$
f(p) = f(\lim_{n \to \infty} x_n) = \lim_{n \to \infty} f(x_n) = \lim_{n \to \infty} x_{n+1} = p
$$
Thus, $p$ is a fixed point of $f$.

**Step 5: Uniqueness**
Suppose there are two fixed points, $p$ and $q$, such that $f(p) = p$ and $f(q) = q$. By the contraction property:
$$
d(p, q) = d(f(p), f(q)) \le k d(p, q)
$$
Since $k < 1$, the only way a number can be less than or equal to a strict fraction of itself is if it is zero. Therefore, $d(p, q) = 0$, implying $p = q$. The fixed point is unique. $\blacksquare$
:::

::: {.callout-note}
## Intuition: A Dynamical System
This proof views the function $f$ as a dynamical system. Imagine throwing a dart anywhere on a map (your $x_0$). If the rules of movement (the map $f$) strictly compress the distance between any two points over time, then following the path of your dart will inevitably pull you toward a single, inescapable "sink" or attractor—the fixed point $p$. 
:::

::: {.callout-warning}
## The Role of Completeness
The requirement that $M$ is complete is non-negotiable. If $M$ has "holes" (e.g., the metric space is the open interval $(0, 1)$), the sequence $x_n$ might march endlessly toward a missing point on the boundary, resulting in no fixed point residing *within* the space itself.
:::
