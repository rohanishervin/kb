# Completeness in Metric Spaces

## 1. Cauchy Sequences

In previous notes, we discussed the [Cauchy criterion](Cauchy-sequences.ipynb) for sequences of real numbers. There is a natural way to carry these ideas over to a general [metric space](metric-spaces.ipynb) $M$.

**Definition:** A sequence $(p_n)$ in $M$ satisfies a **Cauchy condition** (and is called a **Cauchy sequence**) provided that for each $\epsilon > 0$ there is an integer $N$ such that for all $k, n \geq N$ we have $d(p_k, p_n) < \epsilon$. 
In symbols:
$$\forall \epsilon > 0, \exists N \text{ such that } k, n \geq N \implies d(p_k, p_n) < \epsilon$$

Intuitively, the terms of a Cauchy sequence "bunch together" as $n \to \infty$.

**Theorem:** Every convergent sequence $(p_n)$ is Cauchy. (Convergence $\implies$ Cauchy).

**Proof:**
If $(p_n)$ converges to $p$ as $n \to \infty$, then given any $\epsilon > 0$, there is an integer $N$ such that for all $n \geq N$ we have:
$$d(p_n, p) < \frac{\epsilon}{2}$$
By the triangle inequality, if $k, n \geq N$, then:
$$d(p_k, p_n) \leq d(p_k, p) + d(p, p_n) < \frac{\epsilon}{2} + \frac{\epsilon}{2} = \epsilon$$
Thus, the sequence is Cauchy. $\square$

## 2. Complete Metric Spaces

Is the converse true? Does Cauchy $\implies$ Convergent? 
In the metric space $\mathbb{R}$, yes (this is a fundamental property of real numbers). However, in a general metric space, this need not be true. 

**Counterexample (The Rationals):**
Consider the metric space $\mathbb{Q}$ of rational numbers with the inherited metric $d(x, y) = |x - y|$. Consider the sequence:
$$(r_n) = (1.4, 1.41, 1.414, 1.4142, \dots)$$
This sequence is clearly Cauchy (the terms get arbitrarily close together). However, $(r_n)$ refuses to converge in $\mathbb{Q}$. In the larger space of $\mathbb{R}$, it converges to $\sqrt{2}$. By the uniqueness of limits, if it converged to some $r \in \mathbb{Q}$, then $r$ would have to equal $\sqrt{2}$, which is false since $\sqrt{2}$ is irrational. Thus, a sequence can be Cauchy but fail to converge because its "target" limit is missing from the space.

**Definition:** A metric space $M$ is **complete** if *every* Cauchy sequence in $M$ converges to a limit that is contained in $M$. 

**Lemma:** Let $(p_n)$ be a sequence of vectors in $\mathbb{R}^m$, where $p_n = (p_{1n}, p_{2n}, \dots, p_{mn})$. If $(p_n)$ is a Cauchy sequence in $\mathbb{R}^m$, then each of its component sequences $(p_{in})_{n \in \mathbb{N}}$ is a Cauchy sequence in $\mathbb{R}$.

**Proof:**
Let $\epsilon > 0$. Since $(p_n)$ is a Cauchy sequence in $\mathbb{R}^m$, there exists an integer $N$ such that for all $k, n \geq N$, the total distance $d(p_k, p_n) < \epsilon$. 
Recall from our study of product metrics that the distance between individual components is always less than or equal to the total distance between the vectors (e.g., $|p_{ik} - p_{in}| \leq d_{\max}(p_k, p_n) \leq d_E(p_k, p_n)$). 
Therefore, for any specific component $i$ and for all $k, n \geq N$, we have:
$$|p_{ik} - p_{in}| \leq d(p_k, p_n) < \epsilon$$
This perfectly satisfies the Cauchy condition for real numbers. Thus, each component sequence $(p_{in})_{n \in \mathbb{N}}$ is Cauchy in $\mathbb{R}$. $\square$

**Theorem:** $\mathbb{R}^m$ is complete.

**Proof:**
Let $(p_n)$ be a Cauchy sequence in $\mathbb{R}^m$. Express $p_n$ in its vector components as $p_n = (p_{1n}, \dots, p_{mn})$.
By the preceding Lemma, each individual component sequence $(p_{in})_{n \in \mathbb{N}}$ is Cauchy in $\mathbb{R}$. 
Because $\mathbb{R}$ is an established complete metric space, every Cauchy sequence within it must converge. Therefore, each component sequence converges to some real limit $L_i$.
Let $L$ be the vector composed of these limits: $L = (L_1, L_2, \dots, L_m)$. 
Because a vector sequence converges if and only if all of its component sequences converge, the overall vector sequence $(p_n)$ converges to the vector $L \in \mathbb{R}^m$. 
Since every Cauchy sequence in $\mathbb{R}^m$ converges to a limit in $\mathbb{R}^m$, the space is complete. $\square$

## 3. Completeness of Subspaces

**Theorem:** Every closed subset of a complete metric space is a complete metric subspace.

**Proof:**
Let $A$ be a closed subset of the complete metric space $M$, and let $(p_n)$ be a Cauchy sequence in $A$ (with respect to the inherited metric). 
Since the sequence is in $A$ and $A \subset M$, it is also a Cauchy sequence in $M$. 
Because $M$ is a complete space, the sequence is guaranteed to converge to a limit $p \in M$. 
Since $A$ is a closed set, it must contain all of its limits. Therefore, we safely conclude $p \in A$. Since every Cauchy sequence in $A$ converges to a limit inside $A$, $A$ is complete. $\square$

**Corollary:** Every closed subset of Euclidean space ($\mathbb{R}^m$) is a complete metric space.

**Proof:**
This follows obviously from the combination of the previous theorem and the completeness of $\mathbb{R}^m$. $\square$

## 4. Completeness is NOT a Topological Property

**Remark:** Completeness is a property of the *metric* (how distance is measured), not the *topology* (the structure of open sets). 

For example, consider the real line $\mathbb{R}$ with its usual metric, and the open interval $(-1, 1)$ with the metric it inherits from $\mathbb{R}$. These two spaces are **homeomorphic** (they are topologically identical and can be perfectly stretched/mapped onto each other without tearing, e.g., using $f(x) = \tan(\frac{\pi x}{2})$). 

Although they are [homeomorphic](continuity-topology.ipynb), $\mathbb{R}$ is complete, but $(-1, 1)$ is **not**. 

::: {.callout-note}
## Intuition: Why isn't $(-1, 1)$ complete?
To see why $(-1, 1)$ is incomplete, think about a sequence like $x_n = 1 - \frac{1}{n}$. The terms are $(0, 0.5, 0.66, 0.75, \dots)$. As $n$ grows, the terms bunch up infinitely close to each other, meaning the sequence is strictly **Cauchy**. 
However, the point they are bunching toward is exactly $1$, which is *outside* the open interval $(-1, 1)$. The sequence is desperately trying to converge to a "hole" at the boundary. Because the Cauchy sequence never reaches a limit *inside* the space $(-1, 1)$, the space is incomplete!
:::
