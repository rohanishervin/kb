# Functions and Mappings

## 1. Anatomy of a Function
A function $f: A \to B$ maps elements from a **Domain** ($A$) to a **Codomain/Target** ($B$). The actual subset of $B$ that gets "hit" by the function is called the **Image** or Range.

Depending on the context, mathematicians use different terms for functions:

* **Function:** Usually implies mapping to sets of numbers (e.g., $\mathbb{R}$ or $\mathbb{C}$).
* **Mapping:** Used in abstract sets or topology.
* **Transformation:** Used in geometry or linear algebra (e.g., rotations, translations).
* **Operator:** Used when mapping functions to other functions (e.g., the derivative operator).

## 2. Classification of Mappings

1.  **Injection (One-to-One):** Every element in the domain maps to a *unique* element in the codomain. ($f(x) = f(y) \implies x = y$).
2.  **Surjection (Onto):** Every element in the codomain is mapped to by at least one element in the domain. (Image = Codomain).
3.  **Bijection:** A function that is both injective and surjective. It establishes a perfect one-to-one pairing between $A$ and $B$, meaning an inverse function $f^{-1}$ exists.
