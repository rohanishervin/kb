# Set Theory Preliminaries

### 1. Fundamental Sets

**Definition:** A **set** is a collection of elements.

::: {.callout-warning}
## A Note on Naivety
This is a "naive" definition. While sufficient for our current purposes in Analysis, it can lead to logical contradictions such as **Russell's Paradox**. In a more rigorous axiomatic framework (like ZFC), we must be more careful about how sets are constructed.
:::

**Membership:**

* If an element $a$ is in set $A$, we denote it by $a \in A$.
* If it is not in the set, we denote it by $a \notin A$.

**Class:** A collection of sets. Like our definition of a set, this is an intuitive placeholder that serves us well for now.

We build our number systems starting from the most basic:

* **Natural Numbers:** $\mathbb{N} = \{1, 2, 3, \dots\}$
* **Integers:** $\mathbb{Z} = \{0, \pm 1, \pm 2, \dots\}$ (From the German *Zahlen*)
* **Rational Numbers:** $\mathbb{Q} = \{ p/q : p, q \in \mathbb{Z}, q \neq 0 \}$

### 2. Set Operations

* **Disjoint Sets:** $A$ and $B$ are **disjoint** if their intersection is empty: $A \cap B = \emptyset$.
* **Symmetric Difference ($\Delta$):** Represents elements that are in either $A$ or $B$, but not in both.
    $$A \Delta B = (A \setminus B) \cup (B \setminus A)$$


### 3. Equivalence Relations and Classes

An **Equivalence Relation** ($\sim$) on a set $S$ generalizes the concept of equality. To be valid, it must satisfy three properties for all $x, y, z \in S$:

1. **Reflexive:** $x \sim x$.
2. **Symmetric:** $x \sim y \implies y \sim x$.
3. **Transitive:** $x \sim y \land y \sim z \implies x \sim z$.

::: {.callout-note}
## The Quotient Set and Classes
An equivalence relation partitions $S$ into **Equivalence Classes**. 

While these are technically "equivalence sets," mathematical convention almost universally uses the term "classes." The set of all such classes is called the **Quotient Set**, denoted $S/\sim$. In Analysis, we often treat these classes as single "points" or "objects" in a new, more abstract space.
:::
