# Problem Set: Function Spaces

## Problem 1: Pugh, Ch 2, P 1

**Problem Statement:**
Let $M$ and $N$ be metric spaces. 

**Part (a):** Formulate the definitions of [pointwise convergence](function-spaces.ipynb#pointwise-convergence-the-weak-form) and [uniform convergence](function-spaces.ipynb#uniform-convergence-the-strong-form) for a sequence of functions $f_n : M \to N$.

**Part (b):** Under what conditions on the metric spaces are these two types of convergence strictly equivalent?

::: {.callout-tip collapse="true"}
## Solution

**Part (a): Definitions**

Let $(M, d_M)$ and $(N, d_N)$ be metric spaces, and let $f_n : M \to N$ be a sequence of functions. Suppose $f : M \to N$ is a function.

1.  **Pointwise Convergence:** The sequence $f_n$ converges pointwise to $f$ if, for every point in the domain, the sequence of evaluated points converges in $N$.
    $$
    \forall x \in M, \forall \varepsilon > 0, \exists N \in \mathbb{N} \text{ s.t. } \forall n \ge N \implies d_N(f_n(x), f(x)) < \varepsilon
    $$

2.  **Uniform Convergence:** The sequence $f_n$ converges uniformly to $f$ if the rate of convergence depends only on $\varepsilon$ and is independent of the point $x \in M$.
    $$
    \forall \varepsilon > 0, \exists N \in \mathbb{N} \text{ s.t. } \forall x \in M, \forall n \ge N \implies d_N(f_n(x), f(x)) < \varepsilon
    $$

**Part (b): Equivalence of Concepts**

The concepts of pointwise and uniform convergence are equivalent if and only if the domain $M$ is a **finite metric space** (or a union of isolated points in certain discrete formulations, but "finite" is the general requisite for arbitrary sequences).

**Proof Sketch:**

* **If $M$ is finite:** Let $M = \{x_1, x_2, \dots, x_k\}$. For any given $\varepsilon > 0$, pointwise convergence guarantees we can find an index $N_i$ for each $x_i$ such that $d_N(f_n(x_i), f(x_i)) < \varepsilon$ for all $n \ge N_i$. Since there are only finitely many elements in $M$, we can simply define $N = \max(N_1, N_2, \dots, N_k)$. This maximum $N$ will satisfy the condition for uniform convergence across all $x \in M$.
* **If $M$ is infinite:** We can almost always construct a sequence of functions that converges pointwise but not uniformly by letting the "error bump" escape to infinity or concentrate at a specific accumulation point.
:::

::: {.callout-note}
## Intuition: The Order of Quantifiers
The core difference between pointwise and uniform convergence lies entirely in the order of the quantifiers—specifically, whether $N$ depends on $x$.

* **Pointwise:** $N$ is chosen *after* $x$. Thus, $N$ can be customized for every single point (some points may converge very slowly).
* **Uniform:** $N$ is chosen *before* $x$. One single $N$ must work simultaneously for all $x \in M$, meaning there is a global "worst-case" bound on how slowly the sequence converges.
:::

::: {.callout-warning}
## Notation Trap: Absolute Value vs. Metric
A common pitfall when transitioning from introductory real analysis to general topology or metric spaces is blindly using the absolute value notation $|x - y|$ to denote distance. Remember that elements of an arbitrary metric space (like spaces of matrices, sequences, or functions) do not necessarily have defined algebraic operations like subtraction. Always default to the distance function $d(x, y)$ when working in abstract metric spaces.
:::



## Problem 2: Pugh, Ch 2, P 3

**Problem Statement:**
Let $f_n : [a, b] \to \mathbb{R}$ be a sequence of piecewise continuous functions. Assume that every function in the sequence is continuous at a specific point $x_0 \in [a, b]$. Furthermore, suppose the sequence converges uniformly to a limit function, $f_n \rightrightarrows f$ as $n \to \infty$.

**Part (a):** Prove that the limit function $f$ is continuous at $x_0$. 

**Part (b):** Prove or disprove that the limit function $f$ must be piecewise continuous on $[a, b]$.

::: {.callout-tip collapse="true"}
## Solution

**Part (a): Proof of Continuity at $x_0$**

We want to show that for any $\varepsilon > 0$, there exists a $\delta > 0$ such that if $|x - x_0| < \delta$, then $|f(x) - f(x_0)| < \varepsilon$. We will use the classic $\varepsilon/3$ trick, leveraging the triangle inequality to bridge the gap between $f(x)$ and $f(x_0)$ using our approximant functions $f_n$.

Let $\varepsilon > 0$ be given.

**Step 1: Exploit Uniform Convergence**
Because $f_n \rightrightarrows f$, there exists an integer $N \in \mathbb{N}$ such that for all $n \ge N$ and for all $x \in [a, b]$:
$$|f_n(x) - f(x)| < \frac{\varepsilon}{3}$$

Fix this specific $N$. The inequality holds for $x$ and also specifically for our target point $x_0$:
$$|f_N(x_0) - f(x_0)| < \frac{\varepsilon}{3}$$

**Step 2: Exploit Continuity of $f_N$ at $x_0$**
We are given that every $f_n$ is continuous at $x_0$, which means $f_N$ is continuous at $x_0$. Therefore, there exists a $\delta > 0$ such that for all $x \in [a,b]$ satisfying $|x - x_0| < \delta$:
$$|f_N(x) - f_N(x_0)| < \frac{\varepsilon}{3}$$

**Step 3: Combine with the Triangle Inequality**
Now, assume $x \in [a,b]$ such that $|x - x_0| < \delta$. We can rewrite the distance between $f(x)$ and $f(x_0)$ by adding and subtracting $f_N(x)$ and $f_N(x_0)$:
$$
\begin{aligned}
|f(x) - f(x_0)| &= |f(x) - f_N(x) + f_N(x) - f_N(x_0) + f_N(x_0) - f(x_0)| \\
&\le |f(x) - f_N(x)| + |f_N(x) - f_N(x_0)| + |f_N(x_0) - f(x_0)|
\end{aligned}
$$

Substituting the bounds we established from Steps 1 and 2:
$$
\begin{aligned}
|f(x) - f(x_0)| &< \frac{\varepsilon}{3} + \frac{\varepsilon}{3} + \frac{\varepsilon}{3} = \varepsilon
\end{aligned}
$$
Thus, $f$ is continuous at $x_0$. $\blacksquare$

---

**Part (b): Disproof of Piecewise Continuity**

The claim is **false**. We can construct a sequence of piecewise continuous functions that converge uniformly to a function with infinitely many discontinuities. 

**Counterexample:**
Let $[a, b] = [0, 1]$ and let our point of continuity be $x_0 = 1$. 
Define the sequence $f_n : [0, 1] \to \mathbb{R}$ as:
$$
f_n(x) = \begin{cases} 
\frac{1}{k} & \text{if } x = \frac{1}{k} \text{ for an integer } k \in [2, n] \\
0 & \text{otherwise}
\end{cases}
$$

* **$f_n$ is piecewise continuous:** Each $f_n$ only has a finite number of jump discontinuities (exactly $n-1$ discontinuities at $1/2, 1/3, \dots, 1/n$). 
* **$f_n$ is continuous at $x_0 = 1$:** For any $n$, $f_n(1) = 0$. Furthermore, for any $x \in (1/2, 1]$, $f_n(x) = 0$. Thus, the limit as $x \to 1$ is $0$, matching the function value.

Now, define the limit function $f : [0, 1] \to \mathbb{R}$:
$$
f(x) = \begin{cases} 
\frac{1}{k} & \text{if } x = \frac{1}{k} \text{ for any integer } k \ge 2 \\
0 & \text{otherwise}
\end{cases}
$$

**Step 1: Show $f_n \rightrightarrows f$ (Uniform Convergence)**
The difference between $f(x)$ and $f_n(x)$ is simply the "tail" of the sequence of points:
$$
|f(x) - f_n(x)| = \begin{cases} 
\frac{1}{k} & \text{if } x = \frac{1}{k} \text{ for } k > n \\
0 & \text{otherwise}
\end{cases}
$$
The maximum possible difference occurs at the largest remaining point, which is $k = n + 1$. 
Therefore, the supremum norm is $\sup_{x \in [0,1]} |f(x) - f_n(x)| = \frac{1}{n+1}$.
Since $\lim_{n \to \infty} \frac{1}{n+1} = 0$, the sequence $f_n$ converges uniformly to $f$.

**Step 2: Show $f$ is not piecewise continuous**
The limit function $f(x)$ has discontinuities at every point $x = \frac{1}{k}$ for all integers $k \ge 2$. Because the points $\frac{1}{k}$ accumulate at $0$, there are infinitely many discontinuities in the interval $[0, 1]$. By definition, a piecewise continuous function on a closed interval can only have finitely many discontinuities. Therefore, $f$ is not piecewise continuous. $\blacksquare$
:::

::: {.callout-note}
## Intuition: The $\varepsilon/3$ Trick
Part (a) uses the ubiquitous "$\varepsilon/3$ argument." To show that the limit function $f$ doesn't jump too much near $x_0$, we detour through an intermediate function $f_n$ that we *know* behaves well. 

Imagine $f(x)$ and $f(x_0)$ as two points across a river. We can't step between them directly, but we can step from $f(x)$ to a stepping stone $f_n(x)$ (close by uniform convergence), then walk along the continuous path of $f_n$ to the stone $f_n(x_0)$ (close by continuity), and finally step back onto the shore to $f(x_0)$ (close by uniform convergence). If each of the three steps is smaller than $\varepsilon/3$, the total distance is bounded by $\varepsilon$.
:::

::: {.callout-note}
## Intuition: Why Uniform Convergence Fails to Preserve "Piecewise" Continuity
In Part (b), we see a subtle limitation of [uniform convergence](function-spaces.ipynb#uniform-convergence-the-strong-form). Uniform convergence tightly controls the *height* or *size* of any "errors" or "jumps" between $f_n$ and $f$. It forces the sequence to squash down into a tight $\varepsilon$-tube around $f$. 

However, uniform convergence does *not* control the *horizontal density* of those jumps. As long as the height of the newly introduced discontinuities gets arbitrarily small (like the jumps of size $1/k$ as $k \to \infty$), we can pack infinitely many of them into a finite space without breaking the uniform convergence bound.
:::


## Problem 3: Pugh, Ch 2, P 4

**Problem Statement:**
Consider a sequence of functions $f_n$. 

**Part (a):** Let $f_n : \mathbb{R} \to \mathbb{R}$ be a sequence of functions where each $f_n$ is [uniformly continuous](uniform-continuity.ipynb#the-definition-of-uniform-continuity). Suppose the sequence is [uniformly convergent](function-spaces.ipynb#uniform-convergence-the-strong-form) to a limit function $f$, meaning $f_n \rightrightarrows f$ as $n \to \infty$. Prove or disprove that the limit function $f$ must also be uniformly continuous.
**Part (b):** Determine whether the result from Part (a) still holds if the functions map from an arbitrary metric space $M$ to another arbitrary metric space $N$ (i.e., $f_n : M \to N$), rather than mapping from $\mathbb{R}$ to $\mathbb{R}$.

::: {.callout-tip collapse="true"}
## Solution

**Part (a): Proof of Uniform Continuity for $\mathbb{R} \to \mathbb{R}$**

The statement is **true**. We will prove it using an $\varepsilon/3$ argument, heavily relying on the triangle inequality.

We need to show that for any given $\varepsilon > 0$, there exists a $\delta > 0$ such that for all $x, y \in \mathbb{R}$, $|x - y| < \delta \implies |f(x) - f(y)| < \varepsilon$. 

Let $\varepsilon > 0$ be given.

**Step 1: Fix a uniformly close approximant function.**
Because $f_n \rightrightarrows f$, there exists an integer $K \in \mathbb{N}$ such that for all $n \ge K$ and for all $x \in \mathbb{R}$:
$$|f_n(x) - f(x)| < \frac{\varepsilon}{3}$$
We fix this specific $n=K$. This bound holds for all points in $\mathbb{R}$, so it holds for our arbitrary points $x$ and $y$:
$$|f_K(x) - f(x)| < \frac{\varepsilon}{3} \quad \text{and} \quad |f_K(y) - f(y)| < \frac{\varepsilon}{3}$$

**Step 2: Use the uniform continuity of the chosen approximant.**
By the problem's premise, the specific function $f_K$ is uniformly continuous on $\mathbb{R}$. Therefore, there exists a $\delta > 0$ such that for *any* $x, y \in \mathbb{R}$ satisfying $|x - y| < \delta$, we have:
$$|f_K(x) - f_K(y)| < \frac{\varepsilon}{3}$$

**Step 3: Combine bounds via the triangle inequality.**
Let $x, y \in \mathbb{R}$ be any points such that $|x - y| < \delta$. We can bound the distance between $f(x)$ and $f(y)$ by inserting the function $f_K$:
$$
\begin{aligned}
|f(x) - f(y)| &= |f(x) - f_K(x) + f_K(x) - f_K(y) + f_K(y) - f(y)| \\
&\le |f(x) - f_K(x)| + |f_K(x) - f_K(y)| + |f_K(y) - f(y)|
\end{aligned}
$$
Substituting our bounds from Steps 1 and 2:
$$
\begin{aligned}
|f(x) - f(y)| &< \frac{\varepsilon}{3} + \frac{\varepsilon}{3} + \frac{\varepsilon}{3} \\
&= \varepsilon
\end{aligned}
$$
Because this $\delta$ depends only on $\varepsilon$ (and our choice of $K$, which in turn only depended on $\varepsilon$), and it works simultaneously for all $x, y \in \mathbb{R}$, we have proven that $f$ is uniformly continuous. $\blacksquare$

---

**Part (b): Extension to Arbitrary Metric Spaces**

The result **remains true** for mappings between arbitrary metric spaces $(M, d_M)$ and $(N, d_N)$. The proof structure is completely identical to Part (a), with the absolute value notation replaced by the respective metric distance functions.

**Proof:**
Let $\varepsilon > 0$. We want to find a $\delta > 0$ such that for all $x, y \in M$, $d_M(x, y) < \delta \implies d_N(f(x), f(y)) < \varepsilon$.

1. Since $f_n \rightrightarrows f$ uniformly on $M$, there exists a $K \in \mathbb{N}$ such that for all $z \in M$, $d_N(f_K(z), f(z)) < \frac{\varepsilon}{3}$.
2. Since $f_K : M \to N$ is uniformly continuous, there exists a $\delta > 0$ such that for all $x, y \in M$ with $d_M(x, y) < \delta$, we have $d_N(f_K(x), f_K(y)) < \frac{\varepsilon}{3}$.
3. Let $x, y \in M$ such that $d_M(x, y) < \delta$. By the triangle inequality in the metric space $N$:
   $$
   \begin{aligned}
   d_N(f(x), f(y)) &\le d_N(f(x), f_K(x)) + d_N(f_K(x), f_K(y)) + d_N(f_K(y), f(y)) \\
   &< \frac{\varepsilon}{3} + \frac{\varepsilon}{3} + \frac{\varepsilon}{3} \\
   &= \varepsilon
   \end{aligned}
   $$
Thus, $f$ is uniformly continuous on the arbitrary metric space $M$. $\blacksquare$
:::

::: {.callout-note}
## Intuition: Uniformity Preserves Uniformity
Notice the striking similarity between this proof and the proof that uniform convergence preserves standard continuity at a single point (from the previous problem set). 

The $\varepsilon/3$ trick works here to prove *uniform* continuity because neither Step 1 (the uniform convergence bound) nor Step 2 (the uniform continuity of $f_K$) restrict us to a specific anchor point $x_0$. Because both ingredients apply globally to all $x$ and $y$ at once, their combination natively yields a global (uniform) continuity bound for the limit function $f$.
:::


## Problem 4: Pugh, Ch 2, P 8

**Problem Statement:**
Consider the sequence of functions $f_n : \mathbb{R} \to \mathbb{R}$ defined by the equation:
$$
f_n(x) = \cos(n + x) + \log\left(1 + \frac{1}{\sqrt{n + 2}} \sin^2(n^n x)\right)
$$
Determine whether this sequence of functions is [equicontinuous](equicontinuity-Arzela-Ascoli-Thm.ipynb.ipynb#equicontinuity) on $\mathbb{R}$. Provide a full formal proof to justify your conclusion.

::: {.callout-tip collapse="true"}
## Solution

The statement is **true**; the sequence $f_n$ is indeed equicontinuous on $\mathbb{R}$.

To prove this, we will decompose $f_n(x)$ into the sum of two sequences of functions, $h_n(x)$ and $g_n(x)$, and prove that both are equicontinuous. The sum of two equicontinuous sequences is equicontinuous.

Let $f_n(x) = h_n(x) + g_n(x)$, where:

* $h_n(x) = \cos(n + x)$
* $g_n(x) = \log\left(1 + \frac{1}{\sqrt{n + 2}} \sin^2(n^n x)\right)$

**Step 1: Equicontinuity of $h_n(x)$**
For any $n \in \mathbb{N}$ and any $x \in \mathbb{R}$, the derivative of $h_n(x)$ is $h_n'(x) = -\sin(n + x)$.
Because $|h_n'(x)| \le 1$ for all $x$ and all $n$, by the [Mean Value Theorem]() we have:
$$|h_n(x) - h_n(y)| \le 1 \cdot |x - y|$$
This shows that $h_n(x)$ is uniformly Lipschitz continuous with a constant independent of $n$. Thus, for any $\varepsilon > 0$, choosing $\delta_1 = \varepsilon / 2$ guarantees that $|h_n(x) - h_n(y)| < \varepsilon / 2$ whenever $|x - y| < \delta_1$, for all $n$. Therefore, $\{h_n\}$ is equicontinuous.

**Step 2: Equicontinuity of $g_n(x)$**
The function $g_n(x)$ oscillates extremely rapidly due to the $n^n x$ term inside the sine function. However, we must look at its amplitude. Since $0 \le \sin^2(n^n x) \le 1$ for all $x$, we can bound $g_n(x)$ from above and below:
$$0 \le g_n(x) \le \log\left(1 + \frac{1}{\sqrt{n + 2}}\right)$$
Notice that as $n \to \infty$, the maximum amplitude $\log\left(1 + \frac{1}{\sqrt{n + 2}}\right) \to 0$. This means $g_n \rightrightarrows 0$ uniformly on $\mathbb{R}$.

Let $\varepsilon > 0$ be given. Because $g_n$ converges uniformly to $0$, there exists an integer $N \in \mathbb{N}$ such that for all $n \ge N$ and all $x \in \mathbb{R}$:
$$|g_n(x)| < \frac{\varepsilon}{4}$$
By the [triangle inequality](), for all $n \ge N$ and for any $x, y \in \mathbb{R}$:
$$|g_n(x) - g_n(y)| \le |g_n(x)| + |g_n(y)| < \frac{\varepsilon}{4} + \frac{\varepsilon}{4} = \frac{\varepsilon}{2}$$
Notice this holds *regardless* of the distance between $x$ and $y$.

Now we only need to worry about the finite number of functions $g_1, g_2, \dots, g_{N-1}$. Each of these functions has a continuous, bounded derivative on $\mathbb{R}$ (because they are compositions of bounded periodic functions with smooth functions). Thus, each individual $g_k$ (for $k < N$) is uniformly continuous.
Since there are only finitely many of them, we can find a single $\delta_2 > 0$ such that for any $k < N$, $|x - y| < \delta_2 \implies |g_k(x) - g_k(y)| < \varepsilon / 2$. 

Combining the finite case and the tail case, if $|x - y| < \delta_2$, then $|g_n(x) - g_n(y)| < \varepsilon / 2$ for *all* $n \in \mathbb{N}$. Thus, $\{g_n\}$ is equicontinuous.

**Step 3: Combine via Triangle Inequality**
Let $\varepsilon > 0$. Choose $\delta = \min(\delta_1, \delta_2)$ based on the bounds established above. For any $x, y \in \mathbb{R}$ such that $|x - y| < \delta$, and for any $n \in \mathbb{N}$:
$$
\begin{aligned}
|f_n(x) - f_n(y)| &= |(h_n(x) + g_n(x)) - (h_n(y) + g_n(y))| \\
&\le |h_n(x) - h_n(y)| + |g_n(x) - g_n(y)| \\
&< \frac{\varepsilon}{2} + \frac{\varepsilon}{2} = \varepsilon
\end{aligned}
$$
This satisfies the exact definition of equicontinuity. $\blacksquare$
:::

::: {.callout-note}
## Intuition: The "Red Herring" of Rapid Oscillation
At first glance, the term $\sin^2(n^n x)$ looks like a trap designed to break equicontinuity. As $n$ grows, this sine wave wiggles infinitely fast, meaning the derivative (the slope) of $f_n$ blows up to infinity. Usually, wildly oscillating functions fail to be equicontinuous.

However, the amplitude of these wiggles is throttled by the $\log(1 + 1/\sqrt{n+2})$ envelope. The height of the waves is shrinking to exactly zero uniformly. If a function is globally trapped inside an incredibly thin "tube" of height $\varepsilon$, it literally cannot vary by more than $\varepsilon$ between any two points, no matter how violently it wiggles back and forth inside that tube!
:::

::: {.callout-warning}
## Common Misconception: Unbounded Derivatives
A very common mistake in analysis is assuming that a sequence of differentiable functions must have uniformly bounded derivatives to be equicontinuous. 

While uniformly bounded derivatives *guarantee* equicontinuity (via the Mean Value Theorem, as we saw with the $\cos(n+x)$ term), they are **not necessary**. The sequence $g_n(x)$ in this problem is a perfect counterexample: its derivatives $|g_n'(x)| \to \infty$ as $n \to \infty$, yet the sequence remains equicontinuous because the total variation is squashed by the vanishing amplitude.
:::
