# Properties of Probability

By accepting the three axioms of [probability](probability.ipynb) (Non-negativity, Normalization, and Countable Additivity), we can derive several fundamental properties. For any [events](event.ipynb) $A$ and $B$, the following hold:

## 1\. Complement Property

**Theorem:** $P(A^c) = 1 - P(A)$

**Proof:**
Since $A$ and its complement $A^c$ are disjoint ($A \cap A^c = \emptyset$), and their union is the entire sample space ($A \cup A^c = S$), we can apply the Additivity axiom:
$$P(S) = P(A \cup A^c) = P(A) + P(A^c)$$
By the Normalization axiom, $P(S) = 1$. Therefore:
$$1 = P(A) + P(A^c)$$
Rearranging yields $P(A^c) = 1 - P(A)$. $\square$

-----

## 2\. Monotonicity of Probability Measure

**Theorem:** If $A \subseteq B$, then $P(A) \leq P(B)$.

**Proof:**
Given $A \subseteq B$, we can partition $B$ into two disjoint events: $A$, and the part of $B$ that is not in $A$ ($B \cap A^c$).
Because $A$ and $B \cap A^c$ are mutually exclusive:
$$P(B) = P\left(A \cup (B \cap A^c)\right) = P(A) + P(B \cap A^c)$$
By the Non-negativity axiom, probability is never negative, so $P(B \cap A^c) \geq 0$. Therefore, adding this non-negative value to $P(A)$ means:
$$P(B) \geq P(A)$$
$\square$

-----

## 3\. Addition Rule for Two Events

**Theorem:** $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

**Proof:**
We can express the union $A \cup B$ as the union of two disjoint sets: $A$ and $(B \cap A^c)$.
Applying the Additivity axiom gives:
$$P(A \cup B) = P(A) + P(B \cap A^c) \quad \text{--- (Equation 1)}$$

To evaluate $P(B \cap A^c)$, observe that the event $B$ itself can be split into two disjoint pieces: the part of $B$ inside $A$, and the part outside $A$.
$$B = (A \cap B) \cup (B \cap A^c)$$
Applying Additivity again:
$$P(B) = P(A \cap B) + P(B \cap A^c)$$
Rearranging this to solve for the piece we need:
$$P(B \cap A^c) = P(B) - P(A \cap B)$$

Finally, substitute this result back into Equation 1:
$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$
$\square$
