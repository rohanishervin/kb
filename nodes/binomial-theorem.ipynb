# Binomial Theorem


The binomial theorem states that

$$
(x + y)^n = \sum_{k=0}^{n} \binom{n}{k} x^k y^{n-k}
$$

where $\binom{n}{k}$ is the [binomial coefficient](binomial-coefficient.ipynb), given by:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

# Proof:

To prove the binomial theorem for any nonnegative integer $n$, consider expanding the product $(x + y)(x + y)\cdots(x + y)$ where there are $n$ factors of $(x + y)$.

Just as the expansion of $(a + b)(c + d) = ac + ad + bc + bd$ results in terms where we pick either $a$ or $b$ from the first factor and either $c$ or $d$ from the second factor, the 
terms of $(x + y)^n$ are obtained by picking either $x$ or $y$ (but not both) from each factor.

To be more precise, there are $\binom{n}{k}$ ways to choose exactly $k$ factors that contribute an $x$, and the remaining $(n - k)$ factors will contribute a $y$. Each such choice yields 
the term $x^k y^{n-k}$.

Therefore, the binomial theorem follows:
$$
(x + y)^n = \sum_{k=0}^{n} \binom{n}{k} x^k y^{n-k}
$$
