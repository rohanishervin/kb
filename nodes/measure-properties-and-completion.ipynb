# Properties of Measures and Measure Spaces

## 1\. Finite and $\sigma$-Finite Measures

When working with a [measure space](measure.ipynb) $(X, \mathcal{M}, \mu)$, it is often necessary to classify the "size" of the measure itself. 

**Definition (Finite and $\sigma$-Finite Measures):**
Let $(X, \mathcal{M}, \mu)$ be a measure space. 

* The measure $\mu$ is called **finite** provided that $\mu(X) < \infty$. 
* The measure $\mu$ is called **$\sigma$-finite** if $X$ can be expressed as the union of a countable collection of measurable sets, each having finite measure. 

We can apply these exact same classifications to individual sets within the space. A measurable set $E$ is of finite measure if $\mu(E) < \infty$, and it is $\sigma$-finite if $E$ is the union of a countable collection of measurable sets, each of finite measure. 

::: {.callout-note}
## Intuition: Disjoint Covers in $\sigma$-Finiteness
When proving properties about $\sigma$-finite spaces, it is highly convenient to assume the countable cover is pairwise disjoint. If we have a countable cover $\{X_k\}_{k=1}^\infty$ where $\mu(X_k) < \infty$, we can always construct a disjoint cover by letting the first set be $X_1$, the second be $X_2 \setminus X_1$, the third be $X_3 \setminus (X_1 \cup X_2)$, and so on. Because subsets of finite measure sets also have finite measure, this new disjoint collection will still consist entirely of finite measure sets.
:::

**Examples:**

1. [Lebesgue measure]() on the interval $[0, 1]$ is a finite measure. 
2. Lebesgue measure on the entire real line $(-\infty, \infty)$ is not finite, but it is $\sigma$-finite because we can write $\mathbb{R} = \bigcup_{n=1}^\infty [-n, n]$, and each interval $[-n, n]$ has finite measure. 
3. The [counting measure](measure.ipynb#common-examples-of-measures) on an uncountable set is an example of a measure that is **not** $\sigma$-finite.

## 2\. Probability Measures

A special and ubiquitous case of a finite measure is a probability measure.

**Definition (Probability Measure):**
An arbitrary measure $\mu$ on a space $X$ is called a **probability measure** if the measure of the entire space is exactly $1$, i.e., $\mu(X) = 1$. 

In probability theory, this measure is typically denoted by $\mathbb{P}$ rather than $\mu$. 

::: {.callout-note}
## Terminology: Measure Theory vs. Probability Theory
There is a direct dictionary translating measure-theoretic concepts into probabilistic ones. For example, if a property holds everywhere except on a set of measure zero, we say it holds [almost everywhere](continuity-of-measure.ipynb#almost-everywhere-terminology) in measure theory. In probability theory, if an event occurs with probability $1$ (meaning the set where it fails has probability measure zero, $\mathbb{P}(E^c) = 0$), we say the event occurs **almost surely**. 
:::

## 3\. Complete Measure Spaces

A minor nuisance in measure theory is the existence of sets of measure zero that have subsets which are technically not measurable. Completeness fixes this.

**Definition (Complete Measure Space):**
A measure space $(X, \mathcal{M}, \mu)$ is said to be **complete** if the [$\sigma$-algebra](sigma-algebra.ipynb) $\mathcal{M}$ contains all subsets of sets of measure zero. That is, if $E \in \mathcal{M}$ with $\mu(E) = 0$, and $A \subseteq E$, then it is required that $A \in \mathcal{M}$. 

::: {.callout-warning}
## Borel vs. Lebesgue Completeness
Standard Lebesgue measure on the real line is complete. However, if we restrict Lebesgue measure strictly to the $\sigma$-algebra of [Borel sets](), the space is no longer complete. This is because the [Cantor set]() is a Borel set with Lebesgue measure zero, but it contains subsets that are not Borel measurable. 
:::

Fortunately, any measure space that is not complete can be easily extended to one that is.

**Proposition 3 (Completion of a Measure Space):**
Let $(X, \mathcal{M}, \mu)$ be a measure space. Define $\mathcal{M}_0$ to be the collection of all subsets $E \subseteq X$ of the form $E = A \cup B$, where $B \in \mathcal{M}$ and $A \subseteq C$ for some $C \in \mathcal{M}$ for which $\mu(C) = 0$. 

For such a set $E$, define $\mu_0(E) = \mu(B)$. Then:

1. $\mathcal{M}_0$ is a $\sigma$-algebra that contains $\mathcal{M}$. 
2. $\mu_0$ is a measure that extends $\mu$. 
3. $(X, \mathcal{M}_0, \mu_0)$ is a complete measure space. 

The space $(X, \mathcal{M}_0, \mu_0)$ is formally called the **completion** of the measure space $(X, \mathcal{M}, \mu)$.
