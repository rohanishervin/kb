# Moment Generating Functions

## 1\. Introduction and Definition

A generating function acts as a mathematical "clothesline" to display a sequence of numbers, bridging combinatorics, probability, and calculus. In probability, the **moment generating function (MGF)** is a specific type of generating function that encodes the moments of a probability distribution. 

**Definition (Moment Generating Function):**
The moment generating function of a [random variable](random-variables.ipynb) $X$ is defined as the function $M(t)$:
$$
M(t) = E(e^{tX})
$$
if this [expected value](expectation.ipynb#the-concept-of-expected-value) is finite on some open interval $(-a, a)$ containing $0$. If it is not finite on such an interval, we say the MGF of $X$ does not exist.

::: {.callout-note}
## Intuition: The Variable $t$
You might wonder what the variable $t$ represents physically or probabilistically. The answer is nothing in particular; $t$ is purely a bookkeeping device. Introducing $t$ allows us to leverage the tools of continuous calculus (like derivatives) to manage and manipulate discrete sequences of moments.
:::

::: {.callout-warning}
## Checking the MGF
For any valid MGF $M$, it must be true that $M(0) = 1$. This is because $M(0) = E(e^{0 \cdot X}) = E(1) = 1$. Whenever you compute an MGF, plugging in $0$ is an excellent quick check for algebraic mistakes.
:::

::: {.callout-warning}
## Existence of the MGF vs. Characteristic Function
Not all random variables have an MGF. Even if all moments of $X$ exist, the moments might grow too quickly for the MGF to converge. However, introducing an imaginary number $i = \sqrt{-1}$ yields the **characteristic function** $\psi(t) = E(e^{itX})$ (or Fourier transform), which *always* exists.
:::

## 2\. Fundamental Properties and Theorems

The MGF is a powerful tool for three main reasons: it encodes the moments, it uniquely determines the distribution, and it simplifies finding the distribution of sums of independent random variables.

**Theorem 2.1 (Moments via derivatives of the MGF):**
Given the MGF of $X$, the $n$-th moment of $X$ can be found by evaluating the $n$-th derivative of the MGF at $0$:
$$
E(X^n) = M^{(n)}(0)
$$

::: {.callout-tip collapse="true"}
## Proof of Moments via derivatives of the MGF
**Step 1:** Consider the Taylor series expansion of $M(t)$ about $0$:
$$
M(t) = \sum_{n=0}^{\infty} M^{(n)}(0) \frac{t^n}{n!}
$$

**Step 2:** By definition, expand $e^{tX}$ using its Taylor series inside the expectation:
$$
M(t) = E(e^{tX}) = E\left(\sum_{n=0}^{\infty} X^n \frac{t^n}{n!}\right)
$$

**Step 3:** By linearity of expectation (and valid technical conditions ensuring we can swap the expectation and the infinite sum due to finiteness in the interval around $0$), we distribute the expectation:
$$
M(t) = \sum_{n=0}^{\infty} E(X^n) \frac{t^n}{n!}
$$

**Step 4:** Matching the coefficients of $t^n / n!$ in the two expansions gives $E(X^n) = M^{(n)}(0)$. $\blacksquare$
:::

::: {.callout-note}
## Intuition: Bypassing LOTUS
For a continuous random variable $X$, computing moments usually requires evaluating difficult integrals via the [Law of the Unconscious Statistician (LOTUS)](lotus.ipynb#law-of-the-unconscious-statistician-lotus). This theorem is highly surprising: it replaces complex integration with straightforward differentiation.
:::

---

**Theorem 2.2 (MGF determines the distribution):**
The MGF of a random variable uniquely determines its distribution. If two random variables have the same MGF on even a tiny interval $(-a, a)$ containing $0$, they must have the exact same distribution. *(Note: The proof involves advanced analysis and is omitted here)*.

---

**Theorem 2.3 (MGF of a sum of independent random variables):**
If $X$ and $Y$ are independent, then the MGF of $X+Y$ is the product of their individual MGFs:
$$
M_{X+Y}(t) = M_X(t)M_Y(t)
$$

::: {.callout-tip collapse="true"}
## Proof of Theorem 2.3
By definition, $M_{X+Y}(t) = E(e^{t(X+Y)}) = E(e^{tX}e^{tY})$. Because $X$ and $Y$ are independent, $e^{tX}$ and $e^{tY}$ are independent functions of $X$ and $Y$. The expectation of a product of independent random variables is the product of their expectations, yielding $E(e^{tX})E(e^{tY}) = M_X(t)M_Y(t)$. $\blacksquare$
:::

---

**Proposition 2.4 (MGF of location-scale transformation):**
If $X$ has MGF $M(t)$, then the MGF of a linear transformation $a+bX$ is:
$$
E(e^{t(a+bX)}) = e^{at} E(e^{btX}) = e^{at} M(bt)
$$

## 3\. Generating Moments via Taylor Expansions

While Theorem 2.1 states we can differentiate the MGF, sometimes it is easier to read the moments directly from the Taylor series expansion. This allows us to find *all* moments simultaneously.

**Example 1: Moments of the Exponential Distribution**
Let $X \sim \text{Expo}(1)$. The MGF is $M(t) = \frac{1}{1-t}$ for $t < 1$. Recognizing this as a geometric series, we expand it for $|t| < 1$:
$$
M(t) = \frac{1}{1-t} = \sum_{n=0}^{\infty} t^n = \sum_{n=0}^{\infty} n! \frac{t^n}{n!}
$$
Matching coefficients with $M(t) = \sum_{n=0}^{\infty} E(X^n) \frac{t^n}{n!}$, we immediately see that $E(X^n) = n!$ for all $n$.
By scale transformation, for $Y = X/\lambda \sim \text{Expo}(\lambda)$, $E(Y^n) = \frac{n!}{\lambda^n}$.

**Example 2: Moments of the Standard Normal**
For $Z \sim \mathcal{N}(0,1)$, $M(t) = e^{t^2/2}$. Expanding via Taylor series:
$$
M(t) = e^{t^2/2} = \sum_{n=0}^{\infty} \frac{(t^2/2)^n}{n!} = \sum_{n=0}^{\infty} \frac{t^{2n}}{2^n \cdot n!} = \sum_{n=0}^{\infty} \frac{(2n)!}{2^n \cdot n!} \frac{t^{2n}}{(2n)!}
$$
Thus, the even moments are $E(Z^{2n}) = \frac{(2n)!}{2^n \cdot n!}$, and all odd moments are $0$ by symmetry. This yields $E(Z^2)=1$, $E(Z^4)=3$, $E(Z^6)=15$, etc.

**Example 3: Moments of the Log-Normal Distribution**
A random variable $Y \sim \mathcal{LN}(\mu, \sigma^2)$ is defined as $Y = e^X$ where $X \sim \mathcal{N}(\mu, \sigma^2)$. The MGF of $Y$ *does not exist* because the integral diverges for any $t>0$. However, we can use the Normal MGF to find its moments:
$$
E(Y^n) = E(e^{nX}) = M_X(n) = e^{n\mu + \frac{1}{2}n^2\sigma^2}
$$

::: {.callout-warning}
## Nomenclature Pitfall: Log-Normal
"Log-Normal" does **not** mean "the log of a Normal distribution", because a Normal variable can be negative and its log would be undefined. Rather, it means **"the log is Normal"**. Note that the parameters $\mu$ and $\sigma^2$ describe the mean and variance of the *underlying* Normal distribution, not the Log-Normal distribution itself.
:::

## 4\. Deriving the Distribution of Sums via MGFs

Using Theorem 2.3, we can quickly prove that certain distributions are closed under addition. 

**Example: Sum of Independent Poissons**
Let $X \sim \text{Pois}(\lambda)$ and $Y \sim \text{Pois}(\mu)$ be independent. 

1. The MGF of $X$ is $M_X(t) = e^{\lambda(e^t - 1)}$.
2. The MGF of $Y$ is $M_Y(t) = e^{\mu(e^t - 1)}$.
3. The MGF of $X+Y$ is their product:
$$
M_{X+Y}(t) = e^{\lambda(e^t - 1)} e^{\mu(e^t - 1)} = e^{(\lambda+\mu)(e^t - 1)}
$$

Because this is exactly the MGF of a $\text{Pois}(\lambda+\mu)$ distribution, and MGFs uniquely identify distributions, $X+Y \sim \text{Pois}(\lambda+\mu)$.

**Example: Sum of Independent Normals**
Let $X_1 \sim \mathcal{N}(\mu_1, \sigma_1^2)$ and $X_2 \sim \mathcal{N}(\mu_2, \sigma_2^2)$ be independent.
$$
M_{X_1+X_2}(t) = e^{\mu_1 t + \frac{1}{2}\sigma_1^2 t^2} \cdot e^{\mu_2 t + \frac{1}{2}\sigma_2^2 t^2} = e^{(\mu_1+\mu_2)t + \frac{1}{2}(\sigma_1^2+\sigma_2^2)t^2}
$$
This matches the $\mathcal{N}(\mu_1+\mu_2, \sigma_1^2+\sigma_2^2)$ MGF, proving that the sum of independent Normals is Normal, and their means and variances simply add.

::: {.callout-note}
## Intuition: Cramér's Theorem
A fascinating converse holds true: if $X_1$ and $X_2$ are independent and their sum $X_1+X_2$ is Normal, then $X_1$ and $X_2$ *must individually be Normal*. This is known as **Cramér's theorem**.
:::

## 5\. Catalog of Common MGFs

Here is a summary list of common distributions and their Moment Generating Functions derived in the text:

*   **Bernoulli Distribution ($X \sim \text{Bern}(p)$):**
    $M(t) = pe^t + q$, defined on $(-\infty, \infty)$.
*   **Geometric Distribution ($X \sim \text{Geom}(p)$):**
    $M(t) = \frac{p}{1-qe^t}$ for $t \in (-\infty, \log(1/q))$.
*   **Uniform Distribution ($U \sim \text{Unif}(a,b)$):**
    $M(t) = \frac{e^{tb} - e^{ta}}{t(b-a)}$ for $t \ne 0$.
*   **Binomial Distribution ($X \sim \text{Bin}(n,p)$):**
    $M(t) = (pe^t + q)^n$.
*   **Negative Binomial Distribution ($X \sim \text{NBin}(r,p)$):**
    $M(t) = \left(\frac{p}{1-qe^t}\right)^r$ for $qe^t < 1$.
*   **Standard Normal Distribution ($Z \sim \mathcal{N}(0,1)$):**
    $M(t) = e^{t^2/2}$.
*   **Normal Distribution ($X \sim \mathcal{N}(\mu, \sigma^2)$):**
    $M(t) = e^{\mu t + \frac{1}{2}\sigma^2 t^2}$.
*   **Exponential Distribution ($X \sim \text{Expo}(\lambda)$):**
    $M(t) = \frac{\lambda}{\lambda - t}$ for $t < \lambda$.
