## 1\. What is a Difference Equation?

While differential equations describe continuous change (how things change from one infinitesimally small moment to the next), **difference equations** describe discrete change. They model phenomena that evolve in distinct steps or stages—like compounding interest yearly, animal population counts every spring, or stepping through a sequence of coin flips in a game.

Formally, a difference equation is an expression that relates the value of a sequence $x(n)$ at discrete time $n$ to its values at previous times $n-1, n-2, \dots$

The general form of a difference equation is:
$$f(n, x(n), x(n+1), \dots, x(n+k)) = 0$$
where $n \in \mathbb{Z}^+$ and $f:\mathbb{Z}^+ \times \mathbb{R}^k  \to \mathbb{R}$.

## 2\. Order of a Difference Equation

The **order** of a difference equation is determined by the difference between the highest and lowest time arguments of the sequence $x$ present in the equation.

$$\text{Order} = (\text{Highest Argument}) - (\text{Lowest Argument})$$

**Examples:**

1.  $x(n+1) - 3x(n) = 0$
      * Highest: $n+1$, Lowest: $n$. Order = $(n+1) - n = \mathbf{1}$.
2.  $x(n+2) + 5x(n+1) + x(n) = n^2$
      * Highest: $n+2$, Lowest: $n$. Order = $(n+2) - n = \mathbf{2}$.

::: {.callout-note}

## Intuition: Order = Memory

You can intuitively think of the "order" as the system's memory. A first-order system only needs to remember exactly *one* step into the past to know what to do next. A second-order system needs to remember two steps into the past.
:::

## 3\. Linear vs. Nonlinear

A difference equation is **linear** if the sequence terms $x(n), x(n+1), \dots$ appear only to the first power and are not multiplied by each other or passed through non-linear functions (like $\sin$ or $e^x$).

  * **Linear:** $x(n+1) = a(n)x(n) + g(n)$
  * **Nonlinear:** $x(n+1) = a x(n)^2$  or  $x(n+1) = \cos(x(n))$

If the term $g(n) = 0$ for all $n$, the equation is called **homogeneous**. If $g(n) \neq 0$, it is **non-homogeneous**.
