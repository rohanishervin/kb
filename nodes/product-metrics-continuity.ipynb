# Product Metrics and Continuity

When we take the Cartesian product of two [metric spaces](metric-spaces.ipynb), $M = X \times Y$, we need a way to measure the distance between two points $p = (x, y)$ and $p' = (x', y')$ in this new combined space. 

There are three natural ways to define this product metric:

1. **The Euclidean Metric:** $d_E(p, p') = \sqrt{d_X(x, x')^2 + d_Y(y, y')^2}$
2. **The Max Metric:** $d_{\max}(p, p') = \max\{d_X(x, x'), d_Y(y, y')\}$
3. **The Sum (Taxicab/Manhattan) Metric:** $d_{\text{sum}}(p, p') = d_X(x, x') + d_Y(y, y')$

## 1. Equivalence of Product Metrics

Even though these three formulas yield different exact distances, they are bounded by each other. This means they are **topologically equivalent**—if a sequence converges under one, it is mathematically guaranteed to converge under the others.

**Proposition 16:** For any points in the product space, the metrics are bounded by the following inequality:
$$d_{\max} \leq d_E \leq d_{\text{sum}} \leq 2d_{\max}$$

**Proof:**

* $d_{\max} \leq d_E$: Dropping the smaller squared term from inside the square root of $d_E$ leaves exactly $\sqrt{\max^2}$, which is $d_{\max}$.
* $d_E \leq d_{\text{sum}}$: Squaring $d_{\text{sum}}$ yields $d_X^2 + d_Y^2 + 2d_Xd_Y$. Squaring $d_E$ yields $d_X^2 + d_Y^2$. Since distances are always non-negative, the cross term $2d_Xd_Y \geq 0$, proving $d_E^2 \leq d_{\text{sum}}^2$.
* $d_{\text{sum}} \leq 2d_{\max}$: The sum of two non-negative numbers can never be larger than twice the larger of the two numbers. $\square$

## 2. Convergence in Product Spaces

**Theorem 17:** The following are strictly equivalent for a sequence $p_n = (p_{1n}, p_{2n})$ in $M = M_1 \times M_2$:

* (a) $(p_n)$ converges with respect to $d_{\max}$.
* (b) $(p_n)$ converges with respect to $d_E$.
* (c) $(p_n)$ converges with respect to $d_{\text{sum}}$.
* (d) The component sequences $(p_{1n})$ and $(p_{2n})$ converge in $M_1$ and $M_2$ respectively.

**Proof:**
The equivalence of (a), (b), and (c) follows immediately from the squeeze theorem applied to the inequalities in Proposition 16. If distance goes to $0$ in one metric, it is squeezed to $0$ in all of them. The equivalence to (d) directly follows from the definition of the metrics. $\square$

**Corollary 19 (Convergence in $\mathbb{R}^m$):** By extending the metrics to $m$ dimensions, a sequence of vectors $(v_n)$ in $\mathbb{R}^m$ converges if and only if **each of its individual component sequences** converges. 
$$v = \lim_{n \to \infty} v_n = \left(\lim_{n \to \infty} v_{1n}, \lim_{n \to \infty} v_{2n}, \dots, \lim_{n \to \infty} v_{mn}\right)$$

## 3. Continuity of Product Functions

**Corollary 18:** If $f: M \to N$ and $g: X \to Y$ are continuous functions, then their Cartesian product $f \times g$ is also continuous. 
*(Note: $f \times g$ maps a point $(p, x) \in M \times X$ to the point $(fp, gx) \in N \times Y$).*

**Proof:**
Let $(p_n, x_n) \to (p, x)$ in $M \times X$. 
By Theorem 17, this means $p_n \to p$ in $M$ and $x_n \to x$ in $X$. 
Because $f$ and $g$ are individually continuous, they preserve limits: $f(p_n) \to f(p)$ and $g(x_n) \to g(x)$. 
Applying Theorem 17 again in the target space, the combined sequence $(f(p_n), g(x_n))$ converges to $(f(p), g(x))$. Thus, $f \times g$ preserves sequential convergence and is continuous. $\square$

## 4. Continuity of Metrics and Norms

Is distance itself a continuous concept? Yes. A small wiggle in points only causes a small wiggle in their distance. 

The distance function $d: M \times M \to \mathbb{R}$ takes two points and outputs a real number. 

**Theorem 20:** The distance function $d$ is uniformly continuous.

**Proof:**
We will check the $\epsilon-\delta$ continuity using the sum metric $d_{\text{sum}}$ on the domain $M \times M$. 
Given $\epsilon > 0$, we choose $\delta = \epsilon$. 
Suppose the distance between two pairs of points, $(p, q)$ and $(p', q')$, is less than $\delta$:
$$d_{\text{sum}}((p, q), (p', q')) = d(p, p') + d(q, q') < \delta$$

By the triangle inequality:
$$d(p, q) \leq d(p, p') + d(p', q') + d(q', q)$$
Rearranging this to isolate the difference between the two main distances:
$$d(p, q) - d(p', q') \leq d(p, p') + d(q, q') < \delta = \epsilon$$

By swapping the roles of the pairs, we get the same bound for $d(p', q') - d(p, q)$, which gives us the absolute value:
$$|d(p, q) - d(p', q')| < \epsilon$$
Thus, $d$ is continuous. By Theorem 17, since it is continuous under $d_{\text{sum}}$, it is continuous under all valid product metrics. $\square$

**Corollary 21:** The specific product metrics $d_{\max}, d_E,$ and $d_{\text{sum}}$ are all continuous functions.

**Corollary 22:** The absolute value is a continuous mapping from $\mathbb{R} \to \mathbb{R}$. More generally, the norm $\|v\|$ is a continuous mapping from any normed space to $\mathbb{R}$.
**Proof:** The norm is simply the distance to the origin: $\|v\| = d(v, 0)$. Since $d$ is a continuous function, fixing one input to $0$ means the resulting norm function is also continuous. $\square$

## Intuition: Geometry of the Product Metrics
Why are the metrics topologically equivalent? Because their open sets (open balls) can always fit inside one another! 

In $\mathbb{R}^2$:

* The **Euclidean ($d_E$)** unit ball is a standard circle.
* The **Max ($d_{\max}$)** unit ball is a square aligned with the axes.
* The **Sum/Manhattan ($d_{\text{sum}}$)** unit ball is a diamond (a rotated square).

Because you can always perfectly nest a smaller circle inside the diamond, a smaller square inside that circle, and so on, an open set defined by one metric is always an open set defined by the others.
