# Field $(F, +, \cdot)$

A field is a non-empty set $F$ equipped with two binary operations:

* **Addition**: $+: F \times F \rightarrow F$
* **Multiplication**: $\cdot: F \times F \rightarrow F$

that satisfies the following properties (axioms) for all $x, y, z \in F$:

### Addition Axioms

1. **Closure**: $x + y \in F$
2. **Commutativity**: $x + y = y + x$
3. **Associativity**: $(x + y) + z = x + (y + z)$
    * *(Note: $x_1 + x_2 + \dots + x_n$ is well-defined, making parentheses unnecessary).*
4. **Additive Identity**: $\exists 0 \in F$ such that $x + 0 = x$.
    * **Uniqueness Proof**: Suppose $e$ and $e'$ are both additive identities. Then $e' = e' + e = e + e' = e$. Thus, the identity is unique and denoted as $0$.
5. **Additive Inverse**: $\forall x \in F, \exists y \in F$ such that $x + y = 0$.
    * **Uniqueness Proof**: Suppose $y$ and $y'$ are both inverses of $x$. Then $y = y + 0 = y + (x + y') = (y + x) + y' = 0 + y' = y'$. Thus, the inverse is unique and denoted as $-x$.

### Multiplication Axioms

6. **Closure**: $xy \in F$
7. **Commutativity**: $xy = yx$
8. **Associativity**: $(xy)z = x(yz)$
9. **Multiplicative Identity**: $\exists 1 \in F$ (where $1 \neq 0$) such that $x \cdot 1 = x$.
    * *(By similar logic to addition, $1$ is unique).*
10. **Multiplicative Inverse**: $\forall x \neq 0 \in F, \exists y \in F$ such that $xy = 1$.
    * *(Denoted as $x^{-1}$ or $\frac{1}{x}$).*

### Distributivity Axiom
11. **Distributivity**: $x(y + z) = xy + xz$

---

# Standard Examples

1. **Rational Numbers**: $(\mathbb{Q}, +, \cdot)$
2. **Real Numbers**: $(\mathbb{R}, +, \cdot)$
3. **Prime Fields**: $\mathbb{Z}_p = \{0, 1, \dots, p-1\}$ where $p$ is a prime number, with addition and multiplication modulo $p$.

# Fundamental Propositions

## Proposition
For any $a \in F$, $0 \cdot a = 0$.

**Proof:**
By the additive identity axiom (Axiom 4), we know $0 = 0 + 0$.
Substituting this into our expression:
$$0 \cdot a = (0 + 0) \cdot a$$
By distributivity (Axiom 11):
$$0 \cdot a = 0 \cdot a + 0 \cdot a$$
By the additive inverse axiom (Axiom 5), there exists $-(0 \cdot a)$. We add it to both sides:
$$0 \cdot a + -(0 \cdot a) = (0 \cdot a + 0 \cdot a) + -(0 \cdot a)$$
By associativity (Axiom 3) and the definition of the inverse:
$$0 = 0 \cdot a + (0 \cdot a + -(0 \cdot a))$$
$$0 = 0 \cdot a + 0$$
$$0 = 0 \cdot a \quad \blacksquare$$

---

## Proposition
For any $a, b \in F$, $(-a)(-b) = ab$.

**Proof:**
First, we establish a lemma: $(-a)b = -(ab)$. 
Using distributivity and Proposition 1: 
$$ab + (-a)b = (a + (-a))b = 0 \cdot b = 0$$
Since $ab + (-a)b = 0$, by the uniqueness of additive inverses, $(-a)b$ is the inverse of $ab$, meaning $(-a)b = -(ab)$.

Now, we evaluate $(-a)(-b)$ by adding $-(ab)$ to it:
$$(-a)(-b) + -(ab)$$
Substitute $-(ab)$ with $(-a)b$ (from our lemma):
$$= (-a)(-b) + (-a)b$$
Factor out $(-a)$ using distributivity:
$$= (-a)(-b + b)$$
$$= (-a) \cdot 0$$
By Proposition 1, this equals $0$. 
Therefore:
$$(-a)(-b) + -(ab) = 0$$
This means $(-a)(-b)$ is the additive inverse of $-(ab)$. Since the additive inverse of $-(ab)$ is simply $ab$, we conclude:
$$(-a)(-b) = ab \quad \blacksquare$$
