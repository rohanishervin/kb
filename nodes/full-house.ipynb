# Full House

A 5-card hand is dealt from a standard, well-shuffled 52-card deck. The hand is called a full house in poker if it consists of three cards of some rank and two cards of another rank, for example, three 7's 
and two 10's (in any order). What is the probability of obtaining a full house?


# Solution:
All of the $\binom{52}{5}$ possible hands are equally likely by symmetry, so the [naive definition](probability-naive-definition.ipynb) can be applied.
In mathematical terms, the number of ways to get a full house can be calculated as follows by [binomial coefficient formula](binomial-coefficient-formula.ipynb): 
$$
\text{Number of Full House Hands} = \binom{13}{1} \times \binom{4}{3} \times \binom{12}{1} \times \binom{4}{2}
$$

The total number of possible 5-card hands from a standard deck is:
$$
\binom{52}{5}
$$

Therefore, the probability \(P\) of drawing a full house is given by:
$$
P(\text{Full House}) = \frac{\binom{13}{1} \times \binom{4}{3} \times \binom{12}{1} \times \binom{4}{2}}{\binom{52}{5}}
$$
