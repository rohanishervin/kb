# Homeomorphisms and Compactness

A [**homeomorphism**](continuity-topology.ipynb#homeomorphisms) is a "bicontinuous bijection"—a one-to-one and onto mapping between two spaces where both the function $f$ and its inverse $f^{-1}$ are continuous. If a homeomorphism exists between two spaces, they are topologically identical. 

Historically, [compactness](compactness.ipynb) was so intrinsically linked to topological structure that it was originally called *bicompactness*. 

## 1. Compactness is a Topological Property

**Theorem:** If $M$ is a compact metric space, and $M$ is homeomorphic to another space $N$, then $N$ is also compact.

**Proof:** 

Since $M$ and $N$ are homeomorphic, there is a continuous bijection $f: M \to N$. This means $N$ is the continuous image of the entire space $M$ ($N = f(M)$). We previously [proved](continuity-and-compactness.ipynb#the-continuous-image-of-a-compact-set) that the continuous image of any compact set is compact. Therefore, $N$ must be compact. $\square$

This theorem provides a very fast way to prove that two spaces are *not* topologically equivalent.

**Corollary:** The closed interval $[0, 1]$ and the real line $\mathbb{R}$ are not homeomorphic.

**Proof:** 

The interval $[0, 1]$ is closed and bounded, so by the [Heine-Borel theorem](compactness.ipynb#the-heine-borel-theorem), it is compact. The real line $\mathbb{R}$ is not bounded, so it is not compact. Since one is compact and the other isn't, no homeomorphism can possibly exist between them. $\square$

## 2. The "Automatic Homeomorphism" Theorem

Proving a function is a homeomorphism usually requires doing the work twice: proving $f$ is continuous, and then separately proving $f^{-1}$ is continuous. However, if the domain is compact, you get the second half for free!

**Theorem:** If $M$ is compact, then any continuous bijection $f: M \to N$ is automatically a homeomorphism. (That is, its inverse bijection $f^{-1}: N \to M$ is guaranteed to be continuous).

**Proof:**

We must verify that $f^{-1}$ preserves sequential convergence. Suppose we have a sequence $q_n \to q$ in the target space $N$. 
Because $f$ is a bijection, every point in $N$ has exactly one pre-image in $M$. Let $p_n = f^{-1}(q_n)$ and let $p = f^{-1}(q)$. 
To prove $f^{-1}$ is continuous, we must show that $p_n \to p$ in $M$.

We will prove this by contradiction.

1. If $(p_n)$ *refuses* to converge to $p$, there must exist some "rebel" subsequence $(p_{n_k})$ that stays strictly away from $p$. That is, there is some distance $\delta > 0$ such that for all $k$, $d(p_{n_k}, p) \geq \delta$.
2. Because this rebel subsequence lives entirely in $M$, and **$M$ is compact**, the subsequence itself must contain a sub-subsequence $(p_{n_{k(\ell)}})$ that converges to some limit point $p^* \in M$ as $\ell \to \infty$.
3. Because every point in this sub-subsequence was at least $\delta$ away from $p$, the limit $p^*$ must also be at least $\delta$ away from $p$. Therefore, $d(p, p^*) \geq \delta$, which definitively means $p \neq p^*$.
4. Now, we use the fact that the original function $f$ is continuous. Since $p_{n_{k(\ell)}} \to p^*$, applying $f$ means:
   $$f(p_{n_{k(\ell)}}) \to f(p^*)$$
5. However, we defined $f(p_{n_{k(\ell)}})$ to be $q_{n_{k(\ell)}}$. Since the entire "mother sequence" $q_n \to q$, any subsequence of it must also converge to $q$.
6. This forces $f(p^*) = q$. But we already know $f(p) = q$. 
7. This implies $f(p^*) = f(p)$ even though $p \neq p^*$. This contradicts the fact that $f$ is a bijection (one-to-one)!

Because assuming $p_n$ does not converge to $p$ breaks the bijection of $f$, it must be true that $p_n \to p$. Therefore, $f^{-1}$ is continuous. $\square$

## 3. Counterexample: When the Domain is NOT Compact

If $M$ is not compact, the Theorem above is spectacularly false. A function can be a continuous bijection, but its inverse can suddenly "tear" the space.

**Example:**
Consider the half-open interval $M = [0, 2\pi)$. This domain is bounded, but not closed (it lacks $2\pi$), so it is **not compact**. 
Let $N = S^1$ (the unit circle in $\mathbb{R}^2$). 
Define the function $f: [0, 2\pi) \to S^1$ by:
$$f(x) = (\cos x, \sin x)$$

This function simply takes the line segment and wraps it exactly once around the circle. 

* It is **continuous**: Moving slightly on the line segment moves you slightly on the circle.
* It is a **bijection**: Every point on the circle is hit exactly once.

However, $f$ is **not a homeomorphism** because $f^{-1}$ fails to be continuous! 
Take a sequence of points on the circle in the 4th quadrant that creep upward toward $(1,0)$ (e.g., angles of $350^\circ, 355^\circ, 359^\circ \dots$). In the domain $[0, 2\pi)$, these pre-images approach the value $2\pi$. 
But the exact limit point $(1,0)$ on the circle maps to $0$ in the domain. 
The pre-images approach $2\pi$, but the actual value is $0$. The limit of the sequence does not equal the function of the limit. The inverse function has a massive discontinuous "jump" right at the origin! 

Because the domain was not compact, the inverse function could not be guaranteed continuity.
