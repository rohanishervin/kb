# Logic Preliminaries

This notebook serves as a foundational reference for mathematical logic, combining propositional calculus, quantifiers, and formal proof strategies.

## 1. Fundamentals of Propositional Logic

A **proposition** (or statement) is a declarative sentence that is either **True (T)** or **False (F)**. We typically label these with lowercase letters like $p, q, r$.

### 1.1. Basic Logical Operators
| Operator | Notation | English Term | Condition for Truth |
| :--- | :---: | :--- | :--- |
| **Negation** | $\neg p$ | Not $p$ | True if $p$ is false. |
| **Conjunction** | $p \land q$ | $p$ and $q$ | True only if **both** are true. |
| **Disjunction** | $p \lor q$ | $p$ or $q$ | True if **at least one** is true. |

### 1.2. Expanded Truth Table (Logical Equivalence)
The following table demonstrates the consistency of basic operators and proves the identity of the implication.

| $p$ | $q$ | $\neg p$ | $p \land q$ | $p \lor q$ | $p \implies q$ | $\neg p \lor q$ |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| T | T | F | T | T | **T** | **T** |
| T | F | F | F | T | **F** | **F** |
| F | T | T | F | T | **T** | **T** |
| F | F | T | F | F | **T** | **T** |

## 2. Conditionals and Logical Equivalence

### 2.1. The Implication ($p \implies q$)
An implication is only **False** when the hypothesis ($p$) is True but the conclusion ($q$) is False. As shown in the table above:
$$(p \implies q) \equiv (\neg p \lor q)$$

### 2.2. The Biconditional ($p \iff q$)
Known as "**if and only if**" (iff) or **Equivalence**. It is true only when $p$ and $q$ have the same truth value.
$$(p \iff q) \equiv (p \implies q) \land (q \implies p)$$

### 2.3. De Morgan's Laws
These laws describe how to distribute a negation over a conjunction or disjunction:

1.  $\neg(p \land q) \equiv \neg p \lor \neg q$
2.  $\neg(p \lor q) \equiv \neg p \land \neg q$

### 2.4. The Contrapositive
The **Contrapositive** is logically equivalent to the original implication.

* **Original:** $p \implies q$
* **Contrapositive:** $\neg q \implies \neg p$
* **Identity:** $(p \implies q) \equiv (\neg q \implies \neg p)$

## 3. Predicate Logic and Quantifiers

### 3.1. Types of Quantifiers
* **Universal Quantifier ($\forall$):** "For all", "For every".
* **Existential Quantifier ($\exists$):** "There exists", "There is at least one".

### 3.2. Negating Quantified Statements
When negating a statement with quantifiers, the quantifiers flip ($\forall \leftrightarrow \exists$) and the inner statement is negated.

**Example: The Limit Definition ($\epsilon-\delta$)**
Negating the definition of a limit is a classic exercise in real analysis.

* **Statement (Limit exists):**
    $$\forall \epsilon > 0, \exists \delta > 0, \forall x \ (0 < |x - a| < \delta \implies |f(x) - L| < \epsilon)$$
* **Negation (Limit does not exist):**
    Using the identity $\neg(p \implies q) \equiv (p \land \neg q)$:
    $$\exists \epsilon > 0, \forall \delta > 0, \exists x \ (0 < |x - a| < \delta \land |f(x) - L| \geq \epsilon)$$

## 4. Formal Proof Strategies

### 4.1. The English Proof Template
Use this structure to keep your proofs readable and professionally formatted:

* **Theorem/Proposition:** (The statement to be proven)
* **Given / Let:** (Assumptions and variable definitions)
* **To Prove / Goal:** (The desired conclusion)
* **Proof:** (The logical derivation)
    * *Claim:* (A sub-goal)
    * *Lemma:* (A small auxiliary theorem)
* **Conclusion:** (Summary and symbol) $\square$

### 4.2. Common Methods of Proof
1.  **Direct Proof:** Assume $p$ is true, use logic/definitions to show $q$ is true.
2.  **Proof by Contrapositive:** Assume $\neg q$ is true, show that $\neg p$ must follow.
3.  **Proof by Contradiction:** Assume $p$ and $\neg q$ are both true, then show this leads to an impossible result (e.g., $1=0$).
4.  **Proof by Cases:** Exhaustively check all possible scenarios (e.g., if $n$ is even, then... if $n$ is odd, then...).
5.  **Existence Proof:**
    * *Constructive:* Point to a specific value that satisfies the statement.
    * *Non-constructive:* Use a theorem (like the Mean Value Theorem) to prove it *must* exist without necessarily finding it.

### Implication Negation

$$
\begin{aligned}
\neg(p \implies q) &\equiv \neg(\neg p \lor q) & \text{(By Implication Identity)} \\
&\equiv \neg(\neg p) \land \neg q & \text{(By De Morgan's Law)} \\
&\equiv p \land \neg q & \text{(By Double Negation)}
\end{aligned}
$$
