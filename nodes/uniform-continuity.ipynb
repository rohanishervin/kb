# Uniform Continuity and Compactness

In calculus, standard continuity is a "local" property—it is defined point by point. **Uniform continuity** is a much stronger, "global" property that applies to the entire domain at once. 

## 1. The Definition of Uniform Continuity

The definition of uniform continuity in abstract [metric spaces](metric-spaces.ipynb) is completely analogous to the definition for real-valued functions.

**Definition:** A function $f: M \to N$ is **uniformly continuous** if for each $\epsilon > 0$ there exists a single $\delta > 0$ such that for *any* two points $p, q \in M$:
$$d_M(p, q) < \delta \implies d_N(f(p), f(q)) < \epsilon$$

**The Crucial Difference:**

* In standard continuity, $\delta$ depends on both the chosen tolerance $\epsilon$ **and** the specific location $p$. If the function gets exceptionally steep in certain areas, you might need a much smaller $\delta$ to satisfy the same $\epsilon$.
* In uniform continuity, $\delta$ depends **only** on $\epsilon$. One single, universal $\delta$ works simultaneously for every single point across the entire domain $M$.

---

## 2. Compactness Guarantees Uniform Continuity

If a domain is allowed to stretch out to infinity, or approach open boundaries (like asymptotes), a function can steepen infinitely, making a global $\delta$ impossible. However, if the domain is [compact](compactness.ipynb), the function is "trapped", bounding how wildly it can behave.

**Theorem (Heine-Cantor):** Every continuous function defined on a compact metric space is uniformly continuous.

**Proof (By Contradiction):**
Suppose that $f: M \to N$ is continuous, the domain $M$ is compact, but $f$ completely fails to be uniformly continuous. 

If it fails to be uniformly continuous, it means there exists some specific "bad" tolerance $\epsilon > 0$ where no global $\delta$ will work. No matter how incredibly tiny we make $\delta$, we can always find two rogue points $p, q \in M$ that are closer than $\delta$ but whose outputs map further apart than $\epsilon$.

Let us systematically shrink $\delta$ by setting $\delta = 1/n$ for each integer $n \in \mathbb{N}$. 
For each $n$, we can find a pair of points. Let $(p_n)$ and $(q_n)$ be sequences of these points in $M$ such that:
$$d_M(p_n, q_n) < \frac{1}{n} \quad \text{while} \quad d_N(f(p_n), f(q_n)) \geq \epsilon$$

Because $M$ is a compact space, the sequence $(p_n)$ must have a subsequence $(p_{n_k})$ that converges to some limit point $p \in M$ as $k \to \infty$.

Now, let's look at the corresponding subsequence $(q_{n_k})$. Because the distance $d_M(p_n, q_n) < 1/n$, the distance between them shrinks to $0$ as $n \to \infty$. Therefore, the sequence $(q_{n_k})$ is "dragged along" and must converge to the exact same limit point $p$:
$$q_{n_k} \to p \text{ as } k \to \infty$$

Now we use our assumption that $f$ is standardly continuous at $p$. Because it is continuous, it preserves sequential limits:
$$f(p_{n_k}) \to f(p) \quad \text{and} \quad f(q_{n_k}) \to f(p)$$

If both sequences of outputs are converging to the exact same point $f(p)$, the distance between them must eventually drop to zero. By the triangle inequality, if we pick a large enough index $k$:
$$d_N(f(p_{n_k}), f(q_{n_k})) \leq d_N(f(p_{n_k}), f(p)) + d_N(f(p), f(q_{n_k})) < \epsilon$$

But this gives us a massive contradiction! Our sequence construction explicitly guaranteed that $d_N(f(p_n), f(q_n)) \geq \epsilon$ for *all* $n$. 
Because assuming the function is not uniformly continuous leads to an impossible logical contradiction, the function must be uniformly continuous. $\square$

**Corollary:** Any continuous function defined on a closed interval $[a, b]$ is uniformly continuous. (Because $[a,b]$ is compact by the Heine-Borel Theorem).

## 3. When Compactness is Missing

To understand why this requires compactness, look at continuous functions on non-compact domains:

* **Domain bounded but not closed:** $f(x) = \frac{1}{x}$ on $(0, 1)$. The function is continuous. But as you approach $0$, the curve becomes infinitely steep. For a fixed $\epsilon$, the required $\delta$ window gets smaller and smaller the closer you get to 0. No single $\delta$ works everywhere.
* **Domain closed but not bounded:** $f(x) = x^2$ on $\mathbb{R}$. As $x \to \infty$, the parabola gets infinitely steep. The required $\delta$ for a fixed $\epsilon$ shrinks to zero as you move further down the x-axis.
