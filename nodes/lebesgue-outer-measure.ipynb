# Lebesgue Outer Measure

## 1\. Motivation

A [measure](measure.ipynb#measurable-spaces-and-measures) is, informally, a function $\mu$ assigning to subsets of a space a nonnegative "size" that generalizes length, area, and volume. On $\mathbb{R}$ the natural notion of size is the length of an interval; on $\mathbb{R}^2$ it is the area of a rectangle; on $\mathbb{R}^3$ it is the volume of a box; and on $\mathbb{R}^n$ the natural generalization is the volume of an $n$-dimensional box.

Any reasonable measure $\mu$ defined on a collection of sets $\mathcal{M}$ (a [$\sigma$-algebra](sigma-algebra.ipynb)) should satisfy at least:

* $\mu(\varnothing) = 0$,
* **countable additivity**: if $\{E_i\}$ are pairwise disjoint sets in $\mathcal{M}$, then $\mu\left(\bigcup_i E_i\right) = \sum_i \mu(E_i)$.

The natural question this section answers is: starting only from the length of an interval (or the volume of a box), can we extend the notion of "size" to *arbitrary* subsets of $\mathbb{R}^n$ — a triangle, a disk, or worse — in a way that is consistent with the axioms above? The first step toward such an extension is the **outer measure**, which is defined for *every* subset of $\mathbb{R}^n$, but which we will see (in a later note) is only countably additive on a suitably restricted collection of "measurable" sets.

::: {.callout-warning}
## Outer measure is not itself "the" measure
Outer measure $m^*$ is defined on *all* subsets of $\mathbb{R}^n$ and satisfies monotonicity and countable subadditivity, but it fails to be countably additive on all sets — there exist disjoint sets $A, B$ with $m^*(A \cup B) < m^*(A) + m^*(B)$. Restricting $m^*$ to the collection of Lebesgue measurable sets is what recovers full countable additivity; that restriction is the subject of a later note.
:::

## 2\. Definition of Outer Measure

### 2.1. On $\mathbb{R}$

Let $I$ be a nonempty interval of real numbers. Define its length $|I|$ to be $\infty$ if $I$ is unbounded, and otherwise the difference of its endpoints.

For $E \subseteq \mathbb{R}$, consider all countable collections $\{I_k\}_{k=1}^\infty$ of nonempty, open, bounded intervals that **cover** $E$, i.e. $E \subseteq \bigcup_{k=1}^\infty I_k$. The **outer measure** of $E$ is

$$
m^*(E) = \inf\left\{ \sum_{k=1}^\infty |I_k| \; : \; E \subseteq \bigcup_{k=1}^\infty I_k \right\}.
$$

The covers must be *countable* collections of *open* intervals; this is essential to the propositions proved below.

### 2.2. On $\mathbb{R}^n$

The same construction generalizes verbatim once "interval" is replaced by "box." An open **box** (or **rectangle** when $n=2$) is a Cartesian product of $n$ bounded open intervals, e.g. in $\mathbb{R}^2$: $R = (a,b) \times (c,d)$, with volume

$$
|R| = (b-a)(d-c).
$$

For $A \subseteq \mathbb{R}^n$, the outer measure is

$$
m^*(A) = \inf\left\{ \sum_{k=1}^\infty |R_k| \; : \; A \subseteq \bigcup_{k=1}^\infty R_k,\ R_k \text{ open boxes} \right\}.
$$

This defines a set function $m^* : 2^{\mathbb{R}^n} \to [0,\infty]$ on *every* subset of $\mathbb{R}^n$. Since a cover always exists (e.g. a countable union of huge boxes) and lengths/volumes are nonnegative, $m^*(A)$ is a well-defined element of $[0,\infty]$, independent of the order of summation.

It is immediate from the definition that $m^*(\varnothing) = 0$ (a single arbitrarily small interval/box covers $\varnothing$).


## 3\. Monotonicity

**Proposition.** If $A \subseteq B \subseteq \mathbb{R}^n$, then $m^*(A) \le m^*(B)$.

::: {.callout-tip collapse="true"}
## Proof of Monotonicity
By definition,

$$
m^*(B) = \inf\left\{ \sum_k |R_k| \; : \; B \subseteq \bigcup_k R_k \right\}.
$$

Since $A \subseteq B$, every countable open cover $\{R_k\}$ of $B$ is also a cover of $A$. Hence the collection of covers of $B$ is a *subset* of the collection of covers of $A$: every sum $\sum_k |R_k|$ that is a candidate in the infimum defining $m^*(B)$ is also a candidate in the infimum defining $m^*(A)$. Taking the infimum over a smaller collection of candidates can only produce a value that is greater than or equal to the infimum over the larger collection. Therefore

$$
m^*(A) \le m^*(B). \qquad \blacksquare
$$
:::

::: {.callout-note}
## Intuition: Why monotonicity holds
$A$ has *more* covers available to it than $B$ does — anything covering the bigger set $B$ automatically covers the smaller set $A$, but $A$ may also be covered more cheaply using covers that miss parts of $B$. More options to choose from when minimizing can only decrease (or leave unchanged) the infimum, never increase it.
:::

## 4\. Measure-Zero Sets

**Definition.** A set $Z \subseteq \mathbb{R}^n$ has **measure zero** if $m^*(Z) = 0$.

**Equivalent formulation.** $m^*(Z) = 0$ if and only if for every $\varepsilon > 0$ there exists a countable collection of open boxes $\{I_k\}$ with

$$
Z \subseteq \bigcup_k I_k \quad \text{and} \quad \sum_k |I_k| < \varepsilon.
$$

**Proposition.** Every countable set $A = \{x_1, x_2, x_3, \dots\} \subseteq \mathbb{R}^n$ has outer measure zero.

::: {.callout-tip collapse="true"}
## Proof that countable sets have measure zero
Let $\varepsilon > 0$. Our goal is to cover $A$ by intervals whose total length is smaller than $\varepsilon$; since $A$ has countably many points, we can afford to spend an exponentially shrinking budget of length on each one.

For each $n \ge 1$, center an interval of length $\varepsilon / 2^n$ at $x_n$:

$$
I_n = \left(x_n - \frac{\varepsilon}{2^{n+1}}, \; x_n + \frac{\varepsilon}{2^{n+1}}\right), \qquad |I_n| = \frac{\varepsilon}{2^n}.
$$

Then clearly $A \subseteq \bigcup_n I_n$, since $x_n \in I_n$ for every $n$. Moreover

$$
\sum_{n=1}^\infty |I_n| = \sum_{n=1}^\infty \frac{\varepsilon}{2^n} = \varepsilon.
$$

Since $\{I_n\}$ is a valid countable cover of $A$ by open intervals, $m^*(A) \le \varepsilon$. As $\varepsilon > 0$ was arbitrary, $m^*(A) = 0$. $\blacksquare$
:::

::: {.callout-note}
## Intuition: Why countability is exactly enough
A single point always has outer measure zero (cover it with a tiny interval). The obstruction to extending this to infinitely many points is that summing "tiny" lengths infinitely many times could still diverge. The trick is to make the $n$-th interval's length decay *geometrically* ($\varepsilon/2^n$), so that no matter how many points there are — as long as they're countable — the total budget $\sum \varepsilon/2^n = \varepsilon$ stays finite and can be made arbitrarily small. This is why, for instance, $\mathbb{Q}$ has outer measure zero even though it is dense in $\mathbb{R}$: density affects topology, not the total length needed to cover it.
:::



## 5\. Outer Measure of a Box

Before proving that outer measure agrees with volume on boxes, we need a bridge between [Pugh's sequential Heine–Borel Theorem](compactness.ipynb#the-heine-borel-theorem) (closed and bounded $\implies$ compact, in the sequential sense) and the *covering* property we actually need for the proof below.

::: {.callout-warning}
## Sequential vs. covering compactness
Pugh's Heine–Borel Theorem states that a closed, bounded set $K \subseteq \mathbb{R}^m$ is **sequentially compact**: every sequence in $K$ has a subsequence converging to a point of $K$. The proof of $m^*(R) = |R|$ below needs a different-looking property: that any countable open cover of $R$ has a *finite subcollection* that still covers $R$. These two notions of compactness are equivalent in $\mathbb{R}^m$, but the equivalence is not automatic — it needs its own short argument, given in the lemma below.
:::

**Lemma.** Let $K \subseteq \mathbb{R}^m$ be closed and bounded (hence sequentially compact, by Heine–Borel). If $\{U_\alpha\}$ is any collection of open sets with $K \subseteq \bigcup_\alpha U_\alpha$, then some *finite* subcollection $U_{\alpha_1}, \dots, U_{\alpha_N}$ already covers $K$.

::: {.callout-tip collapse="true"}
## Proof of the Lemma (sequential compactness $\implies$ finite subcover)

We argue by contradiction, using a bisection strategy: repeatedly halve $K$ into smaller and smaller pieces, and show that if no piece ever admits a finite subcover, we can extract a convergent sequence whose limit breaks the assumption.

Since $K$ is bounded, enclose it in a closed box $Q_0$. Suppose, for contradiction, that $\{U_\alpha\}$ covers $K$ but **no finite subcollection** covers $K$.

**Step 1: Bisection.** Split $Q_0$ into $2^m$ closed sub-boxes of half the side length (cut each coordinate interval in half). This splits $K \cap Q_0 = K$ into $2^m$ pieces $K \cap Q_0^{(1)}, \dots, K \cap Q_0^{(2^m)}$. If every one of these pieces could be covered by finitely many $U_\alpha$'s, then taking the union of those finite subcollections would give a finite subcover of all of $K$ — contradicting our assumption. So **at least one** piece, call it $K_1 := K \cap Q_1$ (with $Q_1$ a sub-box of half the diameter of $Q_0$), also admits **no finite subcover** from $\{U_\alpha\}$.

**Step 2: Iterate.** Repeat the bisection on $Q_1$: split it into $2^m$ sub-boxes, and by the same argument at least one sub-box $Q_2$ gives a piece $K_2 := K \cap Q_2$ with no finite subcover. Continuing inductively, we obtain a nested sequence of closed boxes

$$
Q_0 \supseteq Q_1 \supseteq Q_2 \supseteq \cdots, \qquad \text{diam}(Q_n) = \frac{\text{diam}(Q_0)}{2^n} \to 0,
$$

such that $K_n := K \cap Q_n$ is nonempty (else the empty finite subcollection would trivially "cover" it) and admits no finite subcover from $\{U_\alpha\}$, for every $n$.

**Step 3: Extract a limit point.** Pick a point $x_n \in K_n$ for each $n$. Since $K$ is sequentially compact, $(x_n)$ has a subsequence converging to some $p \in K$. Because the $Q_n$ are nested with shrinking diameter and $x_k \in Q_n$ for all $k \ge n$, the *entire* sequence $(x_n)$ (not just the subsequence) converges to $p$, and moreover $p \in Q_n$ for every $n$ (each $Q_n$ is closed, and it contains the tail of a sequence converging to $p$).

**Step 4: Contradiction.** Since $p \in K$ and $\{U_\alpha\}$ covers $K$, there is some $U_\alpha$ with $p \in U_\alpha$. Since $U_\alpha$ is open, it contains an open ball $B_r(p)$ for some $r > 0$. Because $\text{diam}(Q_n) \to 0$ and $p \in Q_n$ for every $n$, we can choose $n$ large enough that $\text{diam}(Q_n) < r$, which forces $Q_n \subseteq B_r(p) \subseteq U_\alpha$. But then the single set $U_\alpha$ covers $K_n = K \cap Q_n$ — a finite (in fact one-element) subcover, contradicting the fact that $K_n$ was constructed to admit no finite subcover.

This contradiction shows our assumption was false: $\{U_\alpha\}$ must admit a finite subcover of $K$. $\blacksquare$
:::

::: {.callout-note}
## Intuition: Why bisection bridges the two notions of compactness
The bisection argument is really just Bolzano–Weierstrass in disguise: if you could never finitely cover a box, you could keep drilling down into smaller and smaller "guilty" sub-boxes forever, and sequential compactness guarantees these shrinking regions squeeze down onto an actual point $p \in K$. But $p$ has to be covered by *some* single open set from the cover — and once the sub-boxes get small enough to fit inside that one open set, you've contradicted the idea that they were unfixably hard to cover. The failure at every scale collapses into a single point where, in fact, there was never any real difficulty at all.
:::

### 5.1. The Proposition

**Proposition.** The outer measure of a box equals its volume. On $\mathbb{R}$: $m^*(I) = |I|$ for any interval $I$. On $\mathbb{R}^2$: $m^*(R) = |R|$ for any rectangle $R$. In general, on $\mathbb{R}^n$: $m^*(B) = |B|$ for any box $B$.

We prove the $\mathbb{R}^2$ case; the argument is identical in $\mathbb{R}^n$ with grids replaced by higher-dimensional grids.

::: {.callout-tip collapse="true"}
## Proof that $m^*(R) = |R|$ for a closed rectangle $R = [a,b] \times [c,d]$

**Step 1: $m^*(R) \le |R|$.**

For any $\varepsilon > 0$, enclose $R$ in a slightly larger *open* rectangle $R'$ with $|R'| \le |R| + \varepsilon$ (just fatten each side by a small amount). Since $\{R'\}$ is a valid open cover of $R$,

$$
m^*(R) \le |R'| \le |R| + \varepsilon.
$$

Since $\varepsilon > 0$ was arbitrary, $m^*(R) \le |R|$.

**Step 2: $m^*(R) \ge |R|$.**

Let $\{R_k\}_{k=1}^\infty$ be any countable collection of open rectangles covering $R$. Since $R$ is closed and bounded, the **Lemma** above (which rests on Pugh's sequential Heine–Borel Theorem, not the covering definition) guarantees a *finite* subcollection $R_1, \dots, R_n$ already covers $R$. It suffices to show

$$
\sum_{k=1}^n |R_k| \ge |R|,
$$

since discarding the remaining $R_k$'s only decreases the left side, so proving this for the finite subcover proves $\sum_k |R_k| \ge |R|$ for the full collection, and taking the infimum over all covers gives $m^*(R) \ge |R|$.

*Building a common grid.* Collect all the $x$-coordinates of the endpoints of $R$ and of $R_1, \dots, R_n$ into an increasing list $X = \{x_0 < x_1 < \dots < x_r\}$, and similarly all $y$-coordinates into $Y = \{y_0 < y_1 < \dots < y_s\}$. This grid partitions $R$ into small subrectangles $[x_i, x_{i+1}] \times [y_j, y_{j+1}]$, and

$$
|R| = (b-a)(d-c) = \left(\sum_i (x_{i+1}-x_i)\right)\left(\sum_j (y_{j+1}-y_j)\right) = \sum_{i,j} (x_{i+1}-x_i)(y_{j+1}-y_j). \tag{$*$}
$$

*Each $R_k$ is also a union of grid cells.* Since the endpoints of $R_k$ appear in $X$ and $Y$, $R_k$ is itself exactly the union of the grid cells it contains:

$$
|R_k| = \sum_{\substack{a_k \le x_i < x_{i+1} \le b_k \\ c_k \le y_j < y_{j+1} \le d_k}} (x_{i+1}-x_i)(y_{j+1}-y_j).
$$

*Every grid cell of $R$ lies in some $R_k$.* Because $R_1, \dots, R_n$ cover $R$, every point of $R$ — and hence every grid cell $[x_i,x_{i+1}]\times[y_j,y_{j+1}]$ of $R$ — is contained in at least one $R_k$. Summing over the grid cells of $R$ therefore counts each cell's contribution at least once when we sum $|R_k|$ over all $k$:

$$
(*) = \sum_{i,j} (x_{i+1}-x_i)(y_{j+1}-y_j) \le \sum_{k=1}^n |R_k|.
$$

Combining with $(*)$, we get $|R| \le \sum_{k=1}^n |R_k|$, which is what we needed. Taking the infimum over all covers of $R$ gives $m^*(R) \ge |R|$.

Combining Steps 1 and 2, $m^*(R) = |R|$ for closed rectangles.

**Extension to a general rectangle $R$ (not necessarily closed).**

Let $\varepsilon > 0$. Choose a closed rectangle $S_1 \subseteq R$ with $|S_1| > |R| - \varepsilon$, and a closed rectangle $S_2 \supseteq R$ with $|S_2| < |R| + \varepsilon$ (shrink/expand $R$ slightly). By monotonicity and the closed-rectangle case already proved,

$$
|R| - \varepsilon < |S_1| = m^*(S_1) \le m^*(R) \le m^*(S_2) = |S_2| < |R| + \varepsilon.
$$

Letting $\varepsilon \to 0$ gives $m^*(R) = |R|$. $\blacksquare$
:::

::: {.callout-note}
## Intuition: Why the grid argument works
The heart of the proof is that a finite cover of a rectangle by other rectangles, however messily they overlap, can be "aligned" by drawing every horizontal and vertical line that appears as an edge of any rectangle involved. This common refinement chops $R$ into tiny grid cells, and turns an awkward overlapping-rectangles comparison into simple bookkeeping: since every grid cell of $R$ is claimed by at least one covering rectangle, summing volumes over the covering rectangles cannot undercount the volume of $R$ — it can only double-count due to overlaps, which only helps the inequality $|R| \le \sum |R_k|$ go the right way.
:::

::: {.callout-warning}
## Where the Lemma is essential
Step 2 relies entirely on the Lemma reducing an arbitrary countable open cover to a finite one. This is exactly why the general (non-closed) rectangle case is handled separately by sandwiching between closed rectangles, rather than adapting the grid argument directly — the Lemma requires closedness and boundedness to invoke sequential compactness.
:::


## 6\. Further Properties: Translation Invariance and Subadditivity

**Proposition (Translation Invariance).** For any set $A \subseteq \mathbb{R}^n$ and vector $y$, $m^*(A+y) = m^*(A)$.

::: {.callout-tip collapse="true"}
## Proof of Translation Invariance
A countable collection $\{I_k\}$ covers $A$ if and only if the shifted collection $\{I_k + y\}$ covers $A + y$. Since translating an open box by $y$ produces an open box of the *same volume*, $|I_k + y| = |I_k|$, so

$$
\sum_k |I_k| = \sum_k |I_k + y|.
$$

Every candidate sum for $m^*(A)$ corresponds exactly to a candidate sum for $m^*(A+y)$, and vice versa, with the same value. Hence the two infima coincide: $m^*(A+y) = m^*(A)$. $\blacksquare$
:::

**Proposition (Countable Subadditivity).** If $\{E_k\}_{k=1}^\infty$ is any countable collection of subsets of $\mathbb{R}^n$, disjoint or not, then

$$
m^*\left(\bigcup_{k=1}^\infty E_k\right) \le \sum_{k=1}^\infty m^*(E_k).
$$

::: {.callout-tip collapse="true"}
## Proof of Countable Subadditivity
If any $E_k$ has infinite outer measure, the inequality holds trivially. Assume each $m^*(E_k) < \infty$. Let $\varepsilon > 0$. For each $k$, by definition of infimum there is a countable open cover $\{I_{k,i}\}_{i=1}^\infty$ of $E_k$ with

$$
\sum_{i=1}^\infty |I_{k,i}| < m^*(E_k) + \frac{\varepsilon}{2^k}.
$$

The doubly-indexed collection $\{I_{k,i}\}_{1\le k,i<\infty}$ is countable (a countable union of countable collections) and covers $\bigcup_k E_k$. Hence

$$
m^*\left(\bigcup_k E_k\right) \le \sum_{k,i} |I_{k,i}| = \sum_k \left(\sum_i |I_{k,i}|\right) < \sum_k \left(m^*(E_k) + \frac{\varepsilon}{2^k}\right) = \sum_k m^*(E_k) + \varepsilon.
$$

Since this holds for every $\varepsilon > 0$, it holds with $\varepsilon = 0$:

$$
m^*\left(\bigcup_k E_k\right) \le \sum_k m^*(E_k). \qquad \blacksquare
$$
:::

**Corollary (Finite Subadditivity).** For any finite collection $\{E_k\}_{k=1}^n$, $m^*\left(\bigcup_{k=1}^n E_k\right) \le \sum_{k=1}^n m^*(E_k)$, obtained from countable subadditivity by setting $E_k = \varnothing$ for $k > n$.

::: {.callout-warning}
## Outer measure is *not* countably additive in general
Subadditivity gives $\le$, not $=$. Outer measure fails to be countably (or even finitely) additive on arbitrary sets: there exist disjoint $A, B \subseteq \mathbb{R}$ with $m^*(A \cup B) < m^*(A) + m^*(B)$. Recovering equality requires restricting attention to the Lebesgue measurable sets.
:::

---

**Summary of key results:**

* $m^*: 2^{\mathbb{R}^n} \to [0,\infty]$ is defined on every subset via countable open-box covers.
* $m^*(\varnothing) = 0$.
* [Monotonicity](measure.ipynb#fundamental-properties-of-measures): $A \subseteq B \implies m^*(A) \le m^*(B)$.
* Every countable set has outer measure zero.
* $m^*(B) = |B|$ for any box $B$ (interval, rectangle, or their $n$-dimensional generalization).
* $m^*$ is translation invariant.
* $m^*$ is countably subadditive, but not countably additive on all sets.
