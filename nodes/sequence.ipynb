# Sequences

## 1\. Formal Definition

In mathematics, a **sequence** is fundamentally a [function](def-functions-mappings.ipynb) whose domain is the set of positive integers (often denoted as $\mathbb{Z}^+$ or $\mathbb{N}$).

Instead of using the standard function notation like $f(n)$, the outputs of a sequence are conventionally written using subscript notation, where $n$ represents the "index" or position:
$$a_1, a_2, a_3, \dots, a_n, \dots$$

This infinite ordered list is often denoted compactly as $(a_n)_{n=1}^{\infty}$ or simply $\{a_n\}$.

-----

## 2\. Sequences of Real Numbers

A **sequence of real numbers** is a sequence whose codomain is the set of all real numbers, [$\mathbb{R}$](real-numbers.ipynb).

Formally, it is defined as a mapping:
$$a: \mathbb{Z}^+ \to \mathbb{R}$$

This means that for every positive integer index $n \in \mathbb{Z}^+$, there corresponds exactly one real number $a_n \in \mathbb{R}$.

::: {.callout-note}

## Intuition: Discrete Steps

Unlike continuous functions $f(x)$ where $x$ can be any decimal like $1.5$ or $\pi$, sequences only exist at discrete, whole-number steps ($n = 1, 2, 3 \dots$). You cannot ask for the "1.5th term" of a sequence. If you graph a sequence, it will never be a connected line; it will always be a scatter plot of isolated, distinct dots.
:::

-----

## 3\. Common Examples

Here are two fundamental types of real-valued sequences:

  * **Arithmetic Sequence:** The difference between consecutive terms is constant.
      * Formula: $a_n = 2n$
      * Terms: $2, 4, 6, 8, 10, \dots$
  * **Geometric Sequence:** The ratio between consecutive terms is constant.
      * Formula: $a_n = \frac{1}{2^n}$
      * Terms: $\frac{1}{2}, \frac{1}{4}, \frac{1}{8}, \frac{1}{16}, \dots$

-----

## 4\. Python Visualization

When plotting sequences, it is standard practice to use scatter plots (or stem plots) rather than continuous line plots to emphasize their discrete domain.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the domain (indices n = 1 to 15)
n = np.arange(1, 16)

# Define two sequences of real numbers
a_n = 1 / n              # Harmonic sequence
b_n = (-1)**n * (1 / n)  # Alternating harmonic sequence

plt.figure(figsize=(10, 5))

# Plotting the sequences as discrete points
plt.stem(n, a_n, linefmt='b-', markerfmt='bo', basefmt='k-', label='$a_n = 1/n$')
plt.stem(n, b_n, linefmt='r-', markerfmt='ro', basefmt='k-', label='$b_n = (-1)^n / n$')

plt.title('Visualization of Real Number Sequences')
plt.xlabel('Index (n)')
plt.ylabel('Term Value ($a_n$)')
plt.xticks(n)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
