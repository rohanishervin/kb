# Supremum and Infimum

A set $A \subseteq \mathbb{R}$ is **bounded above** if there exists a number $b \in \mathbb{R}$ such that:

$$\forall a \in A: \quad a \leq b$$

The number $b$ is called an **upper bound** for $A$.

Similarly, the set $A$ is **bounded below** if there exists a lower bound $l \in \mathbb{R}$ satisfying:

$$\forall a \in A: \quad l \leq a$$

A real number $s$ is the **least upper bound** for a set $A \subseteq \mathbb{R}$ if it meets the following two criteria:

**(i) Upper bound condition**
$s$ is an upper bound for $A$, i.e., $\forall x \in A, s \geq x$

**(ii) Leastness property**
If $b$ is any upper bound for $A$, then $s \leq b$, i.e., $s = \inf\{x \in \mathbb{R} : \forall a \in A, x \geq a\}$
