# Squeeze Theorem for Sequences

## 1\. Theorem Statement

Let $(a_n)$, $(b_n)$, and $(c_n)$ be [sequences](sequence.ipynb) of real numbers.

If there exists an integer $K \in \mathbb{N}$ such that for all $n \geq K$:
$$a_n \leq b_n \leq c_n$$

and if the outer sequences converge to the same limit $L$:
$$\lim_{n \to \infty} a_n = \lim_{n \to \infty} c_n = L$$

then the inner sequence also converges to $L$:
$$\lim_{n \to \infty} b_n = L$$


::: {.callout-note}
### Intuition
If sequence $b_n$ is trapped between two sequences that are both zeroing in on the exact same target, $b_n$ has no choice but to hit that target as well.
:::

## 2\. Formal Proof

**Proof:**

Let $\epsilon > 0$ be given. We need to find an integer $N$ such that for all $n \geq N$, $|b_n - L| < \epsilon$.

1.  Because $\lim_{n \to \infty} a_n = L$, there exists an integer $N_1$ such that for all $n \geq N_1$:
    $$|a_n - L| < \epsilon \implies L - \epsilon < a_n < L + \epsilon$$

2.  Because $\lim_{n \to \infty} c_n = L$, there exists an integer $N_2$ such that for all $n \geq N_2$:
    $$|c_n - L| < \epsilon \implies L - \epsilon < c_n < L + \epsilon$$

3.  By the premise of the theorem, we also know there is an integer $K$ where the inequality $a_n \leq b_n \leq c_n$ begins to hold.

Let us choose $N = \max(K, N_1, N_2)$.

For all $n \geq N$, all three conditions hold simultaneously. Therefore, we can chain the inequalities together:
$$L - \epsilon < a_n \leq b_n \leq c_n < L + \epsilon$$

By ignoring the outer terms, we extract the strict relationship for $b_n$:
$$L - \epsilon < b_n < L + \epsilon$$

This is exactly equivalent to:
$$|b_n - L| < \epsilon$$

Since this holds for any arbitrarily chosen $\epsilon > 0$, we conclude that $\lim_{n \to \infty} b_n = L$. $\square$

## 3\. Classic Application

The Squeeze Theorem is particularly useful for dealing with sequences that oscillate wildly but are bounded.

**Example:** Find the limit of $b_n = \frac{\sin(n)}{n}$ as $n \to \infty$.

We know that the sine function is strictly bounded:
$$-1 \leq \sin(n) \leq 1$$

Divide the entire inequality by $n$ (since $n \geq 1$, the inequality signs do not flip):
$$-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}$$

Let $a_n = -\frac{1}{n}$ and $c_n = \frac{1}{n}$.
Since $\lim_{n \to \infty} \left(-\frac{1}{n}\right) = 0$ and $\lim_{n \to \infty} \left(\frac{1}{n}\right) = 0$, the Squeeze Theorem guarantees that:
$$\lim_{n \to \infty} \frac{\sin(n)}{n} = 0$$
