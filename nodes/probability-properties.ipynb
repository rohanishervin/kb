# Properties of Probability

For any events A and B, the following properties hold:

#### 1. Complement Property
$$P(A^c) = 1 - P(A)$$

#### 2. Monotonicity of Probability Measure
If $A \subseteq B$, then $$P(A) \leq P(B)$$

#### 3. Addition Rule for Two Events
$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

# Proof:
#### 1. Complement Property
Since $A$ and $A^c$ are disjoint, their union is $S$. By the second axiom of [probability definition](probability.ipynb),

$$P(S) = P(A \cup A^c) = P(A) + P(A^c),$$
but since $P(S) = 1$ by the first axiom, we have
$$P(A) + P(A^c) = 1.$$

#### 2. Monotonicity of Probability Measure

Given $A \subseteq B$, we can express $B$ as the union of $A$ and $B \cap A^c$. Since $A$ and $B \cap A^c$ are disjoint, we have
$$P(B) = P\left(A \cup (B \cap A^c)\right) = P(A) + P(B \cap A^c).$$

Since probability is nonnegative, we know that $P(B \cap A^c) \geq 0$. Therefore,
$$P(B) \geq P(A),$$
which shows that the probability measure is monotone.


#### 3. Addition Rule for Two Events
Given two events $A$ and $B$, we can express their union as $A \cup B$. We can also write this as the union of $A$ with a new event, $B \cap A^c$, where $A^c$ is the complement of $A$. This leads to the 
equation:

$$P(A \cup B) = P(A \cup (B \cap A^c))$$

Using the second axiom, we can rewrite this as:

$$P(A \cup B) = P(A) + P(B \cap A^c)$$

To show that $P(B \cap A^c) = P(B) - P(A \cap B)$, we need to demonstrate that $A \cap B$ and $B \cap A^c$ are disjoint events whose union is $B$. This allows us to apply the second axiom again:

$$P(A \cap B) + P(B \cap A^c) = P(B)$$

Rearranging this equation gives us the desired result:

$$P(B \cap A^c) = P(B) - P(A \cap B)$$
