# Adjusting for Overcounting

## 1\. Core Concept and Formula

When determining the total number of distinct arrangements (permutations) of $n$ items where some items are physically indistinguishable from one another, we cannot simply use $n!$. Using $n!$ assumes every single item is unique.

To find the true number of distinct arrangements, we must divide the total permutations by the number of ways the identical items can be rearranged among themselves. This principle is known as **adjusting for overcounting**.

If a set of $n$ items contains groups of repetitive elements of sizes $r_1, r_2, \dots, r_k$ (where $r_1 + r_2 + \dots + r_k = n$), the number of unique permutations is given by the **multinomial coefficient**:

$$\frac{n!}{r_1! \cdot r_2! \dots r_k!}$$

-----

## 2\. Intuition: Why Do We Divide?

Imagine you have three letters: **A, B, B**.

If we treat the two B's as completely distinct by labeling them $B_1$ and $B_2$, we have $3! = 6$ total permutations:

1.  A, $B_1$, $B_2$
2.  A, $B_2$, $B_1$
3.  $B_1$, A, $B_2$
4.  $B_2$, A, $B_1$
5.  $B_1$, $B_2$, A
6.  $B_2$, $B_1$, A

However, if we remove the sub-labels (because the B's are indistinguishable), pairs of these arrangements collapse into identical sequences:

  * (1 & 2) become **A B B**
  * (3 & 4) become **B A B**
  * (5 & 6) become **B B A**

For every distinct arrangement, there are exactly $2!$ ways to swap the B's without changing what the word looks like. Because we counted every arrangement $2!$ times, we must *divide* by $2!$ to correct our total.

-----

## 3\. Formal Derivation

**Proof:**

Let $N$ be the true number of visually distinct arrangements of our $n$ items.

Suppose we temporarily paint each identical item a slightly different shade so that all $n$ items become $100\%$ distinct from one another. We know the total number of permutations of $n$ distinct items is $n!$.

Now, take just *one* of our $N$ distinct arrangements. How many ways can we reorder the newly painted items without changing the underlying sequence of elements?

  * We can arrange the $r_1$ items in $r_1!$ ways.
  * We can arrange the $r_2$ items in $r_2!$ ways.
  * $\dots$ and so on.

By the [multiplication-rule](multiplication-rule.ipynb), there are exactly $(r_1! \cdot r_2! \dots r_k!)$ ways to shuffle the identical items within any single observable arrangement.

Therefore, multiplying our distinct arrangements $N$ by the number of hidden internal shuffles gives us the total permutations of $n$ distinct items:
$$N \cdot (r_1! \cdot r_2! \dots r_k!) = n!$$

Solving for $N$ yields the adjusting for overcounting formula:
$$N = \frac{n!}{r_1! \cdot r_2! \dots r_k!}$$
$\square$

-----

## 4\. Classic Example: The MISSISSIPPI Problem

**Question:** How many distinct ways can you arrange the letters in the word MISSISSIPPI?

**Solution:**

1.  First, count the total number of letters: $n = 11$.
2.  Next, group the indistinguishable items and count their frequencies:
      * M: 1
      * I: 4
      * S: 4
      * P: 2
3.  Apply the overcounting formula:

$$\frac{11!}{1! \cdot 4! \cdot 4! \cdot 2!}$$

$$= \frac{39,916,800}{1 \cdot 24 \cdot 24 \cdot 2} = \frac{39,916,800}{1,152} = 34,650$$

There are exactly $34,650$ unique ways to arrange the word MISSISSIPPI.

-----

## 5\. Connection to Combinations

The principle of adjusting for overcounting is exactly how we derive the formula for the [binomial-coefficient](binomial-coefficient.ipynb) (or "n choose k").

When we select $k$ items from a pool of $n$ items where the order of selection *does not matter* (a combination), we first calculate the permutations where order *does* matter:
$$P(n, k) = \frac{n!}{(n-k)!}$$

However, because the order doesn't matter, an arrangement of $\{A, B, C\}$ is the exact same subset as $\{B, C, A\}$. For every subset of $k$ items, there are $k!$ ways to arrange them.

To adjust for this overcounting, we divide the permutation formula by $k!$:
$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$
