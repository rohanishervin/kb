# Functions and Mappings

## 1. Anatomy of a Function
A function $f: A \to B$ maps elements from a **Domain** ($A$) to a **Codomain/Target** ($B$). The actual subset of $B$ that gets "hit" by the function is called the **Image** or Range.

Depending on the context, mathematicians use different terms for functions:

* **Function:** Usually implies mapping to sets of numbers (e.g., $\mathbb{R}$ or $\mathbb{C}$).
* **Mapping:** Used in abstract sets or topology.
* **Transformation:** Used in geometry or linear algebra (e.g., rotations, translations).
* **Operator:** Used when mapping functions to other functions (e.g., the derivative operator).

## 2. Classification of Mappings

1.  **Injection (One-to-One):** Every element in the domain maps to a *unique* element in the codomain. ($f(x) = f(y) \implies x = y$).
2.  **Surjection (Onto):** Every element in the codomain is mapped to by at least one element in the domain. (Image = Codomain).
3.  **Bijection:** A function that is both injective and surjective. It establishes a perfect one-to-one pairing between $A$ and $B$, meaning an inverse function $f^{-1}$ exists.

## 3. Images, Preimages, and Set Operations

When analyzing topological spaces or abstract sets, we frequently map entire sets back and forth using a function $f: X \to Y$. 

* The **Image** of a subset $A \subset X$ is $f(A) = \{f(x) \mid x \in A\}$.
* The **Preimage** (or Inverse Image) of a subset $V \subset Y$ is $f^{-1}(V) = \{x \in X \mid f(x) \in V\}$.
  
*(Note: The notation $f^{-1}$ here does not mean the function is invertible; the preimage set always exists even for many-to-one functions).*

A fundamental rule of mathematics is that **preimages behave perfectly, while images are flawed**. 

### 3.1. Preimages Respect Everything
Preimages perfectly preserve all boolean set operations. For any subsets $A, B \subset Y$:

1.  **Union:** $f^{-1}(A \cup B) = f^{-1}(A) \cup f^{-1}(B)$
2.  **Intersection:** $f^{-1}(A \cap B) = f^{-1}(A) \cap f^{-1}(B)$
3.  **Subset:** $A \subset B \implies f^{-1}(A) \subset f^{-1}(B)$
4.  **Complement:** $f^{-1}(A^c) = (f^{-1}(A))^c$

**Proof of Complement Property:**
Let $x \in f^{-1}(A^c)$. By definition, this means $f(x) \in A^c$. 
Because $f(x)$ is in the complement of $A$, it means $f(x) \notin A$. 
If $f(x) \notin A$, then $x$ cannot be in the preimage of $A$, so $x \notin f^{-1}(A)$. 
Therefore, $x$ must be in the complement of the preimage: $x \in (f^{-1}(A))^c$. 
Since the logic holds perfectly in reverse, the sets are exactly equal. $\square$

### 3.2. Images Have "Gotchas"
Images behave well with "inclusive" operations (unions and subsets) but **fail** with "exclusive" operations (intersections and complements) whenever a function is many-to-one. For any subsets $A, B \subset X$:

1.  **Union (True):** $f(A \cup B) = f(A) \cup f(B)$
2.  **Subset (True):** $A \subset B \implies f(A) \subset f(B)$
3.  **Intersection (FALSE):** $f(A \cap B) \neq f(A) \cap f(B)$
4.  **Complement (FALSE):** $f(A^c) \neq (f(A))^c$

**Proof of Union Property:**
Let $z \in f(A \cup B)$. By definition, there is some $x \in A \cup B$ such that $f(x) = z$. 
Since $x \in A \cup B$, either $x \in A$ or $x \in B$.
If $x \in A$, then $z = f(x) \in f(A)$.
If $x \in B$, then $z = f(x) \in f(B)$.
In either case, $z$ belongs to $f(A) \cup f(B)$. (This proves $f(A \cup B) \subseteq f(A) \cup f(B)$. The reverse direction is identical, proving equality). $\square$

**Counterexample for Intersection:**
Generally, $f(A \cap B) \subseteq f(A) \cap f(B)$, but they are not equal. 
Consider a constant function $f(x) = 0$. 
Let domain sets be $A = \{1\}$ and $B = \{2\}$. 
* The intersection $A \cap B = \emptyset$. Therefore, the image of the intersection is $f(\emptyset) = \mathbf{\emptyset}$.
* However, the individual images are $f(A) = \{0\}$ and $f(B) = \{0\}$. Their intersection is $\{0\} \cap \{0\} = \mathbf{\{0\}}$.
Since $\emptyset \neq \{0\}$, the equality fails. The many-to-one nature of $f$ artificially forced disjoint sets to overlap in the target space!
