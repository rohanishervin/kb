# Equicontinuity and the Arzelà-Ascoli Theorem

## 1. Compactness in Function Spaces

In finite-dimensional spaces such as $\mathbb{R}^m$, the [Heine-Borel Theorem](compactness.ipynb#the-heine-borel-theorem) tells us that a set is compact if and only if it is closed and bounded. However, when we move to infinite-dimensional spaces, such as the function space $C^0([a,b], \mathbb{R})$ equipped with the uniform norm, this characterization breaks down. Closed and bounded subsets in $C^0$ are rarely compact.

Consider the closed unit ball in $C^0([0,1], \mathbb{R})$:
$$
\mathcal{B} = \{f \in C^0([0,1], \mathbb{R}) : \|f\| \le 1\}
$$
To see why $\mathcal{B}$ is not compact, we can examine the sequence of functions $f_n(x) = x^n$. 

* Clearly, $f_n \in \mathcal{B}$ for all $n \in \mathbb{N}$ since $|x^n| \le 1$ for $x \in [0,1]$.
* If this sequence had a subsequence $f_{n_k}$ that converged in $C^0$ to some limit $f$, then $f$ must also be the pointwise limit.
* The pointwise limit is $f(x) = 0$ for $x < 1$ and $f(1) = 1$.
* Since this limit function $f$ is not continuous, it is not in $C^0$. 

Therefore, no subsequence can converge uniformly (with respect to the metric in $C^0$). The core issue is that $C^0$ is infinite-dimensional. In fact, a normed vector space has a compact closed unit ball if and only if it is finite-dimensional. 

To guarantee that a closed and bounded subset of $C^0$ is compact—which is crucial for extracting convergent subsequences of functions—we need an additional condition. This condition is **equicontinuity**.

::: {.callout-note}
## Intuition: Why Boundedness Isn't Enough
In finite dimensions, being "bounded" restricts the points to a finite box, leaving them nowhere to go, forcing them to cluster. In infinite-dimensional function spaces, a sequence of functions can remain bounded in magnitude but become infinitely steep or oscillatory (like higher powers of $x^n$ near $1$, or high-frequency sine waves). They escape compactness not by growing infinitely large, but by wiggling into new dimensions of the space. Equicontinuity explicitly forbids this by putting a global speed limit on how fast the functions can change.
:::

## 2. Equicontinuity

Equicontinuity essentially requires a family of functions to be "equally continuous" uniformly across the entire family. The shape and steepness of their graphs are uniformly controlled.

**Definition (Equicontinuous Sequence):**
A sequence of functions $(f_n)$ in $C^0$ is **equicontinuous** if for every $\epsilon > 0$, there exists a $\delta > 0$ such that for all $s, t$ in the domain and all $n \in \mathbb{N}$:

$$
|s-t| < \delta \implies |f_n(s) - f_n(t)| < \epsilon
$$

**Definition (Equicontinuous Family/Set):**
Similarly, a set of functions $\mathcal{E} \subset C^0$ is **equicontinuous** if for every $\epsilon > 0$, there exists a $\delta > 0$ such that for all $s, t$ in the domain and all $f \in \mathcal{E}$:

$$
|s-t| < \delta \implies |f(s) - f(t)| < \epsilon
$$

::: {.callout-warning}
## Uniform vs. Pointwise Equicontinuity
The definition above might be more accurately called **uniform equicontinuity**, as the single choice of $\delta$ works uniformly for all points $s,t$ and all functions in the family. By contrast, **pointwise equicontinuity** would only require that at each specific point $x$, there is a $\delta > 0$ working for all functions in the family:
$\forall \epsilon > 0$ and $\forall x \in [a,b]$, $\exists \delta > 0$ such that $|x-t| < \delta \text{ and } n \in \mathbb{N} \implies |f_n(x) - f_n(t)| < \epsilon.$
Unless otherwise stated, we assume the stronger uniform definition when discussing subsets of $C^0([a,b], \mathbb{R})$.
:::

## 3. The Arzelà-Ascoli Theorem

The fundamental theorem connecting compactness and equicontinuity allows us to extract a uniformly convergent subsequence from a bounded, equicontinuous sequence of functions.

**Theorem (Arzelà-Ascoli Theorem):**
Every bounded, equicontinuous sequence of functions in $C^0([a,b], \mathbb{R})$ has a uniformly convergent subsequence.

Before proving the theorem, we state a brief but necessary lemma regarding the indices of subsequences.

**Lemma (Subsequence Indices):**
If $(f_k)$ is a subsequence of $(g_n)$, then for each $k \in \mathbb{N}$, we have $f_k = g_r$ for some index $r \ge k$.

::: {.callout-tip collapse="true"}
## Proof of Lemma
By the definition of a subsequence, $f_k = g_{n_k}$ for some sequence of strictly increasing indices $n_1 < n_2 < \dots < n_k$. 
Since the indices are strictly increasing integers, it must be the case that $n_k \ge k$ for all $k$. Thus, setting $r = n_k$, we have $r \ge k$ and $f_k = g_r$. $\blacksquare$
:::

## 4. Proof of the Arzelà-Ascoli Theorem

The proof relies on a classic "diagonalization" argument combined with an $\epsilon/3$ trick.

::: {.callout-tip collapse="true"}
## Proof of the Arzelà-Ascoli Theorem

**Step 1: Pointwise convergence on a dense subset via diagonalization.**
Since the domain $[a, b]$ is a compact metric space, it contains a countable dense set $D = \{d_1, d_2, d_3, \dots\}$ (for example, the rational numbers in the interval). 

Because the original sequence $(f_n)$ is bounded in $C^0$, there exists a constant $M$ such that $|f_n(x)| \le M$ for all $n \in \mathbb{N}$ and all $x \in [a,b]$.
If we evaluate the sequence at the first dense point $d_1$, we get a bounded sequence of real numbers $(f_n(d_1))$. By the [Bolzano-Weierstrass Theorem](compactness.ipynb#the-bolzano-weierstrass-theorem), this sequence has a convergent subsequence. Let's denote this subsequence as $(f_{1, k})$, which converges to some limit $y_1 \in \mathbb{R}$ as $k \to \infty$.

Now, we evaluate this new subsequence $(f_{1, k})$ at the second point $d_2$. This is again a bounded sequence in $\mathbb{R}$, so we can extract a sub-subsequence, denoted $(f_{2, k})$, which converges to some limit $y_2$. 
Notice that $(f_{2, k})$ evaluated at $d_1$ still converges to $y_1$ because it is a subsequence of $(f_{1, k})$.

Continuing this process inductively, we construct a nested family of subsequences $(f_{m, k})$ satisfying:

1.  $(f_{m, k})$ is a subsequence of $(f_{m-1, k})$.
2.  For any $j \le m$, the sequence $f_{m, k}(d_j) \to y_j$ as $k \to \infty$.

We now form the **diagonal sequence** $(g_m)$ by taking the $m$-th term of the $m$-th subsequence: $g_m = f_{m, m}$. We claim $(g_m)$ converges pointwise on all of $D$.
Fix any index $j \in \mathbb{N}$ and consider $m \ge j$. By repeated application of the Lemma regarding subsequence indices:
$$
g_m = f_{m,m} = f_{m-1, r_1} = f_{m-2, r_2} = \dots = f_{j, r}
$$
for some index $r \ge m$. As $m \to \infty$, the index $r \to \infty$ as well. Therefore, evaluating at $d_j$:
$$
g_m(d_j) = f_{j, r}(d_j) \to y_j \text{ as } m \to \infty
$$
Thus, the diagonal sequence $(g_m)$ converges at every point in the dense set $D$.

**Step 2: Upgrading to uniform convergence on the whole domain.**
We will show that $(g_m)$ is a Cauchy sequence in $C^0$, which implies it converges uniformly to a continuous limit since $C^0$ is [complete](completeness.ipynb#complete-metric-spaces).

Let $\epsilon > 0$ be given. Because the original sequence is equicontinuous, there exists a $\delta > 0$ such that for all $s, t \in [a, b]$ and all $m$:
$$
|s - t| < \delta \implies |g_m(s) - g_m(t)| < \frac{\epsilon}{3}
$$

Since $D$ is dense and $[a,b]$ is compact, we can cover $[a,b]$ with a finite number of $\delta$-neighborhoods centered at points in $D$. Let $\{d_1, d_2, \dots, d_J\}$ be the centers of this finite cover. 
Since this set is finite, and $g_m$ converges at each $d_j$, we can find an integer $N$ such that for all $l, m \ge N$ and all $j \le J$:
$$
|g_m(d_j) - g_l(d_j)| < \frac{\epsilon}{3}
$$

Now, pick an arbitrary point $x \in [a,b]$. Because the points $\{d_1, \dots, d_J\}$ form a $\delta$-net, there exists some $d_j$ (with $j \le J$) such that $|x - d_j| < \delta$. For any $l, m \ge N$, we apply the triangle inequality:
$$
\begin{aligned}
|g_m(x) - g_l(x)| &\le |g_m(x) - g_m(d_j)| + |g_m(d_j) - g_l(d_j)| + |g_l(d_j) - g_l(x)| \\
&< \frac{\epsilon}{3} + \frac{\epsilon}{3} + \frac{\epsilon}{3} = \epsilon
\end{aligned}
$$
The first and third terms are bounded by $\epsilon/3$ due to equicontinuity, and the second term is bounded by $\epsilon/3$ due to pointwise convergence on $D$.
Since this bound $\epsilon$ is independent of $x$, the sequence $(g_m)$ is uniformly Cauchy, meaning it converges in $C^0$. $\blacksquare$
:::

::: {.callout-note}
## Intuition: The Diagonal Argument
Why take the diagonal sequence $f_{m,m}$? If we just took the first subsequence $f_{1,k}$, it's only guaranteed to converge at $d_1$. If we took $f_{2,k}$, it converges at $d_1$ and $d_2$, but we need a single sequence that converges everywhere on $D$ simultaneously. Since $D$ is infinite, we can't take the "infinity-th" subsequence. The diagonal sequence $f_{m,m}$ eventually acts like a sub-sequence of *every* nested row, inheriting the convergence behavior at every point in the countable dense set. 
:::

## 5. The Propagation Theorem

Step 2 of the Arzelà-Ascoli proof is powerful enough to be stated as its own theorem. It highlights that under the rigid structure of equicontinuity, local convergence information on a dense subset forcefully extends to uniform global convergence.

**Theorem (Arzelà-Ascoli Propagation Theorem):**
If an equicontinuous sequence of functions converges pointwise on a dense subset of its domain, then that convergence propagates to uniform convergence over the entire domain.

The proof of this theorem is exactly the "$\epsilon/3$ part" (Step 2) of the proof of the main Arzelà-Ascoli theorem shown above.

## 6. Uniform Boundedness of Derivatives

One of the most common ways to demonstrate that a family of functions is equicontinuous is by bounding their derivatives. 

**Corollary:**
Let $f_n : [a,b] \to \mathbb{R}$ be a sequence of differentiable functions whose derivatives are uniformly bounded. If the sequence $(f_n(x_0))$ is bounded for at least one point $x_0 \in [a,b]$, then the sequence $(f_n)$ has a uniformly convergent subsequence on $[a,b]$.

::: {.callout-tip collapse="true"}
## Proof
Let $M$ be the uniform bound for the derivatives, meaning $|f_n'(x)| \le M$ for all $n \in \mathbb{N}$ and all $x \in [a,b]$.

**Step 1: Show equicontinuity.**
By the Mean Value Theorem, for any $s, t \in [a,b]$, there exists a point $c$ between $s$ and $t$ such that:
$$
|f_n(s) - f_n(t)| = |f_n'(c)| |s - t| \le M |s - t|
$$
Given any $\epsilon > 0$, we can choose $\delta = \frac{\epsilon}{M+1}$. Then for any $|s-t| < \delta$, we have:
$$
|f_n(s) - f_n(t)| \le M \left( \frac{\epsilon}{M+1} \right) < \epsilon
$$
This establishes that $(f_n)$ is equicontinuous.

**Step 2: Show boundedness in $C^0$.**
We are given that at $x_0$, the sequence is bounded, so $|f_n(x_0)| \le C$ for all $n$.
For any arbitrary point $x \in [a,b]$, we can use the triangle inequality and our derivative bound:
$$
\begin{aligned}
|f_n(x)| &\le |f_n(x) - f_n(x_0)| + |f_n(x_0)| \\
&\le M|x - x_0| + C \\
&\le M|b - a| + C
\end{aligned}
$$
Since this upper bound $M|b-a| + C$ does not depend on $x$ or $n$, the entire sequence $(f_n)$ is bounded in the $C^0$ norm. 

**Step 3: Apply Arzelà-Ascoli.**
Because $(f_n)$ is both bounded and equicontinuous, the Arzelà-Ascoli Theorem guarantees the existence of a uniformly convergent subsequence. $\blacksquare$
:::

## 7. The Heine-Borel Theorem for Function Spaces

We can summarize the findings above by returning to our initial motivation: finding a true analog of the Heine-Borel theorem for continuous function spaces.

**Theorem (Heine-Borel in $C^0$):**
A subset $\mathcal{E} \subset C^0([a,b], \mathbb{R})$ is compact if and only if it is closed, bounded, and equicontinuous.

::: {.callout-tip collapse="true"}
## Proof

**Forward Direction ($\implies$):**
Assume $\mathcal{E}$ is compact. In any metric space, a compact set is closed and totally bounded. 
Total boundedness implies that for any $\epsilon > 0$, we can cover $\mathcal{E}$ with a finite number of $\epsilon/3$-neighborhoods centered at functions $f_1, f_2, \dots, f_n \in C^0$. 
Since each $f_k$ is continuous on the compact domain $[a,b]$, it is uniformly continuous. Therefore, there exists a $\delta > 0$ such that for all $s,t$ and all $k \in \{1, \dots, n\}$:
$$
|s - t| < \delta \implies |f_k(s) - f_k(t)| < \frac{\epsilon}{3}
$$
Now, take any arbitrary $f \in \mathcal{E}$. Because the finite neighborhoods cover $\mathcal{E}$, $f$ must lie in the $\epsilon/3$-neighborhood of some $f_k$. This means $\|f - f_k\| < \epsilon/3$. 
For any $s, t$ with $|s - t| < \delta$, we apply the triangle inequality:
$$
\begin{aligned}
|f(s) - f(t)| &\le |f(s) - f_k(s)| + |f_k(s) - f_k(t)| + |f_k(t) - f(t)| \\
&< \frac{\epsilon}{3} + \frac{\epsilon}{3} + \frac{\epsilon}{3} = \epsilon
\end{aligned}
$$
Since this $\delta$ works for all $f \in \mathcal{E}$, the set $\mathcal{E}$ is equicontinuous. 

**Reverse Direction ($\impliedby$):**
Assume $\mathcal{E}$ is closed, bounded, and equicontinuous.
Let $(f_n)$ be any sequence in $\mathcal{E}$. Because $\mathcal{E}$ is bounded and equicontinuous, the sequence $(f_n)$ is bounded and equicontinuous. By the Arzelà-Ascoli Theorem, $(f_n)$ possesses a subsequence $(f_{n_k})$ that converges uniformly to some limit function $f$. 
Since $\mathcal{E}$ is a closed set, this limit $f$ must be contained in $\mathcal{E}$. 
Because every sequence in $\mathcal{E}$ has a subsequence converging to a point in $\mathcal{E}$, the set $\mathcal{E}$ is sequentially compact, which in a metric space is equivalent to being compact. $\blacksquare$
:::
