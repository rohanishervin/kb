# Conditional Probability Examples

## Problem 1: The Two Aces Paradox

Draw 2 random cards from a standard deck.

1.  Calculate $P(\text{both are Aces} \mid \text{you have at least one Ace})$.
2.  Calculate $P(\text{both are Aces} \mid \text{you have the Ace of Spades})$.

**Solution:**

1.  The total number of 2-card combinations is $\binom{52}{2} = 1326$.
    The number of pairs containing *no* Aces is $\binom{48}{2} = 1128$.
    Thus, there are $1326 - 1128 = 198$ pairs that contain *at least one* Ace.
    Only $\binom{4}{2} = 6$ of these pairs consist of two Aces.
    $$P(\text{Both} \mid \text{Have an Ace}) = \frac{6}{198} = \frac{1}{33} \approx 3.03\%$$

2.  The number of pairs containing the *Ace of Spades* is exactly $51$ (the Ace of Spades paired with any of the other 51 cards).
    Of those 51 pairs, only $3$ contain a second Ace (Ace of Hearts, Diamonds, or Clubs).
    $$P(\text{Both} \mid \text{Have Ace of Spades}) = \frac{3}{51} = \frac{1}{17} \approx 5.88\%$$

:::{.callout-note}
## Intuition
Why does simply naming the suit nearly double the probability?

When you say "I have an Ace," there is ambiguity. You could have any of the 4 Aces, meaning the denominator of possible pairs is quite large. By specifying the suit, you eliminate the ambiguity, drastically shrinking the sample space (from 198 pairs down to 51 pairs), which concentrates the probability on the remaining outcomes.
:::

## Problem 2: Disease Testing (Base Rate Fallacy)

A patient tested for a rare disease afflicting 1% of the population. The test claims to be 95% accurate. Fred tests positive. What is the probability he actually has the disease?

**Definitions:**

  * **Prevalence** $P(D) = 0.01$: The baseline probability of having the disease.
  * **Sensitivity** $P(T|D) = 0.95$: True Positive Rate (Probability of a positive test given you are sick).
  * **Specificity** $P(T^c|D^c) = 0.95$: True Negative Rate (Probability of a negative test given you are healthy).

**Solution (Bayes' Theorem):**
We want the posterior probability $P(D|T)$.
$$P(D|T) = \frac{P(T|D)P(D)}{P(T)}$$

Using the [Law of Total Probability](law-of-total-probability.ipynb), the unconditional probability of testing positive $P(T)$ is:
$$P(T) = P(T|D)P(D) + P(T|D^c)P(D^c)$$
$$P(T) = (0.95)(0.01) + (1 - 0.95)(0.99) = 0.0095 + 0.0495 = 0.059$$

Now, apply Bayes' Rule:
$$P(D|T) = \frac{0.0095}{0.059} \approx 0.161$$

:::{.callout-note}
## Intuition
Despite a "95% accurate" test, patient only has a \~16% chance of being sick\! This is highly counterintuitive and is known as the Base Rate Fallacy. Because the disease is so rare, the sheer volume of healthy people taking the test means that the 5% false-positive rate generates far more false positives ($0.0495$) than there are actual sick people testing positive ($0.0095$).
:::


## Problem 3: The Prosecutor's Fallacy (Sally Clark Case)

The Prosecutor's Fallacy occurs when the conditional probability of $A$ given $B$ is confused with the probability of $B$ given $A$.
$$P(A|B) \neq P(B|A)$$

**The Case:** Sally Clark was convicted of murdering her two infant sons after both died of SIDS (Sudden Infant Death Syndrome). The prosecution argued that the probability of *one* child dying of SIDS was $1 \text{ in } 8500$. Assuming [independence](independence-of-events.ipynb), the probability of *two* children dying of SIDS was calculated as $(1/8500)^2 \approx 1 \text{ in } 73 \text{ million}$.

The prosecutor argued this proved her guilt. There were two massive mathematical blunders here:

1.  **False Independence:** Genetics and environment mean infant deaths in the same family are highly *conditionally dependent*. The true probability of a second SIDS death given a first is much higher than $1/8500$.
2.  **The Fallacy:** The prosecutor calculated $P(\text{Two SIDS deaths} \mid \text{Innocent})$. However, the court actually needed $P(\text{Innocent} \mid \text{Two infant deaths})$. Because double-infanticide is also exceedingly rare, comparing the relative likelihoods using Bayes' theorem yields a vastly different conclusion.
