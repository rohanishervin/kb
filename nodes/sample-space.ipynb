# Sample Space

## 1\. Definition

In probability theory, the **sample space** (usually denoted by $S$, $\Omega$, or $U$) of an [experiment](experiment.ipynb) is the set of *all* possible, mutually exclusive [outcomes](outcome.ipynb) of that experiment.

An individual outcome $\omega \in S$ is often called a sample point.

## 2\. Classifications of Sample Spaces

Sample spaces are broadly categorized by their cardinality (size):

1.  **Finite / Discrete:** The space contains a finite number of distinct outcomes.
      * *Example:* Rolling a standard six-sided die: $S = \{1, 2, 3, 4, 5, 6\}$.
2.  **Countably Infinite / Discrete:** The space contains an infinite number of outcomes, but they can be put into a one-to-one correspondence with the natural numbers $\mathbb{N}$.
      * *Example:* Flipping a coin until a "Heads" appears. The outcome is the number of flips required: $S = \{1, 2, 3, 4, \dots\}$.
3.  **Uncountably Infinite / Continuous:** The space contains an unbroken continuum of outcomes, typically represented by an interval of real numbers.
      * *Example:* The exact time it takes for a radioactive atom to decay: $S = [0, \infty)$.

By itself, the sample space is just a raw set. To do probability, we must group these outcomes into [events](event.ipynb) and assign probabilities to them using a [probability](probability.ipynb) measure.
