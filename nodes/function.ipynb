# Function
Given two sets $\boxed{A}$ and $\boxed{B}$, a function $f$ from $A$ to $B$ is a rule or mapping that takes each element $x \in A$ and associates with it a single element of $B$. In this case, we write $f : A 
\rightarrow B$.

## Function Notation
Given an element $x \in A$, the expression $f(x)$ is used to represent the element of $B$ associated with $x$ by $f$.

## Domain and Range
The set $A$ is called the **domain** of $f$. The **range** of $f$ is not necessarily equal to $B$ but refers to the subset of $B$ given by $\{y \in B : y = f(x) \text{ for some } x \in A\}$.
