# Adding Inequalities (not sure about title)

for all $a,b,c,d \in F$ when F is an [ordered filed](ordered-field.ipynb) we have:
if $a<b$ and $c<d$ then $a+b<c+d$

# Proof

$a<b$ so $b-a>0$ according to oredered field third property. similarly $d-c>0$. if we add 2 positive numbers it statys positive so:
$(b-a) + (d-c) > 0$ with some rearrangements we get:

$$ (b+d)-(a+c) > 0 $$

which is:
$$a+c<b+d$$
