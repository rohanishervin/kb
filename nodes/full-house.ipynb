# Probability of a Full House

## 1\. Problem Statement

In a standard 52-card deck, there are $4$ suits and $13$ ranks. A **"full house"** is a 5-card poker hand that contains exactly three cards of one rank (a "three-of-a-kind") and exactly two cards of another distinct rank (a "pair").

*Example:* Three Queens and two 4s.

What is the exact probability of being dealt a full house from a perfectly shuffled deck?

-----

## 2\. Mathematical Setup

Because every possible combination of 5 cards is equally likely to be drawn, we can rely entirely on the [naive definition of probability](probability-naive-definition.ipynb):
$$P(\text{Full House}) = \frac{\text{Number of Favorable Outcomes}}{\text{Total Possible 5-Card Hands}}$$

The denominator is the total number of distinct 5-card hands we can draw from 52 cards. Order does not matter, so we use the [binomial coefficient formula](binomial-coefficient-formula.ipynb):
$$|\Omega| = \binom{52}{5} = \frac{52!}{5! \cdot 47!} = 2,598,960$$

-----

## 3\. Calculating Favorable Outcomes

To find the numerator, we construct a full house step-by-step. By the [multiplication rule](multiplication-rule.ipynb), we multiply the number of ways to execute each independent step:

1.  **Choose the rank for the triplet:** We must select 1 rank out of the 13 available ranks (e.g., Queens).
    $$\binom{13}{1} = 13$$

2.  **Choose the specific cards for the triplet:** A rank has 4 suits. We must select 3 of those 4 suits to form our three-of-a-kind.
    $$\binom{4}{3} = 4$$

3.  **Choose the rank for the pair:** We must select 1 rank for the pair. Since we cannot reuse the rank we chose for the triplet, we only have 12 ranks left to choose from (e.g., 4s).
    $$\binom{12}{1} = 12$$

4.  **Choose the specific cards for the pair:** We must select 2 out of the 4 available suits for this new rank.
    $$\binom{4}{2} = 6$$

Multiplying these steps together yields the total number of valid full house combinations:
$$\text{Favorable Outcomes} = \binom{13}{1} \times \binom{4}{3} \times \binom{12}{1} \times \binom{4}{2}$$
$$\text{Favorable Outcomes} = 13 \times 4 \times 12 \times 6 = 3,744$$

-----

## 4\. Final Probability Calculation

Plugging our numerator and denominator back into the naive probability formula:
$$P(\text{Full House}) = \frac{3,744}{2,598,960} \approx 0.00144$$

You have approximately a **$0.144\%$** chance (or about 1 in 694) of being dealt a full house in a 5-card draw.

-----

## 5\. Intuition: The "Asymmetry" Pitfall

**Common Mistake:** Why didn't we just pick the two ranks we need at the same time using $\binom{13}{2}$?

Because a full house is **asymmetric**.

  * Hand A: Three Queens and Two 4s.
  * Hand B: Three 4s and Two Queens.

These are two entirely different poker hands. If we simply used $\binom{13}{2}$ to grab "Queens and 4s", we wouldn't have assigned which one is the triplet and which one is the pair.

By executing Step 1 ($\binom{13}{1}$) and Step 3 ($\binom{12}{1}$) sequentially, we naturally account for this asymmetry. Notice that $\binom{13}{1} \times \binom{12}{1} = 156$. If we incorrectly used $\binom{13}{2}$, we would get $78$ (exactly half), because we would be failing to swap the triplet and the pair\!

*(Note: For a symmetric hand, like "Two Pairs" (e.g., Two Queens, Two 4s, and a junk card), we absolutely MUST use $\binom{13}{2}$ to avoid overcounting\!)*

-----

## 6\. Python Simulation

You can verify the theoretical probability empirically by simulating millions of random poker hands.

```python
import numpy as np
from collections import Counter

def simulate_full_house(trials=1000000):
    # A deck of 52 cards represented just by their 13 ranks (0 through 12)
    # Each rank appears exactly 4 times
    deck = np.repeat(np.arange(13), 4)
    
    full_houses = 0
    
    for _ in range(trials):
        # Draw 5 cards without replacement
        hand = np.random.choice(deck, 5, replace=False)
        
        # Count the frequency of each rank in the hand
        counts = list(Counter(hand).values())
        
        # A full house requires exactly one count of 3 and one count of 2
        if 3 in counts and 2 in counts:
            full_houses += 1
            
    probability = full_houses / trials
    display(f"Simulated Probability: {probability:.5f} ({probability*100:.3f}%)")
    display(f"Theoretical Probability: 0.00144 (0.144%)")

# Run the simulation (Note: 1 million trials takes a few seconds)
simulate_full_house(1000000)
```
