# Vandermonde's Identity

## 1\. The Theorem Statement

**Vandermonde's Identity** (also known as Vandermonde's Convolution) is a fundamental combinatorial theorem relating binomial coefficients. It states that for any non-negative integers $m, n,$ and $k$:

$$\binom{m+n}{k} = \sum_{j=0}^{k} \binom{m}{j} \binom{n}{k-j}$$

At its core, this identity describes how to calculate the number of ways to choose $k$ items from a combined pool of $m + n$ items by breaking the selection down into every possible combination of subgroups.

-----

## 2\. Proof via Story

Using the technique of [story proofs](story-proofs.ipynb), we can intuitively demonstrate why this equality holds without expanding massive factorials.

**The Scenario:** Imagine an organization consists of exactly $m$ men and $n$ women. The board needs to form a special committee of exactly $k$ people. How many distinct committees can be formed?

**Method A: The Direct Approach (Left Side)**

  * We can simply ignore the gender breakdown entirely. We have a combined pool of $m + n$ total people, and we need to choose $k$ of them.
  * The total number of ways to do this is simply $\binom{m+n}{k}$.

**Method B: The Stratified Approach (Right Side)**

  * Alternatively, we can break the selection down by the exact number of men on the committee, which we will call $j$. The number of men $j$ can range anywhere from $0$ up to $k$.
  * If we choose exactly $j$ men (which can be done in $\binom{m}{j}$ ways), we are mathematically forced to choose exactly $k - j$ women to complete the $k$-person committee (which can be done in $\binom{n}{k-j}$ ways).
  * By the [multiplication rule](multiplication-rule.ipynb), there are exactly $\binom{m}{j}\binom{n}{k-j}$ ways to form a committee that contains exactly $j$ men.
  * To account for *every* possible committee composition, we must sum this product across all possible values of $j$ (from $0$ up to $k$).
  * This yields the right side: $\sum_{j=0}^{k} \binom{m}{j} \binom{n}{k-j}$.

**Conclusion:** Because both methods perfectly count the total number of distinct $k$-person committees, the two expressions must be strictly equal. $\square$

-----

## 3\. Proof via the Binomial Theorem (Algebraic)

Vandermonde's Identity can also be rigorously proven using polynomials and the [binomial theorem](binomial-theorem.ipynb).

Consider the polynomial expansion of $(1 + x)^{m+n}$. By the binomial theorem, the coefficient of the $x^k$ term in this expansion is exactly:
$$\binom{m+n}{k}$$

Now, let's rewrite the polynomial using exponent rules:
$$(1 + x)^{m+n} = (1 + x)^m (1 + x)^n$$

Expand both of these smaller binomials:
$$\left( \sum_{i=0}^{m} \binom{m}{i} x^i \right) \left( \sum_{l=0}^{n} \binom{n}{l} x^l \right)$$

When we multiply these two massive sums together, how do we form the $x^k$ term? We must multiply an $x^j$ term from the first sum with an $x^{k-j}$ term from the second sum (since $x^j \cdot x^{k-j} = x^k$).
The coefficient for this specific combination is $\binom{m}{j}\binom{n}{k-j}$.

To get the *total* coefficient for $x^k$, we must add up the coefficients for all possible pairs of $j$ and $k-j$ that multiply to $x^k$. This is exactly:
$$\sum_{j=0}^{k} \binom{m}{j} \binom{n}{k-j}$$

Since both expressions represent the exact same coefficient of $x^k$ in the expansion of $(1+x)^{m+n}$, they must be equal. $\square$

## 4\. Practical Example

**Question:** You have $5$ red balls and $4$ blue balls in a bag. You draw $3$ balls. Show how Vandermonde's Identity calculates the total possible combinations.

**Solution:**
Here, $m = 5$ (red), $n = 4$ (blue), and $k = 3$ (drawn).

**Direct Calculation (LHS):**
$$\binom{5+4}{3} = \binom{9}{3} = \frac{9 \times 8 \times 7}{3 \times 2 \times 1} = 84$$

**Stratified Sum (RHS):**
We sum $\binom{5}{j}\binom{4}{3-j}$ for $j = 0, 1, 2, 3$:

  * $j=0$ (0 red, 3 blue): $\binom{5}{0}\binom{4}{3} = 1 \times 4 = 4$
  * $j=1$ (1 red, 2 blue): $\binom{5}{1}\binom{4}{2} = 5 \times 6 = 30$
  * $j=2$ (2 red, 1 blue): $\binom{5}{2}\binom{4}{1} = 10 \times 4 = 40$
  * $j=3$ (3 red, 0 blue): $\binom{5}{3}\binom{4}{0} = 10 \times 1 = 10$

Total $= 4 + 30 + 40 + 10 = 84$. The identity holds perfectly.

## 5\. Python Verification

```python
import math

def nCr(n, r):
    """Safely calculates the binomial coefficient."""
    if r > n or r < 0: 
        return 0
    return math.comb(n, r)

def verify_vandermonde(m=5, n=4, k=3):
    # Left Hand Side: Direct Selection
    lhs = nCr(m + n, k)
    
    # Right Hand Side: Stratified Sum
    rhs = sum(nCr(m, j) * nCr(n, k - j) for j in range(k + 1))
    
    display(f"Parameters: m={m}, n={n}, k={k}")
    display(f"LHS (Direct): {lhs}")
    display(f"RHS (Stratified Sum): {rhs}")
    display(f"Identity Holds: {lhs == rhs}")

verify_vandermonde(5, 4, 3)
```
