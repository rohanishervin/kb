# Event

## 1\. Definition

In probability theory, an **event** $A$ is a subset of the [sample-space](sample-space.ipynb) $S$ to which a probability is assigned.

$$A \subseteq S$$

We say that the event $A$ has **occurred** if the actual [outcome](outcome.ipynb) $\omega$ of the experiment is an element of $A$ ($\omega \in A$).

## 2\. Set Operations on Events

Because events are subsets, the logical relationships between them are entirely governed by standard set theory:

  * **Intersection ($A \cap B$):** The event that *both* $A$ and $B$ occur.
  * **Union ($A \cup B$):** The event that *at least one* of $A$ or $B$ occurs.
  * **Complement ($A^c$):** The event that $A$ does *not* occur. ($A^c = S \setminus A$).
  * **Subset ($A \subseteq B$):** If event $A$ occurs, then event $B$ *must* also occur.

Two events $A$ and $B$ are **disjoint** (mutually exclusive) if they cannot occur at the same time, meaning $A \cap B = \emptyset$.
