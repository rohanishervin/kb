# Binomial Coefficient Formula

As we defined [binomial coefficients](binomial-coefficient.ipynb) earlier. now we can find the formula for $k \leq n$:
$$\binom{n}{k} = \frac{n(n-1)(n-2)... (n-k+1)}{k!} = \frac{n!}{k!(n-k)!} $$

# Proof

To make an ordered choice of k elements without replacement based on [Sampling without replacement](sampling-without-replacement.ipynb) we have $n(n-1)(n-2)... (n-k+1)$ elements. but we don't care about their ordering. so we can use [adjusting for overcounting](adjusting-for-overcounting.ipynb) and divide the result by $k!$.
