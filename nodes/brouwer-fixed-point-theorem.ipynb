# Brouwer Fixed Point Theorem

The Brouwer Fixed Point Theorem is a fundamental result in real analysis that guarantees the existence of a fixed point for continuous functions mapping a compact, convex set into itself. 

## 1\. Statement of the Theorem

**Theorem (Brouwer Fixed Point Theorem):**
Let $K \subset \mathbb{R}^n$ be a convex and [compact](compactness.ipynb) set. If $f: K \to K$ is a continuous function, then $f$ has at least one fixed point. That is, there exists some $x^* \in K$ such that:
$$f(x^*) = x^*$$

::: {.callout-note}
## Intuition: The "Stirring" Analogy
A classic intuitive explanation for Brouwer's theorem in 2D is the "coffee cup" analogy. If you take a cup of coffee (a compact, convex space) and stir it gently (a continuous transformation $f$), once the liquid comes to rest, there is at least one particle of coffee that is in the exact same absolute location it was before you started stirring. 
:::

## 2\. The 1-Dimensional Case

The proof of Brouwer's theorem for $n=1$ is relatively simple and relies on foundational theorems of real analysis. However, the proof for general dimensions $n > 1$ is significantly more difficult and typically requires tools from algebraic topology (such as homology groups or the fundamental group).

**Theorem (1D Brouwer Fixed Point):**
If $f: [a,b] \to [a,b]$ is continuous, then $f$ has a fixed point.

::: {.callout-tip collapse="true"}
## Proof of 1D Brouwer Fixed Point
We will provide two methods to prove this result. 

**Method 1: Using the Intermediate Value Theorem**

Define a new function $g: [a,b] \to \mathbb{R}$ by:
$$g(x) = f(x) - x$$

Because $f$ is continuous and the identity function $x \mapsto x$ is continuous, their difference $g$ is also continuous. We can evaluate $g$ at the endpoints of the interval:

1. At $x = a$, we have $g(a) = f(a) - a$. Since the codomain of $f$ is $[a,b]$, we know $f(a) \ge a$. Therefore, $g(a) \ge 0$.
2. At $x = b$, we have $g(b) = f(b) - b$. Since the codomain of $f$ is $[a,b]$, we know $f(b) \le b$. Therefore, $g(b) \le 0$.

By the Intermediate Value Theorem, since $g(x)$ is continuous and changes sign (or is zero) between $a$ and $b$, there exists at least one point $x^* \in [a,b]$ such that $g(x^*) = 0$. 

This implies:
$$f(x^*) - x^* = 0 \implies f(x^*) = x^*$$
Thus, $x^*$ is a fixed point. $\blacksquare$

**Method 2: Using the[Supremum**

Consider the set of points where the function lies above the identity line. We can locate the fixed point by looking at the supremum of such points:
$$x^* = \sup \{ x \in [a,b] : \forall y < x, f(y) > y \}$$
By the Axiom of Completeness, this supremum exists. Using the continuity of $f$, one can formally show that the limit of this process forces $f(x^*) = x^*$. $\blacksquare$
:::

## 3\. Necessity of the Conditions (Counterexamples)

The requirements that $K$ must be both **compact** and **convex** are strictly necessary. If either condition is dropped, the theorem fails to hold.

**Counterexample 1: Non-Compact Space**
Let $K = (0,1)$, which is convex but **not compact** (it is not closed). 
Define a continuous function $f: (0,1) \to (0,1)$ as:
$$f(x) = \frac{x}{2}$$
For a fixed point to exist, we would need $x = \frac{x}{2} \implies x = 0$. However, $0 \notin (0,1)$. Therefore, $f$ has no fixed point in $K$. 

**Counterexample 2: Non-Convex Space**
Let $K$ be an annulus (a ring-shaped region), which is compact but **not convex** (it has a hole in the middle). 
Define $f: K \to K$ as a rotation of $90^\circ$ around the origin. 
Since every point is shifted along a circular path, no point maps to itself. Therefore, $f$ has no fixed point. 

::: {.callout-warning}
## Common Misconception
People often forget that the domain and codomain must be exactly the *same* space $K$. If $f$ maps $K$ to a different space $Y$, the concept of a fixed point $f(x) = x$ is not even well-defined!
:::

## 4\. The Fixed Point Property (FPP) and Homeomorphisms

We can generalize the concept of "always having a fixed point" to any topological space. 

**Definition (Fixed-Point Property):**
A [topological space](metric-space-topology.ipynb#the-axioms-of-topology) $K$ is said to have the **fixed-point property (FPP)** if every continuous function $f: K \to K$ has at least one fixed point. 

A powerful result in topology is that the fixed-point property is a **topological invariant**; it is preserved under [homeomorphisms](continuity-topology.ipynb#homeomorphisms).

**Theorem:**
If a space $K$ is homeomorphic to a set $K'$ that has the fixed-point property (such as a compact, convex set in $\mathbb{R}^n$), then $K$ also has the fixed-point property.

::: {.callout-tip collapse="true"}
## Proof of Topological Invariance of FPP
Let $K'$ be a compact and convex set in $\mathbb{R}^n$. By the Brouwer Fixed Point Theorem, $K'$ has the fixed-point property. 
Assume there exists a homeomorphism $\phi: K \to K'$. This means $\phi$ is a continuous bijection with a continuous inverse $\phi^{-1}: K' \to K$.

Let $f: K \to K$ be an arbitrary continuous function on $K$. We want to show that $f$ has a fixed point.

**Step 1:** Construct a function on $K'$
Define a new function $f': K' \to K'$ as the composition:
$$f' = \phi \circ f \circ \phi^{-1}$$
Since $f'$, $\phi$, and $\phi^{-1}$ are all continuous functions, their composition $f'$ is continuous on $K'$.

**Step 2:** Apply FPP on $K'$
Because $K'$ has the fixed-point property, the continuous function $f'$ must have a fixed point $x \in K'$. Thus:
$$f'(x) = x$$

**Step 3:** Translate the fixed point back to $K$
Substitute our definition of $f'$ into the fixed point equation:
$$\phi(f(\phi^{-1}(x))) = x$$
Apply $\phi^{-1}$ to both sides of the equation:
$$\phi^{-1}(\phi(f(\phi^{-1}(x)))) = \phi^{-1}(x)$$
$$f(\phi^{-1}(x)) = \phi^{-1}(x)$$

Let $y = \phi^{-1}(x)$. Since $x \in K'$ and $\phi^{-1}$ maps into $K$, we know $y \in K$. Substituting $y$ into the equation gives:
$$f(y) = y$$
Thus, $f$ has a fixed point in $K$. Because $f$ was arbitrary, $K$ has the fixed-point property. $\blacksquare$
:::
