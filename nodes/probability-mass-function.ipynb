# The Probability Mass Function (PMF)

**Definition (Probability mass function):** The probability mass function (PMF) of a discrete [r.v.](random-variables.ipynb) $X$ is the function $p_X$ given by:
$$p_X(x) = P(X = x)$$
Note that this function evaluates to a positive number if $x$ is in the support of $X$, and exactly 0 otherwise.

::: {.callout-note title="Remark: X = x is an Event"}
In writing $P(X = x)$, we are using the expression $X = x$ as shorthand to denote an *event*. It consists of all underlying outcomes $s$ to which the random variable $X$ assigns the specific number $x$. 

Formally, $\{X = x\}$ is defined as the set $\{s \in S : X(s) = x\}$, but writing $\{X = x\}$ is shorter and more intuitive. 

**Example:** If $X$ counts the number of Heads in two fair coin tosses, then the event $\{X = 1\}$ consists of the sample outcomes $HT$ and $TH$, which are the two outcomes to which $X$ assigns the number 1. Since $\{HT, TH\}$ is a valid subset of the sample space, it is an event. So it makes perfect sense to talk about $P(X = 1)$, or more generally, $P(X = x)$. 

If $\{X = x\}$ were anything other than an event, it would make no sense to calculate its probability! Consequently, it does not make sense to write “$P(X)$”; we can only take the probability of an event, not of a random variable itself.
:::

## Properties of a Valid PMF

A function qualifies as a valid Probability Mass Function if it satisfies two specific criteria.

**Theorem:** For any discrete random variable $X$ with support values $x_j$, the PMF must satisfy:

1.  **Non-negativity:** $p_X(x_j) \geq 0$
2.  **Sums to 1:** $\sum_{j=1}^{\infty} p_X(x_j) = 1$

**Proof:**
The first criterion is true simply since probability is inherently nonnegative. 

The second criterion is true since $X$ must take on *some* value when an experiment is run, and the events $\{X = x_j\}$ are disjoint (mutually exclusive, as the function $X$ cannot map a single outcome $s$ to two different numbers). Therefore, by the axioms of probability:
$$\sum_{j=1}^{\infty} P(X = x_j) = P\left(\bigcup_{j=1}^{\infty} \{X = x_j\}\right) = P(X = x_1 \text{ or } X = x_2 \text{ or } \dots) = 1$$
$\square$
