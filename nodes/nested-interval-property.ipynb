# Nested Interval Property

## 1\. Theorem Statement

For each $n \in \mathbb{N}$, assume we are given a closed and bounded [interval](interval.ipynb) $I_n = [a_n, b_n] = \{x \in \mathbb{R} : a_n \leq x \leq b_n\}$.

Assume also that the intervals are **nested**, meaning each $I_n$ contains $I_{n+1}$:
$$I_1 \supseteq I_2 \supseteq I_3 \supseteq \dots \supseteq I_n \supseteq I_{n+1} \supseteq \dots$$

Then, the resulting nested sequence of closed intervals has a nonempty intersection. That is, there exists at least one point $x \in \mathbb{R}$ that belongs to every interval:
$$\bigcap_{n=1}^\infty I_n \neq \emptyset$$

## 2\. Formal Proof

The proof relies fundamentally on the [**Axiom of Completeness**](axiom-of-completeness.ipynb) (the property that every non-empty subset of $\mathbb{R}$ that is bounded above has a supremum).

**Proof:**

Let $A = \{a_n : n \in \mathbb{N}\}$ be the set of all left endpoints.

Because the intervals are nested ($I_1 \supseteq I_2 \supseteq \dots$), we know that the left endpoints are monotonically increasing and the right endpoints are monotonically decreasing:
$$a_1 \leq a_2 \leq a_3 \leq \dots \leq b_3 \leq b_2 \leq b_1$$

In fact, *every* left endpoint $a_n$ is strictly less than or equal to *every* right endpoint $b_m$, regardless of whether $n \leq m$ or $n > m$.
*(If $n \leq m$, then $a_n \leq a_m \leq b_m$. If $n > m$, then $a_n \leq b_n \leq b_m$.)*

This means the set $A$ is bounded above by any right endpoint, such as $b_1$.

By the Axiom of Completeness, since $A$ is non-empty and bounded above, it must have a supremum. Let $x = \sup A$.

Since $x$ is an upper bound for $A$, it must be greater than or equal to every element in $A$:
$$a_n \leq x \quad \text{for all } n \in \mathbb{N}$$

Since every right endpoint $b_m$ is also an upper bound for $A$, and $x$ is the *least* upper bound, $x$ must be less than or equal to every right endpoint:
$$x \leq b_n \quad \text{for all } n \in \mathbb{N}$$

Combining these two facts gives:
$$a_n \leq x \leq b_n \quad \text{for all } n \in \mathbb{N}$$

This is the exact definition of $x \in [a_n, b_n]$. Therefore, $x \in I_n$ for all $n$, which means $x \in \bigcap_{n=1}^\infty I_n$.

Thus, the intersection is nonempty. $\square$

## 3\. Corollary: Length Approaching Zero

If we add the condition that the lengths of the intervals converge to zero, the intersection contains exactly one point.

**Theorem:** If $\lim_{n \to \infty} (b_n - a_n) = 0$, then $\bigcap_{n=1}^\infty I_n$ consists of a single, unique point.

*(Intuition: The intervals clamp down tighter and tighter until they crush into a single real number. This is often used to prove the uncountability of $\mathbb{R}$.)*

## 4\. Why Closed and Bounded? (Counterexamples)

The Nested Interval Property fails if the intervals are either open or unbounded. The "closed and bounded" criteria are absolutely required.

  * **Fails for Open Intervals:** Let $I_n = \left(0, \frac{1}{n}\right)$.
    The intervals are nested: $(0, 1) \supseteq (0, 1/2) \supseteq (0, 1/3) \dots$
    However, $\bigcap_{n=1}^\infty \left(0, \frac{1}{n}\right) = \emptyset$. There is no real number strictly greater than $0$ that is also strictly less than $\frac{1}{n}$ for *all* $n$.

  * **Fails for Unbounded Intervals:** Let $I_n = [n, \infty)$.
    The intervals are nested: $[1, \infty) \supseteq [2, \infty) \supseteq [3, \infty) \dots$
    However, $\bigcap_{n=1}^\infty [n, \infty) = \emptyset$. [Archimedes' principle](archimedean-property-of-R.ipynb) guarantees there is no real number greater than every natural number $n$.
