#### 1\. Defining the "Cut"

To construct $\mathbb{R}$, we assume the [Ordered Field](ordered-field.ipynb) $(\mathbb{Q}, +, \cdot)$ is given. We define the real numbers using a partition method. A **Dedekind Cut** is a pair of subsets $(A, B)$ of $\mathbb{Q}$ that satisfies the following conditions:

  * **Non-triviality:** $A \cup B = \mathbb{Q}$, $A \cap B = \emptyset$, and $A, B \neq \emptyset$.
  * **Downward Closure:** For every $a \in A$ and $b \in B$, it must hold that $a < b$.
  * **No Largest Element:** $A$ contains no largest element.

**Definition:** A **real number** $x$ is defined as a Dedekind cut $A|B$ in $\mathbb{Q}$.

**Definition:** The set $\mathbb{R}$ is the [class](pre-set-theory.ipynb) of all Dedekind cuts.

::: {.callout-note}

## Understanding the Partition

In practice, since $B$ is completely determined by $A$ (where $B = \mathbb{Q} \setminus A$), we often refer to the set $A$ itself as the "cut." The condition $a < b$ ensures that $A$ lies entirely to the left of $B$ on the rational line.
:::

#### 2\. Identifying Rationals in $\mathbb{R}$

Strictly speaking, $\mathbb{Q}$ is not a subset of $\mathbb{R}$ because $\mathbb{R}$ is a collection of sets (cuts), while $\mathbb{Q}$ is a set of numbers. To resolve this, we create an **isomorphic correspondence** between the existing rational numbers and our new construction. For every $r \in \mathbb{Q}$, we define a **rational cut** $r^*$:

$$r^* = \{ q \in \mathbb{Q} : q < r \}$$

Hence, we can state $\mathbb{Q} \subseteq \mathbb{R}$ by identifying each $r \in \mathbb{Q}$ with its corresponding cut $r^* \in \mathbb{R}$.

::: {.callout-important}

## Why the "No Largest Element" Condition?

This condition is essential for **uniqueness**. Without it, a rational number like $1$ could be represented by two distinct lower sets:

1.  $A_1 = \{q \in \mathbb{Q} : q < 1\}$
2.  $A_2 = \{q \in \mathbb{Q} : q \leq 1\}$

Both sets represent the same "gap" or "point" on the line. By mandating that $A$ has no largest element, we disqualify $A_2$, ensuring that every real number has a **unique** representation as a cut.
:::
