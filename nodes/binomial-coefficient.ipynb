# Binomial Coefficient Formula

## 1\. Definition and Formula

Building on the concept of [binomial coefficients](binomial-coefficient.ipynb), the algebraic formula for calculating the number of ways to choose $k$ elements from a set of $n$ distinct elements (where $0 \leq k \leq n$) without regard to order is denoted as $\binom{n}{k}$ (read as "$n$ choose $k$").

The explicit formula is:
$$\binom{n}{k} = \frac{n(n-1)(n-2)\dots(n-k+1)}{k!} = \frac{n!}{k!(n-k)!}$$

*(Note: By definition, $0! = 1$. If $k > n$, then $\binom{n}{k} = 0$, because you cannot choose more items than exist in the set).*

-----

## 2\. Intuition and Derivation

**Why does this formula work?** Imagine you are forming a 3-person committee from a group of 10 people.
If the *order* of selection mattered (e.g., President, Vice President, Treasurer), this would be [sampling without replacement](sampling-without-replacement.ipynb). You would have $10$ choices for the first spot, $9$ for the second, and $8$ for the third.
$$P(10, 3) = 10 \times 9 \times 8 = 720 \text{ sequences}$$

Algebraically, this permutation formula is $\frac{n!}{(n-k)!}$.

However, for a simple committee, the subset $\{\text{Alice, Bob, Charlie}\}$ is exactly the same as $\{\text{Bob, Charlie, Alice}\}$. Because a subset is *unordered*, we have wildly overcounted the number of unique groups.

How much did we overcount by? Every unique group of $3$ people can be arranged among themselves in $3! = 6$ different ways.

By applying the principle of [adjusting for overcounting](adjusting-for-overcounting.ipynb), we must divide our permutation total by the number of internal arrangements ($k!$).
$$\text{Combinations} = \frac{\text{Permutations}}{k!} = \frac{\frac{n!}{(n-k)!}}{k!} = \frac{n!}{k!(n-k)!}$$

Dividing $720$ sequences by $6$ internal arrangements yields exactly $120$ unique committees. $\square$

-----

## 3\. Key Properties and Proofs

The binomial coefficient possesses several elegant mathematical properties that are highly useful in probability calculations.

### Property 1: Symmetry

Choosing $k$ items to *keep* from a set of $n$ is mathematically identical to choosing $(n-k)$ items to *reject* (or leave behind).
$$\binom{n}{k} = \binom{n}{n-k}$$

**Algebraic Proof:**
Apply the factorial definition to the right-hand side:
$$\binom{n}{n-k} = \frac{n!}{(n-k)!(n - (n-k))!} = \frac{n!}{(n-k)!(k!)} = \binom{n}{k}$$
$\square$

### Property 2: Extreme Boundaries

There is exactly $1$ way to choose nothing, and exactly $1$ way to choose everything.
$$\binom{n}{0} = 1 \quad \text{and} \quad \binom{n}{n} = 1$$

**Algebraic Proof:**
$$\binom{n}{0} = \frac{n!}{0!(n-0)!} = \frac{n!}{1 \cdot n!} = 1$$
$$\binom{n}{n} = \frac{n!}{n!(n-n)!} = \frac{n!}{n! \cdot 0!} = 1$$
$\square$

## 4\. Classic Example

**Question:** A standard deck of cards contains $52$ cards. How many unique $5$-card poker hands can be dealt?

**Solution:**
Because the order in which the cards are dealt to your hand does not matter (you can rearrange them in your hand freely), we use the binomial coefficient:
$$\binom{52}{5} = \frac{52!}{5!(52-5)!} = \frac{52!}{5! \cdot 47!}$$

To calculate this by hand efficiently, we expand the numerator until the $47!$ cancels out:
$$\binom{52}{5} = \frac{52 \times 51 \times 50 \times 49 \times 48}{5 \times 4 \times 3 \times 2 \times 1}$$
$$= \frac{311,875,200}{120} = 2,598,960$$

There are exactly $2,598,960$ distinct $5$-card poker hands.
