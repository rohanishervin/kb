# Problem Set: Real Numbers

## Problem 1: Pugh, Ch. 1, P. 9

**Problem Statement:**
Let $x = A|B$, $x' = A'|B'$ be cuts in $\mathbb{Q}$. We defined $x + x' = (A + A') \mid \text{rest of } \mathbb{Q}$.

(a) Show that although $B + B'$ is disjoint from $A + A'$, it may happen in degenerate cases that $\mathbb{Q}$ is not the union of $A + A'$ and $B + B'$.
(b) Infer that the definition of $x + x'$ as $(A + A') \mid (B + B')$ would be incorrect.
(c) Why did we not define $x \cdot x' = (A \cdot A') \mid \text{rest of } \mathbb{Q}$?

**Mathematical Solution:**

**Part (a):**
First, we show that $(A + A') \cap (B + B') = \emptyset$.
By the definition of a [cut](def-dedekind-cuts.ipynb#defining-the-cut), every element in the lower class is strictly less than every element in the upper class. Thus, for any $a \in A, a' \in A'$ and $b \in B, b' \in B'$, we have $a < b$ and $a' \le b'$ (or $a \le b$ and $a' < b'$). 
Adding these inequalities gives $a + a' < b + b'$. Since every element of $A + A'$ is strictly smaller than every element of $B + B'$, their intersection must be empty.

To show that their union might not be $\mathbb{Q}$, consider the cuts for $x = \sqrt{2}$ and $x' = -\sqrt{2}$:

* $x = A|B$ where $A = \{q \in \mathbb{Q} : q < \sqrt{2}\}$ and $B = \{q \in \mathbb{Q} : q > \sqrt{2}\}$
* $x' = A'|B'$ where $A' = \{q \in \mathbb{Q} : q < -\sqrt{2}\}$ and $B' = \{q \in \mathbb{Q} : q > -\sqrt{2}\}$

If we form the sums:

* $A + A' = \{a + a' : a < \sqrt{2}, a' < -\sqrt{2}\} = \{q \in \mathbb{Q} : q < 0\}$
* $B + B' = \{b + b' : b > \sqrt{2}, b' > -\sqrt{2}\} = \{q \in \mathbb{Q} : q > 0\}$

The union $(A + A') \cup (B + B')$ is $\mathbb{Q} \setminus \{0\}$. The rational number $0$ is omitted, so the union is not $\mathbb{Q}$.

**Part (b):**
A Dedekind cut must be a partition of $\mathbb{Q}$, meaning the union of its left and right sets must equal exactly $\mathbb{Q}$. As shown in part (a), $(A + A') \cup (B + B')$ can omit a rational number (in this case, 0). Therefore, defining $x + x'$ as $(A + A') \mid (B + B')$ fails the basic definition of a cut, as it would leave "gaps" in the rationals.

**Part (c):**
Defining multiplication this way fails because the product of two negative numbers is positive, which violates the "downward closed" property required for the left half of a cut. 

For a counterexample, consider the cut for $x = -1$:
$$x = A|B \text{ where } A = \{q \in \mathbb{Q} : q < -1\} \text{ and } B = \{q \in \mathbb{Q} : q \ge -1\}$$

If we attempt to define $x \cdot x$ using $A \cdot A$:
$$A \cdot A = \{a_1 \cdot a_2 : a_1 < -1 \text{ and } a_2 < -1\}$$

Since multiplying two numbers less than $-1$ always yields a number strictly greater than $1$, we get $A \cdot A = \{q \in \mathbb{Q} : q > 1\}$. 
This set is not downward closed (e.g., $2 \in A \cdot A$, but $0 < 2$ and $0 \notin A \cdot A$). Thus, it cannot form the left half of a valid Dedekind cut.

## Problem 2: Pugh, Ch. 1, P. 15

**Problem Statement:**
Given $y \in \mathbb{R}$, $n \in \mathbb{N}$, and $\epsilon > 0$, show that for some $\delta > 0$, if $u \in \mathbb{R}$ and $|u - y| < \delta$ then $|u^n - y^n| < \epsilon$. 
*(Hint: Prove the inequality when $n = 1$, $n = 2$, and then do induction on $n$ using the identity $u^n - y^n = (u - y)(u^{n-1} + u^{n-2}y + \dots + y^{n-1})$).*

**Mathematical Solution (Direct Bounding Method):**

While the hint suggests induction, a direct algebraic bounding method is highly efficient and perfectly valid.

**Case $n = 1$:**
We want to show that if $|u - y| < \delta$, then $|u^1 - y^1| < \epsilon$. 
We simply choose $\delta = \epsilon$. 

**Case $n > 1$:**
We begin with the provided algebraic identity:
$$u^n - y^n = (u - y)(u^{n-1} + u^{n-2}y + \dots + u y^{n-2} + y^{n-1})$$

Taking the absolute value of both sides gives:
$$|u^n - y^n| = |u - y| \cdot |u^{n-1} + u^{n-2}y + \dots + y^{n-1}|$$

To control the size of $u$, we place a preliminary restriction on $\delta$. Let us restrict $\delta \le 1$. 
If we assume $|u - y| < 1$, we can use the reverse triangle inequality to bound $|u|$:
$$|u| - |y| \le |u - y| < 1 \implies |u| < |y| + 1$$

Now we apply the standard [triangle inequality](triangle-inequality.ipynb) to the second polynomial factor:
$$|u^{n-1} + u^{n-2}y + \dots + y^{n-1}| \le |u|^{n-1} + |u|^{n-2}|y| + \dots + |y|^{n-1}$$

Because we know that $|u| < |y| + 1$, and trivially $|y| < |y| + 1$, we can strictly bound every single term in this expansion by $(|y| + 1)^{n-1}$. Since there are exactly $n$ terms in the expansion:
$$|u^{n-1} + u^{n-2}y + \dots + y^{n-1}| < n(|y| + 1)^{n-1}$$

Substitute this bound back into our original absolute value equation:
$$|u^n - y^n| < |u - y| \cdot n(|y| + 1)^{n-1}$$

We want this entire expression to be strictly less than $\epsilon$. Therefore, we need:
$$|u - y| \cdot n(|y| + 1)^{n-1} < \epsilon \implies |u - y| < \frac{\epsilon}{n(|y| + 1)^{n-1}}$$

To satisfy both our preliminary restriction ($|u - y| < 1$) and our $\epsilon$-bound, we define $\delta$ as the minimum of the two:
$$\delta = \min\left(1, \frac{\epsilon}{n(|y| + 1)^{n-1}}\right)$$

With this choice of $\delta$, if $|u - y| < \delta$, it guarantees that $|u^n - y^n| < \epsilon$, completing the proof.

## Problem 3: Pugh, Ch. 1, P. 16

**Problem Statement:**
Given $x > 0$ and $n \in \mathbb{N}$, prove that there is a unique $y > 0$ such that $y^n = x$. That is, the $n^{\text{th}}$ root of $x$ exists and is unique. 
*(Hint: Consider $y = \text{l.u.b.}\{s \in \mathbb{R} : s^n \le x\}$. Then use Exercise 15 to show that $y^n$ can be neither $< x$ nor $> x$.)*

**Mathematical Solution:**

**Part 1: Existence of the Supremum**
Let $S = \{s \in \mathbb{R} : s^n \le x\}$. To use the Least Upper Bound property, we must show $S$ is non-empty and bounded above.

* **Non-empty:** Since $x > 0$, $0^n = 0 \le x$, so $0 \in S$.
* **Bounded above:** * If $x \le 1$, any $s > 1$ yields $s^n > 1 \ge x$, so $1$ is an upper bound. 
  * If $x > 1$, any $s > x$ yields $s^n > x^n > x$, so $x$ is an upper bound. 
Since $S$ is non-empty and bounded above, by the [Axiom of Completeness](axiom-of-completeness.ipynb#the-axiom-of-completeness-1), the supremum $y = \text{l.u.b.} S$ exists. Since $S$ contains positive numbers (e.g., small fractions), $y > 0$.

**Part 2: Proving $y^n = x$ (by contradiction)**

We will use the continuity property proven in Problem 15: For any $\epsilon > 0$, there exists a $\delta > 0$ such that $|u - y| < \delta \implies |u^n - y^n| < \epsilon$.

* **Case 1: Suppose $y^n < x$.**
  Let $\epsilon = x - y^n > 0$. By Problem 15, there exists a $\delta > 0$ satisfying the continuity condition. 
  Choose a specific test point $u = y + \frac{\delta}{2}$. 
  Since $|u - y| = \frac{\delta}{2} < \delta$, we know $|u^n - y^n| < \epsilon$. 
  Dropping the absolute value gives $u^n - y^n < x - y^n \implies u^n < x$.
  Because $u^n < x$, the number $u$ must belong to the set $S$. However, $u = y + \frac{\delta}{2} > y$. This is a contradiction, because $y$ is supposed to be an upper bound for all elements in $S$. Thus, $y^n \not< x$.

* **Case 2: Suppose $y^n > x$.**
  Let $\epsilon = y^n - x > 0$. Again, by Problem 15, there exists an associated $\delta > 0$.
  Choose a specific test point $u = y - \frac{\delta}{2}$. 
  Since $|u - y| = \frac{\delta}{2} < \delta$, we know $|u^n - y^n| < \epsilon$. 
  This means $y^n - u^n < y^n - x \implies x < u^n$.
  Now, take any arbitrary element $s \in S$. By definition, $s^n \le x$. Therefore, $s^n \le x < u^n$, which implies $s < u$. 
  Because every $s \in S$ is strictly less than $u$, $u$ is an upper bound for $S$. However, $u = y - \frac{\delta}{2} < y$. This is a contradiction, because $y$ is supposed to be the *least* upper bound. Thus, $y^n \not> x$.

Since $y^n$ is neither strictly less than nor strictly greater than $x$, we conclude exactly $y^n = x$.

**Part 3: Proving Uniqueness**
Suppose there exist two distinct positive numbers $y_1 > 0$ and $y_2 > 0$ such that $y_1^n = x$ and $y_2^n = x$. 
If $y_1 < y_2$, multiplying the inequality by itself $n$ times (which preserves the strict inequality since both are positive) yields $y_1^n < y_2^n \implies x < x$, which is a contradiction.
By symmetry, $y_1 > y_2$ also yields a contradiction. Therefore, $y_1 = y_2$, proving the root is unique.
