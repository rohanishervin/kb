# Probability


A **probability space** consists of a sample space $\mathbf{S}$ and a probability function $P$ which takes an event $A \subseteq S$ as input and returns $P(A)$, a real number between 0 and 1, as output.

The function $P$ must satisfy the following axioms:

### Axioms for Probability Function

#### 1. Non-Negativity and Normalization
$$\begin{align}
P(\emptyset) &= 0 \\
P(S) &= 1
\end{align}$$

#### 2. Countable Additivity
If $A_1, A_2, \ldots$ are disjoint events (meaning $A_i \cap A_j = \emptyset$ for $i \neq j$), then:
$$ P \left( \bigcup_{j=1}^{\infty} A_j \right) = \sum_{j=1}^{\infty} P(A_j) $$
