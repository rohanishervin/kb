# The Lebesgue Measure

The [outer measure](lebesgue-outer-measure.ipynb) $m^*$ is defined for *all* subsets of $\mathbb{R}$, but it suffers from a critical flaw: it is only countably subadditive. To build a robust theory of integration, we need a measure that is countably additive (where the measure of a union of disjoint pieces perfectly equals the sum of their individual measures).

We achieve this by restricting the outer measure strictly to the well-behaved [$\sigma$-algebra](sigma-algebra-of-measurable-sets.ipynb) of [measurable sets](measurable-sets.ipynb) $\mathcal{M}$.

## 1. Definition and Countable Additivity

**Definition:** The **Lebesgue measure**, denoted $m$, is defined as the restriction of the outer measure $m^*$ to the domain of measurable sets $\mathcal{M}$. 
That is, for any $E \in \mathcal{M}$:
$$
m(E) = m^*(E)
$$

**Proposition:** The Lebesgue measure is countably additive. Let $\{E_i\}_{i=1}^\infty$ be a countable sequence of pairwise disjoint measurable sets. Then:
$$
m\left(\bigcup_{i=1}^\infty E_i\right) = \sum_{i=1}^\infty m(E_i)
$$

::: {.callout-tip collapse="false"}
## Proof of Proposition

We already know from a previous lemma that finite additivity holds for disjoint measurable sets under the outer measure:
$$
m^*\left(\bigcup_{i=1}^n E_i\right) = \sum_{i=1}^n m^*(E_i)
$$

Since the finite union is a subset of the infinite union, i.e., $\bigcup_{i=1}^n E_i \subseteq \bigcup_{i=1}^\infty E_i$, the monotonicity of the outer measure implies:
$$
m^*\left(\bigcup_{i=1}^\infty E_i\right) \ge m^*\left(\bigcup_{i=1}^n E_i\right) = \sum_{i=1}^n m^*(E_i)
$$

This inequality holds for any arbitrary integer $n$. Taking the limit as $n \to \infty$, we obtain:
$$
m^*\left(\bigcup_{i=1}^\infty E_i\right) \ge \sum_{i=1}^\infty m^*(E_i)
$$

Simultaneously, we universally have countable subadditivity for the outer measure on any sets:
$$
m^*\left(\bigcup_{i=1}^\infty E_i\right) \le \sum_{i=1}^\infty m^*(E_i)
$$

Since both the "$\ge$" and "$\le$" inequalities hold, we must have exact equality. Because the countable union of measurable sets is itself measurable, we can replace the notation $m^*$ with the formally restricted Lebesgue measure $m$, yielding:
$$
m\left(\bigcup_{i=1}^\infty E_i\right) = \sum_{i=1}^\infty m(E_i)
$$
This establishes the countable additivity of the Lebesgue measure. $\blacksquare$
:::

## 2. A Complete Measure Space

With the Lebesgue measure formally defined, we have constructed a complete mathematical structure known as a [measure space](measure.ipynb).

The triplet $(\mathbb{R}, \mathcal{M}, m)$ forms a measure space because it satisfies three defining axioms:

1.  $\mathbb{R}$ is the underlying set.
2.  $\mathcal{M}$ is a $\sigma$-algebra over $\mathbb{R}$.
3.  $m: \mathcal{M} \to [0, \infty]$ is a countably additive measure where $m(\emptyset) = 0$.

Furthermore, this specific measure space is a **[complete measure](measure-properties-and-completion.ipynb) space**. 

::: {.callout-note}
## Intuition: Completeness
Completeness means that the measure space has no "holes" when it comes to negligible sets. If a set $B \in \mathcal{M}$ has measure zero ($m(B) = 0$), then *any* subset $A \subseteq B$ is automatically measurable and also has measure zero. 

We proved this previously when establishing the $\sigma$-algebra properties: a set bounded by a measure-zero set has an outer measure of zero, and any set with an outer measure of zero perfectly satisfies Carathéodory's criterion.
:::
