# Borel Sets and Measurability

Having established that the collection of [measurable sets](measurable-sets.ipynb), denoted $\mathcal{M}$, forms a [$\sigma$-algebra](sigma-algebra.ipynb), we must verify that this abstract collection actually contains the standard topological sets we use in analysis. In this note, we show that intervals, open sets, closed sets, and ultimately all Borel sets are Lebesgue measurable.

## 1. Intervals and Open Sets

Earlier, we proved via Carathéodory's criterion that the ray $[0, \infty)$ is measurable. Using similar algebraic manipulations, one can prove that any basic interval $(a, b)$, $[a, b]$, or rectangle in higher dimensions is a measurable set. From this foundation, we can measure any open set.

**Proposition:** Every open set in $\mathbb{R}$ is Lebesgue measurable.

::: {.callout-tip collapse="false"}
## Proof of Proposition

A fundamental theorem of the topology of the real line states that any open set $O \subseteq \mathbb{R}$ can be expressed as a countable union of disjoint open intervals:
$$O = \bigcup_{i=1}^\infty (a_i, b_i)$$
We know that every individual open interval $(a_i, b_i)$ is measurable. Furthermore, because $\mathcal{M}$ is a $\sigma$-algebra, it is closed under countable unions. Therefore, the countable union of these measurable intervals, which is exactly $O$, must also be measurable. $\blacksquare$
:::

<!-- CELL BREAK -->

## 2. Closed Sets

**Proposition:** Every closed set in $\mathbb{R}$ is Lebesgue measurable.

::: {.callout-tip collapse="false"}
## Proof of Proposition

By definition, a set $C \subseteq \mathbb{R}$ is closed if and only if its complement $C^c$ is open. 

We just proved that all open sets are measurable. Because the $\sigma$-algebra $\mathcal{M}$ is closed under complements, the complement of the open set $C^c$ (which is $(C^c)^c = C$) must also be measurable. $\blacksquare$
:::

<!-- CELL BREAK -->

## 3. The Borel $\sigma$-Algebra

The Borel $\sigma$-algebra, denoted $\mathcal{B}$, is defined abstractly as the *smallest* $\sigma$-algebra that contains all open sets in a given topological space. 

Because $\mathcal{M}$ is a $\sigma$-algebra, and we have just proven that $\mathcal{M}$ contains all open sets, it inherently satisfies the conditions of being a $\sigma$-algebra containing the open sets. Since $\mathcal{B}$ is the *intersection* of all such $\sigma$-algebras (making it the smallest one), it must be true that $\mathcal{B}$ is a subset of $\mathcal{M}$:

$$\mathcal{B} \subseteq \mathcal{M}$$

**Conclusion:** Every Borel set is Lebesgue measurable. This includes not just open and closed sets, but countable intersections of open sets ($G_\delta$ sets), countable unions of closed sets ($F_\sigma$ sets), and any further countable combinations of these.

## Are all subsets measurable?

Since $\mathcal{B} \subseteq \mathcal{M}$, it is natural to ask if $\mathcal{M}$ is simply the power set $\mathcal{P}(\mathbb{R})$ (i.e., every possible subset). It is not. Assuming the [Axiom of Choice](https://en.wikipedia.org/wiki/Axiom_of_choice), it is possible to construct pathological sets that cannot be measured consistently, such as the [Vitali set](https://en.wikipedia.org/wiki/Vitali_set). Therefore, the hierarchy is strict: $\mathcal{B} \subseteq \mathcal{M} \subset \mathcal{P}(\mathbb{R})$.
:::
