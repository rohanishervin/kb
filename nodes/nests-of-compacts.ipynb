# Nests of Compact Sets

When analyzing decreasing sequences of sets, [compactness](compactness.ipynb) provides a powerful guarantee that the sets won't simply "vanish" into nothingness.

## 1. Nested Sequences and Intersections

**Definition:** If $A_1 \supset A_2 \supset \dots \supset A_n \supset A_{n+1} \supset \dots$, then $(A_n)$ is called a **nested sequence of sets**. 

The infinite intersection of this sequence is the set of all points that survive in *every* step of the sequence:
$$\bigcap_{n=1}^\infty A_n = \{p : \text{for each } n \text{ we have } p \in A_n\} \text{}$$

---

## 2. The Intersection Theorem

**Theorem:** The intersection of a nested sequence of compact nonempty sets is both compact and nonempty.

**Proof (Compactness of the Intersection):**

* Let $(A_n)$ be a nested sequence of compact nonempty sets. 
* From our fundamental theorems of compactness, we know that every compact set $A_n$ is also a closed set. 
* A fundamental property of topology is that the arbitrary intersection of closed sets is always closed. 
* Therefore, the total intersection $\bigcap A_n$ is a closed subset trapped inside the first compact set, $A_1$. 
* Since any closed subset of a compact set is itself compact, the intersection is guaranteed to be compact.

**Proof (Nonemptiness of the Intersection):**

* It remains to show that the intersection is not completely empty.
* Because each $A_n$ is nonempty, we can construct a sequence by picking one point $a_n \in A_n$ for each $n \in \mathbb{N}$. 
* Because the sets are strictly nested ($A_1 \supset A_2 \dots$), every single point in this sequence $(a_n)$ ultimately lies inside the largest set, $A_1$. 
* Since $A_1$ is compact, this sequence $(a_n)$ must possess a subsequence $(a_{n_k})$ that converges to some limit point $p \in A_1$.
* Now consider $A_2$. Except for potentially the very first term, the entire tail of the subsequence $(a_{n_k})$ lies entirely inside $A_2$. Because $A_2$ is a closed set, it must contain the limit of any convergent sequence inside it, meaning $p \in A_2$. 
* By induction, this same logic holds true for $A_3$, $A_4$, and all subsequent sets in the nested sequence. 
* Since the limit point $p$ is contained within every single set $A_n$, it must be true that $p \in \bigcap A_n$. Thus, the intersection is definitively nonempty. $\square$

## 3. Set Diameter and Convergence to a Point

**Definition:** The **diameter** of a nonempty set $S \subset M$ is the supremum (least upper bound) of the distances $d(x, y)$ between all possible pairs of points inside $S$.
$$\text{diam}(S) = \sup \{d(x, y) : x, y \in S\}$$

**Corollary:** If, in addition to being nested, nonempty, and compact, the sets $A_n$ have a diameter that tends to 0 as $n \to \infty$, then the total intersection $A = \bigcap A_n$ consists of exactly one single point.

**Proof:**

* Let $A = \bigcap A_n$. For every individual $n \in \mathbb{N}$, the intersection $A$ is a subset of $A_n$ ($A \subset A_n$). 
* Because $A$ is a subset, its diameter can never exceed the diameter of the parent set $A_n$. Since $\text{diam}(A_n) \to 0$, it implies that $\text{diam}(A)$ must be exactly zero. 
* By the axioms of a metric space, distinct points must lie at a strictly positive distance from each other. Because the diameter is zero, $A$ can contain at most one point. 
* Since we proved in the previous theorem that $A$ contains at least one point, it must consist of exactly one single point. $\square$

## 4. Counterexample: Why Compactness is Strictly Required

It is vital to understand that nested, nonempty sets can easily have an empty intersection if they are *not* compact. 

Consider a sequence of **open** discs in $\mathbb{R}^2$ where the $n$-th disc is centered at $(1/n, 0)$ on the x-axis and has a radius of exactly $1/n$.
$$A_n = \{(x,y) \in \mathbb{R}^2 : (x - 1/n)^2 + y^2 < (1/n)^2\}$$

* These sets are strictly nested ($A_1 \supset A_2 \supset A_3 \dots$) and nonempty.
* However, because they are open sets, they are not compact.
* The intersection of all these open discs is completely **empty**. 
* Why? Because they are open discs, they do *not* contain their boundary. The origin $(0,0)$ is geometrically exactly on the boundary of every single disc, meaning it is never actually inside any of them! Any point strictly to the right of the origin will eventually be excluded as the discs shrink toward zero. If we had used closed discs (which are compact), their boundaries would be included, and their closures would perfectly intersect at exactly the origin.
