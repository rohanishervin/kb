# Limit Inferior and Limit Superior


## 1\. Definitions

For any [sequence](sequence.ipynb) $\{x_n\}$ of [real numbers](real-numbers.ipynb), we are often interested in the eventual behavior of its "tails." We define the sets representing these tails as $S_n = \{x_k : k \geq n\}$.

### Limit Inferior ($\liminf$)

If $\{x_n\}$ is bounded below, the **limit inferior** is the limit of the infima of the tails:
$$\liminf_{n \to \infty} x_n = \lim_{n \to \infty} (\inf S_n) = \lim_{n \to \infty} \left( \inf \{x_k : k \geq n\} \right)$$

### Limit Superior ($\limsup$)

If $\{x_n\}$ is bounded above, the **limit superior** is the limit of the suprema of the tails:
$$\limsup_{n \to \infty} x_n = \lim_{n \to \infty} (\sup S_n) = \lim_{n \to \infty} \left( \sup \{x_k : k \geq n\} \right)$$

-----

## 2\. The Monotonicity of Tails

The power of these definitions lies in the fact that they transform a potentially chaotic sequence into two **monotone** sequences.

Let $a_n = \inf S_n$ and $b_n = \sup S_n$. As $n$ increases, the tail $S_n$ becomes smaller because we are removing elements from the set: $S_{n+1} \subseteq S_n$.

  * **$\liminf$ is increasing:** Since $S_{n+1} \subseteq S_n$, the greatest lower bound of $S_{n+1}$ must be greater than or equal to that of $S_n$. Formally, $A \subseteq B \implies \inf A \geq \inf B$. Thus, $a_n \leq a_{n+1}$.
  * **$\limsup$ is decreasing:** Similarly, removing elements from a set cannot increase its supremum. Formally, $A \subseteq B \implies \sup A \leq \sup B$. Thus, $b_n \geq b_{n+1}$.

Because these sequences are monotone and bounded (assuming $\{x_n\}$ is bounded), they are guaranteed to converge in $\mathbb{R}$ by the [Monotone Convergence Theorem](Cauchy-sequences.ipynb).

-----

## 3\. Proposition: Convergence via Envelopes

**Theorem:** Let $\{x_n\}$ be a bounded sequence. Then $\{x_n\}$ converges to a limit $L$ if and only if:
$$\liminf_{n \to \infty} x_n = \limsup_{n \to \infty} x_n = L$$

**Proof:**

**($\implies$) Direction:**

Suppose the sequence $\{x_n\}$ converges to $L$. By the definition of a limit, for any arbitrarily small $\epsilon > 0$, there exists an integer $N$ such that for all $k \geq N$, we have:
$$L - \epsilon < x_k < L + \epsilon$$

Now, consider the tail of the sequence, $S_n = \{x_k : k \geq n\}$, for any $n \geq N$. Since every element in this tail satisfies the inequality above, it follows that:

  * $L - \epsilon$ is a valid lower bound for $S_n$. Because the infimum is the *greatest* lower bound, we must have $\inf S_n \geq L - \epsilon$.
  * $L + \epsilon$ is a valid upper bound for $S_n$. Because the supremum is the *least* upper bound, we must have $\sup S_n \leq L + \epsilon$.

Furthermore, it is a basic property of sets that the infimum is always less than or equal to the supremum ($\inf S_n \leq \sup S_n$). Therefore, for all $n \geq N$, we have successfully trapped both sequences:
$$L - \epsilon \leq \inf S_n \leq \sup S_n \leq L + \epsilon$$

Because this holds for any arbitrarily small $\epsilon > 0$, as $n \to \infty$, both $\inf S_n$ and $\sup S_n$ are squeezed strictly toward $L$. Thus:
$$\liminf_{n \to \infty} x_n = \limsup_{n \to \infty} x_n = L$$

**($\impliedby$) Direction:**
Suppose $\liminf x_n = \limsup x_n = L$. Let $a_n = \inf S_n$ and $b_n = \sup S_n$.
By definition, for any index $n$, we have:
$$a_n \leq x_n \leq b_n$$
We are given that $\lim_{n \to \infty} a_n = L$ and $\lim_{n \to \infty} b_n = L$.
By the **Squeeze Theorem**, since $x_n$ is trapped between two sequences that both converge to $L$, the sequence $\{x_n\}$ must also converge to $L$. $\square$

-----

::: {.callout-note}

## Intuition

Think of $\limsup$ and $\liminf$ as the "ceiling" and "floor" of the sequence's eventual behavior. A sequence might bounce around forever (like $(-1)^n$), having different floor and ceiling values. But if the ceiling and floor eventually meet at the same point, the sequence has nowhere else to go—it *must* converge there.
:::
