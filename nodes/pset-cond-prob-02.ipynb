# Problem Set: Conditional Probability & Combinatorics

## Problem 1: Resource Distribution (Inclusion-Exclusion)

**Problem Statement:**
A load balancer routes 7 incoming requests randomly across 4 available server clusters. Assuming every request is equally likely to be routed to any cluster, what is the probability that *every* server cluster receives at least one request?

**Mathematical Solution:**
We find the probability of the complement: the event that at least one server cluster receives *zero* requests. Let $A_i$ be the event that cluster $i$ is empty. Using the [Principle of Inclusion-Exclusion (PIE)](inclusion-exclusion.ipynb) and symmetry:
$$P\left(\bigcup_{i=1}^{4} A_i\right) = \binom{4}{1}P(A_1) - \binom{4}{2}P(A_1 \cap A_2) + \binom{4}{3}P(A_1 \cap A_2 \cap A_3) - \binom{4}{4}P(A_1 \cap \dots \cap A_4)$$

* $P(A_1)$: All 7 requests go to the other 3 clusters: $(3/4)^7$
* $P(A_1 \cap A_2)$: All 7 requests go to the remaining 2: $(2/4)^7$
* $P(A_1 \cap A_2 \cap A_3)$: All requests go to the last 1: $(1/4)^7$
* $P(A_1 \cap A_2 \cap A_3 \cap A_4)$: Impossible: $0$

$$P(\text{At least one empty}) = 4\left(\frac{3}{4}\right)^7 - 6\left(\frac{2}{4}\right)^7 + 4\left(\frac{1}{4}\right)^7 \approx 0.487$$
$$P(\text{All active}) = 1 - 0.487 = \mathbf{0.513}$$

## Problem 2: Course Scheduling (Inclusion-Exclusion)

**Problem Statement:**
A student selects 7 electives randomly from a pool of 30 non-overlapping available options. The options are evenly distributed across 5 days of the week (6 per day). What is the probability the student has at least one elective every single day?

**Mathematical Solution:**
Let $B_i$ be the event that the student has *no* classes on day $i$. We want $1 - P(B_1 \cup \dots \cup B_5)$. Using PIE:
$$P(\cup B_i) = \binom{5}{1}P(B_1) - \binom{5}{2}P(B_1 \cap B_2) + \binom{5}{3}P(B_1 \cap B_2 \cap B_3)$$
*(Intersections of 4 or 5 days are impossible because choosing 7 classes requires at least 2 days).*

* $P(B_1) = \binom{24}{7}/\binom{30}{7}$
* $P(B_1 \cap B_2) = \binom{18}{7}/\binom{30}{7}$
* $P(B_1 \cap B_2 \cap B_3) = \binom{12}{7}/\binom{30}{7}$

$$P(\cup B_i) = 5\frac{\binom{24}{7}}{\binom{30}{7}} - 10\frac{\binom{18}{7}}{\binom{30}{7}} + 10\frac{\binom{12}{7}}{\binom{30}{7}} = \frac{263}{377}$$
$$P(\text{Class every day}) = 1 - \frac{263}{377} = \mathbf{\frac{114}{377} \approx 0.302}$$

## Problem 3: Self-Independence

**Problem Statement:**
Under what conditions, if any, can a specific event be statistically independent of itself?

**Mathematical Solution:**
By the definition of independence, an event $A$ is [independent](independence-of-events.ipynb) of itself if and only if:
$$P(A \cap A) = P(A)P(A)$$
Since $A \cap A$ is simply $A$, this simplifies to:
$$P(A) = P(A)^2$$
The only real numbers that satisfy $x = x^2$ are $0$ and $1$. Therefore, an event is independent of itself **only if its probability is exactly 0 or 1** (deterministic events).

## Problem 4: Independence of Complements

**Problem Statement:**
Prove that if $A$ and $B$ are independent events, their complements $A^c$ and $B^c$ are also strictly independent.

**Mathematical Solution:**
We need to prove that $P(A^c \cap B^c) = P(A^c)P(B^c)$.
Using De Morgan's Laws, the intersection of complements is the complement of the union:
$$P(A^c \cap B^c) = P((A \cup B)^c) = 1 - P(A \cup B)$$
Expanding the union:
$$= 1 - [P(A) + P(B) - P(A \cap B)]$$
Because $A$ and $B$ are independent, substitute $P(A \cap B) = P(A)P(B)$:
$$= 1 - P(A) - P(B) + P(A)P(B)$$
Factoring this polynomial:
$$= (1 - P(A))(1 - P(B)) = \mathbf{P(A^c)P(B^c)}$$

## Problem 5: Pairwise vs. Mutual Independence

**Problem Statement:**
Construct a scenario with three events that are pairwise independent but fail to be mutually independent.

**Mathematical Solution:**
Consider tossing two independent, fair coins. Define three events:

* $A$: The first toss is Heads. ($P(A) = 1/2$)
* $B$: The second toss is Heads. ($P(B) = 1/2$)
* $C$: Both tosses yield the exact same result (HH or TT). ($P(C) = 1/2$)

Checking pairwise independence:

* $P(A \cap B) = 1/4 = P(A)P(B)$
* $P(A \cap C) = P(\text{HH}) = 1/4 = P(A)P(C)$
* $P(B \cap C) = P(\text{HH}) = 1/4 = P(B)P(C)$
They are completely pairwise independent. 

Checking mutual independence:
$$P(A \cap B \cap C) = P(\text{HH}) = 1/4$$
$$P(A)P(B)P(C) = (1/2)(1/2)(1/2) = 1/8$$
Since $1/4 \neq 1/8$, they are **not mutually independent**.

## Problem 6: The Multiplication Anomaly

**Problem Statement:**
Construct a scenario where three events $A, B,$ and $C$ are clearly dependent, yet the multiplication rule $P(A \cap B \cap C) = P(A)P(B)P(C)$ accidentally holds.

**Mathematical Solution:**
To satisfy the equation trivially, we can force both sides to equal 0. 
Let $P(A) = 0$. By definition, $P(A)P(B)P(C) = 0$. Furthermore, because $A \cap B \cap C \subseteq A$, $P(A \cap B \cap C) \leq P(A) = 0$. Both sides are now exactly 0.
To ensure the system is *dependent*, we simply make $B$ and $C$ dependent. Let $B$ and $C$ be the exact same event with a probability of $1/2$. Because $P(B \cap C) = 1/2 \neq 1/4$, the events are dependent, yet the three-way multiplication rule holds.

## Problem 7: The Mystery Box (Conditional Updates)

**Problem Statement:**
A box contains one item that is equally likely to be Type X or Type Y. A second item, strictly of Type X, is added to the box. The items are shuffled, and one is drawn at random. If the drawn item is Type X, what is the probability that the item remaining in the box is also Type X?

**Mathematical Solution:**
Let $I_X$ be the event that the initial item was Type X. Let $D_X$ be the event that the drawn item was Type X. We want the probability that the remaining item is X, which is exactly the probability that the initial item was X (since the added item was X). We need $P(I_X | D_X)$.

By [Bayes' Rule](conditional-probability.ipynb):
$$P(I_X | D_X) = \frac{P(D_X | I_X)P(I_X)}{P(D_X | I_X)P(I_X) + P(D_X | I_Y)P(I_Y)}$$

* If initial was X, there are 2 X's. Prob drawing X is 1.
* If initial was Y, there is 1 X and 1 Y. Prob drawing X is 1/2.
$$P(I_X | D_X) = \frac{(1)(1/2)}{(1)(1/2) + (1/2)(1/2)} = \frac{1/2}{3/4} = \mathbf{\frac{2}{3}}$$

## Problem 8: The Phishing Filter (Bayes' Rule)

**Problem Statement:**
A cybersecurity system analyzes incoming network packets. Historically, 80% of flagged traffic is malicious (M), while 20% is benign. The system detects a malicious byte signature (S). This signature is found in 10% of all true malicious packets, but it is also accidentally found in 1% of benign packets. A new packet triggers the signature alert. What is the probability it is malicious?

**Mathematical Solution:**
By Bayes' Rule:

$$P(M|S) = \frac{P(S|M)P(M)}{P(S|M)P(M) + P(S|M^c)P(M^c)}$$
$$P(M|S) = \frac{(0.10)(0.80)}{(0.10)(0.80) + (0.01)(0.20)} = \frac{0.08}{0.08 + 0.002} = \frac{0.08}{0.082}$$
$$P(M|S) = \frac{80}{82} \approx \mathbf{0.9756}$$

## Problem 9: Conflicting Evidence

**Problem Statement:**
In a hypothesis scenario, we evaluate a hypothesis $H$ against two pieces of evidence $E_1$ and $E_2$.

* a) Is it possible for $E_1$ and $E_2$ to individually increase the probability of $H$, but combined, they decrease it?
* b) Prove mathematically that updating the probability sequentially (first with $E_1$, then $E_2$) yields the exact same result as updating simultaneously.

**Mathematical Solution:**
**Part a:** Yes. Consider the "Alibi" scenario. A crime happens between 1pm-3pm. $E_1$: suspect seen near the scene 1pm-2pm. $E_2$: suspect seen near the scene 2pm-3pm. Individually, these place the suspect nearby, increasing probability of guilt. Together, they prove the suspect was in a coffee shop for the entire duration of the crime, providing a complete alibi and reducing guilt to near 0.

**Part b:** Sequential update using a new probability measure $P_{new}(X) = P(X|E_1)$:
$$P_{new}(H|E_2) = \frac{P_{new}(H \cap E_2)}{P_{new}(E_2)} = \frac{P(H \cap E_2 | E_1)}{P(E_2 | E_1)}$$
Expanding the conditional probabilities:
$$= \frac{P(H \cap E_1 \cap E_2)/P(E_1)}{P(E_1 \cap E_2)/P(E_1)}$$
$$= \frac{P(H \cap E_1 \cap E_2)}{P(E_1 \cap E_2)} = \mathbf{P(H | E_1, E_2)}$$

## Problem 10: Faulty Components

**Problem Statement:**
A fault was caused by either Component A or Component B (equal initial probability). The fault leaves a specific trace found in only 10% of all components. Component A is tested and matches the trace. Component B's trace status is unknown. 
a) What is the new probability that Component A caused the fault?
b) What is the probability that Component B also has this trace?

**Mathematical Solution:**
**Part a:** Let $G_A, G_B$ be the guilt of A and B. Let $E$ be the evidence (A matches, guilty party matches). 
$$P(G_A|E) = \frac{P(E|G_A)P(G_A)}{P(E|G_A)P(G_A) + P(E|G_B)P(G_B)}$$

* If A is guilty, $E$ occurs with probability $1$ (since A definitely matches).
* If B is guilty, A matches by pure population chance ($0.1$).
$$P(G_A|E) = \frac{1(1/2)}{1(1/2) + 0.1(1/2)} = \frac{1/2}{0.55} = \mathbf{\frac{10}{11}}$$

**Part b:** Let $M_B$ be the event B matches the trace. Condition on who is guilty:
$$P(M_B|E) = P(M_B|G_A, E)P(G_A|E) + P(M_B|G_B, E)P(G_B|E)$$

* If A is guilty, B is just a random component, so $P(M_B|G_A) = 1/10$.
* If B is guilty, B MUST match the trace, so $P(M_B|G_B) = 1$.
$$P(M_B|E) = \left(\frac{1}{10}\right)\left(\frac{10}{11}\right) + (1)\left(\frac{1}{11}\right) = \mathbf{\frac{2}{11}}$$

## Problem 11: The AI Opponent (Conditional Independence)

**Problem Statement:**
You face an AI opponent for 2 consecutive rounds. The AI is equally likely to be set to Easy, Normal, or Hard. Your chances of winning an individual round are 90%, 50%, and 30%, respectively.
a) What is your probability of winning the first round?
b) You won the first round! Given this, what is your probability of winning the second round?

**Mathematical Solution:**
**Part a:** By the Law of Total Probability across difficulties ($E, N, H$):
$$P(W_1) = (0.90)(1/3) + (0.50)(1/3) + (0.30)(1/3) = \mathbf{\frac{17}{30}}$$

**Part b:** We want $P(W_2 | W_1) = \frac{P(W_1 \cap W_2)}{P(W_1)}$. 
Assuming rounds are conditionally independent given the difficulty:
$$P(W_1 \cap W_2) = (0.90)^2(1/3) + (0.50)^2(1/3) + (0.30)^2(1/3) = \frac{1.15}{3} = \frac{23}{60}$$
$$P(W_2 | W_1) = \frac{23/60}{17/30} = \frac{23}{60} \times \frac{30}{17} = \mathbf{\frac{23}{34}}$$
