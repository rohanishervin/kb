# Sigma-Algebras

To build a rigorous foundation for measure theory, we cannot simply assign a "size" or "volume" to every possible subset of an arbitrary space (due to pathologies like the [Banach-Tarski paradox](https://en.wikipedia.org/wiki/Banach%E2%80%93Tarski_paradox)). Instead, we must restrict our attention to a collection of "well-behaved" sets. This collection is formalized as a **$\sigma$-algebra**.

## 1\. Definition and Core Properties

**Definition:** Let $X$ be a set, and let $\mathcal{P}(X)$ represent its [power set](), the set of all subsets of $X$. A **$\sigma$-algebra** (or $\sigma$-field) $\mathcal{M} \subseteq \mathcal{P}(X)$ on $X$ is a collection of subsets that satisfies the following three properties:

1.  **Contains the universal set:** $X \in \mathcal{M}$.
2.  **Closure under complementation:** If some set $E \in \mathcal{M}$, then its complement $E^c$ (or $X \setminus E$) is also in $\mathcal{M}$.
3.  **Closure under countable unions:** If $E_1, E_2, E_3, \dots \in \mathcal{M}$, then their union $\bigcup_{i=1}^\infty E_i \in \mathcal{M}$.

::: {.callout-note}
## Intuition: The Empty Set and "Sigma"
Because $X \in \mathcal{M}$ (Property 1) and $\mathcal{M}$ is closed under complementation (Property 2), the empty set $\emptyset = X^c$ is automatically in $\mathcal{M}$. Some authors use $\emptyset \in \mathcal{M}$ as the first axiom instead, which is entirely equivalent. 

Furthermore, the prefix "$\sigma$" (sigma) typically denotes operations that are *countably infinite*, whereas "$\delta$" (delta) refers to finite operations. A standard algebra of sets only requires closure under *finite* unions. A $\sigma$-algebra upgrades this requirement to *countable* unions, which is strictly necessary for limits and calculus-like operations in measure theory.
:::

Because $\sigma$-algebras are closed under both complements and countable unions, De Morgan's laws immediately guarantee that they are also closed under countable intersections.

::: {.callout-tip collapse="false"}
## Proof of Closure under Countable Intersections

**Proposition:** If $\mathcal{M}$ is a $\sigma$-algebra on $X$, and $E_1, E_2, \dots \in \mathcal{M}$, then $\bigcap_{i=1}^\infty E_i \in \mathcal{M}$.

**Proof:** 
Let $E_1, E_2, \dots \in \mathcal{M}$. 
By the definition of a $\sigma$-algebra, it is closed under complementation, so $E_1^c, E_2^c, \dots \in \mathcal{M}$. 
Because $\mathcal{M}$ is closed under countable unions, their union must also be in $\mathcal{M}$:
$$
\bigcup_{i=1}^\infty E_i^c \in \mathcal{M}
$$
Applying complementation one more time to this resulting set gives:
$$
\left( \bigcup_{i=1}^\infty E_i^c \right)^c \in \mathcal{M}
$$
By De Morgan's laws, the complement of a union is the intersection of the complements:
$$
\left( \bigcup_{i=1}^\infty E_i^c \right)^c = \bigcap_{i=1}^\infty (E_i^c)^c = \bigcap_{i=1}^\infty E_i
$$
Therefore, $\bigcap_{i=1}^\infty E_i \in \mathcal{M}$. $\blacksquare$
:::

## 2\. Examples of Sigma-Algebras

For any given set $X$, there are two trivial extremes of $\sigma$-algebras:

1.  **The Power Set:** $\mathcal{M} = \mathcal{P}(X)$ (the collection of all subsets of $X$). This is the largest possible $\sigma$-algebra on $X$.
2.  **The Trivial $\sigma$-algebra:** $\mathcal{M} = \{\emptyset, X\}$. This is the smallest possible $\sigma$-algebra on $X$.

Often, we want a $\sigma$-algebra that is somewhere between these two extremes. We can construct one by starting with a specific family of sets and "generating" a $\sigma$-algebra from it.

**Definition:** Let $S$ be an arbitrary family of subsets of $X$. The **$\sigma$-algebra generated by $S$**, denoted $\sigma(S)$, is defined as the intersection of all $\sigma$-algebras on $X$ that contain $S$. It is the smallest $\sigma$-algebra containing every set in $S$.

## 3\. The Borel Sigma-Algebra on the Reals

One of the most important $\sigma$-algebras in analysis is the [Borel sigma-algebra]() on the real line $\mathbb{R}$. We can define it as the $\sigma$-algebra generated by the collection of all open intervals. 

Let $S = \{(a, b) : a < b, a,b \in \mathbb{R}\}$. The resulting generated $\sigma$-algebra $\sigma(S)$ contains far more than just open intervals. 

::: {.callout-tip collapse="false"}
## Proof that Singletons are Borel Sets

**Proposition:** For every real number $x \in \mathbb{R}$, the singleton set $\{x\}$ belongs to $\sigma(S)$.

**Proof:** 
We can construct the singleton $\{x\}$ as a countable intersection of shrinking open intervals centered at $x$:
$$
\{x\} = \bigcap_{n=1}^\infty \left( x - \frac{1}{n}, x + \frac{1}{n} \right)
$$
Since each interval $\left( x - \frac{1}{n}, x + \frac{1}{n} \right)$ is an open interval, it belongs to $S$, and thus belongs to $\sigma(S)$. Because $\sigma$-algebras are closed under countable intersections, the intersection of this countable family of sets must also be in $\sigma(S)$. Therefore, $\{x\} \in \sigma(S)$. $\blacksquare$
:::

**Corollary:** Because singletons are in $\sigma(S)$ and $\sigma$-algebras are closed under countable unions, **every countable subset of $\mathbb{R}$** belongs to $\sigma(S)$. 
For example, a countable set $E = \{x_1, x_2, \dots\}$ can be written as $E = \bigcup_{n=1}^\infty \{x_n\}$, which is a countable union of sets in $\sigma(S)$.

::: {.callout-tip collapse="false"}
## Proof that Closed Intervals are Borel Sets

**Proposition:** Every closed interval $[a, b]$ belongs to $\sigma(S)$.

**Proof:**
There are two elegant ways to show this:

**Method 1 (Intersection of Open Intervals):** 
We can express $[a, b]$ as the intersection of a sequence of slightly larger open intervals:
$$
[a, b] = \bigcap_{n=1}^\infty \left( a - \frac{1}{n}, b + \frac{1}{n} \right)
$$
Since $\sigma(S)$ is closed under countable intersections and each open interval is in $\sigma(S)$, the closed interval $[a, b] \in \sigma(S)$.

**Method 2 (Union with Singletons):**
We can construct the closed interval by taking the open interval $(a, b)$ and uniting it with its endpoints:
$$
[a, b] = \{a\} \cup (a, b) \cup \{b\}
$$
We already proved that singletons $\{a\}$ and $\{b\}$ are in $\sigma(S)$. By definition, $(a, b) \in S \subset \sigma(S)$. Since $\sigma(S)$ is closed under finite unions (as a special case of countable unions), $[a, b] \in \sigma(S)$. $\blacksquare$
:::

Even incredibly complex sets, such as the [Cantor set](cantor-set.ipynb#definition-and-construction), are Borel sets. The Cantor set $C$ is constructed by iteratively removing the middle thirds of closed intervals. It can be written as $C = \bigcap_{n=1}^\infty C_n$, where each $C_n$ is a finite union of closed intervals. Since closed intervals are in $\sigma(S)$, their finite unions $C_n$ are in $\sigma(S)$, and the countable intersection of these unions guarantees $C \in \sigma(S)$.
