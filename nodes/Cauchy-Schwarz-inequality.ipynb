# Inner Product Spaces and the Cauchy-Schwarz Inequality

## 1. The $\epsilon$-Principle
Before diving into spaces, a quick foundational tool for proofs in $\mathbb{R}$. It is similar to [zero squeeze](zero-squeeze-in-R.ipynb):

For any $a, b \in \mathbb{R}$, if $\forall \epsilon > 0, |a - b| < \epsilon$, then $a = b$.
*(Proof Intuition: If two numbers are closer than any strictly positive distance, they must be the exact same number.)*

## 2. Inner Product Spaces
Let $V$ be a vector space. An **inner product** on $V$ (often called a dot product in $\mathbb{R}^n$) is a function $\langle x, y \rangle$ that takes two vectors and returns a real number, satisfying three properties for all $x, y, z \in V$ and $c \in \mathbb{R}$:

1.  **Bilinearity:** $\langle x, y + cz \rangle = \langle x, y \rangle + c\langle x, z \rangle$
2.  **Symmetry:** $\langle x, y \rangle = \langle y, x \rangle$
3.  **Positive Definite:** $\langle x, x \rangle \geq 0$, and $\langle x, x \rangle = 0 \iff x = 0$

The **length** (or induced norm) of a vector is defined as $|x| = \sqrt{\langle x, x \rangle}$.

## 3. The Cauchy-Schwarz Inequality

**Theorem:** For any $x, y \in V$ in an inner product space, 
$$|\langle x, y \rangle| \leq |x| |y|$$

**Proof (Using Quadratic Forms):**
Let $w = x + ty$ for any $t \in \mathbb{R}$. By the positive definite property, we know that the inner product of $w$ with itself must be non-negative:
$$Q(t) = \langle w, w \rangle = \langle x + ty, x + ty \rangle \geq 0$$
Expanding this using bilinearity and symmetry:
$$Q(t) = \langle x, x \rangle + 2t\langle x, y \rangle + t^2\langle y, y \rangle \geq 0$$
Notice that $Q(t)$ is a quadratic function of $t$. For a quadratic function $at^2 + bt + c \geq 0$ to always be non-negative, its discriminant must be less than or equal to zero ($\Delta = b^2 - 4ac \leq 0$). 
Here, $a = \langle y, y \rangle$, $b = 2\langle x, y \rangle$, and $c = \langle x, x \rangle$.
$$(2\langle x, y \rangle)^2 - 4(\langle y, y \rangle)(\langle x, x \rangle) \leq 0$$
$$4\langle x, y \rangle^2 \leq 4\langle x, x \rangle\langle y, y \rangle$$
$$\langle x, y \rangle^2 \leq |x|^2 |y|^2$$
Taking the square root yields the Cauchy-Schwarz inequality: $|\langle x, y \rangle| \leq |x||y|$. $\square$

::: {.callout-note}
## Algebraic Version in $\mathbb{R}^n$
If we write vectors as $x = (a_1, \dots, a_n)$ and $y = (b_1, \dots, b_n)$, the inequality takes its famous algebraic form:
$$\left( \sum_{k=1}^{n} a_k b_k \right)^2 \leq \left( \sum_{k=1}^{n} a_k^2 \right) \left( \sum_{k=1}^{n} b_k^2 \right)$$
Equality holds if and only if one vector is a scalar multiple of the other ($b_k = c a_k$).
:::

## 4. The Triangle Inequality (Euclidean Distance)
We can use Cauchy-Schwarz to prove the [Triangle Inequality](triangle-inequality.ipynb): $|x + y| \leq |x| + |y|$.
**Proof:**
$$|x + y|^2 = \langle x + y, x + y \rangle = |x|^2 + 2\langle x, y \rangle + |y|^2$$
Since $\langle x, y \rangle \leq |x||y|$ (by Cauchy-Schwarz):
$$|x + y|^2 \leq |x|^2 + 2|x||y| + |y|^2 = (|x| + |y|)^2$$
Taking the square root gives $|x + y| \leq |x| + |y|$. $\square$

## 5. Infinite-Dimensional Example: Continuous Functions
The Cauchy-Schwarz inequality holds for *any* inner product space, not just finite-dimensional ones like $\mathbb{R}^n$.

Consider $V = C([a, b], \mathbb{R})$, the space of all continuous functions from $[a,b] \to \mathbb{R}$. 

* Unlike $\mathbb{R}^n$, which has finite degrees of freedom (a finite basis $e_1, \dots, e_n$), this space has an independent value at every single point in the interval, making it **infinite-dimensional**.
* We can define an inner product here using an integral:
    $$\langle f, g \rangle = \int_a^b f(x)g(x)dx$$
Applying Cauchy-Schwarz to this space yields a powerful integral inequality:
$$\int_a^b f(x)g(x)dx \leq \sqrt{\int_a^b f(x)^2dx} \sqrt{\int_a^b g(x)^2dx}$$
