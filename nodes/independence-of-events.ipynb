# Independence of Events

## 1\. Definition

Two [events](event.ipynb) $A$ and $B$ are **independent** if the occurrence of one provides absolutely no information about the probability of the occurrence of the other.

Formally, $A$ and $B$ are independent if and only if their joint probability is the product of their marginal probabilities:
$$P(A \cap B) = P(A)P(B)$$

If $P(A) > 0$ and $P(B) > 0$, this definition is strictly equivalent to stating that their [conditional probabilities](conditional-probability.ipynb) equal their unconditional probabilities:
$$P(B|A) = P(B)$$
$$P(A|B) = P(A)$$

::: {.callout-warning}

## Independence $\neq$ Disjointness

It is a common trap to confuse "independent" events with "disjoint" (mutually exclusive) events. They are fundamentally different\!

  * **Disjoint:** $A$ and $B$ cannot happen together ($A \cap B = \emptyset$). If $A$ happens, you know with 100% certainty that $B$ *did not* happen. Therefore, disjoint events are highly **dependent**.
  * **Independent:** The outcome of $A$ gives no information about $B$.

**Counter-example:** Draw a single card from a standard deck.

Let $A$ be drawing a Spade ($P(A) = 13/52 = 1/4$).

Let $B$ be drawing an Ace ($P(B) = 4/52 = 1/13$).

They are *not* disjoint because they intersect at the Ace of Spades ($P(A \cap B) = 1/52$).

However, $P(A)P(B) = (1/4)(1/13) = 1/52$. Because $P(A \cap B) = P(A)P(B)$, drawing a Spade and drawing an Ace are perfectly **independent**.
:::

## 2\. Independence of Complements

**Theorem:** If events $A$ and $B$ are independent, then their complements are also independent. Specifically: $A$ and $B^c$ are independent, $A^c$ and $B$ are independent, and $A^c$ and $B^c$ are independent.

**Proof (for $A$ and $B^c$):**

By the [law of total probability](law-of-total-probability.ipynb), $P(A) = P(A \cap B) + P(A \cap B^c)$.

Rearranging this:

$$P(A \cap B^c) = P(A) - P(A \cap B)$$

Because $A$ and $B$ are independent, we substitute $P(A \cap B) = P(A)P(B)$:

$$P(A \cap B^c) = P(A) - P(A)P(B) = P(A)(1 - P(B))$$

Since $1 - P(B) = P(B^c)$, we get:

$$P(A \cap B^c) = P(A)P(B^c)$$

Thus, $A$ and $B^c$ are independent. $\square$

## 3\. Independence of Multiple Events

Extending independence beyond two events requires more than just checking pairs.

**Definition (Mutual Independence):** A collection of events $A_1, A_2, \dots, A_n$ is mutually independent if the probability of the intersection of *any* sub-collection of these events equals the product of their individual probabilities.

For exactly three events ($A, B, C$), all four of these conditions must hold simultaneously:

1.  $P(A \cap B) = P(A)P(B)$
2.  $P(A \cap C) = P(A)P(C)$
3.  $P(B \cap C) = P(B)P(C)$
4.  $P(A \cap B \cap C) = P(A)P(B)P(C)$

::: {.callout-warning}

## Pairwise Independence is Not Enough

Just because events are independent in pairs does not mean they are mutually independent\!
**Example:** Flip two fair coins.

Let $A$ = First coin is Heads.

Let $B$ = Second coin is Heads.

Let $C$ = The coins match (both Heads or both Tails).

$P(A) = P(B) = P(C) = 1/2$.

Checking pairs: $P(A \cap B) = 1/4 = P(A)P(B)$. The same holds for $(A,C)$ and $(B,C)$. They are **pairwise independent**.

However, $P(A \cap B \cap C) = P(\text{Both are Heads}) = 1/4$.

But $P(A)P(B)P(C) = (1/2)(1/2)(1/2) = 1/8$.

Since $1/4 \neq 1/8$, the three events are **not** mutually independent. Knowing $A$ and $B$ guarantees $C$.
:::
