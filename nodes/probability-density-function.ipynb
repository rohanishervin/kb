# Probability Density Functions (PDF)

Because a continuous [random variable](random-variables.ipynb) can take on any real value in an interval, the probability of it landing on one exact, infinitely precise point is $0$. Therefore, we measure the *density* of probability over an interval using the Probability Density Function.

## 1. Definition and Fundamental Theorem of Calculus

**Definition:** For a continuous random variable $X$ with a differentiable CDF $F(x)$, the Probability Density Function (PDF) is the derivative of the CDF:

$$f(x) = \frac{d}{dx} F(x) = F'(x)$$

By the Fundamental Theorem of Calculus, we can reconstruct the [CDF](cumulative-distribution-functions.ipynb) and compute probabilities by integrating the PDF:

$$F(x) = P(X \le x) = \int_{-\infty}^{x} f(t) dt$$

To find the probability that $X$ falls within a specific interval $(a, b]$, we integrate the PDF over that range:

$$P(a < X \le b) = F(b) - F(a) = \int_{a}^{b} f(x) dx$$

::: {.callout-note title="Intuition: The Units of $f(x)$"}
The value $f(x)$ is **not** a probability. It is a density, and it can mathematically exceed $1$. 

To interpret $f(x)$, imagine an infinitesimally small interval of length $dx$ centered at $x$. The probability that $X$ falls within this tiny interval is approximately the area of a rectangle with height $f(x)$ and width $dx$:
$$P\left(X \in \left[x - \frac{dx}{2}, x + \frac{dx}{2}\right]\right) \approx f(x)dx$$
If $X$ represents distance in centimeters, $f(x)$ has units of $cm^{-1}$, making the product $f(x)dx$ a dimensionless probability.
:::

## 2. Properties of Valid PDFs

**Theorem:** A function $f(x)$ is a valid PDF if and only if it satisfies two conditions:

1. **Non-negativity:** $f(x) \ge 0$ for all $x \in \mathbb{R}$.
2. **Normalizes to 1:** $\int_{-\infty}^{\infty} f(x) dx = 1$.

**Proof:**

1. Since $F(x)$ is a valid CDF, it is monotonically non-decreasing. The derivative of a non-decreasing function cannot be negative, hence $f(x) = F'(x) \ge 0$.
2. By the properties of a valid CDF, the limit as $x \to \infty$ covers the entire sample space. Therefore, $\int_{-\infty}^{\infty} f(x) dx = F(\infty) - F(-\infty) = 1 - 0 = 1$. $\blacksquare$

## 3. Standard Continuous Examples

To become comfortable with PDFs, here are two standard continuous distributions defined strictly by their CDFs and PDFs.

**Example 1: The Logistic Distribution**

* **CDF:** $F(x) = \frac{e^x}{1 + e^x}$ for $x \in \mathbb{R}$
* **PDF (via differentiation):** $f(x) = \frac{e^x}{(1 + e^x)^2}$

**Example 2: The Rayleigh Distribution**

* **CDF:** $F(x) = 1 - e^{-x^2/2}$ for $x > 0$
* **PDF (via differentiation):** $f(x) = x e^{-x^2/2}$ for $x > 0$
