# Nested Interval Property

For each $n \in \mathbb{N}$, assume we are given a closed [interval](interval.ipynb) $I_n = [a_n, b_n] = \{x \in \mathbb{R} : a_n \leq x \leq b_n\}$. Assume also that each $I_n$ contains $I_{n+1}$. Then, the resulting nested sequence of closed intervals  $I_1 \supseteq I_2 \supseteq I_3 \supseteq ...$  has a nonempty intersection; that is, $\cap_{n=1}^\infty  I_n \neq \emptyset$.
