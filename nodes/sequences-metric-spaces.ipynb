# Sequences in Metric Spaces


## 1\. Sequences and Convergence

A sequence in a metric space $M$ is a function $f: \mathbb{N} \to M$, denoted as $(p_n)$.
A sequence converges to a limit $p \in M$ if:
$$\forall \epsilon > 0, \exists N \in \mathbb{N} \text{ s.t. } n \geq N \implies d(p_n, p) < \epsilon$$
If a limit exists, it is unique.

## 2\. Cauchy Sequences in General Spaces

We previously defined [Cauchy sequences in $\mathbb{R}$](Cauchy-sequences.ipynb). The definition easily generalizes to any metric space $(M, d)$:
A sequence $(p_n)$ is Cauchy if:
$$\forall \epsilon > 0, \exists N \in \mathbb{N} \text{ s.t. } \forall j, k \geq N \implies d(p_j, p_k) < \epsilon$$

**Theorem:** Every convergent sequence in a metric space is a Cauchy sequence.

**Proof:** Let $p_n \to p$. For any $\epsilon > 0$, choose $N$ such that for all $n \geq N$, $d(p_n, p) < \frac{\epsilon}{2}$.
For any $j, k \geq N$, apply the triangle inequality:
$$d(p_j, p_k) \leq d(p_j, p) + d(p, p_k) < \frac{\epsilon}{2} + \frac{\epsilon}{2} = \epsilon$$
Thus, $(p_n)$ is Cauchy. $\square$

## 3\. Complete Metric Spaces
::: {.callout-warning}
The converse of the above theorem is *not* always true\! A Cauchy sequence does not have to converge if the space has a "hole" in it.

  * **Example:** In $M = \mathbb{Q}$, the sequence $1, 1.4, 1.414 \dots$ is Cauchy, but it does not converge because $\sqrt{2} \notin \mathbb{Q}$.
:::
A metric space $(M, d)$ is called **Complete** if *every* Cauchy sequence in $M$ converges to a limit that is also in $M$. $\mathbb{R}$ is complete; $\mathbb{Q}$ is not.

## 4. Subsequences

Just as a set can have a subset, a sequence can have a **subsequence**. 

**Definition:** Let $(p_n)_{n \in \mathbb{N}}$ be a sequence in a metric space. If $n_1 < n_2 < n_3 < \dots$ is a strictly increasing sequence of positive integers, then the sequence $(q_k)_{k \in \mathbb{N}}$ defined by $q_k = p_{n_k}$ is called a subsequence of $(p_n)$. 

*Note: The terms in a subsequence must occur in the exact same order as they do in the "mother" sequence.*

**Examples:**

* The sequence of even numbers $2, 4, 6, 8, \dots$ is a subsequence of the natural numbers $1, 2, 3, 4, \dots$. Here, the index sequence is $n_k = 2k$.
* The sequence of primes $3, 5, 7, 11, 13, \dots$ is a subsequence of the odd numbers $1, 3, 5, 7, 9, \dots$, which in turn is a subsequence of the natural numbers.

**Theorem:** Every subsequence of a convergent sequence converges to the same limit as the "mother" sequence.

**Proof:** Assume $\lim_{n \to \infty} p_n = p$. Let $(p_{n_k})$ be a subsequence.
For any $\epsilon > 0$, there exists an $N$ such that for all $n \geq N$, $d(p_n, p) < \epsilon$.
Because $n_k$ is a strictly increasing sequence of integers, $n_k \geq k$. 
If we choose $k \geq N$, it guarantees $n_k \geq N$, which means $d(p_{n_k}, p) < \epsilon$. Thus, $\lim_{k \to \infty} p_{n_k} = p$. $\square$


## 5. Topological Definition of Convergence

**Proposition:** In a metric space, a sequence $(p_n)$ converges to $p$ if and only if for every open set $V$ containing $p$, there exists an $N \in \mathbb{N}$ such that for all $n \geq N$, $p_n \in V$.

**Proof:**
We must prove both directions of this equivalence to show that the metric and [topological definitions](metric-space-topology.ipynb) are identical.

**($\implies$ Metric implies Topological):**
Assume $p_n \to p$ in the standard metric sense. Let $V$ be an arbitrary open set containing $p$. 
Because $V$ is open, there exists some radius $r > 0$ such that the entire neighborhood $M_r p$ is strictly contained in $V$ (i.e., $d(q, p) < r \implies q \in V$).
By the metric definition of sequence convergence, since $r > 0$, there exists an integer $N$ such that for all $n \geq N$, $d(p_n, p) < r$.
This means $p_n \in M_r p$. Since $M_r p \subset V$, it follows that $p_n \in V$ for all $n \geq N$.

**($\impliedby$ Topological implies Metric):**
Assume that for *every* open set $V$ containing $p$, the sequence is eventually trapped in $V$. We must prove $p_n \to p$ metrically.
Let $\epsilon > 0$ be given. Let us define the specific set $V = \{q \in M \mid d(q, p) < \epsilon\}$. 
This set is the open $\epsilon$-neighborhood of $p$ ($M_\epsilon p$). Because [open balls are open sets](topology-examples.ipynb), and it clearly contains $p$, it qualifies as our valid open set $V$.
By our initial topological assumption, there exists an integer $N$ such that for all $n \geq N$, $p_n \in V$.
By the definition of how we built $V$, stating $p_n \in V$ means exactly that $d(p_n, p) < \epsilon$. This perfectly satisfies the metric definition of convergence. $\square$
