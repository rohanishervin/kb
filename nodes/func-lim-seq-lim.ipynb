# The Sequential Criterion for Function Limits

This theorem establishes a perfect, bidirectional bridge between the continuous concept of a [function limit](limit.ipynb) and the discrete concept of a [sequence limit](sequence-limit.ipynb). It allows us to translate any theorem proven for sequences directly into a theorem for functions.

## 1\. The Main Theorem (Equivalence)

Let $I$ be an open interval that contains the point $c$, and suppose that $f$ is a function defined on $I$ (except possibly at the point $c$ itself).

The function $f$ has a limit $L$ as $x$ approaches $c$:
$$\lim_{x \to c} f(x) = L$$

**If and only if:**
For every sequence $\{x_n\}$ strictly in $I \setminus \{c\}$ that converges to $c$, the corresponding sequence of outputs $\{f(x_n)\}$ converges to $L$.

$$\Big( \forall \{x_n\} \subset I \setminus \{c\}, \lim_{n \to \infty} x_n = c \Big) \implies \lim_{n \to \infty} f(x_n) = L$$

::: {.callout-note}

## Intuition: The "No Surprises" Rule

Imagine $x$ walking toward $c$ along a continuous path. This theorem says that if the limit truly is $L$, it doesn't matter if you "skip" along the path like a stone (a sequence $x_n$). No matter what discrete stepping stones you choose to approach $c$, your altitudes $f(x_n)$ must eventually lock in on the exact same target altitude $L$.
:::

## 2\. The Divergence Criterion (Non-Existence)

While Part 1 is theoretically profound, Part 2 is the ultimate practical tool. Proving that a function limit *does not exist* using the standard $\epsilon-\delta$ definition is notoriously difficult. The sequential criterion provides a much easier way: finding two disagreeing paths.

**Theorem (Sequential Criterion for Divergence):**
Suppose there exist two sequences $\{x_n\}$ and $\{y_n\}$ in $I \setminus \{c\}$, both converging to $c$, such that:

  * $\{f(x_n)\}$ converges to $L_1$
  * $\{f(y_n)\}$ converges to $L_2$

If $L_1 \neq L_2$, then the overall function limit $\lim_{x \to c} f(x)$ **does not exist**.

*(Furthermore, if you can find just one sequence $\{x_n\} \to c$ where $\{f(x_n)\}$ diverges or oscillates wildly, the function limit also does not exist).*

-----

## 3\. Classic Example: The Topologist's Sine Curve

**Question:** Prove that $\lim_{x \to 0} \sin\left(\frac{1}{x}\right)$ does not exist.

**Solution:**
We will use the Divergence Criterion. We need to invent two sequences, $\{x_n\}$ and $\{y_n\}$, that both converge to $c = 0$, but produce different limits when plugged into $f(x)$.

**Sequence 1 (Riding the Peaks):**
Let $x_n = \frac{1}{\frac{\pi}{2} + 2\pi n}$.
As $n \to \infty$, the denominator goes to infinity, so $x_n \to 0$.
Now look at the function values:
$$f(x_n) = \sin\left(\frac{\pi}{2} + 2\pi n\right) = 1$$
So, the sequence $\{f(x_n)\}$ is constantly $1, 1, 1 \dots$, which converges to **$L_1 = 1$**.

**Sequence 2 (Riding the Roots):**
Let $y_n = \frac{1}{\pi n}$.
As $n \to \infty$, $y_n \to 0$.
Now look at the function values:
$$f(y_n) = \sin(\pi n) = 0$$
So, the sequence $\{f(y_n)\}$ is constantly $0, 0, 0 \dots$, which converges to **$L_2 = 0$**.

**Conclusion:** Since $x_n \to 0$ and $y_n \to 0$, but $L_1 \neq L_2$ ($1 \neq 0$), the limit does not exist. The function oscillates infinitely fast between $1$ and $-1$ near zero.
