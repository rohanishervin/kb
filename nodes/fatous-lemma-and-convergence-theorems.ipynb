# Fatou's Lemma and Convergence Theorems

In measure theory, establishing when we can interchange limits and integrals is of fundamental importance. The cornerstone of these convergence theorems is **Fatou's Lemma**, which provides an inequality for the limit infimum of integrals of non-negative [measurable functions](measurable-functions.ipynb). From it, we derive two other powerful results: the Monotone Convergence Theorem and the Lebesgue Dominated Convergence Theorem.

## 1\. Fatou's Lemma

**Fatou's Lemma** establishes that the integral of the pointwise limit of a sequence of non-negative functions is bounded above by the limit infimum of their integrals.

**Theorem (Fatou's Lemma):**
Let $(X, \mathcal{M}, \mu)$ be a [measure space](measure.ipynb) and $\{f_n\}$ be a sequence of non-negative measurable functions on $X$ such that $f_n \to f$ [almost everywhere](limits-of-measurable-functions.ipynb#convergence-and-equality-almost-everywhere) on $X$. Then:
$$ \int_X f \, d\mu \le \liminf_{n \to \infty} \int_X f_n \, d\mu $$

Alternatively, expressed using the $\epsilon$-$N$ definition of the limit infimum, for any $\epsilon > 0$, there exists an $N$ such that for all $n > N$:
$$ \int_X f_n \, d\mu > \int_X f \, d\mu - \epsilon $$

::: {.callout-note}
## Intuition: Missing Mass
Why is it an inequality and not an equality? Mass can "escape to infinity" (either horizontally or vertically) in the limit, meaning the functions $f_n$ might contain area that disappears when taking the pointwise limit $f$. Therefore, the sequence of integrals can remain large even if the limit function has a small (or zero) integral. 
:::

::: {.callout-tip collapse="true"}
## Proof of Fatou's Lemma

**Step 1: Reduction to everywhere convergence**
Since $f_n \to f$ almost everywhere, there is a set $X_0 \subset X$ with $\mu(X_0^c) = 0$ where pointwise convergence holds exactly. We can restrict our attention to $X_0$ without loss of generality. 

**Step 2: Approximation by Simple Functions**
By the definition of the [Lebesgue integral](lebesgue-integral.ipynb), the integral of $f$ is the supremum of the integrals of all non-negative simple functions bounded by $f$:
$$ \int_X f \, d\mu = \sup_{0 \le \varphi \le f} \int_X \varphi \, d\mu $$
Fix an arbitrary simple function $\varphi$ such that $0 \le \varphi \le f$, and fix a small constant $\delta \in (0, 1)$.

**Step 3: Constructing an increasing sequence of sets**
Define the sequence of sets $A_n$ as follows:
$$ A_n = \{x \in X : f_k(x) > (1-\delta)\varphi(x) \text{ for all } k \ge n\} $$
Because $f_n(x) \to f(x)$ and $f(x) \ge \varphi(x)$, for every $x \in X$, there eventually exists an integer $N$ such that $f_k(x) > (1-\delta)\varphi(x)$ for all $k \ge N$. This implies that the sequence $\{A_n\}$ is an increasing sequence of sets ($A_1 \subset A_2 \subset \dots$) and their union is the entire space $X$:
$$ \bigcup_{n=1}^\infty A_n = X $$

**Step 4: Lower bound on the integral**
We can bound the integral of $f_n$ over the whole space by its integral over the subset $A_n$:
$$ \int_X f_n \, d\mu \ge \int_{A_n} f_n \, d\mu \ge \int_{A_n} (1-\delta)\varphi \, d\mu = (1-\delta)\int_{A_n} \varphi \, d\mu $$

Let the simple function be represented in its standard form as $\varphi = \sum_{k=1}^m c_k \mathbf{1}_{E_k}$. Then:
$$ \int_{A_n} \varphi \, d\mu = \sum_{k=1}^m c_k \mu(A_n \cap E_k) $$

Since $A_n \nearrow X$, the sequence of sets $(A_n \cap E_k)$ increases to $E_k$. By the [continuity of measure from below](continuity-of-measure.ipynb), $\mu(A_n \cap E_k) \to \mu(E_k)$ as $n \to \infty$. Therefore:
$$ \lim_{n \to \infty} \int_{A_n} \varphi \, d\mu = \sum_{k=1}^m c_k \mu(E_k) = \int_X \varphi \, d\mu $$

Taking the limit infimum of our lower bound yields:
$$ \liminf_{n \to \infty} \int_X f_n \, d\mu \ge (1-\delta)\int_X \varphi \, d\mu $$

**Step 5: Taking the suprema**
Since $\delta \in (0, 1)$ was arbitrary, we can let $\delta \to 0$ to obtain:
$$ \liminf_{n \to \infty} \int_X f_n \, d\mu \ge \int_X \varphi \, d\mu $$
Finally, taking the supremum over all simple functions $\varphi \le f$ gives:
$$ \liminf_{n \to \infty} \int_X f_n \, d\mu \ge \int_X f \, d\mu $$
$\blacksquare$
:::

## 2\. Monotone Convergence Theorem

The Monotone Convergence Theorem (MCT) provides a simple condition under which the limit of integrals equals the integral of the limit: the functions must be non-negative and increasing. It is a direct corollary of Fatou's Lemma.

**Theorem (Monotone Convergence Theorem):**
Let $\{f_n\}$ be a sequence of non-negative measurable functions such that $f_1 \le f_2 \le f_3 \le \dots$ almost everywhere, and let $f_n \to f$ almost everywhere. Then:
$$ \lim_{n \to \infty} \int_X f_n \, d\mu = \int_X f \, d\mu $$

::: {.callout-tip collapse="true"}
## Proof of the Monotone Convergence Theorem

Since the sequence is monotonically increasing, we have $f_n \le f$ almost everywhere for all $n$. By the monotonicity of the integral, this implies:
$$ \int_X f_n \, d\mu \le \int_X f \, d\mu $$

Taking the limit supremum of both sides gives:
$$ \limsup_{n \to \infty} \int_X f_n \, d\mu \le \int_X f \, d\mu $$

On the other hand, by Fatou's Lemma, we have:
$$ \int_X f \, d\mu \le \liminf_{n \to \infty} \int_X f_n \, d\mu $$

Combining these two inequalities, we get:
$$ \limsup_{n \to \infty} \int_X f_n \, d\mu \le \int_X f \, d\mu \le \liminf_{n \to \infty} \int_X f_n \, d\mu $$

Since the limit supremum is bounded above by the limit infimum, the limit must exist and equal $\int_X f \, d\mu$.
$\blacksquare$
:::

::: {.callout-warning}
## Decreasing Sequences Require Integrability
The Monotone Convergence Theorem applies to *increasing* sequences of functions. If we instead have a monotonically *decreasing* sequence ($f_1 \ge f_2 \ge f_3 \dots \ge 0$) converging to $f$, we **must** have $\int_X f_1 \, d\mu < \infty$ to guarantee that $\lim \int f_n = \int f$. 

**Counterexample:**
Consider the measure space $(\mathbb{R}, \mathcal{B}, m)$ where $m$ is the Lebesgue measure. Let $f_n = \mathbf{1}_{[n, \infty)}$. 
The sequence decreases pointwise to the zero function $f = 0$. However, for every $n$, $\int_{\mathbb{R}} f_n \, dm = \infty$. Thus:
$$ \lim_{n \to \infty} \int f_n = \infty \neq 0 = \int f $$
:::

## 3\. Lebesgue Dominated Convergence Theorem

The Lebesgue Dominated Convergence Theorem (LDCT) is arguably the most practically useful of the convergence theorems. It allows us to pass a limit inside an integral as long as the sequence of functions is uniformly bounded by a single integrable function.

**Theorem (Lebesgue Dominated Convergence Theorem):**
Let $(X, \mathcal{M}, \mu)$ be a measure space and $\{f_n\}$ a sequence of measurable functions on $X$ for which $f_n \to f$ pointwise almost everywhere on $X$ and the function $f$ is measurable. Assume there is a non-negative function $g$ that is integrable over $X$ and dominates the sequence $\{f_n\}$ on $X$ in the sense that:
$$ |f_n| \le g \text{ a.e. on } X \text{ for all } n $$
Then $f$ is integrable over $X$ and:
$$ \lim_{n \to \infty} \int_X f_n \, d\mu = \int_X f \, d\mu $$

::: {.callout-tip collapse="true"}
## Proof of the Lebesgue Dominated Convergence Theorem

From the hypothesis $|f_n| \le g$, we know that for all $n$:
$$ -g \le f_n \le g $$

This allows us to construct two non-negative sequences of measurable functions:
1. $0 \le g - f_n$
2. $0 \le g + f_n$

Since $f_n \to f$ pointwise almost everywhere, it follows that $(g - f_n) \to (g - f)$ and $(g + f_n) \to (g + f)$ almost everywhere. We can now apply the $\epsilon$-$N$ formulation of Fatou's Lemma to both sequences. 

For any arbitrary $\epsilon > 0$, there exists an integer $N$ such that for all $n > N$:
$$ \int_X (g - f_n) \, d\mu > \int_X (g - f) \, d\mu - \epsilon $$
$$ \int_X (g + f_n) \, d\mu > \int_X (g + f) \, d\mu - \epsilon $$

By the linearity of the integral, we can expand these expressions:
$$ \int_X g \, d\mu - \int_X f_n \, d\mu > \int_X g \, d\mu - \int_X f \, d\mu - \epsilon $$
$$ \int_X g \, d\mu + \int_X f_n \, d\mu > \int_X g \, d\mu + \int_X f \, d\mu - \epsilon $$

Because $g$ is an integrable function (its integral is finite, $\int_X g \, d\mu < \infty$), we can safely subtract $\int_X g \, d\mu$ from both sides of each inequality. This yields:
$$ -\int_X f_n \, d\mu > -\int_X f \, d\mu - \epsilon $$
$$ \int_X f_n \, d\mu > \int_X f \, d\mu - \epsilon $$

Multiplying the first inequality by $-1$ (which flips the sign direction) gives:
$$ \int_X f_n \, d\mu < \int_X f \, d\mu + \epsilon $$

Combining these two bounded results, we obtain:
$$ \int_X f \, d\mu - \epsilon < \int_X f_n \, d\mu < \int_X f \, d\mu + \epsilon $$

This statement is precisely the definition of convergence for the sequence of integrals. Because $\epsilon > 0$ was arbitrary, we conclude:
$$ \lim_{n \to \infty} \int_X f_n \, d\mu = \int_X f \, d\mu $$
$\blacksquare$
:::
