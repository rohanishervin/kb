# Problem Set: Discrete Random Variables

## Problem 1: The World Series

**Problem Statement:**

In the World Series of baseball, two teams (Team A and Team B) play a sequence of [independent](independence-of-events.ipynb) games. The series concludes as soon as one team wins four games. Suppose that the probability of Team A winning any individual game is $p$. 

**Part (a):** What is the probability that Team A wins the overall series?

**Part (b):** Provide a clear, intuitive explanation of whether the answer to part (a) changes if the rules were modified. Instead of stopping when a team reaches 4 wins, suppose the teams *always* play exactly 7 games, and the winner of the series is the team that wins the majority of these games. Does this affect the probability of Team A winning?

::: {.callout-tip collapse="true"}
## Solution

**Part (a):**

To find the probability that Team A wins the series, we can use a clever conceptual trick. Because a team needs exactly 4 wins to take the series, the absolute maximum number of games that can be played is $4 + 3 = 7$ (if the series goes to a 4-3 finish). 

Imagine that the teams always play out all 7 games, even if one team has already clinched the 4 wins needed. We can call the unnecessary extra games "phantom games." 

Crucially, playing these phantom games does not change the winner of the series. If Team A gets 4 wins early, they will finish the 7-game set with *at least* 4 wins. Conversely, if they finish a 7-game set with at least 4 wins, it guarantees they reached 4 wins before Team B did. 

Thus, the probability of Team A winning the series is equivalent to the probability of Team A winning 4 or more games in a fixed 7-game series. 

Let $Y$ be the number of games Team A wins out of 7. The random variable $Y$ follows a [Binomial distribution](bernoulli-and-binomial-distributions.ipynb) with $n=7$ trials and success probability $p$. Let $q = 1-p$. The probability that Team A wins is:

$$
\begin{aligned}
P(\text{Team A wins}) &= P(Y \ge 4) \\
&= \sum_{k=4}^{7} \binom{7}{k} p^k q^{7-k} \\
&= \binom{7}{4}p^4 q^3 + \binom{7}{5}p^5 q^2 + \binom{7}{6}p^6 q^1 + \binom{7}{7}p^7 q^0
\end{aligned}
$$

**Part (b):**

Based on the logic used in Part (a), the probability of Team A winning the series **does not depend** on whether they stop early or play all 7 games. 

Intuitively, once a team has secured a 4-win advantage, it becomes impossible for the opposing team to reach 4 wins in the remaining games (since there are only 7 games total). Winning or losing the remaining "dead rubber" games won't change the final series winner. Because deciding the winner does not mathematically depend on those final games once the threshold is crossed, the probability remains exactly the same under either rule set.
:::

::: {.callout-note}
## Intuition: The Phantom Games Trick
The argument used in Part (b) is a powerful tool in probability known as extending the sample space or using "phantom games." When a process stops early based on a condition (like reaching 4 wins), the variable length of the process can make calculations messy. By imagining that the process continues to a fixed length (7 games) and realizing that the additional outcomes don't alter the fundamental success/failure condition of the original process, we convert a Negative Binomial distribution problem into a much simpler Binomial distribution problem.
:::
