# Naive Definition of Probability

## 1\. Definition

Let $A$ be an [event](event.ipynb) for an [experiment](experiment.ipynb) with a finite [sample-space](sample-space.ipynb) $S$.

The naive probability of $A$, denoted as $P_{\text{naive}}(A)$, is given by the ratio of the number of [outcomes](outcome.ipynb) favorable to $A$ to the total number of outcomes in $S$:

$$P_{\text{naive}}(A) = \frac{|A|}{|S|}$$

where $|A|$ denotes the cardinality (size) of event $A$.

## 2\. Strict Requirements

For this definition to be mathematically valid, two absolute conditions must be met:

1.  **Finite Sample Space:** $|S| = n < \infty$. The definition breaks down (division by infinity) if there are infinitely many outcomes.
2.  **Equiprobability:** Every individual outcome $\omega \in S$ must have the exact same likelihood of occurring.

## 3\. Limitations

The naive definition fails in continuous sample spaces (e.g., throwing a dart at a board) and in scenarios where outcomes are biased or weighted (e.g., a loaded die or predicting the weather). For these situations, the axiomatic definition of [probability](probability.ipynb) is required.
