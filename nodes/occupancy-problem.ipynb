# The Occupancy Problem

## 1\. Problem Statement

The **Occupancy Problem** (often visualized as "Balls and Bins") asks a fundamental question: Given $n$ independent trials with $m$ equally likely outcomes (or randomly throwing $n$ balls into $m$ distinct bins), what is the probability that **all $m$ outcomes have been observed at least once** (no bin is empty)?

This is deeply related to the "Coupon Collector's Problem", but instead of asking *how many* trials it takes to get everything, we are asking for the probability of getting everything in a *fixed* number of trials.

-----

## 2\. Mathematical Derivation (Using PIE)

Calculating this directly is incredibly difficult because there are countless ways to satisfy "at least one in every bin." Instead, we calculate the complement: the probability that **at least one bin is empty**.

We define $A_i$ as the event that a specific bin $i$ is empty. We want to find the probability of the union of all these events, $P(A_1 \cup A_2 \cup \dots \cup A_m)$, which requires the [Principle of Inclusion-Exclusion](inclusion-exclusion.ipynb).

**Step 1: Total Outcomes**
By [sampling with replacement](sampling-with-replacement.ipynb), the total number of ways to distribute $n$ balls into $m$ bins is $m^n$.

**Step 2: Individual Empty Bins (1st PIE Term)**
If bin $i$ is empty, the $n$ balls must all go into the remaining $(m-1)$ bins. There are $(m-1)^n$ ways to do this.
Because there are $\binom{m}{1}$ specific bins we could choose to be empty, the first sum in our PIE formula is:
$$\binom{m}{1} (m-1)^n$$

**Step 3: Pairwise Empty Bins (2nd PIE Term)**
If two specific bins, $i$ and $j$, are both empty, the balls must go into the remaining $(m-2)$ bins. There are $(m-2)^n$ ways to do this.
Because there are $\binom{m}{2}$ pairs of bins, the second sum is:
$$\binom{m}{2} (m-2)^n$$

**Step 4: The General PIE Formula**
Continuing this alternating pattern of inclusion and exclusion, the total number of distributions where *at least one* bin is empty is:
$$\text{Empty Bins} = \sum_{k=1}^{m} (-1)^{k-1} \binom{m}{k} (m-k)^n$$

**Step 5: The Final Probability**
To find the probability that *no* bin is empty, we subtract the "Empty Bins" arrangements from the total $m^n$, and then divide by $m^n$:
$$P(\text{All Observed}) = \sum_{k=0}^{m} (-1)^k \binom{m}{k} \left(1 - \frac{k}{m}\right)^n$$

-----

## 3\. Classic Example Calculation

**Question:** If you distribute $7$ items randomly into $4$ distinct categories, what is the exact probability that every category receives at least one item?

**Solution:**
Here, $n = 7$ and $m = 4$. Apply the final formula:

  * **k=0:** $\binom{4}{0} (1 - 0)^7 = (1)(1) = 1$
  * **k=1:** $-\binom{4}{1} (1 - 1/4)^7 = -4(3/4)^7 = -4(2187/16384) = -8748/16384$
  * **k=2:** $+\binom{4}{2} (1 - 2/4)^7 = +6(2/4)^7 = +6(128/16384) = +768/16384$
  * **k=3:** $-\binom{4}{3} (1 - 3/4)^7 = -4(1/4)^7 = -4(1/16384) = -4/16384$
  * **k=4:** $+\binom{4}{4} (1 - 4/4)^7 = 0$

Summing these terms:
$$P(\text{All Observed}) = 1 - \frac{8748}{16384} + \frac{768}{16384} - \frac{4}{16384}$$
$$P(\text{All Observed}) = \frac{16384 - 8748 + 768 - 4}{16384} = \frac{8400}{16384} \approx 0.5127$$

There is exactly a $51.27\%$ chance that all 4 categories contain at least one item.

-----

## 4\. Python Simulation

We can empirically verify this $51.27\%$ mathematical truth by simulating the distribution using `numpy`.

```python
import numpy as np

def sim_occupancy(balls=7, bins=4, trials=100000):
    """
    Simulates distributing n balls into m bins.
    Returns the probability that all bins have at least one ball.
    """
    # np.random.randint(0, bins, balls) generates 'balls' random numbers between 0 and bins-1
    # len(np.unique(...)) counts how many distinct bins received at least one ball
    results = [len(np.unique(np.random.randint(0, bins, balls))) for _ in range(trials)]
    
    # Calculate the mean of boolean values (True if all bins hit, False otherwise)
    success_rate = np.mean([r == bins for r in results])
    
    return success_rate

# Run the simulation
empirical_prob = sim_occupancy(balls=7, bins=4, trials=100000)
display(f"Empirical Success rate: {empirical_prob:.4f}")
display(f"Theoretical Exact rate: 0.5127")
```
