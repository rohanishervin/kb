# Partnerships

Show that: 

$$\frac{(2n)!}{2^n \cdot n!} = (2n-1)(2n-3)\cdots 3\cdot 1$$

# Proof:
we can use [story proofs](story-proofs.ipynb)

### Method 1: Lining up People
Take $2n$ people and assign them ID numbers from 1 to $2n$. We can form partnerships by lining up the people in some order and then grouping them into pairs. This process counts the number of ways to break 
$2n$ people into $n$ partnerships, but it **overcounts** by a factor of $$(\text{\( n! \) × \( 2n \)}),$$ since the order of pairs doesn't matter, nor does the order within each pair.

### Method 2: Partner Selection
Alternatively, count the number of possibilities as follows by [multiplication rule](multiplication-rule.ipynb):

* There are $(2n - 1)$ choices for the partner of person $1$.
* For each choice made by person $1$, there are $(2n - 3)$ remaining choices for person $2$ (or person $3$, if person $2$ was already paired to person $1$).
* This process continues until all people have been assigned a partner.
