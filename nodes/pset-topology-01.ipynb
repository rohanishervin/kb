# Problem Set: Metric Space Topology

## Problem 1: Open Interval $(a, b)$ is an Open Set in $\mathbb{R}$

**Problem Statement:**
Prove that the open interval $(a, b)$ is an open set in the metric space $(\mathbb{R}, |\cdot|)$.

**Proof:**
Let $S = (a, b)$ and choose an arbitrary point $p \in S$. By definition, $a < p < b$.

By [definition](metric-space-topology.ipynb), we must find a radius $r > 0$ such that the neighborhood $M_r p \subset S$. 
Choose $r = \min(p - a, b - p)$. Since $p$ is strictly between $a$ and $b$, $r$ is strictly positive.

For any point $q \in M_r p$, we have $|q - p| < r$, which means:
$$p - r < q < p + r$$

Substituting our bounds for $r$:
$$a = p - (p - a) \le p - r < q < p + r \le p + (b - p) = b$$

Thus, $a < q < b$, meaning $q \in S$. Therefore, the entire neighborhood $M_r p$ is contained in $S$, proving $(a, b)$ is an open set. $\square$


## Problem 2: Clopen Sets

**Problem Statement:**
What are clopen sets? Give the canonical examples in any metric space.

**Solution:**
A set is **clopen** if it is both open and closed. 

In any metric space $M$, the empty set $\emptyset$ and the whole space $M$ are always clopen.


## Problem 3: Clopen Subsets of $\mathbb{R}$

**Problem Statement:**
What are all the clopen subsets of $\mathbb{R}$ with the standard Euclidean metric? Explain why.

**Solution:**
In the metric space $\mathbb{R}$ with the standard Euclidean metric, the **only** clopen sets are $\emptyset$ and $\mathbb{R}$. There are no other subsets of $\mathbb{R}$ that are both open and closed. This is a fundamental property related to the **connectivity** of the real line.

::: {.callout-note}
## Intuition: The Relativity of Topology
Openness and closedness are **not intrinsic properties** of a set. They are relative to the background metric space $M$. A set $S$ might be open when viewed in one space, but not open when viewed as a subset of a larger space. 
:::


## Problem 4: Clopen Sets in $\mathbb{Q}$

**Problem Statement:**
Identify a nontrivial clopen set in $\mathbb{Q}$ (with the metric inherited from $\mathbb{R}$) and justify why it is both open and closed.

**Solution:**
If we change the background metric space to $M = \mathbb{Q}$, the "gaps" left by irrational numbers allow for more clopen sets:

1. The set $\mathbb{Q}$ is both open and closed in $\mathbb{Q}$ (by the general property of $M$).
2. The set $S = \{x \in \mathbb{Q} : x^2 < 2\}$ is **clopen** in $\mathbb{Q}$.
   * **It is open:** For any $q \in S$, since $q^2 < 2$, there is a rational neighborhood $(q-r, q+r) \cap \mathbb{Q}$ that stays within the bounds of $\sqrt{2}$.
   * **It is closed:** In $\mathbb{R}$, the limit points are $\pm\sqrt{2}$, which are irrational. In the world of $\mathbb{Q}$, these points do not exist. Therefore, every sequence in $S$ that converges *to a rational number* must converge to a point already in $S$.


## Problem 5: Infinite Intersection of Open Sets

**Problem Statement:**
Show by example that an infinite intersection of open sets need not be open.

**Solution:**
While the [axiom of topology](metric-space-topology.ipynb) states that the intersection of any *finite* number of open sets is open, this property does not hold for *infinite* intersections.

Consider the metric space $\mathbb{R}$ with the standard metric. Define an infinite sequence of open intervals:
$$U_n = \left(-\frac{1}{n}, \frac{1}{n}\right) \quad \text{for } n = 1, 2, 3, \dots$$

Every $U_n$ is an open set (as proven in Problem 1). Now, let $S$ be the intersection of all these sets:
$$S = \bigcap_{n=1}^{\infty} U_n = \bigcap_{n=1}^{\infty} \left(-\frac{1}{n}, \frac{1}{n}\right)$$

By the Archimedean property, the only real number that is strictly between $-1/n$ and $1/n$ for *every* positive integer $n$ is $0$. Therefore, the infinite intersection collapses to a single point:
$$S = \{0\}$$

**Proof that $S$ is not open:**
For $S = \{0\}$ to be open, there must exist some radius $r > 0$ such that the entire neighborhood $M_r(0) = (-r, r)$ is contained within $S$. 
However, for any strictly positive $r$, the interval $(-r, r)$ contains infinitely many non-zero numbers, meaning $(-r, r) \not\subset \{0\}$. 
Thus, $\{0\}$ is not an open set. (It is, in fact, a closed set). $\square$


## Problem 6: Every Open Ball is an Open Set

**Problem Statement:**
Prove that in any metric space $(M, d)$, the open ball $M_r p$ is an open set.

**Proof:**
Let $S = M_r p = \{x \in M \mid d(x, p) < r\}$. 
To prove $S$ is open, we must show that for any arbitrary point $q \in S$, there exists some radius $s > 0$ such that the neighborhood $M_s q$ is entirely contained within $S$.

Since $q \in S$, we know that the distance $d(p, q) < r$. 
Let us define a new radius:
$$s = r - d(p, q)$$
Because $d(p, q) < r$, it is strictly true that $s > 0$.

Now, we must verify that $M_s q \subset S$. Let $x$ be any point in $M_s q$. 
By definition, $d(x, q) < s$. 

To determine if $x \in S$, we check its distance to the center $p$ using the **triangle inequality**:
$$d(x, p) \leq d(x, q) + d(q, p)$$

Substitute our known strict bound $d(x, q) < s$:
$$d(x, p) < s + d(q, p)$$

Now, substitute our definition of $s$:
$$d(x, p) < (r - d(p, q)) + d(q, p)$$
$$d(x, p) < r$$

Since $d(x, p) < r$, the point $x$ belongs to $S$. Because this holds for any arbitrary point $x \in M_s q$, we conclude that $M_s q \subset S$. 
Therefore, $M_r p$ is an open set. $\square$


## Problem 7: Pugh, Ch. 2, P. 9

**Problem Statement:**
A sequence $(x_n)$ in $\mathbb{R}$ increases if $n < m$ implies $x_n \le x_m$. It strictly increases if $n < m$ implies $x_n < x_m$. It decreases or strictly decreases if $n < m$ always implies $x_n \ge x_m$ or always implies $x_n > x_m$. A sequence is monotone if it increases or it decreases. Prove that every sequence in $\mathbb{R}$ which is monotone and bounded converges in $\mathbb{R}$.

**Mathematical Solution:**

Suppose $(x_n)$ is a sequence in $\mathbb{R}$ that is monotone and bounded. Because it is bounded, it is bounded both above and below. Because it is monotone, it must be either increasing or decreasing. We address these two cases.

### Case 1: The sequence $(x_n)$ is increasing.

Since the sequence is bounded, it is bounded above by some real number $M$. 

Let $S = \{x_n : n \in \mathbb{N}\}$.
Because $x_1 \in S$, we know $S \neq \emptyset$.
Furthermore, $\forall n \in \mathbb{N}, x_n \le M$, which means $S$ is bounded above.

By the [Axiom of Completeness](axiom-of-completeness.ipynb#the-axiom-of-completeness-1), the supremum of this set exists in $\mathbb{R}$. Let $x^* = \sup(S)$.
By the definition of the [supremum](supremum-infimum.ipynb#the-epsilon-characterization), for any $\epsilon > 0$, there exists an integer $N \in \mathbb{N}$ such that:
$$x^* - \epsilon < x_N$$

Because the sequence $(x_n)$ is increasing, for all $n \ge N$, we know that $x_N \le x_n$. Therefore:
$$x^* - \epsilon < x_N \le x_n \quad \text{--- (1)}$$

Additionally, because $x^*$ is an upper bound for the entire set $S$, it is always true that:
$$x_n \le x^* < x^* + \epsilon \quad \text{--- (2)}$$

Combining inequalities (1) and (2), we get:
$$x^* - \epsilon < x_n < x^* + \epsilon$$
$$|x_n - x^*| < \epsilon$$

Since this holds for any $\epsilon > 0$ whenever $n \ge N$, by definition, $x_n \to x^*$. The sequence converges.

---

### Case 2: The sequence $(x_n)$ is decreasing.

Initial logic and parameters are symmetrically validated for the decreasing case, bounded below by some real number $m$. 

By the Axiom of Completeness, the infimum exists. Let $x^* = \inf(S)$. Applying the definition of the infimum and the decreasing property for all $n \ge N$, we jump directly to the final transformation:
$$x^* - \epsilon < x^* \le x_n \le x_N < x^* + \epsilon$$

Which immediately yields:
$$|x_n - x^*| < \epsilon$$

Hence, $x_n \to x^*$. The sequence converges. $\blacksquare$

## Problem 8: Pugh, Ch. 2, P. 11

**Problem Statement:**
Let $(x_n)$ be a sequence in $\mathbb{R}$.

(a) Prove that $(x_n)$ has a monotone subsequence.
(b) How can you deduce that every bounded sequence in $\mathbb{R}$ has a convergent subsequence?
(c) Infer that you have a second proof of the Bolzano-Weierstrass Theorem in $\mathbb{R}$.
(d) What about the Heine-Borel Theorem?

**Mathematical Solution:**

### Part (a): Existence of a Monotone Subsequence

**Proof:**
To prove this, we introduce the concept of a "peak" in a sequence.
We call an index $m$ a **peak** of the sequence $(x_n)$ if, for every future index $n > m$, it is true that $x_m \ge x_n$. In other words, the sequence never again reaches a value strictly greater than $x_m$.

Let $P \subset \mathbb{N}$ be the set of all peak indices of our sequence. There are exactly two mutually exclusive possibilities for the size of $P$:

* **Case 1: $P$ is infinite.** If there are infinitely many peak indices, we can list them in ascending order as $m_1 < m_2 < m_3 < \dots$
By the very definition of a peak, since $m_1$ is a peak and $m_2 > m_1$, it must be that $x_{m_1} \ge x_{m_2}$. Similarly, since $m_2$ is a peak and $m_3 > m_2$, we have $x_{m_2} \ge x_{m_3}$.
Continuing this indefinitely yields $x_{m_1} \ge x_{m_2} \ge x_{m_3} \ge \dots$
Thus, $(x_{m_k})$ is a **monotone non-increasing subsequence**.

* **Case 2: $P$ is finite (or empty).**
If there are only finitely many peaks, there must be a maximum peak index $M$ (if $P$ is empty, let $M=0$).
Choose any index $n_1 > M$. Because $n_1 > M$, $n_1$ is *not* a peak. Since it is not a peak, there must exist some later index $n_2 > n_1$ such that $x_{n_2} > x_{n_1}$.
Now consider $n_2$. Because $n_2 > n_1 > M$, $n_2$ is also *not* a peak. Therefore, there must exist some $n_3 > n_2$ such that $x_{n_3} > x_{n_2}$.
Continuing this indefinitely yields $x_{n_1} < x_{n_2} < x_{n_3} < \dots$
Thus, $(x_{n_k})$ is a **strictly increasing subsequence**.

Since one of these two cases must be true, every sequence in $\mathbb{R}$ possesses at least one monotone subsequence. $\square$

---

### Part (b): Bounded Sequences Have Convergent Subsequences

**Proof:**
Suppose $(x_n)$ is a bounded sequence in $\mathbb{R}$.

1. By the theorem proven in Part (a), we know $(x_n)$ must contain a monotone subsequence, which we will call $(x_{n_k})$.
2. Because the original "parent" sequence $(x_n)$ is bounded (meaning there exists an $M$ such that $|x_n| \le M$ for all $n$), every subsequence drawn from it is inherently bounded by that exact same $M$.
3. Therefore, $(x_{n_k})$ is a bounded, monotone sequence.
4. By the **Monotone Convergence Theorem** (proved previously in Problem 9), every sequence in $\mathbb{R}$ that is both bounded and monotone is guaranteed to converge.

Thus, the subsequence $(x_{n_k})$ converges. $\square$

---

### Part (c): Connection to the Bolzano-Weierstrass Theorem

The standard statement of the **Bolzano-Weierstrass Theorem (Sequential Formulation)** states exactly what we just proved in Part (b): *"Every bounded sequence in $\mathbb{R}$ has a convergent subsequence."*

The traditional proof of [Bolzano-Weierstrass](Bolzano-Weierstrass-theorem.ipynb) (often found earlier in analysis textbooks) relies on the **Nested Interval Property** or the method of continuous bisection—cutting a bounded interval in half repeatedly and tracking the half containing infinitely many terms.

By combining the "Peak Point" lemma from Part (a) with the Monotone Convergence Theorem in Part (b), we have constructed a completely distinct, highly elegant **second proof** of the Bolzano-Weierstrass theorem that completely bypasses interval bisection.

---

### Part (d): Implications for the Heine-Borel Theorem

The **Heine-Borel Theorem** states that a subset of $\mathbb{R}$ is topologically compact (every open cover has a finite subcover) if and only if it is closed and bounded.

In metric spaces, topological compactness is perfectly equivalent to **sequential compactness** (a space where every sequence has a subsequence that converges to a limit *within that space*).

Our new proof of Bolzano-Weierstrass gives us an immediate proof of [Heine-Borel](compactness.ipynb#the-heine-borel-theorem) in the context of sequential compactness:

* Let $S \subset \mathbb{R}$ be a closed and bounded set.
* Take any sequence $(x_n)$ in $S$. Because $S$ is bounded, the sequence is bounded.
* By our Bolzano-Weierstrass proof from (b) and (c), the sequence has a convergent subsequence $(x_{n_k}) \to L$.
* Because $S$ is closed, it contains all its limit points, guaranteeing $L \in S$.
* Therefore, any sequence in a closed and bounded set $S$ has a convergent subsequence whose limit is in $S$, proving $S$ is sequentially compact.

Thus, our new proof of BW serves as the foundational structural pillar for proving Heine-Borel using sequential logic instead of the traditional Lebesgue-number or open-cover contradictions.



## Problem 9: Pugh, Ch. 2, P. 12

**Problem Statement:**
Let $(p_n)$ be a sequence and $f : \mathbb{N} \to \mathbb{N}$ be a bijection. The sequence $(q_k)_{k\in\mathbb{N}}$ with $q_k = p_{f(k)}$ is a **rearrangement** of $(p_n)$.
(a) Are limits of a sequence unaffected by rearrangement?

**Mathematical Solution:**

### Part (a): Rearrangement via Bijection

**Yes, the limits are unaffected.** If $(p_n)$ converges to a limit $L$, then any rearrangement $(q_k)$ created via a bijection will also converge to $L$.

**Proof:**
Suppose the original sequence $(p_n)$ converges to a limit $L$. 
By the formal definition of convergence, for any chosen error tolerance $\epsilon > 0$, there exists a threshold index $N \in \mathbb{N}$ such that:
$$n \ge N \implies |p_n - L| < \epsilon$$

This means that *at most* a finite number of terms from the sequence $(p_n)$ fall outside the $\epsilon$-neighborhood of $L$. Specifically, the only terms that could possibly violate the condition $|p_n - L| < \epsilon$ are those whose indices belong to the finite set:
$$I_{bad} = \{1, 2, \dots, N-1\}$$

We are given a rearranged sequence $(q_k)$, defined as $q_k = p_{f(k)}$, where $f : \mathbb{N} \to \mathbb{N}$ is a bijection. 
Because $f$ is a bijection, it has a well-defined inverse function $f^{-1} : \mathbb{N} \to \mathbb{N}$. 

To guarantee that a term $q_k$ falls *inside* the $\epsilon$-neighborhood, we need its underlying original index, $f(k)$, to be $\ge N$. Equivalently, we need $f(k) \notin I_{bad}$.

Let us identify the exact set of new indices $k$ that map to the "bad" original indices. Because $f^{-1}$ is a well-defined function mapping finite sets to finite sets, the pre-image of $I_{bad}$ under $f$ is exactly:
$$K_{bad} = \{ f^{-1}(1), f^{-1}(2), \dots, f^{-1}(N-1) \}$$

Since $I_{bad}$ is finite and $f$ is a bijection, $K_{bad}$ is also a strictly finite set of integers. 
Because any finite set of integers has a maximum, we can define a new threshold index $K$:
$$K = \max(K_{bad}) + 1$$
*(Note: If $N=1$, then $I_{bad}$ is empty, $K_{bad}$ is empty, and we simply choose $K=1$.)*

Now, consider any index $k \ge K$. 
Because $k > \max(K_{bad})$, we know with absolute certainty that $k \notin K_{bad}$. 
By definition of the sets, if $k \notin K_{bad}$, then it must be true that $f(k) \notin I_{bad}$. 
Since $f(k)$ is a natural number not in $\{1, 2, \dots, N-1\}$, we conclude that $f(k) \ge N$.

Finally, because $f(k) \ge N$, it satisfies the original convergence condition:
$$|q_k - L| = |p_{f(k)} - L| < \epsilon$$

Since we have shown that for any $\epsilon > 0$, there exists a threshold $K$ such that for all $k \ge K$, $|q_k - L| < \epsilon$, we have formally proven that:
$$\lim_{k \to \infty} q_k = L$$
$\blacksquare$


## Problem 10: Pugh, Ch. 2, P. 13

**Problem Statement:**
Assume that $f : M \to N$ is a function from one metric space to another which satisfies the following condition: If a sequence $(p_n)$ in $M$ converges then the sequence $(f(p_n))$ in $N$ converges. Prove that $f$ is continuous. this improves [this](continuity-topology.ipynb#the-epsilon-delta-condition) theorem.

**Mathematical Solution:**

To prove that $f$ is continuous, we must show that it completely preserves sequential convergence: for any sequence $(p_n)$ in $M$ such that $p_n \to p$, the mapped sequence $(f(p_n))$ converges specifically to $f(p)$ in $N$.

Let $(p_n)$ be an arbitrary sequence in $M$ that converges to a limit $p \in M$. 

We construct a new, intertwined sequence $(x_n)$ by alternating the terms of $(p_n)$ with the constant sequence $(p, p, p, \dots)$. We formally define $(x_n)$ as:
$$x_n = \begin{cases} 
p & \text{if } n \text{ is odd} \\
p_{n/2} & \text{if } n \text{ is even} 
\end{cases}$$
Writing out the first few terms, we see $(x_n) = (p, p_1, p, p_2, p, p_3, \dots)$.

**Step 1: Prove $(x_n)$ converges to $p$**

Because $(p_n)$ converges to $p$, for any $\epsilon > 0$, there exists an integer $N$ such that for all $n \ge N$, $d_M(p_n, p) < \epsilon$.
Let us set a new threshold $N' = 2N$. If we choose any index $n \ge N'$ for our new sequence $(x_n)$:
* If $n$ is odd, $x_n = p$, meaning $d_M(x_n, p) = d_M(p, p) = 0 < \epsilon$.
* If $n$ is even, $x_n = p_{n/2}$. Since $n \ge 2N$, it must be that $n/2 \ge N$. Therefore, $d_M(x_n, p) = d_M(p_{n/2}, p) < \epsilon$.

Because all terms past $N'$ are strictly within $\epsilon$ of $p$, the sequence $(x_n)$ converges to $p$ in $M$.

**Step 2: Apply the given hypothesis**

The premise of the problem states that if *any* sequence converges in $M$, its image under $f$ must converge to some limit in $N$. Since we proved $(x_n)$ converges, the sequence $(f(x_n))$ is guaranteed to converge to some unknown limit $L \in N$.

**Step 3: Determine the limit $L$**

A fundamental property of convergent sequences is that every infinite subsequence extracted from it must converge to the exact same limit as the [mother](sequences-metric-spaces.ipynb#subsequences) sequence. 

Let us examine the odd-indexed subsequence of $(f(x_n))$:
$$(f(x_{2k-1})) = (f(p), f(p), f(p), \dots)$$
This is a constant sequence, which trivially converges to $f(p)$. Since this is a subsequence of $(f(x_n))$, its limit must be the parent sequence's limit. Therefore, $L = f(p)$.

Now let us examine the even-indexed subsequence of $(f(x_n))$:
$$(f(x_{2k})) = (f(p_1), f(p_2), f(p_3), \dots) = (f(p_n))$$
Because this is also a subsequence of $(f(x_n))$, it is strictly bound to converge to that exact same limit $L$. 

Therefore, we can conclude that:
$$\lim_{n \to \infty} f(p_n) = L = f(p)$$

Because an arbitrary convergent sequence $p_n \to p$ maps to $f(p_n) \to f(p)$, the function $f$ perfectly satisfies the [sequential definition of continuity](continuity-topology.ipynb#sequential-continuity). $\blacksquare$


## Problem 11: Pugh, Ch. 2, P. 24

**Problem Statement:**
For which intervals $[a, b]$ in $\mathbb{R}$ is the intersection $[a, b] \cap \mathbb{Q}$ a clopen subset of the metric space $\mathbb{Q}$?

**Answer:**
The intersection $[a, b] \cap \mathbb{Q}$ is clopen in $\mathbb{Q}$ if and only if both boundary points are irrational, i.e., $a \notin \mathbb{Q}$ and $b \notin \mathbb{Q}$.

---

**Mathematical Solution:**

Let $E = [a, b] \cap \mathbb{Q}$. We must prove both directions: that irrational boundaries make $E$ clopen, and that rational boundaries prevent $E$ from being clopen.

### Part 1: If $a, b \notin \mathbb{Q}$, then $E$ is clopen in $\mathbb{Q}$.

**Proof that $E$ is closed in $\mathbb{Q}$:**
By using this [theorem](relative-topology-inheritance.ipynb#inheritance-of-closed-sets). The interval $[a, b]$ is closed in $\mathbb{R}$. Therefore, $E = [a, b] \cap \mathbb{Q}$ is strictly closed in $\mathbb{Q}$.

**Proof that $E$ is open in $\mathbb{Q}$:**
Let $p \in E$. Since $p \in \mathbb{Q}$ and the boundaries $a, b$ are strictly irrational, it is impossible for $p$ to equal $a$ or $b$. Thus, $a < p < b$.
Let us define a radius $r$ as the shortest distance from $p$ to the boundaries:
$$r = \min(|p - a|, |b - p|)$$
Because $p \neq a$ and $p \neq b$, we are guaranteed that $r > 0$. 
Consider the open ball of radius $r$ around $p$ in our metric space $\mathbb{Q}$:
$$B_\mathbb{Q}(p, r) = \{q \in \mathbb{Q} : |q - p| < r\}$$
By our choice of $r$, every rational number $q$ in this ball satisfies $a < q < b$. Therefore, $B_\mathbb{Q}(p, r) \subset E$. 
Because we can draw a valid open ball with $r > 0$ strictly inside $E$ around every single point $p \in E$, the set $E$ is open in $\mathbb{Q}$. 

Since $E$ is both closed and open, it is clopen.

---

### Part 2: If $a \in \mathbb{Q}$ or $b \in \mathbb{Q}$, then $E$ is NOT clopen in $\mathbb{Q}$.

Suppose for contradiction that $E$ is clopen, but at least one boundary point is rational. Assume without loss of generality that $b \in \mathbb{Q}$.

Because $b \in [a, b]$ and $b \in \mathbb{Q}$, we know $b \in E$. 
If $E$ is genuinely open in $\mathbb{Q}$, there must exist some radius $r > 0$ such that the open ball $B_\mathbb{Q}(b, r)$ is entirely contained within $E$. 
$$B_\mathbb{Q}(b, r) = (b - r, b + r) \cap \mathbb{Q}$$
However, any open interval $(b - r, b + r)$ inherently contains rational numbers strictly greater than $b$ (for example, the rational number $b + r/2$). 
These numbers are in $\mathbb{Q}$, but they are strictly greater than $b$, meaning they are not in the interval $[a, b]$. 

Thus, the open ball $B_\mathbb{Q}(b, r)$ spills outside of $E$. It is impossible to contain an open ball around $b$ entirely within $E$. Therefore, $E$ is **not open** in $\mathbb{Q}$, and consequently cannot be clopen. The exact same logic applies if $a \in \mathbb{Q}$. $\blacksquare$


## Problem 12: Pugh, Ch. 2, P. 27

**Problem Statement:**
If $S, T \subset M$, a metric space, and $S \subset T$, prove that:

(a) $\overline{S} \subset \overline{T}$
(b) $\text{int}(S) \subset \text{int}(T)$

*(In other words, the operations of taking the closure and taking the interior preserve subset relations. They are monotonic.)*

---

**Mathematical Solution:**

### Part (a): Prove $\overline{S} \subset \overline{T}$

**Proof:**
We will use the sequential characterization of the closure. 

Let $p \in \overline{S}$ be an arbitrary point. By the definition of the closure in a metric space, there must exist a sequence of points $(p_n)$ entirely contained within $S$ that converges to $p$:
$$p_n \in S \quad \text{and} \quad \lim_{n \to \infty} p_n = p$$

We are given the hypothesis that $S \subset T$. Because every term of the sequence $(p_n)$ belongs to $S$, it must also be true that every term belongs to $T$:
$$p_n \in T \quad \text{for all } n \in \mathbb{N}$$

We now have a sequence $(p_n)$ that is entirely contained in $T$ and converges to $p$. By the exact same sequential definition of closure, this guarantees that $p$ belongs to the closure of $T$:
$$p \in \overline{T}$$

Since choosing an arbitrary $p \in \overline{S}$ implies $p \in \overline{T}$, we have proven that $\overline{S} \subset \overline{T}$. $\blacksquare$

---

### Part (b): Prove $\text{int}(S) \subset \text{int}(T)$

**Proof:**
We will use the open ball definition of interior points.

Let $p \in \text{int}(S)$ be an arbitrary point. By the definition of the interior, $p$ must be surrounded by an open ball (neighborhood) of some radius $r > 0$ that is entirely contained within $S$:
$$\exists r > 0 \text{ such that } M_r(p) \subset S$$

We are given the hypothesis that $S \subset T$. Because the open ball $M_r(p)$ is a subset of $S$, and $S$ is a subset of $T$, the transitivity of subsets guarantees that the open ball is also a subset of $T$:
$$M_r(p) \subset S \subset T \implies M_r(p) \subset T$$

By definition, the interior of $T$, denoted $\text{int}(T)$, is the union of all open sets contained within $T$. Since $M_r(p)$ is an open set entirely contained within $T$, it must be a subset of $\text{int}(T)$. 

Because $p$ is the center of $M_r(p)$, we have $p \in M_r(p)$. Therefore:
$$p \in M_r(p) \subset \text{int}(T) \implies p \in \text{int}(T)$$

Since choosing an arbitrary $p \in \text{int}(S)$ implies $p \in \text{int}(T)$, we have proven that $\text{int}(S) \subset \text{int}(T)$. $\blacksquare$


## Problem 13: Pugh, Ch. 2, P. 32

**Problem Statement:**
Show that every subset of $\mathbb{N}$ is clopen. What does this tell you about every function $f : \mathbb{N} \to M$, where $M$ is a metric space?

---

**Mathematical Solution:**

### Part 1: Every subset of $\mathbb{N}$ is clopen

Let $S \subseteq \mathbb{N}$ be an arbitrary subset. We must show that $S$ is both open and closed under the standard metric $d(x, y) = |x - y|$.

**1. Proof that $S$ is open:**
To show $S$ is open, we must show that every point $p \in S$ is an interior point. 
Let $p \in S$. Consider the open ball around $p$ with radius $r = 1/2$:
$$B_{1/2}(p) = \{x \in \mathbb{N} : |x - p| < 1/2\}$$
Because the distance between any two distinct natural numbers is at least $1$, the only natural number strictly within a distance of $1/2$ from $p$ is $p$ itself. Therefore:
$$B_{1/2}(p) = \{p\}$$
Since $p \in S$, it follows that $B_{1/2}(p) \subseteq S$. Because we can fit a valid open ball entirely inside $S$ for every point in $S$, the set $S$ is open.

**2. Proof that $S$ is closed:**
By the definition of topology, a set is closed if and only if its complement is open.
Consider the complement of our set, $S^c = \mathbb{N} \setminus S$. 
Because $S^c$ is simply another subset of $\mathbb{N}$, it immediately follows from the proof in Step 1 that $S^c$ is an open set.
Since its complement $S^c$ is open, $S$ is closed.

Because $S$ is both closed and open, every arbitrary subset $S \subseteq \mathbb{N}$ is **clopen**. $\blacksquare$

*(Note: A space where every subset is open is said to have the **discrete topology**).*

### Part 2: Implications for Functions $f: \mathbb{N} \to M$

This property tells us that **every possible function from $\mathbb{N}$ to any metric space $M$ is inherently continuous.**

**Topological Proof:**
Recall the topological definition of continuity: A function $f : A \to B$ is continuous if and only if the preimage of every open set in $B$ is open in $A$.
Let $V$ be an arbitrary open set in $M$. We must examine its preimage, $f^{-1}(V) \subseteq \mathbb{N}$.
Because we just proved that *every* subset of $\mathbb{N}$ is open, the subset $f^{-1}(V)$ is automatically open. 
Since the preimage of every open set is open, $f$ is perfectly continuous. $\blacksquare$

**Sequential Intuition:**
Why does this happen sequentially? In $\mathbb{N}$, points are isolated. The only way a sequence $(p_n)$ can converge to a limit $p$ in $\mathbb{N}$ is if the sequence eventually becomes perfectly constant (e.g., $p_n = p$ for all $n \ge N$). If we apply a function to an eventually constant sequence, the mapped sequence $f(p_n)$ also becomes eventually constant at $f(p)$, which trivially converges to $f(p)$. Thus, sequential limits are always preserved!
