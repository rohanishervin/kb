# Supremum and Infimum


## 1\. Boundedness

Before defining the "best" bounds, we must define what it means for a set to be constrained at all.

  * **Bounded Above:** A set $A \subseteq \mathbb{R}$ is bounded above if there exists a number $b \in \mathbb{R}$ such that $a \leq b$ for all $a \in A$. The number $b$ is an **upper bound**.
  * **Bounded Below:** A set $A \subseteq \mathbb{R}$ is bounded below if there exists a number $l \in \mathbb{R}$ such that $l \leq a$ for all $a \in A$. The number $l$ is a **lower bound**.
  * **Bounded:** A set is bounded if it is bounded both above and below.

-----

## 2\. The Supremum (Least Upper Bound)

The **supremum** of a set $A$, denoted $\sup(A)$ or $\text{LUB}(A)$, is the "tightest" possible upper bound. A real number $s$ is the supremum of $A$ if it satisfies two conditions:

1.  **It is an upper bound:** $\forall a \in A, a \leq s$.
2.  **It is the *least* such bound:** If $b$ is any upper bound for $A$, then $s \leq b$.

### The $\epsilon$-Characterization

In proofs, we rarely use the "leastness" property directly. Instead, we use this equivalent logical statement:

> $s = \sup(A)$ if and only if $s$ is an upper bound and for every $\epsilon > 0$, there exists some element $a_\epsilon \in A$ such that:
> $$s - \epsilon < a_\epsilon$$

**Intuition:** If you nudge the supremum even a tiny bit to the left ($s - \epsilon$), it is no longer an upper bound because you've "uncovered" at least one element of the set.

-----

## 3\. The Infimum (Greatest Lower Bound)

Similarly, the **infimum** of a set $A$, denoted $\inf(A)$ or $\text{GLB}(A)$, is the "tightest" possible lower bound. A real number $l$ is the infimum of $A$ if:

1.  **It is a lower bound:** $\forall a \in A, l \leq a$.
2.  **It is the *greatest* such bound:** If $m$ is any lower bound for $A$, then $m \leq l$.

### The $\epsilon$-Characterization

> $l = \inf(A)$ if and only if $l$ is a lower bound and for every $\epsilon > 0$, there exists some element $a_\epsilon \in A$ such that:
> $$a_\epsilon < l + \epsilon$$

-----

## 4\. Supremum vs. Maximum

It is vital to remember that the **supremum does not have to be an element of the set.**

| Term | Must it be in the set $A$? | Example: $A = \{x \in \mathbb{R} : 0 < x < 1\}$ |
| :--- | :--- | :--- |
| **Supremum** | No | $\sup(A) = 1$ |
| **Maximum** | Yes | None exists (the set is open) |
| **Infimum** | No | $\inf(A) = 0$ |
| **Minimum** | Yes | None exists |

If the $\sup(A) \in A$, then we say $\sup(A) = \max(A)$.

-----

::: {.callout-tip}

## Relationship to Completeness

The [Axiom of Completeness](axiom-of-completeness.ipynb) states that every non-empty set of real numbers that is bounded above **must** have a supremum in $\mathbb{R}$. This is what separates $\mathbb{R}$ from $\mathbb{Q}$. In the rationals, the set $S = \{x \in \mathbb{Q} : x^2 < 2\}$ is bounded above, but its "supremum" ($\sqrt{2}$) is missing from the set of rationals.
:::
