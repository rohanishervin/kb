# Independence of Random Variables

**Definition (Independence of Two r.v.s):** [Random variables](random-variables.ipynb) $X$ and $Y$ are independent if knowing the value of one provides absolutely zero information about the value of the other. Mathematically, for all real numbers $x, y$:

$$P(X \le x, Y \le y) = P(X \le x) P(Y \le y)$$

For discrete variables, this translates to $P(X=x, Y=y) = P(X=x)P(Y=y)$.

**Definition (Mutual Independence):** A collection of random variables $X_1, \dots, X_n$ is mutually independent if:

$$P(X_1 \le x_1, \dots, X_n \le x_n) = P(X_1 \le x_1) \dots P(X_n \le x_n)$$

This must hold for *all* possible combinations of $x_1, \dots, x_n$.

**Theorem:** If $X$ and $Y$ are independent random variables, then any function of $X$ is independent of any function of $Y$.

::: {.callout-note title="Clarification: i.i.d. vs i.d."}
"Independent" and "Identically Distributed" are two completely different concepts.

* **Identically Distributed (i.d.):** Two variables share the exact same PMF/CDF blueprint.
* **Independent:** Two variables do not influence each other.

Examples:

* **Independent and identically distributed (i.i.d.):** $X$ and $Y$ are independent die rolls.
* **Dependent and identically distributed:** Let $X$ be the number of Heads in $10$ coin tosses, and $Y$ be the number of Tails in those *same* tosses. Both follow $\text{Bin}(10, 0.5)$, but if we know $X$, we perfectly know $Y=10-X$.
:::

## Sums of Independent Binomials

**Theorem:** If $X \sim \text{Bin}(n, p)$ and $Y \sim \text{Bin}(m, p)$, and $X$ and $Y$ are independent, then their sum is also Binomial: $X + Y \sim \text{Bin}(n+m, p)$.

**Proof 1 (Story):**
$X$ counts the number of successes in $n$ independent trials, and $Y$ counts the successes in $m$ additional independent trials. Since the success probability $p$ is identical, $X+Y$ simply counts the total successes across all $n+m$ independent trials, which perfectly matches the story of a $\text{Bin}(n+m, p)$ distribution. $\square$

**Proof 2 (Representation):**
We can represent $X = X_1 + \dots + X_n$ and $Y = Y_1 + \dots + Y_m$ where all $X_i$ and $Y_j$ are i.i.d. $\text{Bern}(p)$ variables. Adding them yields a sum of $n+m$ i.i.d. Bernoulli random variables, which intrinsically defines a $\text{Bin}(n+m, p)$ distribution. $\square$

**Proof 3 (Law of Total Probability):**


$$P(X+Y = k) = \sum_{j=0}^{k} P(X+Y=k | X=j)P(X=j)$$

Because $X$ and $Y$ are independent, conditioning on $X=j$ simply leaves $P(Y = k-j)$.
Substituting the known Binomial PMFs:

$$= \sum_{j=0}^{k} \binom{m}{k-j} p^{k-j} (1-p)^{m-k+j} \binom{n}{j} p^j (1-p)^{n-j}$$

$$= p^k (1-p)^{n+m-k} \sum_{j=0}^{k} \binom{m}{k-j}\binom{n}{j}$$

By Vandermonde's Identity, the sum of combinations simplifies to $\binom{n+m}{k}$, resulting exactly in the $\text{Bin}(n+m, p)$ PMF. $\square$

## Conditional Independence

Random variables can be unconditionally independent, but conditionally dependent (and vice versa).
**Definition:** $X$ and $Y$ are conditionally independent given $Z$ if $P(X=x, Y=y | Z=z) = P(X=x | Z=z) P(Y=y | Z=z)$.

* **Independent but Conditionally Dependent (Matching Pennies):** Let $X$ and $Y$ be independent fair coin flips. Let $Z=XY$ be an indicator that the pennies match ($Z=1$ if they match). Unconditionally, $X$ and $Y$ are independent. But conditional on $Z=1$, $X$ and $Y$ are instantly locked together (knowing $X$ tells us $Y$ exactly).
* **Dependent but Conditionally Independent (Mystery Opponent):** You play two games against a mystery twin who is either terrible ($p=0.9$ to win) or amazing ($p=0.1$ to win). Your wins in game 1 ($X$) and game 2 ($Y$) are unconditionally *dependent*, because winning game 1 strongly implies you are playing the terrible twin, which raises your probability for game 2. However, *conditional* on knowing exactly which twin you face ($Z$), the games $X$ and $Y$ become completely independent Bernoulli trials.
