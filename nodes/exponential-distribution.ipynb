# Exponential Distribution

## 1. Introduction and Definition

The **Exponential distribution** is the continuous counterpart to the [Geometric distribution](geometric-and-negative-binomial-dist.ipynb#the-geometric-distribution). While the Geometric distribution models the discrete number of trials required to observe a success, the Exponential distribution models the continuous waiting time until the first arrival or success, given that these successes occur at a constant average rate. 

**Definition (Exponential Distribution):**
A positive [continuous random variable](random-variables.ipynb#continuous-random-variables) $X$ is said to follow an Exponential distribution with rate parameter $\lambda > 0$ if its probability density function (PDF) is given by:

$$
f(x) = \lambda e^{-\lambda x}, \quad \text{for } x > 0
$$

We denote this relationship as $X \sim \text{Expo}(\lambda)$.

The corresponding cumulative distribution function (CDF) is calculated by integrating the PDF:

$$
F(x) = 1 - e^{-\lambda x}, \quad \text{for } x > 0
$$

<!-- CELL BREAK -->

## 2. Properties and Scale Transformations

Unlike distributions that require both location and scale shifts (like the [Normal distribution](normal-distribution.ipynb)), the Exponential distribution is bounded at zero and thus strictly relies on scale transformations to map between different rate parameters.

**Proposition (Scale Transformation):**
If $X \sim \text{Expo}(1)$, we can scale it to any rate $\lambda$ by defining $Y = \frac{X}{\lambda}$. 

::: {.callout-tip collapse="true"}
## Proof of Scale Transformation
By applying the CDF method, we can find the distribution of $Y$:

$$
\begin{aligned}
P(Y \le y) &= P\left(\frac{X}{\lambda} \le y\right) \\
&= P(X \le \lambda y) \\
&= 1 - e^{-\lambda y}
\end{aligned}
$$

This is precisely the CDF of an $\text{Expo}(\lambda)$ distribution. Conversely, if $Y \sim \text{Expo}(\lambda)$, then $\lambda Y \sim \text{Expo}(1)$. $\blacksquare$
:::

Using the standard integration by parts for the base case $X \sim \text{Expo}(1)$, we know $E(X) = 1$ and $Var(X) = 1$. Applying the scale transformation yields the mean and variance for any $\lambda$:

*   **Mean:** $E(Y) = \frac{1}{\lambda} E(X) = \frac{1}{\lambda}$
*   **Variance:** $Var(Y) = \frac{1}{\lambda^2} Var(X) = \frac{1}{\lambda^2}$

<!-- CELL BREAK -->

## 3. The Memoryless Property

The most distinguishing feature of the Exponential distribution is its lack of memory. 

**Definition (Memoryless Property):**
A continuous probability distribution is memoryless if a random variable $X$ drawn from it satisfies the following condition for all $s, t \ge 0$:

$$
P(X \ge s + t \mid X \ge s) = P(X \ge t)
$$

::: {.callout-note}
## Intuition: What does "Memoryless" mean?
In plain English, the memoryless property means that the distribution "forgets" how long you have already been waiting. If you are waiting for an event (like a machine breaking down or a bus arriving), the probability of having to wait an additional $t$ minutes is exactly the same whether you just started waiting right now or if you have already been waiting for $s$ minutes. The system does not experience wear-and-tear or become "due" for an arrival.
:::

**Proposition:**
The Exponential distribution satisfies the memoryless property.

::: {.callout-tip collapse="true"}
## Proof of Exponential Memorylessness
Let $X \sim \text{Expo}(\lambda)$. We wish to compute the conditional probability $P(X \ge s + t \mid X \ge s)$. Using the definition of conditional probability:

$$
\begin{aligned}
P(X \ge s+t \mid X \ge s) &= \frac{P(X \ge s+t \text{ and } X \ge s)}{P(X \ge s)} \\
&= \frac{P(X \ge s+t)}{P(X \ge s)} \\
&= \frac{e^{-\lambda(s+t)}}{e^{-\lambda s}} \\
&= e^{-\lambda t} \\
&= P(X \ge t)
\end{aligned}
$$

Thus, the additional waiting time $X - s$, given $X \ge s$, is still distributed as $\text{Expo}(\lambda)$. $\blacksquare$
:::

## 4. Characterizing the Exponential Distribution

The Exponential distribution is not just *a* memoryless continuous distribution; it is the *only* one defined on $(0, \infty)$.

**Theorem (Uniqueness of the Memoryless Property):**
If $X$ is a positive continuous random variable exhibiting the memoryless property, then $X$ must possess an Exponential distribution.

::: {.callout-tip collapse="true"}
## Proof of Theorem
Let $F(x)$ be the CDF of $X$, and define the survival function as $G(x) = 1 - F(x)$. The goal is to prove that $G(x) = e^{-\lambda x}$ for some constant $\lambda$. 

**Step 1:** Use the memoryless condition.
The memoryless property $P(X \ge s+t \mid X \ge s) = P(X \ge t)$ implies that for all $s, t \ge 0$:

$$
G(s+t) = G(s)G(t)
$$

**Step 2:** Extend to integers.
Setting $s = t$, we get $G(2t) = G(t)^2$. By induction, for any positive integer $m$:

$$
G(mt) = G(t)^m
$$

**Step 3:** Extend to rational numbers.
Replace $t$ with $\frac{t}{n}$ where $n$ is a positive integer. We see that $G(t) = G\left(n \frac{t}{n}\right) = G\left(\frac{t}{n}\right)^n$, which implies $G\left(\frac{t}{n}\right) = G(t)^{1/n}$. Combining these results yields:

$$
G\left(\frac{m}{n}t\right) = G\left(\frac{t}{n}\right)^m = G(t)^{m/n}
$$

So for any positive rational number $x$, we have $G(xt) = G(t)^x$. 

**Step 4:** Extend to real numbers via continuity.
Since $G$ is continuous, any positive real number $x$ can be approximated by a sequence of rational numbers, allowing us to state $G(xt) = G(t)^x$ for all real $x > 0$. Letting $t = 1$, we define our function as:

$$
G(x) = G(1)^x = e^{-\lambda x}
$$

where we define $\lambda = -\log(G(1))$. Because $G(1)$ represents a probability strictly between 0 and 1, $\lambda > 0$. This perfectly matches the survival function of the Exponential distribution. $\blacksquare$
:::

## 5. Connection to Poisson Processes

The Exponential distribution inherently describes the interarrival times in a specific stochastic process known as a Poisson process. 

**Definition (Poisson Process):**
A sequence of arrivals occurring in continuous time forms a Poisson process with rate $\lambda$ if it meets two criteria:

1. The total number of arrivals within any time interval of duration $t$ is described by a [Poisson distribution](poisson-distribution.ipynb) with parameter $\lambda t$, denoted as $\text{Pois}(\lambda t)$.
2. The number of arrivals that happen within mutually disjoint time intervals are strictly independent of one another.

::: {.callout-warning}
## Warning: Length-Biased Sampling
When modeling wait times (e.g., waiting for a bus in a process governed by Exponential interarrival times), you cannot simply replace a random variable with its expected value. If the average interval between buses is 10 minutes, a random observer arriving uniformly over time does *not* wait 5 minutes on average. 

This paradox, known as **length-biased sampling**, occurs because an observer is proportionally much more likely to show up during a long gap between arrivals than a short gap. As a result of the memoryless property, the observer's expected wait time remains identical to the unconditional expected wait time (e.g., a full 10 minutes).
:::
