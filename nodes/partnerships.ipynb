# Partnerships (The Double Factorial Identity)

## 1\. The Identity

In combinatorics, calculating the number of ways to divide $2n$ distinct items into exactly $n$ pairs (partnerships) results in a famous algebraic identity.

For any integer $n \ge 1$:
$$\frac{(2n)!}{2^n \cdot n!} = (2n-1)(2n-3)\cdots 3\cdot 1$$

*(Note: The expression on the right-hand side—the product of all odd integers up to $2n-1$—is commonly denoted mathematically as the **double factorial**, written as $(2n-1)!!$)*.

-----

## 2\. Proof via Story

We can use a [story proof](story-proofs.ipynb) to demonstrate why this equality is absolutely true without relying on complex algebraic expansion.

**The Scenario:** Imagine we have exactly $2n$ distinct individuals in a room. We need to pair everyone up to form exactly $n$ partnerships. How many distinct ways can we do this?

**Method A: Grouping a Lineup (The Overcounting Approach)**

1.  First, we arrange all $2n$ individuals in a straight line. There are $(2n)!$ distinct ways to form this line.
2.  We then group them into pairs based on their physical position: the 1st and 2nd person form a pair, the 3rd and 4th form a pair, and so on.
3.  However, this method vastly overcounts the true number of partnerships.
      * The order of the $n$ distinct pairs does not matter (Pair A and Pair B is the same as Pair B and Pair A). We must divide by $n!$.
      * The internal order of the two individuals within each specific pair does not matter (Alice & Bob is the same as Bob & Alice). Because there are $n$ pairs, we must divide by $2$ exactly $n$ times, yielding an overcount factor of $2^n$.
4.  By the principle of [adjusting for overcounting](adjusting-for-overcounting.ipynb), dividing the total arrangements by these two factors gives the left side of the equation: $\frac{(2n)!}{2^n \cdot n!}$

**Method B: Sequential Selection (The Direct Approach)**

1.  Let us pick a person at random—or simply the tallest person in the room. They currently have $(2n - 1)$ possible partners to choose from.
2.  Once they pair up and step aside, $2n - 2$ people remain.
3.  We pick the next unmatched person. They will have $(2n - 3)$ possible partners left to choose from.
4.  This pattern continues sequentially down to the final two people, who have exactly $1$ choice (each other).
5.  Using the [multiplication rule](multiplication-rule.ipynb), we multiply these sequential independent choices together, giving the right side of the equation: $(2n-1)(2n-3)\cdots 3\cdot 1$

**Conclusion:** Since both Method A and Method B perfectly count the exact same physical outcome (the total number of ways to form $n$ partnerships from $2n$ people), the two mathematical expressions must be equal. $\square$

## 3\. Practical Example

**Question:** A tennis coach has $6$ players and wants to split them into $3$ practice pairs. How many different pairing combinations can be made?

**Solution:**
Here, $2n = 6$, which means $n = 3$.
Using the sequential method (right side):
$$(2(3)-1)!! = 5 \times 3 \times 1 = 15$$

Let's verify with the lineup method (left side):
$$\frac{6!}{2^3 \cdot 3!} = \frac{720}{8 \cdot 6} = \frac{720}{48} = 15$$

There are exactly $15$ unique ways to pair up the $6$ tennis players.
