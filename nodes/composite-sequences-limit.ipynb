# Algebraic Limit Theorem for Sequences

## 1\. Theorem Statement

Let $(a_n)$ and $(b_n)$ be convergent [sequences](sequence.ipynb) such that $\lim_{n \to \infty} a_n = a$ and $\lim_{n \to \infty} b_n = b$. Let $c \in \mathbb{R}$ be an arbitrary constant.

Then, the limits of their algebraic combinations exist and evaluate as follows:

1.  **Constant Multiple:** $\lim_{n \to \infty} (c a_n) = c a$
2.  **Sum:** $\lim_{n \to \infty} (a_n + b_n) = a + b$
3.  **Difference:** $\lim_{n \to \infty} (a_n - b_n) = a - b$
4.  **Product:** $\lim_{n \to \infty} (a_n b_n) = a b$
5.  **Quotient:** $\lim_{n \to \infty} \left(\frac{a_n}{b_n}\right) = \frac{a}{b}$, provided that $b_n \neq 0$ for all $n$ and $b \neq 0$.

*(Note: The proofs for 1, 2, and 3 rely on direct applications of the triangle inequality and are omitted here.)*

-----

## 2\. Proof of the Product Limit (Property 4)

**Goal:** We want to show that for any $\epsilon > 0$, there exists an $N$ such that for all $n \geq N$, $|a_n b_n - ab| < \epsilon$.

**Proof:**

First, we use a standard algebraic trick: we add and subtract $a_n b$ inside the absolute value to allow us to factor and use the triangle inequality:
$$|a_n b_n - ab| = |a_n b_n - a_n b + a_n b - ab|$$

Apply the triangle inequality:
$$|a_n b_n - ab| \leq |a_n(b_n - b)| + |b(a_n - a)| = |a_n| |b_n - b| + |b| |a_n - a|$$

To control this expression, we need to bound $|a_n|$. Because the sequence $(a_n)$ is convergent, we know it must be bounded. Therefore, there exists some real number $M > 0$ such that $|a_n| \leq M$ for all $n$.

Let $K = \max(M, |b|)$. We can now strictly bound our error:
$$|a_n b_n - ab| \leq K |b_n - b| + K |a_n - a|$$

Because $a_n \to a$ and $b_n \to b$, we can make $|a_n - a|$ and $|b_n - b|$ as small as we want.
Let $\epsilon > 0$ be given.

1.  Choose $N_1$ such that for all $n \geq N_1$, $|a_n - a| < \frac{\epsilon}{2K}$.
2.  Choose $N_2$ such that for all $n \geq N_2$, $|b_n - b| < \frac{\epsilon}{2K}$.

Let $N = \max(N_1, N_2)$. For all $n \geq N$, both conditions hold. Substituting these into our inequality gives:
$$|a_n b_n - ab| < K \left( \frac{\epsilon}{2K} \right) + K \left( \frac{\epsilon}{2K} \right) = \frac{\epsilon}{2} + \frac{\epsilon}{2} = \epsilon$$

Thus, $\lim_{n \to \infty} (a_n b_n) = a b$. $\square$

-----

## 3\. Proof of the Quotient Limit (Property 5)

To prove the quotient rule, it is much easier to first prove a lemma: if $b_n \to b$ (where $b \neq 0$), then $\frac{1}{b_n} \to \frac{1}{b}$. Once we prove this, we can simply apply the Product Limit theorem to $a_n \cdot \frac{1}{b_n}$.

**Lemma: Limit of a Reciprocal**

We want to bound the following expression:
$$\left| \frac{1}{b_n} - \frac{1}{b} \right| = \left| \frac{b - b_n}{b_n b} \right| = \frac{|b_n - b|}{|b_n| |b|}$$

The danger here is the $|b_n|$ in the denominator. If $b_n$ gets arbitrarily close to $0$, the fraction explodes. We must bound $|b_n|$ *away* from $0$.

Because $b_n \to b \neq 0$, the sequence eventually gets very close to $b$. Specifically, let's pick an error margin of $\frac{|b|}{2}$.
There exists an integer $N_1$ such that for all $n \geq N_1$, $|b_n - b| < \frac{|b|}{2}$.
By the [reverse triangle inequality](triangle-inequality.ipynb), this guarantees that for all $n \geq N_1$:
$$|b_n| > \frac{|b|}{2}$$

Now we can replace $|b_n|$ in our denominator with $\frac{|b|}{2}$ to get a strict upper bound on our error expression:
$$\frac{|b_n - b|}{|b_n| |b|} \leq \frac{|b_n - b|}{\left(\frac{|b|}{2}\right) |b|} = \frac{2}{|b|^2} |b_n - b|$$

Now, let $\epsilon > 0$ be given. Because $b_n \to b$, there exists an integer $N_2$ such that for all $n \geq N_2$:
$$|b_n - b| < \frac{\epsilon |b|^2}{2}$$

Let $N = \max(N_1, N_2)$. For all $n \geq N$, our bounding logic holds, yielding:
$$\left| \frac{1}{b_n} - \frac{1}{b} \right| \leq \frac{2}{|b|^2} |b_n - b| < \frac{2}{|b|^2} \left( \frac{\epsilon |b|^2}{2} \right) = \epsilon$$

Thus, $\lim_{n \to \infty} \frac{1}{b_n} = \frac{1}{b}$.

**Final Step:**

By rewriting $\frac{a_n}{b_n}$ as $a_n \cdot \frac{1}{b_n}$, we can apply the Product Limit (Property 4) we just proved:
$$\lim_{n \to \infty} \left( \frac{a_n}{b_n} \right) = \lim_{n \to \infty} \left( a_n \cdot \frac{1}{b_n} \right) = \left( \lim_{n \to \infty} a_n \right) \cdot \left( \lim_{n \to \infty} \frac{1}{b_n} \right) = a \cdot \frac{1}{b} = \frac{a}{b}$$
$\square$
