# The Monty Hall Problem

## 1. Problem Statement and Assumptions

You are on a game show and given the choice of three doors. Behind one door is a car; behind the others, goats. You pick a door (say, Door 1). The host, Monty Hall, examines the other doors, opens one of them (say, Door 2), reveals a goat, and says to you: *"Do you want to switch your pick to Door 3?"*

**Question: Should you switch?**

To solve this rigorously, we must rely on these strict assumptions about the game:

* There is exactly $1$ car and $2$ goats.
* The contestant initially has no idea where the car is (all doors are equally likely, $P = 1/3$).
* Monty **knows** exactly which door has the car.
* Monty **always** opens a goat door after you make your initial pick.
* Monty **always** gives you the option to switch.
* If Monty has a choice of two goat doors to open (i.e., you picked the car on your first try), he chooses between them with equal probability ($1/2$). He is not "lazy" and does not inherently favor the left or right door.

---

## 2. The Crux: The Power of Targeted Information

A common mathematical stumbling block in this problem is failing to distinguish between two seemingly identical pieces of information:

1. "Door 2 has a goat."
2. "Monty *opened* Door 2 to reveal a goat."

Why does the *fact that Monty opened it* change the mathematics so drastically?

::: {.callout-important}
## Random Chance vs. Deliberate Choice
Imagine a random gust of wind blew Door 2 open and revealed a goat. In that scenario, the probability of the car being behind Door 1 or Door 3 would genuinely be $50/50$. 

But Monty is **not** a gust of wind; he is a restricted agent. His choice of which door to open is entirely dependent on where the car actually is.

* If the car is behind Door 3, Monty is **forced** to open Door 2. 
* If the car is behind Door 1, Monty has a free choice.

Therefore, the event we are conditioning on is not simply "Door 2 is a goat." The event is "Monty deliberately avoided the car and selected Door 2." This targeted filtering of information by an all-knowing agent is what breaks the $50/50$ illusion.
:::

---

## 3. Four Mathematical Proofs

### Solution 1: Conditional Probability (Bayes' Rule)
Let us prove this using strict conditional probability. 
Let $C_i$ be the event "The car is behind Door $i$". Initially, $P(C_i) = 1/3$.
Assume you pick **Door 1**. 
Let $M_2$ be the event "Monty opens Door 2". 

We want to find the probability the car is behind Door 3, given Monty opened Door 2: $P(C_3 | M_2)$.

First, let's establish Monty's conditional probabilities (how he behaves depending on where the car is):

* $P(M_2 | C_1) = 1/2$ (If you picked the car, he randomly picks Door 2 or 3).
* $P(M_2 | C_2) = 0$ (He cannot reveal the car).
* $P(M_2 | C_3) = 1$ (He is **forced** to open Door 2 to protect the car behind Door 3).

::: {.callout-note}
## Symmetry and "Loss of Generality"
You might wonder if assuming we picked Door 1 and Monty opened Door 2 narrows the problem too much. In mathematics, it is called **Symmetry**. 

Because all three doors are initially identical (the car is placed randomly), the label "Door 1" is arbitrary. Whatever logic applies to Door 1 applies to Door 2 or 3 as well. Furthermore, once we have picked a door, the remaining two doors are also symmetric. 

By calculating the case for $P(C_3 | M_2)$, we are finding the representative probability for *any* door we might have picked and *any* door Monty might have opened. The math is invariant under relabeling the doors.
:::

Using the [Law of Total Probability](law-of-total-probability.ipynb), the total probability Monty opens Door 2 is:

$$P(M_2) = P(M_2 | C_1)P(C_1) + P(M_2 | C_2)P(C_2) + P(M_2 | C_3)P(C_3)$$
$$P(M_2) = \left(\frac{1}{2}\right)\left(\frac{1}{3}\right) + (0)\left(\frac{1}{3}\right) + (1)\left(\frac{1}{3}\right) = \frac{1}{6} + \frac{1}{3} = \frac{1}{2}$$

Now, apply [Bayes' Theorem](conditional-probability.ipynb):
$$P(C_3 | M_2) = \frac{P(M_2 | C_3)P(C_3)}{P(M_2)} = \frac{(1)(1/3)}{1/2} = \frac{2}{3}$$

By switching, your chance of winning doubles to **$2/3$**.

### Solution 2: Law of Total Probability (Wishful Thinking)
Let's analyze the absolute **Always Switch** strategy using the Law of Total Probability.
Let $S$ be the event "You succeed (win the car) by switching".
Let $D_j$ be the event "The car is behind Door $j$".

Assume you initially pick Door 1. We condition our success on where the car actually is:
$$P(S) = P(S | D_1)P(D_1) + P(S | D_2)P(D_2) + P(S | D_3)P(D_3)$$

* If the car is at $D_1$, Monty opens 2 or 3. You switch away from the car. $P(S | D_1) = 0$.
* If the car is at $D_2$, Monty is forced to open 3. You switch to 2 and win. $P(S | D_2) = 1$.
* If the car is at $D_3$, Monty is forced to open 2. You switch to 3 and win. $P(S | D_3) = 1$.

$$P(S) = (0)\left(\frac{1}{3}\right) + (1)\left(\frac{1}{3}\right) + (1)\left(\frac{1}{3}\right) = \frac{2}{3}$$

### Solution 3: The Complement Logic (Intuitive)
When you first pick Door 1, you have a $1/3$ chance of being right.
This means there is a **$2/3$ chance the car is somewhere else** (behind Door 2 or Door 3).

When Monty opens Door 2 to reveal a goat, he does not change the fact that your initial guess had a $1/3$ chance of being right. The "group" of Door 2 and Door 3 still holds a collective $2/3$ probability. But since Door 2 is eliminated, that entire $2/3$ probability collapses strictly onto Door 3. 

### Solution 4: The Extreme Case (1 Million Doors)
If human psychology still makes the $50/50$ split feel intuitive, scale the problem up asymptotically.

1.  Imagine there are **1,000,000 doors**. 
2.  You pick Door 1. Your chance of being right is literally one in a million.
3.  Monty, who knows where the car is, opens **999,998** other doors, all revealing goats. 
4.  Only Door 1 and Door 743,992 are left closed.

Did your random guess at Door 1 magically become a $50/50$ coin flip? **No.** You are still almost certainly wrong. Monty *had* to leave Door 743,992 closed because it has the car. You switch every single time.

## 4. Python Simulation

We can empirically prove that switching yields a $\approx 66.6\%$ win rate by simulating the game via Monte Carlo methods.

```python
import numpy as np

def simulate_monty_hall(trials=10000):
    switch_wins = 0
    stay_wins = 0
    
    for _ in range(trials):
        # 0, 1, 2 represent the doors
        doors = [0, 1, 2]
        
        car_door = np.random.choice(doors)
        player_choice = np.random.choice(doors)
        
        # Monty must pick a door that is NOT the car and NOT the player's choice
        monty_options = [d for d in doors if d != car_door and d != player_choice]
        monty_choice = np.random.choice(monty_options)
        
        # The door the player would switch to
        switch_choice = [d for d in doors if d != player_choice and d != monty_choice][0]
        
        # Tally wins
        if player_choice == car_door:
            stay_wins += 1
        if switch_choice == car_door:
            switch_wins += 1
            
    print(f"Win percentage if STAYING: {stay_wins / trials * 100:.2f}%")
    print(f"Win percentage if SWITCHING: {switch_wins / trials * 100:.2f}%")

simulate_monty_hall(10000)
```
