# Composite Sequences Limit
 
Suppose for [sequences](sequence.ipynb) that $\{a_n\}$ converges to $a$ and $\{b_n\}$ converges to $b$. Then
 1. the sequence $\{ca_n\}$ converges to $ca$, where $c$ is any constant;
 2. the sequence $\{a_n+b_n\}$ converges to $a + b$;
 3. the sequence $\{a_n-b_n\}$ converges to $a - b$;
 4. the sequence $\{a_nb_n\}$ converges to $ab$;
 5. the sequence $\{\frac{a_n}{b_n}\}$ converges to $a/b$, provided that $b_n\neq 0$ for all $n$ and $b\neq 0$.
