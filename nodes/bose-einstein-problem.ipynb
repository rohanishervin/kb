# Bose-Einstein Counting (Combinations with Replacement)

## 1\. Problem Statement

The **Bose-Einstein Counting Problem** asks us to determine the total number of distinct multi-sets of size $k$ that can be formed from $n$ distinct categories of elements.

This is essentially [sampling with replacement](sampling-with-replacement.ipynb), but with a critical twist: **the order of selection is completely irrelevant**. Only the final tally of items in each category matters.

*(Note: In combinatorics, this is frequently called "combinations with replacement" or "multichoose," and is sometimes denoted by the symbol $\left(\!\binom{n}{k}\!\right)$).*

-----

## 2\. Counter-Example: Why Naive Counting Fails

A common mistake is assuming we can just calculate the ordered permutations $n^k$ and divide by $k!$ using the principle of [adjusting for overcounting](adjusting-for-overcounting.ipynb). This **fails entirely** because identical items do not possess symmetrical $k!$ arrangements.

**The Setup:** Suppose we want to choose $k=2$ donuts from $n=2$ flavors (Chocolate and Vanilla).

  * **The reality:** There are exactly $3$ possible valid orders: (Choc, Choc), (Choc, Van), and (Van, Van).
  * **Naive ordered count:** $2^2 = 4$ sequences (CC, CV, VC, VV).
  * **Naive adjustment:** If we try to divide by $k!$ ($2! = 2$) to remove order, we get $4 / 2 = 2$ outcomes.

Because $3 \neq 2$, the naive math breaks. The formula fails because the sequence "CC" only appears once and cannot be "un-ordered", whereas "CV" and "VC" appear twice. We need a fundamentally different mathematical approach to count these uniformly.

-----

## 3\. Derivation: The "Stars and Bars" Method

To solve this mathematically, we map the problem to a visual system known as **Stars and Bars** (originally popularized by William Feller).

Imagine we want to distribute $k$ indistinguishable items into $n$ distinct categories.

  * We represent the $k$ items as indistinguishable stars: $\star$
  * We represent the boundaries between our $n$ distinct categories as vertical bars: $|$

To separate a space into $n$ distinct categories (bins), we only need exactly $n - 1$ interior bars.

Therefore, *any* valid selection of $k$ items from $n$ categories can be represented as a linear sequence of stars and bars.
For example, if we buy $k=5$ donuts from $n=3$ flavors (A, B, C), the sequence `** | * | **` means exactly:

  * 2 of flavor A
  * 1 of flavor B
  * 2 of flavor C

**The Math:**
Every unique arrangement of this sequence corresponds to a unique combination of items.

  * We have $k$ stars.
  * We have $n - 1$ bars.
  * The total length of our sequence string is $(n + k - 1)$ positions.

We simply need to choose which of the $(n + k - 1)$ positions will hold the $k$ stars. The remaining positions will automatically be bars. Because the stars are identical, this is a standard [binomial coefficient](binomial-coefficient.ipynb) problem:

$$\binom{n+k-1}{k}$$

*(By symmetry, we could also choose the positions for the bars instead, yielding the identical result $\binom{n+k-1}{n-1}$.)*

-----

## 4\. Algebraic Equivalence (Integer Solutions)

The Bose-Einstein counting problem is mathematically identical to finding the number of non-negative integer solutions to a linear equation.

Consider the equation:
$$x_1 + x_2 + \dots + x_n = k$$

Where:

  * $k$ is the total number of items to distribute.
  * $n$ is the number of distinct categories (variables).
  * Each $x_i$ represents the number of items assigned to category $i$ (with the strict constraint that $x_i \geq 0$ and $x_i \in \mathbb{Z}$).

**The Mapping:**
Using the Stars and Bars method, we can perfectly map this algebraic equation to our physical sequence. The $k$ items are the stars. The $n-1$ plus signs ($+$) separating the variables act exactly as our separating bars\!

For example, if we need to solve $x_1 + x_2 + x_3 = 5$, a valid algebraic solution is $x_1=2, x_2=0, x_3=3$.
In Stars and Bars notation, this is exactly: `** | | ***`.

Because the structural mapping is perfectly identical, the total number of non-negative integer solutions to this equation is exactly:

$$\binom{n+k-1}{k}$$

-----

## 5\. Origin of the Name (Physics Context)

Why is a combinatorial method named after Albert Einstein and Satyendra Nath Bose?

In quantum statistical mechanics, particles like photons and gluons are known as **bosons**. Unlike macroscopic objects, bosons of the same species are fundamentally indistinguishable from one another. If you have $k$ photons and $n$ distinct energy states they can occupy, there is absolutely no physical difference between swapping Photon A with Photon B.

Therefore, to calculate the probability of a specific energy configuration, physicists must count the total number of possible macroscopic states. Distributing $k$ indistinguishable bosons into $n$ distinct energy levels requires this exact formula.

## 6\. Practical Example

**Question:** A coffee shop sells $4$ different types of pastries. You want to buy a box of $6$ pastries for your office. How many different box combinations can you create?

**Solution:**

  * $n = 4$ (distinct pastry categories)
  * $k = 6$ (indistinguishable slots in your box to fill)

Apply the Bose-Einstein formula:
$$\binom{n+k-1}{k} = \binom{4+6-1}{6} = \binom{9}{6}$$

Using the [binomial-coefficient-formula](binomial-coefficient-formula.ipynb):
$$\binom{9}{6} = \frac{9!}{6!(9-6)!} = \frac{9!}{6! \cdot 3!} = \frac{9 \times 8 \times 7}{3 \times 2 \times 1} = \frac{504}{6} = 84$$

There are exactly $84$ distinct ways to fill your pastry box.
