# Limit

if $f$ from $I$ to [$\mathbb{R}$](real-numbers.ipynb) is a [real-valued](real-valued-functions.ipynb) function which is defined around $a$, then:

$$
\forall \epsilon > 0,\quad \exists \delta > 0, \quad \forall x\in I: \quad 0 < |x - a| < \delta \Rightarrow |f(x) - L| < \epsilon
$$

It's easy to show it's unique so we can say:

$$
\lim_{x \to a} f(x) = L 
$$
