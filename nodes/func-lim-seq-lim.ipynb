# Limit of a [Function](limit.ipynb) \ Limit of a [sequence](sequence-limit.ipynb)

## Part (a)

Let $I$ be an open interval that contains the point $c$, and suppose that $f$ is a function that is defined on $I$ except possibly at the point $c$. 

The function $f$ has limit $L$ at $c$ if and only if for each sequence $\{x_n\}$ in $I \setminus \{c\}$ that converges to $c$, the sequence $\{f(x_n)\}$ converges to $L$.

## Part (b)

Suppose that there are sequences $\{x_n\}$ and $\{y_n\}$ in $I \setminus \{c\}$ that converge to $c$ such that:

* $\{f(x_n)\}$ converges to $L_1$
* $\{f(y_n)\}$ converges to $L_2$

If $L_1 \neq L_2$, then the function $f$ does not have a limit at $c$.
