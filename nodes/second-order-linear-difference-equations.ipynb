To solve complex systems, we must understand how to solve second-order, linear, homogeneous [difference equations](intro-to-difference-equations.ipynb) with constant coefficients.

The general form is:
$$x(n+2) + p_1 x(n+1) + p_2 x(n) = 0$$

## The Characteristic Equation Method

To solve this, we guess that the solution has the form $x(n) = \lambda^n$ for some constant $\lambda$. Substituting this guess into the equation gives:
$$\lambda^{n+2} + p_1 \lambda^{n+1} + p_2 \lambda^n = 0$$

Dividing through by $\lambda^n$ yields the **Characteristic Equation**:
$$\lambda^2 + p_1 \lambda + p_2 = 0$$

This is just a quadratic equation\! We find its roots, $\lambda_1$ and $\lambda_2$. There are two main cases we care about for probability problems:

### Case 1: Distinct Real Roots ($\lambda_1 \neq \lambda_2$)

If the roots are different, the general solution is a linear combination of the two:
$$x(n) = c_1 \lambda_1^n + c_2 \lambda_2^n$$
where $c_1$ and $c_2$ are constants determined by the initial conditions (boundary values).

### Case 2: Repeated Real Roots ($\lambda_1 = \lambda_2 = \lambda$)

If the characteristic equation has a double root, the first solution is $\lambda^n$. The second linearly independent solution is $n \lambda^n$. The general solution is:
$$x(n) = c_1 \lambda^n + c_2 n \lambda^n = (c_1 + c_2 n) \lambda^n$$
