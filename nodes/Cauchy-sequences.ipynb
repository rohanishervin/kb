# Cauchy Sequences and Completeness

## 1. Context: The Completeness of $\mathbb{R}$

In previous notes, we built the real numbers $\mathbb{R}$ using [**Dedekind cuts**](def-dedekind-cuts.ipynb) and proved that this construction guarantees the [**Least Upper Bound (LUB) Property**](axiom-of-completeness.ipynb). Therefore, we already know $\mathbb{R}$ is complete in the order-theoretic sense. 

Now, we approach completeness from a metric perspective. We want to show that $\mathbb{R}$ has no "gaps" when it comes to sequences converging. To do this, we formalize what it means for the elements of a sequence to get "closer together" independently of a known limit.

## 2. The Cauchy Condition

**Definition:** A sequence $(a_n)$ is a **Cauchy sequence** if, for every arbitrarily small distance $\epsilon > 0$, there exists a point in the sequence $N$ after which *any two arbitrary terms* are within $\epsilon$ of each other:
$$\forall \epsilon > 0, \exists N \in \mathbb{N} \text{ s.t. } \forall j, k \geq N \implies |a_j - a_k| < \epsilon$$

::: {.callout-warning}
## Danger: "Consecutive" Closeness is Not Enough!
Notice we use $j$ and $k$ to emphasize that we are picking *any* two indices past $N$, not just consecutive ones. A sequence where consecutive terms get infinitely close (i.e., $|a_{n+1} - a_n| \to 0$) is **not** necessarily Cauchy.

**Counterexample: The Harmonic Series**

Let $a_n = \sum_{i=1}^n \frac{1}{i} = 1 + \frac{1}{2} + \frac{1}{3} + \dots + \frac{1}{n}$.

The distance between consecutive terms is just the next term added:

$$|a_{n+1} - a_n| = \frac{1}{n+1}$$

As $n \to \infty$, this distance goes to exactly $0$. However, this sequence is **not** Cauchy and diverges to infinity. To see why it fails the Cauchy condition, pick $k = 2n$ and $j = n$. The difference is:
$$|a_{2n} - a_n| = \frac{1}{n+1} + \frac{1}{n+2} + \dots + \frac{1}{2n}$$
There are exactly $n$ terms in this sum, and the smallest one is $\frac{1}{2n}$. Therefore:
$$|a_{2n} - a_n| \geq n \times \left(\frac{1}{2n}\right) = \frac{1}{2}$$
No matter how far out we go in the sequence, we can always find two terms that are at least $\frac{1}{2}$ apart. Thus, it fails the Cauchy condition for $\epsilon = \frac{1}{2}$.
:::

## 3. Preliminary Lemmas

To prove our main theorem linking Cauchy sequences to our previously established LUB property, we need two foundational lemmas.

**Lemma 1: Every Cauchy sequence is bounded.**

**Proof:**
Let $(a_n)$ be a Cauchy sequence. By the Cauchy condition, if we choose $\epsilon = 1$, there exists an $N \in \mathbb{N}$ such that for all $j, k \geq N$, $|a_j - a_k| < 1$. 
If we fix $k = N$, we have $|a_j - a_N| < 1$ for all $j \geq N$. This implies that $|a_j| < |a_N| + 1$.
To find a bound for the *entire* sequence, we just need to account for the finite number of terms before $N$. We set $M$ to be the maximum of those early terms and our new bound:
$$M = \max(|a_1|, |a_2|, \dots, |a_{N-1}|, |a_N| + 1)$$
Thus, $\forall n \in \mathbb{N}, -M \leq a_n \leq M$. $\square$

**Lemma 2 (Monotone Convergence): An increasing sequence converges if it is bounded above.**

**Proof:**
Let $(a_n)$ be an increasing sequence bounded above by some $M$. Because the set of its elements $S = \{a_n\}$ is non-empty and bounded above, the Least Upper Bound Property (which we have via Dedekind cuts) guarantees that a supremum exists in $\mathbb{R}$. Let $b = \sup(a_n)$.
By the definition of a supremum, for any $\epsilon > 0$, $b - \epsilon$ is *not* an upper bound. Therefore, there exists some $N$ such that $b - \epsilon < a_N$.
Because the sequence is increasing, for all $n \geq N$, $a_N \leq a_n$. Furthermore, because $b$ is the upper bound, $a_n \leq b$.
Chaining these inequalities gives:
$$b - \epsilon < a_N \leq a_n \leq b < b + \epsilon$$
This means $|a_n - b| < \epsilon$ for all $n \geq N$. Therefore, $\lim_{n \to \infty} a_n = b$. $\square$

## 4. Theorem: Cauchy Completeness of $\mathbb{R}$

**Theorem:** $\mathbb{R}$ is complete with respect to Cauchy sequences. A sequence $(a_n)$ in $\mathbb{R}$ is convergent if and only if it is a Cauchy sequence.

**Proof (Cauchy $\implies$ Convergent):**

Suppose $(a_n)$ is a Cauchy sequence. By **Lemma 1**, it is bounded by some $M$. 
We define a new sequence $(b_n)$ designed to track the "greatest lower bounds" of the tail ends of $(a_n)$:
$$b_n = \inf \{a_n, a_{n+1}, a_{n+2}, \dots\}$$

Because the set $\{a_i\}_{i \geq n}$ contains $\{a_i\}_{i \geq n+1}$, taking the infimum of the smaller set can only result in a number that is greater than or equal to the previous infimum. Thus, $b_1 \leq b_2 \leq b_3 \dots$, meaning $(b_n)$ is an **increasing** sequence. 
By definition, $b_n \leq a_n \leq M$, so $(b_n)$ is bounded above.
By **Lemma 2**, $(b_n)$ converges to some limit $b \in \mathbb{R}$.

We claim that $\lim_{n \to \infty} a_n = b$. 
Let $\epsilon > 0$. We will use the triangle inequality: $|a_n - b| \leq |a_n - b_n| + |b_n - b|$. We must bound both halves by $\frac{\epsilon}{2}$.

1.  **Bounding $|b_n - b|$:** Since $b_n \to b$, there exists $N_1$ such that for $n \geq N_1$, $|b_n - b| < \frac{\epsilon}{2}$.
2.  **Bounding $|a_n - b_n|$:** Since $a_n$ is Cauchy, there exists $N_2$ such that for $j, k \geq N_2$, $|a_j - a_k| < \frac{\epsilon}{2}$. 
    This means $a_j - \frac{\epsilon}{2} < a_k < a_j + \frac{\epsilon}{2}$. 
    Because $a_k > a_j - \frac{\epsilon}{2}$ for all $k \geq j$, the value $a_j - \frac{\epsilon}{2}$ acts as a lower bound for the tail set $\{a_j, a_{j+1}, \dots\}$.
    Since $b_j$ is the *greatest* lower bound (infimum) of that set, it must be that:
    $$a_j - \frac{\epsilon}{2} \leq b_j \leq a_j$$
    Which implies $|a_j - b_j| \leq \frac{\epsilon}{2}$.

Finally, let $N = \max(N_1, N_2)$. For all $n \geq N$:
$$|a_n - b| \leq |a_n - b_n| + |b_n - b| < \frac{\epsilon}{2} + \frac{\epsilon}{2} = \epsilon$$
Therefore, $(a_n)$ converges to $b \in \mathbb{R}$. $\square$
