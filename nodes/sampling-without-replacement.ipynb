# Sampling Without Replacement

## 1\. Concept and Definition

**Sampling without replacement** occurs when you select $k$ items sequentially from a collection of $n$ distinct objects, and **each chosen item is permanently removed from the pool**.

Because the items are drawn one after another, the **order of selection strictly matters**.

*(Example: Drawing the sequence "Alice, Bob" is considered a distinct outcome from drawing "Bob, Alice").*

-----

## 2\. The Formula and Derivation

The total number of possible ordered sequences of length $k$ drawn from a population of $n$ items without replacement is mathematically known as a **Permutation**, denoted as $P(n, k)$.

$$P(n, k) = n(n - 1)(n - 2) \dots (n - k + 1) = \frac{n!}{(n-k)!}$$

**Proof by the Multiplication Rule:**
This formula is a direct application of the [multiplication-rule](multiplication-rule.ipynb). We view the drawing process as a sequence of $k$ distinct stages:

  * **Stage 1:** The pool is full. We have exactly $n$ choices.
  * **Stage 2:** One item has been removed. We have $n - 1$ choices.
  * **Stage 3:** Two items have been removed. We have $n - 2$ choices.
  * $\dots$
  * **Stage $k$:** $k-1$ items have been removed. We have $n - (k - 1) = n - k + 1$ choices.

Multiplying these stages together yields the expanded formula.

*(Note: If $k > n$, it is physically impossible to draw more unique items than exist in the set. The formula naturally breaks down, and the number of outcomes is strictly $0$.)*

-----

## 3\. Key Property: Statistical Dependence

The defining mathematical feature of sampling *without* replacement is **Dependence**.

Because the chosen item is not returned to the pool, every single draw fundamentally alters the composition of the population. The outcome of the first draw changes both the total number of items remaining (the denominator) and the specific items available (the numerator). Therefore, the probabilities of all subsequent draws are heavily dependent on what happened previously.

This is the exact opposite of [sampling-with-replacement](sampling-with-replacement.ipynb), where probabilities remain statically independent.

-----

## 4\. Classic Example

**Question:** A local chess club has $10$ members. They need to elect a President, a Vice President, and a Treasurer. No person can hold more than one office. How many distinct ways can these officers be chosen?

**Solution:**
Because the roles are distinct, the order of selection matters (President Alice and VP Bob is different from President Bob and VP Alice). Because a person cannot hold two roles, this is sampling *without* replacement.

  * $n = 10$ (Total members)
  * $k = 3$ (Roles to fill)

$$P(10, 3) = 10 \times 9 \times 8 = 720$$

Alternatively, using factorials:
$$P(10, 3) = \frac{10!}{(10-3)!} = \frac{10!}{7!} = 10 \times 9 \times 8 = 720$$

There are exactly $720$ distinct ways to elect the officers.
