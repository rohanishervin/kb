# Ordinary Differential Equations and Picard's Theorem

The qualitative theory of ordinary differential equations (ODEs) fundamentally rests on establishing that solutions actually exist and are unique. This is achieved via **Picard's Theorem**, which elegantly applies the [Banach Contraction Principle](banach-fp-thm.ipynb) to a space of functions.

## 1\. ODEs as Vector Fields

Let $U$ be an open subset of $\mathbb{R}^m$. A system of $m$ simultaneous scalar differential equations can be expressed geometrically as a vector field.

We combine $m$ real-valued functions $f_i$ into a single [vector field]() mapping $F: U \to \mathbb{R}^m$ given by:
$$
F(x) = (f_1(x), f_2(x), \dots, f_m(x))
$$
where $x = (x_1, \dots, x_m)$. 

To solve the ODE, we seek a **trajectory** of $F$. A trajectory is a differentiable curve $\gamma: (a, b) \to U$ (with $a < 0 < b$) describing the path of a particle whose velocity at any given position is prescribed by the vector field $F$. 

Formally, the curve must satisfy the initial value problem:
$$
\begin{aligned}
\gamma'(t) &= F(\gamma(t)) \quad \text{for all } t \in (a, b) \\
\gamma(0) &= p
\end{aligned}
$$

::: {.callout-note}
## Intuition: Following the Wind
Think of $U$ as a weather map, and $F(x)$ assigns a wind velocity vector to every point $x$. An ODE trajectory $\gamma(t)$ is simply the path traced over time by a lightweight leaf dropped at an initial point $p$ at time $t=0$, moving exactly as the wind dictates at each point it visits.
:::

## 2\. The Integral Equation Formulation

Finding a differentiable curve $\gamma(t)$ that solves the differential equation is mathematically equivalent to finding a continuous curve that solves a corresponding integral equation.

By integrating the derivative $\gamma'(t)$ from time $0$ to $t$, and using the [Fundamental Theorem of Calculus]() for vector-valued functions, we obtain:
$$
\gamma(t) = p + \int_0^t F(\gamma(s)) \, ds
$$
Because the vector field $F$ is continuous, any continuous solution to this integral equation is automatically differentiable, and its derivative matches the ODE. We have shifted the problem from "finding a derivative" to "finding a fixed function under an integral operator."

## 3\. Picard's Theorem

Before stating the theorem, we need a condition slightly stronger than continuity. 

**Definition (Lipschitz Condition):**
A vector field $F$ satisfies a [Lipschitz condition]() on $U$ if there exists a constant $L$ such that for all $x, y \in U$:
$$
|F(x) - F(y)| \le L |x - y|
$$
where $|\cdot|$ denotes the Euclidean norm in $\mathbb{R}^m$. (Note: Any differentiable vector field with bounded derivatives is Lipschitz.)

**Theorem (Picard's Existence and Uniqueness Theorem):**
Given a point $p \in U$ and a vector field $F$ that satisfies a Lipschitz condition, there exists a locally unique $F$-trajectory $\gamma(t)$ in $U$ passing through $p$. 

::: {.callout-tip collapse="true"}
## Proof of Picard's Theorem

**Step 1: Setting up the function space**
Because $F$ is continuous, we can find a compact neighborhood $N \subset U$ around $p$ and a constant $M$ such that $|F(x)| \le M$ for all $x \in N$.
We want to define an interval $[-\tau, \tau]$ for time $t$. Choose $\tau > 0$ small enough such that two conditions hold:

1. $\tau M \le r$ (where $r$ is the radius of the neighborhood $N$)
2. $\tau L < 1$ (where $L$ is the Lipschitz constant of $F$)

Let $\mathcal{C}$ be the space of all continuous functions $\gamma: [-\tau, \tau] \to N$. Equipped with the [supremum metric](function-spaces.ipynb#the-metric-space-of-bounded-functions-c_b) $d(\gamma, \sigma) = \sup_{t} |\gamma(t) - \sigma(t)|$, $\mathcal{C}$ forms a complete metric space.

**Step 2: Defining the operator**
Define an integral operator $\Phi: \mathcal{C} \to \mathcal{C}$ by:
$$
\Phi(\gamma)(t) = p + \int_0^t F(\gamma(s)) \, ds
$$
Solving our ODE is equivalent to finding a curve $\gamma$ such that $\Phi(\gamma) = \gamma$. That is, we need a fixed point for $\Phi$.

**Step 3: Proving $\Phi$ maps $\mathcal{C}$ to itself**
We must ensure that applying $\Phi$ doesn't map a curve outside of our neighborhood $N$. For any $t \in [-\tau, \tau]$:
$$
|\Phi(\gamma)(t) - p| = \left| \int_0^t F(\gamma(s)) \, ds \right| \le \tau M \le r
$$
Thus, $\Phi(\gamma)$ remains within $N_r(p)$, so $\Phi$ is a valid map from $\mathcal{C}$ to $\mathcal{C}$.

**Step 4: Proving $\Phi$ is a contraction**
We evaluate the distance between $\Phi(\gamma)$ and $\Phi(\sigma)$ for any two curves in $\mathcal{C}$:
$$
\begin{aligned}
d(\Phi(\gamma), \Phi(\sigma)) &= \sup_{t} \left| \int_0^t F(\gamma(s)) - F(\sigma(s)) \, ds \right| \\
&\le \tau \sup_{s} |F(\gamma(s)) - F(\sigma(s))|
\end{aligned}
$$
Using the Lipschitz condition on $F$:
$$
\begin{aligned}
d(\Phi(\gamma), \Phi(\sigma)) &\le \tau \sup_{s} L |\gamma(s) - \sigma(s)| \\
&= \tau L \, d(\gamma, \sigma)
\end{aligned}
$$
Since we explicitly chose $\tau$ such that $\tau L < 1$, the operator $\Phi$ is a contraction on the complete metric space $\mathcal{C}$.

**Step 5: Applying Banach**
By the Banach Contraction Principle, $\Phi$ has a unique fixed point $\gamma \in \mathcal{C}$. Because $\Phi(\gamma) = \gamma$, $\gamma(t)$ is a continuous solution to the integral equation, and therefore a differentiable solution to the ODE. Local uniqueness is guaranteed by the uniqueness of the fixed point. $\blacksquare$
:::

## 4\. Flows

If we assume trajectories never escape $U$ (for instance, if $U = \mathbb{R}^m$), we can define trajectories for all time $t \in \mathbb{R}$. 

Let $\gamma(t, p)$ denote the trajectory passing through $p$ at time $t=0$. Imagine all points in the space moving simultaneously along their trajectories. A point $p$ travels to $p_1 = \gamma(t_1, p)$ over time $t_1$. Its continued journey from $p_1$ for an additional time $t$ is exactly the trajectory starting from $p_1$. This gives rise to the **flow equation**:
$$
\gamma(t, p_1) = \gamma(t + t_1, p)
$$
This is a direct consequence of the local uniqueness of trajectories established by Picard's theorem.

By defining the **$t$-advance map** $\varphi_t(p) = \gamma(t, p)$, we can rewrite the flow equation elegantly as:
$$
\varphi_{t+s}(p) = \varphi_t(\varphi_s(p))
$$
This demonstrates that the mapping $t \mapsto \varphi_t$ acts as a [group homomorphism]() from the real numbers $(\mathbb{R}, +)$ into the group of homeomorphisms of $\mathbb{R}^m$, where the time-zero map $\varphi_0$ is simply the identity map.
