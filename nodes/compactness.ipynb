# Compactness

Compactness is widely considered the single most important concept in real analysis. Its primary mathematical power lies in reducing the infinite to the finite.

**Definition:** A subset $A$ of a [metric space](metric-spaces.ipynb) $M$ is (sequentially) **compact** if every sequence $(a_n)$ entirely contained in $A$ has a [subsequence](sequences-metric-spaces.ipynb#subsequences) $(a_{n_k})$ that converges to a limit point located inside $A$.

The empty set and finite sets are trivial examples of compact sets. For any sequence $(a_n)$ trapped within a finite set, at least one term must repeat infinitely often, meaning the sequence contains a constant subsequence that trivially converges.

---

## 1. Necessary Conditions for Compactness

Compactness is a highly desirable "good feature" of a set, but before proving a set is compact, we usually check two mandatory prerequisites. 

**Theorem:** Every compact set is both [closed](metric-space-topology.ipynb#limits-of-a-set-and-closed-sets) and bounded.

**Proof that a compact set is closed:**
Suppose $A$ is a compact subset of $M$, and $p$ is a limit point of $A$. 
Because $p$ is a limit, there exists a sequence $(a_n)$ in $A$ converging to $p$. 
By the definition of compactness, some subsequence $(a_{n_k})$ must converge to some limit $q \in A$. 
However, every subsequence of a convergent sequence must converge to the exact same limit as its "mother" sequence. 
Therefore, $q = p$, which means $p \in A$. Since $A$ contains all its limit points, $A$ is closed. $\square$

**Proof that a compact set is bounded:**
Choose and fix any arbitrary point $p \in M$. 
If $A$ is not bounded, then for every integer $n \in \mathbb{N}$, we can find a point $a_n \in A$ such that the distance $d(p, a_n) \geq n$. 
Because $A$ is compact, this sequence $(a_n)$ must possess a subsequence $(a_{n_k})$ that converges. 
All convergent sequences are mathematically bounded. 
However, this directly contradicts our construction where the distance $d(p, a_{n_k}) \to \infty$ as $k \to \infty$. 
Therefore, such an unbounded sequence $(a_n)$ cannot exist, meaning $A$ must be contained in some finite neighborhood $M_r p$. Thus, $A$ is bounded. $\square$

## 2. Compactness in Euclidean Space

We know compact sets must be closed and bounded. In Euclidean space ($\mathbb{R}^m$), the converse is also perfectly true. We build this up starting from 1D intervals.

**Theorem:** The closed interval $[a, b] \subset \mathbb{R}$ is compact.

**Proof:**
Let $(x_n)$ be an arbitrary sequence in $[a,b]$. We define a set $C$:
$$C = \{x \in [a,b] : x_n < x \text{ only finitely often}\}$$
Equivalently, for all but finitely many $n$, $x_n \geq x$. 
Since $a \in C$, we know $C$ is not empty ($C \neq \emptyset$). 
Clearly, $b$ acts as an upper bound for $C$. 
By the least upper bound (L.U.B.) property of the real numbers, there exists a supremum $c = \text{l.u.b.} C$, where $c \in [a,b]$. 
We claim a subsequence of $(x_n)$ converges to $c$. Suppose it does not. Then for some radius $r > 0$, the terms $x_n$ fall into the neighborhood $(c - r, c + r)$ only finitely often. 
This implies that $c + r \in C$, which contradicts the fact that $c$ is an upper bound for $C$. 
Hence, a subsequence must converge to $c$, proving $[a,b]$ is compact. $\square$

### 2.1. Products of Compact Sets
To step from $\mathbb{R}$ up to $\mathbb{R}^m$, we use Cartesian products.

**Theorem:** The Cartesian product of two compact sets is compact.

**Proof:** 
Let $(a_n, b_n) \in A \times B$ be a sequence where both $A$ and $B$ are compact. 
Because $A$ is compact, there exists a subsequence $(a_{n_k})$ that converges to a point $a \in A$. 
Now look at the corresponding $B$-components of that specific subsequence: $(b_{n_k})$. Because $B$ is compact, this sequence has a sub-subsequence $(b_{n_{k_\ell}})$ that converges to a point $b \in B$. 
Crucially, the sub-subsequence of the $A$-components, $(a_{n_{k_\ell}})$, continues to converge to the same point $a$. 
Thus, the joint sub-subsequence $(a_{n_{k_\ell}}, b_{n_{k_\ell}}) \to (a, b)$. This proves $A \times B$ is compact. $\square$

**Corollary:** By mathematical induction, the Cartesian product of any $m$ finite compact sets is compact.

**Corollary:** Every geometric "box" $[a_1, b_1] \times \dots \times [a_m, b_m]$ in $\mathbb{R}^m$ is compact. This is obvious based on the previous proofs.

### 2.2. The Bolzano-Weierstrass Theorem
An equivalent formulation of these geometric facts yields one of the most famous theorems in analysis.

**Theorem (Bolzano-Weierstrass):** Every bounded sequence in $\mathbb{R}^m$ has a convergent subsequence.

**Proof:** If a sequence is bounded, it can be entirely contained within a sufficiently large box. Because we proved boxes are compact, the sequence inside it must inherently possess a subsequence that converges to a limit inside that box. $\square$

## 3. The Heine-Borel Theorem

Before stating the final theorem, we need one simple fact:

**Theorem:** Every closed subset of a compact set is compact.

**Proof:** If $A$ is a closed subset of a compact set $K$, any sequence $(a_n)$ in $A$ is also a sequence in $K$. Because $K$ is compact, a subsequence converges to a limit $p \in K$. Because $A$ is closed, it contains all its limits, so $p \in A$, proving $A$ is compact. $\square$

**Theorem (Heine-Borel):** Every closed and bounded subset of $\mathbb{R}^m$ is compact.

**Proof:** 

Let $A \subset \mathbb{R}^m$ be closed and bounded. 
Because $A$ is bounded, it is contained within some box. 
We established that boxes are compact. 
Because $A$ is a closed subset trapped inside a compact box, the previous theorem implies $A$ itself must be compact. $\square$

---

## 4. Danger: When Closed and Bounded is NOT Compact

The Heine-Borel Theorem is a luxury of Euclidean space. It is absolutely vital to remember that in a general abstract metric space, a set can be both closed and bounded but completely fail to be compact.

**Counterexample 1: The Discrete Metric**
Let the space be the natural numbers $\mathbb{N}$ equipped with the discrete metric (where the distance between any two distinct points is exactly $1$). 
This space is a complete metric space. It is closed in itself, and it is strictly bounded (the maximum distance between any points is $1$). 
However, it is *not* compact. Take the sequence $(1, 2, 3, \dots)$. Because every point is distance $1$ apart, the sequence never bunches together; it is impossible to extract a convergent subsequence.

**Counterexample 2: Function Spaces**
A more striking example happens in $C([0,1], \mathbb{R})$, the metric space of continuous functions with the max metric $d(f, g) = \max\{|f(x) - g(x)|\}$.
This space is complete, but its closed unit ball is not compact. The sequence of functions $f_n = x^n$ exists inside this bounded ball, but it possesses no subsequence that converges with respect to the max metric. In infinite-dimensional spaces, every closed ball is noncompact.

## 5. Examples of Compact Sets
To build intuition, here are standard examples of compact sets:

1. Any finite subset of a metric space (e.g., the empty set).
2. Any closed subset of a compact set.
3. The union of finitely many compact sets.
4. The Cartesian product of finitely many compact sets.
5. The intersection of arbitrarily many compact sets.
6. The closed unit ball in $\mathbb{R}^3$.
7. The boundary of a compact set (like the unit 2-sphere in $\mathbb{R}^3$).
8. The set $\{x \in \mathbb{R} : \exists n \in \mathbb{N} \text{ and } x = 1/n\} \cup \{0\}$.
9. The Hawaiian earring (a specific topological space of shrinking tangent circles).
10. The Cantor set.
