# The Cantor Set

The Cantor set is one of the most fascinating and counter-intuitive constructs in real analysis. It is a set that is "maximally disconnected" (often called "Cantor dust"), yet it contains an uncountable number of points, takes up zero space, and holds the geometric key to creating space-filling curves.

## 1. Definition and Construction

The standard middle-thirds Cantor set is constructed through an infinite iterative process of removing segments.

1. Start with the closed unit interval $C^0 = [0, 1]$.
2. Remove the open middle third $(1/3, 2/3)$. What remains is $C^1 = [0, 1/3] \cup [2/3, 1]$.
3. Remove the open middle third from *each* of the remaining intervals. This leaves $C^2 = [0, 1/9] \cup [2/9, 1/3] \cup [2/3, 7/9] \cup [8/9, 1]$.
4. Continue this process infinitely. In general, $C^n$ is the union of exactly $2^n$ closed intervals, each having a length of $1/3^n$.

The **Cantor Set** $C$ is defined as the infinite nested intersection of all these stages:
$$C = \bigcap_{n=0}^\infty C^n$$

---

## 2. Compactness and Being Non-Empty 

Is there anything actually left in this set after infinitely removing the middle sections? Yes. 

**Theorem:** The Cantor set is a compact, nonempty metric space.

**Proof of Compactness:**

Every stage $C^n$ consists of a finite union of closed, bounded intervals in $\mathbb{R}$, meaning each $C^n$ is a compact set. Because the Cantor set $C$ is the infinite intersection of these [nested compact sets](nests-of-compacts.ipynb), $C$ itself is necessarily compact.

**Proof of Being Non-Empty & Countable Endpoints:**

Let $E$ be the set of all the literal endpoints of the intervals created at every stage $C^n$.
$$E = \{0, 1, 1/3, 2/3, 1/9, 2/9, 7/9, 8/9, \dots\}$$
When we remove the *open* middle thirds, the solid endpoints are explicitly left behind. Therefore, every single endpoint in $E$ survives the infinite intersection and is contained in $C$. 
Because the number of endpoints doubles at each finite stage, $E$ is an infinite but **denumerable (countable)** set. Since $E \subset C$, the Cantor set $C$ is definitively non-empty and infinite.

---

## 3. The Cantor Set is Uncountable

Here we encounter the first major paradox: Even though the set of endpoints $E$ is countable, the actual Cantor set $C$ is **uncountable**. There are uncountably many more points in $C$ than just the endpoints!

**Geometric Coding Proof:**

We can pinpoint any specific location in the Cantor set using an infinite "address string".

* Let $0$ mean "choose the left sub-interval."
* Let $2$ mean "choose the right sub-interval."

For example, the address $C_0$ refers to $[0, 1/3]$ and $C_2$ refers to $[2/3, 1]$. The address $C_{20}$ refers to the left sub-interval of the right chunk, which is $[2/3, 7/9]$.
If we create an *infinite* address string $\omega = \omega_1\omega_2\omega_3\dots$ consisting entirely of $0$s and $2$s, it defines a uniquely nested sequence of ever-shrinking intervals. 
By the nested intersection theorem, this infinite sequence converges to exactly one unique point $p \in C$.

Because every point in $C$ has a unique infinite address made of two choices, $C$ is in a perfect one-to-one correspondence (bijection) with the set of all infinite binary-style sequences. This set of addresses is mathematically proven to be uncountable (it corresponds bijectively to the real numbers).

## 4. Measure Zero 

Despite having uncountably many points (just as many as the entire real number line), the Cantor set takes up exactly zero space. 

**Theorem:** The Cantor set is a zero set; it has an "outer measure" of zero.

**Proof:**
By definition, a set has outer measure zero if, for any $\epsilon > 0$, we can cover the entire set with a countable sequence of open intervals whose total combined length is less than $\epsilon$.
At the $n$-th stage of construction, $C^n$ consists of $2^n$ intervals, each of length $1/3^n$. 
The total length (or measure) of $C^n$ is simply:
$$\text{Total Length} = 2^n \times \left(\frac{1}{3^n}\right) = \left(\frac{2}{3}\right)^n$$
Since $C \subset C^n$ for all $n$, the length of $C$ must be less than or equal to $(2/3)^n$. 
As $n \to \infty$, the value $(2/3)^n$ shrinks to exactly $0$. Therefore, if we make $n$ large enough, the total length falls below any arbitrary $\epsilon > 0$. The Cantor set contains no intervals and its total length is zero.

## 5. Space-Filling Peano Curves 

Perhaps the most astonishing use of the Cantor set is proving the existence of a continuous path that completely fills a 2-dimensional area, like a solid square or disc. 

**Theorem:** There exists a Peano curve, which is a continuous path from an interval that is space-filling (its image has a nonempty interior, like the closed unit disc $B^2$).

**Proof Construction:**

1. **The Cantor Surjection:** There is a remarkable theorem stating that for *any* compact nonempty metric space (like a solid 2D disc $B^2$), there exists a continuous surjection (onto mapping) $\sigma$ from the Cantor set $C$ onto that space. This maps the "dust" of the Cantor set perfectly to every point in the solid disc.
2. **Linear Interpolation:** To make this a curve from the full interval $[0, 1]$, we define a new map $\tau: [0,1] \to B^2$. 
3. If $x$ is in the Cantor set $C$, we let $\tau(x) = \sigma(x)$.
4. If $x$ falls in one of the empty "gap intervals" $(a, b)$ that we removed during construction, we simply draw a straight continuous line connecting $\sigma(a)$ to $\sigma(b)$. 
5. Because $\sigma$ is continuous, the resulting extended path $\tau$ is continuous across the entire interval $[0,1]$. Its image fills the entire target disc, proving that a 1-dimensional line can be continuously folded to perfectly fill a 2-dimensional space.
