# Continuity and Homeomorphisms

## 1. Sequential Continuity
In real analysis, [functions](def-functions-mappings.ipynb) are viewed as devices that map points from a [metric space](metric-spaces.ipynb) $M$ to a metric space $N$. 

**Definition:** A function $f: M \to N$ is continuous if it preserves [sequential convergence](sequences-metric-spaces.ipynb). That is, if $p_n \to p$ in $M$, then the image sequence $f(p_n) \to f(p)$ in $N$. Limits are sent to limits.

**Theorem (Composites):** The composite of continuous functions is continuous.

**Proof:** Let $f: M \to N$ and $g: N \to P$ be continuous, and let $p_n \to p$ in $M$. Because $f$ is continuous, $f(p_n) \to f(p)$. Because $g$ is continuous, applying $g$ to this convergent sequence in $N$ yields $g(f(p_n)) \to g(f(p))$. Therefore, $g \circ f$ preserves convergence. $\square$

**Proposition (Basic Functions):**

* **Identity Map:** The map $\text{id}: M \to M$ is continuous. (If $p_n \to p$, then $\text{id}(p_n) = p_n \to p = \text{id}(p)$).
* **Constant Function:** If $f(x) = q \in N$ for all $x$, $f$ is continuous. (If $p_n \to p$, then $f(p_n) = q \to q = f(p)$).

## 2. Homeomorphisms
Just as vector spaces are isomorphic if there is a linear bijection between them, metric spaces are "isomorphic" if they "look the same" topologically.

A function $f: M \to N$ is a **homeomorphism** if:

1.  It is a bijection (one-to-one and onto).
2.  $f$ is continuous.
3.  $f^{-1}$ is continuous.

If such a function exists, $M$ and $N$ are **homeomorphic** ($M \cong N$). This forms an equivalence relation.
Geometrically, a homeomorphism can bend, twist, stretch, and wrinkle $M$ to make it coincide with $N$, but it cannot rip, puncture, shred, or pulverize $M$.

::: {.callout-warning}
## Danger: Is $f^{-1}$ automatically continuous?
It is a common trap to assume that if a bijection $f$ is continuous, its inverse $f^{-1}$ must also be continuous. **It is not.**

**Counterexample:** Let $f: [0, 2\pi) \to S^1$ (the unit circle) be defined by $f(x) = (\cos x, \sin x)$. 
This maps a half-open interval bijectively onto the circle. $f$ is continuous. However, its inverse is *not*. If we take a sequence of points $(z_n)$ on the circle in the 4th quadrant converging to $(1,0)$ from below, their preimages under $f^{-1}$ approach $2\pi$. But $f^{-1}(1,0) = 0$. Since the limits don't match, $f^{-1}$ fails to preserve sequential convergence!
:::

## 3. The $\epsilon-\delta$ Condition
The sequential definition is completely equivalent to the classic $\epsilon-\delta$ definition used in calculus.

**Theorem:** $f: M \to N$ is continuous if and only if it satisfies the $\epsilon-\delta$ condition:
$$\forall \epsilon > 0, \forall p \in M, \exists \delta > 0 \text{ s.t. } x \in M \text{ and } d_M(x, p) < \delta \implies d_N(f(x), f(p)) < \epsilon$$

**Proof ($\epsilon-\delta \implies$ Sequential):**
Assume $f$ satisfies $\epsilon-\delta$ at $p$. Let $(x_n)$ converge to $p$ in $M$. For any $\epsilon > 0$, there is a $\delta > 0$ satisfying the condition. Because $x_n \to p$, there exists an integer $K$ such that for all $n \geq K$, $d_M(x_n, p) < \delta$. This directly implies $d_N(f(x_n), f(p)) < \epsilon$, meaning $f(x_n) \to f(p)$. 

**Proof (Sequential $\implies \epsilon-\delta$ by Contradiction):**
Assume $f$ is sequentially continuous but fails the $\epsilon-\delta$ condition at $p$. 
This means there exists some $\epsilon > 0$ where *no* $\delta$ works. If we pick $\delta = 1/n$, there must be a point $x_n$ such that $d(x_n, p) < 1/n$ but $d(f(x_n), f(p)) \geq \epsilon$. 
This creates a sequence $(x_n)$ where $x_n \to p$, yet $f(x_n)$ does *not* converge to $f(p)$ (since every term is at least $\epsilon$ away). This contradicts our assumption that $f$ is sequentially continuous. $\square$

## 4. The Topological Definition of Continuity
This bridges the gap into point-set topology by removing the need to measure distances entirely.

**Theorem:** A function $f: M \to N$ is continuous *everywhere* on $M$ if and only if the preimage of every open set in $N$ is an open set in $M$.
Formally: For every open set $V \subseteq N$, the set $f^{-1}(V) = \{x \in M \mid f(x) \in V\}$ is open in $M$.

::: {.callout-note}
## Intuition
Continuous functions don't "tear" spaces. If you take a nice, breathable open set in the target space and pull it back through the function, it remains a nice, breathable open set in the domain.
:::
