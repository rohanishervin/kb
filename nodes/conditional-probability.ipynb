# Conditional Probability

## 1\. Definition

If $A$ and $B$ are [events](event.ipynb) in a [sample space](sample-space.ipynb) and $P(B) > 0$, the **conditional probability of $A$ given $B$** is defined as:
$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$


:::{.callout-note}
### Intuition: Shrinking the Universe

When you are given the conditional information that "$B$ has occurred," your entire universe of possibilities shrinks. The original [sample-space](sample-space.ipynb) $S$ is discarded, and $B$ becomes your new sample space.

To find the probability that $A$ also occurred, you look at the area where $A$ overlaps with $B$ ($A \cap B$). However, probabilities must sum to $1$. Because the area of $B$ is less than $1$, you must mathematically "zoom in" or **normalize** this new universe by dividing by $P(B)$.
:::

## 2\. The Multiplication Rule (Chain Rule)

By simply rearranging the definition of conditional probability, we can calculate the probability of intersections:
$$P(A \cap B) = P(A|B)P(B) = P(B|A)P(A)$$

This naturally extends to the intersection of $n$ events. You can calculate the joint probability by chaining conditionals:
$$P(A_1 \cap A_2 \cap \dots \cap A_n) = P(A_1)P(A_2|A_1)P(A_3|A_1 \cap A_2) \dots P(A_n|A_1 \cap \dots \cap A_{n-1})$$

## 3\. Bayes' Theorem

Bayes' theorem is a mechanism for reversing conditional probabilities. If you know $P(B|A)$, you can find $P(A|B)$ using:
$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$

**Proof:**

From the multiplication rule, we know $P(A \cap B) = P(B|A)P(A)$.
Substitute this into the standard definition of conditional probability $P(A|B) = \frac{P(A \cap B)}{P(B)}$, and the theorem immediately follows. $\square$

### Coherence of Bayesian Updates

Bayes' rule allows us to update our belief in an event $A$ as new evidence arrives. A profound property of this math is that it is **coherent**: the order in which you receive evidence does not matter.
If you observe evidence $E_1$ and then $E_2$, updating your prior sequentially ($P(A) \to P(A|E_1) \to P(A|E_1, E_2)$) yields the exact same final probability as processing both pieces of evidence simultaneously.
