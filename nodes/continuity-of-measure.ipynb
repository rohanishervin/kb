# Continuity of Measure

In measure theory, the measure of a limit set of a monotone sequence of sets is intimately connected to the limit of their individual measures. This property, known as **continuity of measure**, guarantees that measures behave nicely with respect to countable monotone operations, effectively allowing us to exchange limits and measures. 

## 1. The Main Proposition

For a [measure space](measure.ipynb) $(X, \mathcal{M}, \mu)$, continuity of measure holds for both ascending (expanding) and descending (contracting) sequences of measurable sets. 

**Proposition (Continuity of Measure):**
Let $(X, \mathcal{M}, \mu)$ be a measure space. 

1. **Continuity from Below (Ascending):** If $\{A_k\}_{k=1}^\infty$ is an ascending sequence of measurable sets (i.e., $A_1 \subseteq A_2 \subseteq \dots$), then
   $$\mu\left(\bigcup_{k=1}^{\infty} A_k\right) = \lim_{k\to\infty} \mu(A_k)$$
.

2. **Continuity from Above (Descending):** If $\{B_k\}_{k=1}^\infty$ is a descending sequence of measurable sets (i.e., $B_1 \supseteq B_2 \supseteq \dots$) for which $\mu(B_1) < \infty$, then
   $$\mu\left(\bigcap_{k=1}^{\infty} B_k\right) = \lim_{k\to\infty} \mu(B_k)$$
.

<!-- CELL BREAK -->

## 2. Proofs

### 2.1. Proof of Continuity from Below

::: {.callout-tip collapse="false"}
## Proof of Continuity from Below (Ascending)
Let $A = \bigcup_{k=1}^{\infty} A_k$. To use the countable additivity of our measure, we must transform the ascending sequence $\{A_k\}$ into a sequence of *disjoint* sets. 

**Step 1:** Define the disjoint sequence $\{G_k\}$. 
Let $A_0 = \emptyset$. For each $k \ge 1$, define
$$G_k = A_k \setminus A_{k-1}$$
.
By construction, $G_i \cap G_j = \emptyset$ for $i \neq j$, and the partial unions perfectly reconstruct the original sequence: 
$$\bigcup_{i=1}^n G_i = A_n$$
. 
Consequently, the countable union is preserved: $\bigcup_{i=1}^\infty G_i = \bigcup_{k=1}^\infty A_k = A$.

**Step 2:** Apply countable additivity.
Because the sets $G_i$ are disjoint and measurable, we can apply the countable additivity property of $\mu$:
$$\mu(A) = \mu\left(\bigcup_{i=1}^{\infty} G_i\right) = \sum_{i=1}^{\infty} \mu(G_i)$$
.

**Step 3:** Evaluate the infinite series.
By the definition of an infinite series, we can evaluate it as the limit of its partial sums:
$$\begin{aligned} \sum_{i=1}^{\infty} \mu(G_i) &= \lim_{n \to \infty} \sum_{i=1}^n \mu(G_i) \\ &= \lim_{n \to \infty} \sum_{i=1}^n \left( \mu(A_i) - \mu(A_{i-1}) \right) \end{aligned}$$
.
This is a telescoping sum. The partial sum collapses to simply $\mu(A_n)$. Therefore,
$$\mu(A) = \lim_{n \to \infty} \mu(A_n)$$
.
$\blacksquare$
:::

::: {.callout-note}
## Intuition: The "Disjointification" Trick
The core challenge in the ascending proof is that measure theory's foundational axiom (countable additivity) only applies to *mutually disjoint* sets. The sets $A_k$ are nested, not disjoint. The standard trick—often called "disjointification"—is to look at the "rings" or "layers" added at each step ($A_k \setminus A_{k-1}$). The total measure is simply the sum of the measures of all these individual layers. 
:::

### 2.2. Proof of Continuity from Above

::: {.callout-tip collapse="false"}
## Proof of Continuity from Above (Descending)
Let $B = \bigcap_{k=1}^{\infty} B_k$. 

**Step 1:** Convert the descending sequence to an ascending one.
Because $\mu(B_1) < \infty$, we can meaningfully subtract measures. We look at the complements of $B_k$ relative to $B_1$. Define
$$C_k = B_1 \setminus B_k$$
Since $B_1 \supseteq B_2 \supseteq \dots$, it follows that $C_1 \subseteq C_2 \subseteq \dots$, making $\{C_k\}$ an ascending sequence of measurable sets.

**Step 2:** Apply continuity from below.
The union of these ascending sets is:
$$\bigcup_{k=1}^{\infty} C_k = \bigcup_{k=1}^{\infty} (B_1 \setminus B_k) = B_1 \setminus \left( \bigcap_{k=1}^{\infty} B_k \right) = B_1 \setminus B$$
By Continuity from Below (proven above), we have:
$$\mu(B_1 \setminus B) = \lim_{n \to \infty} \mu(B_1 \setminus B_n)$$
.

**Step 3:** Use the finiteness condition to conclude.
Because $B \subseteq B_1$ and $B_n \subseteq B_1$, and critically because $\mu(B_1) < \infty$, we can split the measures of the set differences:
$$\mu(B_1 \setminus B) = \mu(B_1) - \mu(B)$$
.
$$\mu(B_1 \setminus B_n) = \mu(B_1) - \mu(B_n)$$
.
Substituting these into our limit equation gives:
$$\mu(B_1) - \mu(B) = \lim_{n \to \infty} \left( \mu(B_1) - \mu(B_n) \right)$$
.
Since $\mu(B_1)$ is a finite constant, we can pull it out of the limit:
$$\mu(B_1) - \mu(B) = \mu(B_1) - \lim_{n \to \infty} \mu(B_n)$$
.
Subtracting $\mu(B_1)$ from both sides and multiplying by $-1$ yields:
$$\mu(B) = \lim_{n \to \infty} \mu(B_n)$$
.
$\blacksquare$
:::




```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# --- Ascending Sequence ---
ax1.set_title(r"Ascending Sequence" + "\n" + r"$A_1 \subseteq A_2 \subseteq A_3 \dots$")
ax1.set_xlim(-5, 5)
ax1.set_ylim(-5, 5)
ax1.axis('off')

colors_asc = ['#e6f2ff', '#b3d9ff', '#80bfff', '#4da6ff']
radii_asc = [4, 3, 2, 1]
labels_asc = ['A_4', 'A_3', 'A_2', 'A_1']

for r, c, l in zip(radii_asc, colors_asc, labels_asc):
    circle = patches.Circle((0, 0), r, facecolor=c, edgecolor='black', linewidth=1.5)
    ax1.add_patch(circle)
    ax1.text(0, r - 0.5, rf"${l}$", ha='center', va='center', fontsize=12)

ax1.text(0, -4.5, r"Rings represent $G_k = A_k \setminus A_{k-1}$", ha='center', fontsize=10, style='italic')

# --- Descending Sequence ---
ax2.set_title(r"Descending Sequence" + "\n" + r"$B_1 \supseteq B_2 \supseteq B_3 \dots$")
ax2.set_xlim(-5, 5)
ax2.set_ylim(-5, 5)
ax2.axis('off')

colors_desc = ['#ffcccc', '#ff9999', '#ff6666', '#ff3333']
radii_desc = [4, 3, 2, 1]
labels_desc = [r'B_1', r'B_2', r'B_3', r'B = \bigcap B_k']

for r, c, l in zip(radii_desc, colors_desc, labels_desc):
    circle = patches.Circle((0, 0), r, facecolor=c, edgecolor='black', linewidth=1.5)
    ax2.add_patch(circle)
    if r == 1:
        ax2.text(0, 0, rf"${l}$", ha='center', va='center', fontsize=10, color='white', weight='bold')
    else:
        ax2.text(0, r - 0.5, rf"${l}$", ha='center', va='center', fontsize=12)
        
ax2.text(0, -4.5, r"Red core is the limit set $B$", ha='center', fontsize=10, style='italic')

plt.tight_layout()
plt.show()
```



::: {.callout-warning}
## The Finiteness Condition is Strictly Necessary
The condition that $\mu(B_1) < \infty$ in continuity from above is **not** optional. Consider the measure space $(\mathbb{R}, \mathcal{B}, m)$ where $m$ is the [Lebesgue measure](). 
Let $B_k = [k, \infty)$. This is a descending sequence of sets. 
For every $k$, $m(B_k) = \infty$, so $\lim_{k \to \infty} m(B_k) = \infty$. 
However, their intersection is empty: $\bigcap B_k = \emptyset$, and $m(\emptyset) = 0$. 
$0 \neq \infty$, so continuity from above fails here precisely because $m(B_1) = \infty$.
:::

## 3. "Almost Everywhere" Terminology

In measure theory, sets of measure zero are often considered negligible. We formalize this using the phrase **almost everywhere**.

For a measure space $(X, \mathcal{M}, \mu)$ and a measurable subset $E \subseteq X$, we say that a property holds **almost everywhere** on $E$ (or that it holds **for almost all** $x \in E$) if the property holds on $E \setminus E_0$, where $E_0$ is a measurable subset of $E$ satisfying $\mu(E_0) = 0$.
