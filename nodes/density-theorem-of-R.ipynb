# The Density of Rational and Irrational Numbers

## 1\. The Density Theorem of Real Numbers

**Theorem:** Between any two distinct real numbers, there exists a rational number and an irrational number. Because this holds for *any* interval, it implies that every interval $(a, b)$ contains infinitely many rational and infinitely many irrational numbers. In mathematical terms, we say both $\mathbb{Q}$ and $\mathbb{R} \setminus \mathbb{Q}$ are **dense** in $\mathbb{R}$.

Below, we provide two distinct proofs for this theorem. The first relies on the Archimedean property of $\mathbb{R}$, and the second relies on the foundational construction of $\mathbb{R}$ via [Dedekind cuts](def-dedekind-cuts.ipynb).

-----

## 2\. Proof 1: Using the Archimedean Property

Without loss of generality, let $x, y \in \mathbb{R}$ such that $x < y$. Our goal is to find a rational number $q \in \mathbb{Q}$ and an irrational number $s \in \mathbb{R} \setminus \mathbb{Q}$ strictly between them.

### Finding a Rational Number

Since $x < y$, we know $y - x > 0$. By the [Archimedean Property of $\mathbb{R}$](archimedean-property-of-R.ipynb), there exists a positive integer $q \in \mathbb{N}$ such that:
$$q(y - x) > 1 \implies qy - qx > 1$$
Because the distance between $qx$ and $qy$ is strictly greater than $1$, there must be an integer trapped between them. Formally, let $p = \lfloor qx \rfloor + 1$. By the definition of the floor function, we know:
$$p - 1 \leq qx < p$$
This gives us $qx < p$. Furthermore, since $p \leq qx + 1$ and $qx + 1 < qy$ (from our Archimedean gap), we can chain the inequalities together:
$$qx < p < qy$$
Dividing the entire inequality by the positive integer $q$ yields:
$$x < \frac{p}{q} < y$$
Since $p, q \in \mathbb{Z}$ and $q \neq 0$, the number $\frac{p}{q}$ is our required rational number.

### Finding an Irrational Number

We apply a similar logic. If $x < y$, then we also have:
$$\frac{x}{\sqrt{2}} < \frac{y}{\sqrt{2}}$$
By the exact same argument used above, there exists a rational number $r \in \mathbb{Q}$ strictly between these two values:
$$\frac{x}{\sqrt{2}} < r < \frac{y}{\sqrt{2}}$$
Multiplying through by $\sqrt{2}$ gives:
$$x < r\sqrt{2} < y$$
As long as $r \neq 0$ (which we can ensure by choosing a suitably large $q$ that shifts the integer $p$ away from zero), the product $r\sqrt{2}$ is irrational. Thus, we have found an irrational number between $x$ and $y$. $\square$

## 3\. Proof 2: Dedekind Cuts and Transformations (Pugh's Method)

This proof relies on the construction of real numbers as cuts in $\mathbb{Q}$. Before proving the main theorem, we must establish two foundational lemmas.

### Lemma 1: Preservation of Irrationality

**Statement:** If $a \in \mathbb{Q}$ and $b \in \mathbb{R} \setminus \mathbb{Q}$, then their sum $a+b$ and product $a \cdot b$ (for $a \neq 0$) are irrational.

**Proof:**

We use a simple [proof by contradiction](pre-logic.ipynb).
Suppose $a + b = q$ for some $q \in \mathbb{Q}$. Then $b = q - a$. Since the rational numbers are closed under subtraction, $q - a \in \mathbb{Q}$, which implies $b \in \mathbb{Q}$. This contradicts the premise that $b$ is irrational.
Similarly, if $a \cdot b = q$, then $b = \frac{q}{a}$. Since rationals are closed under division (for non-zero divisors), this again implies $b \in \mathbb{Q}$, which is a contradiction. $\square$

### Lemma 2: Density in the Unit Interval

**Statement:** The interval $[0, 1]$ contains infinitely many rational and irrational numbers.

**Proof:**

Consider the sequence of numbers $\frac{1}{n}$ for $\forall n \in \mathbb{N}$. Since $n \geq 1$, $0 < \frac{1}{n} \leq 1$. This generates an infinite set of distinct rational numbers in $[0, 1]$.
Similarly, consider the sequence $\frac{1}{\sqrt{2}n}$ for $\forall n \in \mathbb{N}$. Since $\sqrt{2}n \geq \sqrt{2} > 1$, these values also fall strictly within $[0, 1]$. By Lemma 1, dividing a rational ($1/n$) by an irrational ($\sqrt{2}$) yields an irrational number. Thus, this generates an infinite set of distinct irrational numbers in $[0, 1]$. $\square$

### Main Proof

Let $a, b \in \mathbb{R}$ be represented as Dedekind cuts in $\mathbb{Q}$, such that $a = A|A'$ and $b = B|B'$.
Assume $a < b$. By the definition of the order of cuts, this implies $A \subsetneq B$. Therefore, the set $B \setminus A$ is a nonempty set of rational numbers.

Choose a rational number $r \in B \setminus A$. Because a Dedekind cut $B$ contains no largest element, there must exist another rational number $s \in B$ such that $r < s$.
Because $r \notin A$ (it is in the "upper" part relative to cut $a$), we know $a < r$. Because $s \in B$ (it is in the "lower" part relative to cut $b$), we know $s < b$. Thus, we have trapped two rational numbers between our real cuts:
$$a < r < s < b \quad \text{where } r, s \in \mathbb{Q}$$

Now, consider the affine transformation $T: \mathbb{R} \to \mathbb{R}$ defined by:
$$T(t) = r + (s - r)t$$
Notice the behavior of this transformation:

1.  $T(0) = r$
2.  $T(1) = s$
    Therefore, $T$ maps the interval $[0, 1]$ bijectively onto the interval $[r, s]$.

Because $r$ and $s$ are both rational, the difference $(s - r)$ is rational. By **Lemma 1**, if we plug a rational $t$ into $T(t)$, the result is rational. If we plug an irrational $t$ into $T(t)$, the result is irrational.
Thus, $T$ sends rationals to rationals, and irrationals to irrationals.

By **Lemma 2**, the interval $[0, 1]$ contains infinitely many rationals and irrationals. Because $T$ preserves both cardinality and rationality/irrationality, the mapped interval $[r, s]$ must also contain infinitely many rational and irrational numbers.
Since $[r, s] \subset (a, b)$, it follows that the interval $(a, b)$ contains infinitely many rational and irrational numbers. $\square$
