# Normed Vector Spaces


## 1. Definition of a Norm

A **norm** on a vector space $V$ is a function $| \cdot | : V \to \mathbb{R}$ that generalizes the concept of "length." It must satisfy three properties for all $v, w \in V$ and $\lambda \in \mathbb{R}$:

1.  **Positive Definite:** $|v| \geq 0$, and $|v| = 0 \iff v = 0$
2.  **Absolute Homogeneity:** $|\lambda v| = |\lambda| |v|$
3.  **Triangle Inequality:** $|v + w| \leq |v| + |w|$

We are very used to these properties holding for the standard Euclidean norm.

## 2. Relationship to Inner Product Spaces

Every inner product space is automatically a normed space. 
If we have an inner product $\langle \cdot, \cdot \rangle$, we can simply define the norm as:
$$|v| = \sqrt{\langle v, v \rangle}$$
Because of the [Cauchy-Schwarz inequality](Cauchy-Schwarz-inequality.ipynb), this induced norm is guaranteed to satisfy the Triangle Inequality. 

Therefore: **Inner Product Spaces $\subseteq$ Normed Spaces**

### The Converse is False
Not every normed space is an inner product space! There are valid norms that cannot be generated by *any* inner product. 

**Counterexample:** Consider $V = \mathbb{R}^2$ equipped with the **Max Norm**:

$$|v|_{\max} = \max(|v_1|, |v_2|)$$
This satisfies all three rules of a norm, but it is impossible to construct an inner product $\langle x, y \rangle$ that perfectly induces this specific measurement of length.
