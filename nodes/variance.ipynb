# Variance and Standard Deviation

While the [Expected Value](expectation.ipynb) $E(X)$ provides the "center of mass" of a distribution, the **variance** measures the dispersion, or how "spread out" the random variable is from that center.

## 1. Definition

**Definition:** The variance of a [random variable](random-variables.ipynb) $X$ (whether discrete or continuous), denoted $Var(X)$, is the expected value of the squared deviation of $X$ from its own mean $\mu = E(X)$:

$$Var(X) = E[(X - \mu)^2]$$

Because variance is measured in the squared units of $X$, we define the **standard deviation** ($\sigma$) as the square root of the variance to return the metric to the original units:

$$\text{Standard Deviation } (SD) = \sqrt{Var(X)}$$

## 2. The Alternate Formulation

Directly computing $E[(X - \mu)^2]$ using sums or integrals is often tedious. We can expand this expression using the Linearity of Expectation to derive a more computationally efficient formula.

**Theorem:** For any random variable $X$, 
$$Var(X) = E(X^2) - (E(X))^2$$

**Proof:**

Let $\mu = E(X)$. Remember that $\mu$ is a fixed scalar constant, not a random variable.
$$Var(X) = E[(X - \mu)^2]$$
Expand the quadratic expression:
$$Var(X) = E[X^2 - 2\mu X + \mu^2]$$
Apply the Linearity of Expectation ($E[aX + Y] = aE[X] + E[Y]$):
$$Var(X) = E(X^2) - E(2\mu X) + E(\mu^2)$$
Since $2\mu$ and $\mu^2$ are constants, they factor cleanly out of the expectation:
$$Var(X) = E(X^2) - 2\mu E(X) + \mu^2$$
Substitute $\mu$ back in place of $E(X)$:
$$Var(X) = E(X^2) - 2\mu(\mu) + \mu^2$$
$$Var(X) = E(X^2) - 2\mu^2 + \mu^2$$
$$Var(X) = E(X^2) - \mu^2 = E(X^2) - (E(X))^2 \quad \blacksquare$$

## 3. A Consequence of Non-Negativity

**Theorem:** For any random variable $X$, $E(X^2) \ge (E(X))^2$. 

**Proof:**

By definition, $Var(X) = E[(X - \mu)^2]$.
Because the square of any real number is non-negative, the random variable $(X - \mu)^2$ is strictly non-negative.
The expected value of a non-negative random variable must itself be non-negative. Therefore:
$$Var(X) \ge 0$$
Using our alternate formula for variance:
$$E(X^2) - (E(X))^2 \ge 0$$
$$E(X^2) \ge (E(X))^2 \quad \blacksquare$$
