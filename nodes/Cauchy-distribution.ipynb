# Lemmas on Independent Normal Variables

Before discussing the Cauchy distribution, we establish two important lemmas regarding independent [normal distributions](normal-distribution.ipynb), which form the foundation for analyzing random variables built from normal components.

## 1\. Sum of Independent Normals

**Theorem:**
Let $X$ and $Y$ be independent normal random variables where $X \sim \mathcal{N}(\mu_1, \sigma_1^2)$ and $Y \sim \mathcal{N}(\mu_2, \sigma_2^2)$. Then their sum is normally distributed:
$$X + Y \sim \mathcal{N}(\mu_1 + \mu_2, \sigma_1^2 + \sigma_2^2)$$

::: {.callout-tip collapse="true"}
## Proof of Sum of Independent Normals
We prove this using the [Moment Generating Function](moment-generating-functions.ipynb) (MGF). The MGF of a normal variable $W \sim \mathcal{N}(\mu, \sigma^2)$ is known to be $M_W(t) = e^{\mu t + \frac{1}{2}t^2\sigma^2}$.

**Step 1:** Since $X$ and $Y$ are independent, the MGF of their sum $X+Y$ is the product of their individual MGFs:
$$\begin{aligned} M_{X+Y}(t) &= M_X(t) \cdot M_Y(t) \\ &= \left( e^{\mu_1 t + \frac{1}{2}t^2\sigma_1^2} \right) \left( e^{\mu_2 t + \frac{1}{2}t^2\sigma_2^2} \right) \end{aligned}$$

**Step 2:** Combine the exponents:
$$\begin{aligned} M_{X+Y}(t) &= e^{(\mu_1 + \mu_2)t + \frac{1}{2}t^2(\sigma_1^2 + \sigma_2^2)} \end{aligned}$$

**Step 3:** We recognize this resulting expression as the MGF of a normal distribution with mean $\mu_1 + \mu_2$ and variance $\sigma_1^2 + \sigma_2^2$. Because MGFs uniquely identify distributions, $X+Y \sim \mathcal{N}(\mu_1 + \mu_2, \sigma_1^2 + \sigma_2^2)$.
$\blacksquare$
:::

## 2\. Expected Distance Between Standard Normals

**Proposition:**
Let $Z_1, Z_2 \stackrel{i.i.d.}{\sim} \mathcal{N}(0, 1)$. The expected distance between them is $\mathbb{E}[\vert{}Z_1 - Z_2\vert{}] = \frac{2}{\sqrt{\pi}}$.

::: {.callout-warning}
## Note on the Original Problem Statement
Your handwritten notes asked to find "the distance $E[Z_1 - Z_2]$". Note that $\mathbb{E}[Z_1 - Z_2]$ is simply $0$ by linearity of expectation. The term "distance" mathematically implies the absolute value of the difference, meaning we must evaluate $\mathbb{E}[\vert{}Z_1 - Z_2\vert{}]$. The steps in your notes correctly solved for the absolute difference.
:::

::: {.callout-tip collapse="true"}
## Proof of Expected Distance
**Step 1:** Define $W = Z_1 - Z_2$. By the previous theorem, $W \sim \mathcal{N}(0, 1^2 + 1^2)$, so $W \sim \mathcal{N}(0, 2)$. 

**Step 2:** Because the normal distribution is symmetric, we can represent $W$ as $\sqrt{2}Z$, where $Z \sim \mathcal{N}(0,1)$. We want to find:
$$\mathbb{E}[\vert{}W\vert{}] = \mathbb{E}[\vert{}\sqrt{2}Z\vert{}] = \sqrt{2}\mathbb{E}[\vert{}Z\vert{}]$$

**Step 3:** Use the LOTUS to compute $\mathbb{E}[\vert{}Z\vert{}]$:
$$\sqrt{2}\mathbb{E}[\vert{}Z\vert{}] = \sqrt{2} \int_{-\infty}^{\infty} \vert{}z\vert{} \frac{1}{\sqrt{2\pi}} e^{-z^2/2} dz$$

**Step 4:** Since $\vert{}z\vert{}e^{-z^2/2}$ is an even function, we can change the bounds to $[0, \infty)$ and multiply by 2:
$$\begin{aligned} &= 2\sqrt{2} \int_0^{\infty} z \frac{1}{\sqrt{2\pi}} e^{-z^2/2} dz \\ &= \frac{2\sqrt{2}}{\sqrt{2\pi}} \int_0^{\infty} z e^{-z^2/2} dz \\ &= \frac{2}{\sqrt{\pi}} \int_0^{\infty} z e^{-z^2/2} dz \end{aligned}$$

**Step 5:** Evaluate the integral by substituting $u = z^2/2$, so $du = z dz$:
$$\frac{2}{\sqrt{\pi}} \int_0^{\infty} e^{-u} du = \frac{2}{\sqrt{\pi}} (1) = \frac{2}{\sqrt{\pi}}$$
$\blacksquare$
:::

# The Cauchy Distribution

The **Cauchy distribution** is a pathological continuous probability distribution famous for having undefined moments (no expected value, no variance). 

## 1\. Definition and Story

The standard Cauchy distribution arises naturally as the ratio of two independent standard normal random variables. 

Let $X, Y \stackrel{i.i.d.}{\sim} \mathcal{N}(0,1)$. If we define $T = \frac{X}{Y}$, then $T$ follows the standard Cauchy distribution:
$$T \sim \text{Cauchy}(0, 1)$$

::: {.callout-warning}
## Breakdowns of Standard Theorems
Because $\mathbb{E}[T]$ does not exist, the Cauchy distribution does not obey the Law of Large Numbers or the standard Central Limit Theorem (CLT). If you average a million independent standard Cauchy random variables, the sample mean $\bar{T}_{1,000,000}$ does not converge to a constant; it is *still* exactly a standard Cauchy random variable!
:::

<!-- CELL BREAK -->

## 2\. Deriving the Cauchy PDF

We can derive the probability density function (PDF) of the Cauchy distribution directly from its definition $T = X/Y$ using the cumulative distribution function (CDF) technique.

**Theorem (Cauchy PDF):**
The PDF of $T = \frac{X}{Y}$ where $X,Y \sim \mathcal{N}(0,1)$ is given by:
$$f_T(t) = \frac{1}{\pi(1+t^2)}$$

::: {.callout-tip collapse="true"}
## Proof of the Cauchy PDF
Let $F_T(t)$ be the CDF of $T$. 

**Step 1:** Formulate the CDF probability.
$$F_T(t) = P\left(\frac{X}{Y} \le t\right)$$

Because $X \sim \mathcal{N}(0,1)$ is symmetric around zero ($X \sim -X$), the random variable $X/Y$ has the exact same distribution as $X/\vert{}Y\vert{}$. This symmetry trick allows us to write:
$$P\left(\frac{X}{Y} \le t\right) = P(X \le t\vert{}Y\vert{})$$

**Step 2:** Integrate over the joint distribution of $X$ and $Y$. Since they are independent standard normals, their joint PDF is $\frac{1}{2\pi} e^{-x^2/2} e^{-y^2/2}$. 
$$\begin{aligned} F_T(t) &= \int_{-\infty}^{\infty} \int_{-\infty}^{t\vert{}y\vert{}} \left(\frac{1}{\sqrt{2\pi}} e^{-x^2/2}\right) \left(\frac{1}{\sqrt{2\pi}} e^{-y^2/2}\right) dx dy \\ &= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} e^{-y^2/2} \left( \int_{-\infty}^{t\vert{}y\vert{}} \frac{1}{\sqrt{2\pi}} e^{-x^2/2} dx \right) dy \end{aligned}$$
The inner integral is exactly the standard normal CDF, $\Phi(t\vert{}y\vert{})$.
$$F_T(t) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} e^{-y^2/2} \Phi(t\vert{}y\vert{}) dy$$

**Step 3:** Use symmetry to simplify the outer integral. Since the integrand is an even function of $y$, we can integrate from $0$ to $\infty$ and multiply by 2:
$$F_T(t) = \frac{2}{\sqrt{2\pi}} \int_{0}^{\infty} e^{-y^2/2} \Phi(ty) dy = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty} e^{-y^2/2} \Phi(ty) dy$$

**Step 4:** Differentiate $F_T(t)$ with respect to $t$ to find the PDF $f_T(t) = F_T'(t)$. We apply Leibniz's rule to bring the derivative inside the integral:
$$f_T(t) = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty} e^{-y^2/2} \left[ \frac{\partial}{\partial t} \Phi(ty) \right] dy$$
Using the chain rule, $\frac{\partial}{\partial t} \Phi(ty) = y \cdot \phi(ty) = y \frac{1}{\sqrt{2\pi}} e^{-(ty)^2 / 2}$. Substituting this back in:
$$\begin{aligned} f_T(t) &= \sqrt{\frac{2}{\pi}} \int_{0}^{\infty} e^{-y^2/2} \left( y \frac{1}{\sqrt{2\pi}} e^{-t^2 y^2 / 2} \right) dy \\ &= \frac{1}{\pi} \int_{0}^{\infty} y e^{-y^2/2} e^{-t^2 y^2 / 2} dy \\ &= \frac{1}{\pi} \int_{0}^{\infty} y e^{-\frac{y^2}{2}(1+t^2)} dy \end{aligned}$$

**Step 5:** Evaluate the final integral using the substitution $u = \frac{y^2}{2}(1+t^2)$, which means $du = y(1+t^2)dy$, or $y dy = \frac{du}{1+t^2}$:
$$\begin{aligned} f_T(t) &= \frac{1}{\pi} \int_{0}^{\infty} e^{-u} \frac{du}{1+t^2} \\ &= \frac{1}{\pi(1+t^2)} \left[ -e^{-u} \right]_0^\infty \\ &= \frac{1}{\pi(1+t^2)} \end{aligned}$$
$\blacksquare$
:::

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy, norm

# Create an array of x values
x = np.linspace(-10, 10, 1000)

# Calculate PDF values for Cauchy and Standard Normal
pdf_cauchy = cauchy.pdf(x)
pdf_norm = norm.pdf(x)

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, pdf_cauchy, label='Cauchy(0,1)', color='red', linewidth=2)
ax.plot(x, pdf_norm, label='Normal(0,1)', color='blue', linestyle='--', linewidth=2)

ax.set_title('Cauchy Distribution vs Standard Normal Distribution')
ax.set_xlabel('Value')
ax.set_ylabel('Probability Density')
ax.set_xlim(-10, 10)
ax.legend()
plt.grid(alpha=0.4)

# Add an inset or text to explain the tails
ax.text(-9, 0.35, "Note the 'heavy tails' of\nthe Cauchy distribution,\nwhich cause expected\nvalue to be undefined.", 
        bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))

plt.show()
```
