#### 1\. Order in $\mathbb{R}$

From the axiomatic definition of [real numbers](real-numbers.ipynb), we already saw all [field](field.ipynb) related properties of $\mathbb{R}$. Here, we will just scratch the surface of the field properties of the $\mathbb{R}$ constructed by [Dedekind cuts](def-dedekind-cuts.ipynb).

::: {.callout-note}

## A Note on Rigor

It is entirely possible to formally prove all algebraic field properties (addition, multiplication, distributivity) strictly using Dedekind cuts. However, the proofs—especially for multiplication of negative cuts—are notoriously long and tedious. If you are curious, you can find the exhaustive construction in *Principles of Mathematical Analysis* (Baby Rudin).
:::

Let $x$ and $y$ be real numbers represented by cuts $A|B$ and $C|D$ respectively. We define their order via set inclusion:

  * **$x \leq y$** if and only if $A \subseteq C$.
  * **$x < y$** if and only if $A \subsetneq C$.

#### 2\. The Least Upper Bound Property

**Definition:** A set $S \subset \mathbb{R}$ is **bounded above** if there exists an $M \in \mathbb{R}$ (an upper bound) such that for all $s \in S$, $s \leq M$.

**Theorem (Completeness of $\mathbb{R}$):** The set $\mathbb{R}$ created by Dedekind cuts is **complete**. It satisfies the **Least Upper Bound (LUB) Property**: every non-empty set of real numbers that is bounded above has a least upper bound (supremum) in $\mathbb{R}$.

**Proof:**

**Given:** A non-empty set $S \subset \mathbb{R}$ that is bounded above by a real number $M = A_M|B_M$. 
Every element $x \in S$ is a real number represented by a Dedekind cut, which we denote as $x = A_x|B_x$. Because $M$ is an upper bound, $x \leq M$, which means $A_x \subseteq A_M$ for all $x \in S$.

**Goal:** Construct a cut $z = C|D$ and prove that $z = \sup S$.

**Construction:** We define the lower set $C$ as the union of all lower sets in $S$:
$$C = \bigcup_{x \in S} A_x$$
We define the upper set $D$ as the remainder of the rationals:
$$D = \mathbb{Q} \setminus C$$

**Part 1: Prove $z = C|D$ is a valid Dedekind cut.**

* **Non-triviality:** By construction, $C \cup D = \mathbb{Q}$ and $C \cap D = \emptyset$. 
  Since $S$ is non-empty, there exists at least one $x = A_x|B_x \in S$. Because $A_x \neq \emptyset$, we know $C \neq \emptyset$. 
  To show $D \neq \emptyset$, recall that $A_x \subseteq A_M$ for all $x \in S$. Therefore, the union $C \subseteq A_M$. Because $M$ is a valid cut, $B_M \neq \emptyset$. By definition of a partition, any element in $B_M$ is not in $A_M$, and thus cannot be in $C$. Therefore, $B_M \subseteq D$, which proves $D \neq \emptyset$.
  
* **Downward Closure:** Let $c \in C$ and $d \in D$. We must show $c < d$. 
  Suppose for contradiction that $c \geq d$. Since $c \in C$, $c$ must belong to some lower set $A_x$ for an $x \in S$. Because $A_x|B_x$ is a valid cut, its lower set is downward closed. If $c \in A_x$ and $d \leq c$, then $d$ must also be in $A_x$. But if $d \in A_x$, then $d \in C$, which contradicts the fact that $d \in D = \mathbb{Q} \setminus C$. Therefore, $c < d$.

* **No Largest Element:** Let $c \in C$. By definition, $c \in A_x$ for some $x \in S$. Because $A_x$ is a valid cut, it has no largest element, meaning there exists a rational $c' \in A_x$ such that $c < c'$. Since $c' \in A_x$, it follows that $c' \in C$. Thus, $C$ has no largest element.

Since $C|D$ satisfies all three conditions, it is a valid real number $z \in \mathbb{R}$.

**Part 2: Prove $z$ is the Least Upper Bound.**

* **Upper Bound:** For any $x \in S$, its lower set $A_x$ is a subset of the union $C$ ($A_x \subseteq C$). By our definition of order, $x \leq z$ for all $x \in S$. Therefore, $z$ is an upper bound.
* **Least Upper Bound:** Let $y = A_y|B_y$ be any other upper bound for $S$. By definition, $x \leq y$ for all $x \in S$, meaning $A_x \subseteq A_y$. Since every $A_x$ is contained in $A_y$, their union $C$ must also be contained in $A_y$ ($C \subseteq A_y$). By the definition of order, $z \leq y$.

**Conclusion:** Since $z$ is an upper bound and $z \leq y$ for any other upper bound $y$, $z = \sup S$. $\square$


#### 3\. Absolute Value and Triangle Inequality

The **magnitude** or **absolute value** of $x \in \mathbb{R}$ is defined as:

$$|x| = \begin{cases} x & \text{if } x \geq 0 \\ -x & \text{if } x < 0 \end{cases}$$

We have already proved the general [triangle inequality](triangle-inequality.ipynb). However, we can prove it again here specifically for $\mathbb{R}$, this time relying solely on the [proof by cases](pre-logic.ipynb) method derived from the definition of absolute value.

**Theorem (The Triangle Inequality):** For all $x, y \in \mathbb{R}$, $|x + y| \leq |x| + |y|$.

**Proof:**

We consider the cases for the sign of $x+y$:

  * **Case 1 ($x+y \geq 0$):** By definition, $|x+y| = x+y$. Since $x \leq |x|$ and $y \leq |y|$, it follows that $x+y \leq |x| + |y|$.
  * **Case 2 ($x+y < 0$):** By definition, $|x+y| = -(x+y) = (-x) + (-y)$. Since $-x \leq |x|$ and $-y \leq |y|$, it follows that $(-x) + (-y) \leq |x| + |y|$.

::: {.callout-tip}

## Conclusion

Since the inequality holds in both possible logical cases, we conclude that $|x+y| \leq |x| + |y|$ for all $x, y \in \mathbb{R}$. $\square$
:::
