# Real Numbers ($\mathbb{R}$)

## Axiomatic Definition

The real numbers, denoted by $\mathbb{R}$, can be defined axiomatically as follows:

* They satisfy all the axioms of an [ordered field](ordered-field.ipynb).
* In addition to these axioms, they also satisfy the [**Axiom of Completeness**](axiom-of-completeness.ipynb)
