# Binomial Coefficient Formula

## 1\. Combinatorial Definition

The **binomial coefficient**, denoted mathematically as $\binom{n}{k}$ and read aloud as "$n$ choose $k$", is a foundational concept in combinatorics and probability.

For any non-negative integers $n$ and $k$ (where $0 \leq k \leq n$), it represents the total number of distinct subsets of size $k$ that can be formed from a larger set of $n$ distinct elements.

Crucially, because it defines *subsets*, the **order of selection does not matter**. Selecting the elements $\{A, B\}$ is considered identical to selecting $\{B, A\}$.

## 2\. The Formula

As derived in the [binomial-coefficient-formula](binomial-coefficient-formula.ipynb) node (which uses the principle of [adjusting for overcounting](adjusting-for-overcounting.ipynb)), the algebraic formula is:

$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$

*(Note: If $k > n$, the coefficient is strictly defined as $0$, because it is impossible to choose more unique items than are available in the set).*

-----

## 3\. Pascal's Identity (The Addition Rule)

One of the most important properties of the binomial coefficient is its recursive nature. This relationship is what generates the famous **Pascal's Triangle**.

**Theorem (Pascal's Identity):**
For any positive integers $n$ and $k$ (where $1 \leq k \leq n$):

$$\binom{n}{k} = \binom{n-1}{k-1} + \binom{n-1}{k}$$

**Combinatorial Proof:**
Suppose we want to choose a subset of size $k$ from a group of $n$ people.

Let us single out one specific person in the group, say "Alice." Every possible subset of size $k$ can be perfectly partitioned into two mutually exclusive categories:

1.  **Subsets that INCLUDE Alice:** If we put Alice into our subset, we still need to choose $k-1$ more people to complete the group, and we must choose them from the remaining $n-1$ people. There are exactly $\binom{n-1}{k-1}$ ways to do this.

2.  **Subsets that EXCLUDE Alice:** If we completely ban Alice from our subset, we must choose all $k$ people from the remaining $n-1$ people. There are exactly $\binom{n-1}{k}$ ways to do this.

Since these are the only two logical possibilities and they do not overlap, the total number of subsets $\binom{n}{k}$ must be the sum of these two scenarios:

$$\binom{n}{k} = \binom{n-1}{k-1} + \binom{n-1}{k}$$
$\square$

-----

## 4\. Why is it called the "Binomial" Coefficient?

The name originates from algebra, specifically from the [binomial-theorem](binomial-theorem.ipynb). When you expand a binomial expression raised to a power, like $(x + y)^n$, the coefficients of the resulting polynomial terms are exactly these combinatorial numbers:

$$(x + y)^n = \sum_{k=0}^{n} \binom{n}{k} x^{n-k} y^k$$

**Intuition:** Why does counting subsets apply to algebra? When multiplying out the expansion $(x+y)(x+y)\dots(x+y)$ for $n$ terms, creating the specific term $x^{n-k}y^k$ requires you to choose exactly $k$ of the $n$ parentheses to contribute a $y$ (forcing the other $n-k$ parentheses to contribute an $x$). The number of ways to "choose" those $k$ parentheses is exactly $\binom{n}{k}$.

## 5\. Example Application

**Question:** A pizza parlor offers $8$ different toppings. How many unique $3$-topping pizzas can you order?

**Solution:**
Because the order the toppings are placed on the pizza does not matter, this is a strict combination problem. We need to calculate "8 choose 3":

$$\binom{8}{3} = \frac{8!}{3!(8-3)!} = \frac{8!}{3! \cdot 5!}$$

$$= \frac{8 \times 7 \times 6}{3 \times 2 \times 1} = \frac{336}{6} = 56$$

You can order exactly $56$ distinct $3$-topping pizzas.
