# Field $(F, +, \cdot)$
A field is a non-empty set $F$ of objects that is equipped with two binary operations:

*   **Addition**: $+: F \times F \rightarrow F$
*   **Multiplication**: $\cdot: F \times F \rightarrow F$

that has the following properties:

1. $$\forall x, y \in F: x + y \in F$$
2. $$\forall x, y \in F: x + y = y + x$$
3. $$\forall x, y, z \in F: (x + y) + z = x + (y + z)$$
    *   $x_1 + x_2 + \dots + x_n$ is well-defined and we don't need to write all parentheses.

4. $$\exists e \in F, \forall x \in F: x + e = x$$
    *   The element $e$ is unique. Suppose we have $e$ and $e'$ that both are additive identities. Then:
        $e' = e' + e$ (because $e$ is an additive identity)
        on the other hand, $e = e + e'$ because $e'$ is an additive identity.
        So: $e = e'$, and we can have a symbol for additive identity which is $0$.

5. $$\forall x \in F, \exists y: x + y = 0$$
    *   The element $y$ is unique and we call it $-x$. Suppose $y$ is not unique; then we have two elements $y$ and $y'$ that both satisfy this property.
        So: $y = y + 0 = y + (x + y') = (y + x) + y' = 0 + y' = y'$.

6. $$\forall x, y \in F: xy \in F$$
7. $$\forall x, y: xy = yx$$
8. $$\forall x, y, z: (xy)z = x(zy)$$
9. $$\forall x \in F, \exists e: xe = x$$
    *   The element $e$ is unique and we call it $1$.
10. $$\forall x \neq 0 \in F, \exists y: xy = 1$$

11. $$x(y + z) = xy + xz$$

# Examples:
1. $(\mathbb{Q}, +, \cdot)$
2. $\mathbb{Z}_p$ when p is a prime number. $\mathbb{Z}_p = \{{0,1,...,p-1}\}$
   - it's also called **Prime Field**
