# Triangle Inequality and Its Reverse

for all $x,y \in \mathbb{R}$ we have:

1. $$ |a+b| \leq |a| + |b| $$
2. $$ ||a|-|b|| \leq |a - b| $$



# Proof 
by using [absolute value properties](absolute-value.ipynb) we can write:

$$ -|a|\leq a \leq |a| $$
$$ -|b|\leq b \leq |b| $$

by using [adding inequalities](adding-inequalities.ipynb):
$$ -|a| - |b| \leq a+b \leq |a| + |b| $$

we have $|a| + |b| > 0$ if it's 0, $a=b=0$ which is trivial. so by using (4) in [absolute value properties](absolute-value.ipynb):
$$ ||a| + |b|| \leq |a+b| \leq ||a| + |b|| $$
which is:
$$ |a+b| \leq |a| + |b| $$
