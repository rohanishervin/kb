# The Density Theorem of Real Numbers

Between any two distinct real numbers there is a rational number
and an irrational number.

# Proof:

without loss of generality: $x,y \in \mathbb{R}, x<y$

by [archimedean property of R](archimedean-property-of-R.ipynb):
exists a positive integer $q$ such that: $q(y-x)>1$

it says that there exists an integer $p$ such that: 
$qx<p<qy$ so we can get the rational number $p/q$ since $ x<p/q<y$

similarly there exists an $r\neq0$ which is $\sqrt2x <r< \sqrt2 y$ so: $x<r/\sqrt2 <y$
