# Subsequence
Let $\{x_n\}$ be a [sequence](sequence.ipynb) and let $\{p_n\}$ be a strictly increasing
sequence of positive integers. The sequence $\{ x_{p_n} \}$ is called a subsequence of $\{x_n\}$.

Let $\{x_n\}$ be a sequence of real numbers.
 - If $\{x_n\}$ converges to $L$, then every subsequence of $\{x_n\}$ converges to $L$.
 - If $\{x_n\}$ has two subsequences that converge to different limits, then $\{x_n\}$ does not converge.

Every sequence of real numbers has a monotone subsequence.
