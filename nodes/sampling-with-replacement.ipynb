# Sampling With Replacement

## 1\. Definition and Concept

**Sampling with replacement** is a fundamental method of selecting items from a population. If we select $k$ items sequentially from a set of $n$ available distinct objects, and **each item is placed back into the pool after being chosen**, we are sampling with replacement.

Because the item is returned to the pool, the exact same item can be drawn multiple times in a single sequence. Furthermore, because the items are drawn sequentially, the **order of selection strictly matters**.

*(Example: Drawing the sequence "Red, Blue" is considered a distinct outcome from drawing "Blue, Red").*

-----

## 2\. The Formula and Derivation

The total number of possible ordered sequences of length $k$ drawn from a population of $n$ items with replacement is:

$$\text{Total Outcomes} = n^k$$

**Proof by the Multiplication Rule:**

This formula is a direct application of the [multiplication rule](multiplication-rule.ipynb). We view the drawing process as a sequence of $k$ distinct stages:

  * **Stage 1:** We draw the first item. Since the pool is full, we have exactly $n$ choices.
  * **Stage 2:** Because we replaced the first item, the pool is fully restored. We again have exactly $n$ choices.
  * **Stage 3:** The pool is restored again. We have $n$ choices.
  * $\dots$
  * **Stage $k$:** We still have $n$ choices.

By the multiplication rule, the total number of combined outcomes is the product of the choices at each stage:
$$\text{Total Outcomes} = \underbrace{n \times n \times \dots \times n}_{k \text{ times}} = n^k$$
$\square$

-----

## 3\. Key Property: Statistical Independence

The defining mathematical feature of sampling with replacement is **Independence**.

Because the initial state of the population is completely restored after every single draw, the outcome of any given draw provides absolutely zero information about the outcome of future draws. The probabilities remain perfectly static.

This is in direct contrast to [sampling without replacement](sampling-without-replacement.ipynb), where every draw permanently alters the composition of the pool, making subsequent draws mathematically dependent on the previous ones.

-----

## 4\. Classic Example

**Question:** A standard combination lock has 3 dials, and each dial contains the digits 0 through 9. How many unique "combinations" exist?

**Solution:**
Despite being called a "combination" lock, the order of the numbers strictly matters (the passcode `1-2-3` is different from `3-2-1`), and digits can be repeated (e.g., `7-7-7`).

Therefore, setting the lock is mathematically equivalent to drawing $k=3$ numbers from a pool of $n=10$ numbers (the digits 0-9) *with replacement*.

$$\text{Total Outcomes} = n^k = 10^3 = 1,000$$

There are exactly $1,000$ unique passcodes (ranging from `000` to `999`).

-----

## 5\. Python Simulation

You can verify the $n^k$ formula empirically by generating all possible samples with replacement using Python's `itertools` library.

```python
import itertools

def count_samples_with_replacement(population, k):
    """
    Generates and counts all ordered samples with replacement.
    """
    n = len(population)
    
    # itertools.product is the mathematical equivalent of Cartesian product (n^k)
    all_outcomes = list(itertools.product(population, repeat=k))
    
    display(f"Population (n): {n} items -> {population}")
    display(f"Sample Size (k): {k} draws")
    display(f"Theoretical Total (n^k): {n**k}")
    display(f"Actual Count Generated: {len(all_outcomes)}")
    
    # Print the first few to show order matters and repeats happen
    display(f"First 5 outcomes: {all_outcomes[:5]}")

# Example: Draw 3 times from a set of 4 colors
count_samples_with_replacement(['Red', 'Blue', 'Green', 'Yellow'], 3)
```
