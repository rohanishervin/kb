# Normal Distribution

## 1\. Introduction and the Standard Normal

The **Normal distribution** (often referred to as the Gaussian distribution) is a fundamental continuous probability distribution characterized by its symmetric, bell-shaped [probability density function](probability-density-function.ipynb). Its ubiquity in statistics is largely due to the [Central Limit Theorem](), which states that the sum of a large number of independent and identically distributed [random variables](random-variables.ipynb#continuous-random-variables) will approximately follow a Normal distribution, regardless of the underlying distribution of the individual variables.

To systematically study the Normal distribution, we use a location-scale transformation strategy. We first define the simplest case—the standard Normal distribution—and later generalize it by shifting and scaling.

### 1.1\. The Standard Normal Distribution

**Definition (Standard Normal Distribution):**
A continuous random variable $Z$ follows the **standard Normal distribution** if its probability density function (PDF) is given by:

$$
\varphi(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}, \quad -\infty < z < \infty
$$

We denote this as $Z \sim \mathcal{N}(0, 1)$, indicating a mean of $0$ and a variance of $1$. The constant factor $\frac{1}{\sqrt{2\pi}}$ is the **normalizing constant** required to ensure the total area under the PDF integrates to $1$.

The corresponding cumulative distribution function (CDF) represents the accumulated area under the PDF curve:

$$
\Phi(z) = \int_{-\infty}^{z} \varphi(t) dt = \int_{-\infty}^{z} \frac{1}{\sqrt{2\pi}} e^{-t^2/2} dt
$$

::: {.callout-note}
## Intuition: The Normal CDF
By convention, the standard Normal PDF and CDF are assigned specific Greek letters, $\varphi$ and $\Phi$, respectively. This highlights the importance of the distribution. The CDF $\Phi(z)$ yields an S-shaped curve, mapping the entire real line to the interval $(0, 1)$.
:::

::: {.callout-warning}
## No Closed-Form Antiderivative
It is a mathematically proven fact that the antiderivative of $e^{-t^2/2}$ cannot be expressed in a **closed form** using finite sums of elementary functions (like polynomials or exponentials). Therefore, the CDF $\Phi(z)$ must be left as an integral. Despite this, it is a perfectly well-defined function that can be computed numerically to high precision.
:::

## 2\. Symmetry Properties of the Standard Normal

The standard Normal PDF and CDF exhibit several vital symmetry properties that greatly simplify probability calculations.

1.  **Symmetry of the PDF (Even Function):**
    The PDF is symmetric around $0$. For any real $z$, $\varphi(z) = \varphi(-z)$. In mathematical terms, $\varphi$ is an even function.

2.  **Symmetry of Tail Areas:**
    Because the PDF is perfectly symmetric, the area to the left of $-z$ is exactly equal to the area to the right of $z$.
    $$
    \Phi(-z) = 1 - \Phi(z)
    $$
    *Proof:*

    
    Using the substitution $u = -t$, $du = -dt$:
    $$
    \begin{aligned}
    \Phi(-z) &= \int_{-\infty}^{-z} \varphi(t) dt \\
    &= \int_{z}^{\infty} \varphi(-u) du \\
    &= \int_{z}^{\infty} \varphi(u) du \quad \text{(since } \varphi \text{ is even)} \\
    &= 1 - \int_{-\infty}^{z} \varphi(u) du \\
    &= 1 - \Phi(z)
    \end{aligned}
    $$
    $\blacksquare$

4.  **Symmetry of $Z$ and $-Z$:**

    If $Z \sim \mathcal{N}(0, 1)$, then $-Z \sim \mathcal{N}(0, 1)$ as well.

    
    *Proof:*
    We evaluate the CDF of $-Z$:
    $$
    \begin{aligned}
    P(-Z \le z) &= P(Z \ge -z) \\
    &= 1 - P(Z \le -z) \\
    &= 1 - \Phi(-z) \\
    &= \Phi(z) \quad \text{(using the tail area symmetry property)}
    \end{aligned}
    $$
    Since the CDF of $-Z$ is $\Phi(z)$, $-Z$ is a standard Normal random variable. $\blacksquare$

## 3\. Validity, Mean, and Variance of the Standard Normal

To confirm that the standard Normal distribution is mathematically sound and matches its notation $\mathcal{N}(0, 1)$, we must prove three things: the PDF integrates to $1$, the expectation is $0$, and the variance is $1$.

### 3.1\. Validity of the PDF

We must show that the total area under $e^{-z^2/2}$ is exactly $\sqrt{2\pi}$. Since there is no closed-form antiderivative, we use a clever multivariable calculus technique.

::: {.callout-tip collapse="true"}
## Proof of the Normalizing Constant
We wish to evaluate the integral $I = \int_{-\infty}^{\infty} e^{-z^2/2} dz$.

**Step 1:** Write the integral squared, using two different dummy variables ($x$ and $y$).
$$
I^2 = \left( \int_{-\infty}^{\infty} e^{-x^2/2} dx \right) \left( \int_{-\infty}^{\infty} e^{-y^2/2} dy \right) = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} e^{-\frac{x^2+y^2}{2}} dx dy
$$

**Step 2:** Convert the double integral from Cartesian to polar coordinates.
Let $x = r \cos \theta$ and $y = r \sin \theta$. The area element $dx dy$ becomes $r dr d\theta$ (where $r$ is the Jacobian determinant of the transformation). The limits of integration become $0$ to $\infty$ for $r$, and $0$ to $2\pi$ for $\theta$.

$$
I^2 = \int_{0}^{2\pi} \int_{0}^{\infty} e^{-r^2/2} r dr d\theta
$$

**Step 3:** Evaluate the integral using u-substitution.
Let $u = r^2/2$, which means $du = r dr$.

$$
\begin{aligned}
I^2 &= \int_{0}^{2\pi} \left( \int_{0}^{\infty} e^{-u} du \right) d\theta \\
&= \int_{0}^{2\pi} \left[ -e^{-u} \right]_{0}^{\infty} d\theta \\
&= \int_{0}^{2\pi} 1 \, d\theta \\
&= 2\pi
\end{aligned}
$$

Taking the square root of both sides gives $I = \sqrt{2\pi}$. Therefore, multiplying by $1/\sqrt{2\pi}$ normalizes the integral to $1$. $\blacksquare$
:::

### 3.2\. Mean and Variance

**Theorem:** For $Z \sim \mathcal{N}(0, 1)$, $E(Z) = 0$ and $Var(Z) = 1$.

::: {.callout-tip collapse="true"}
## Proof of Expectation and Variance
**Part 1: Expectation**
$$
E(Z) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} z e^{-z^2/2} dz
$$
The integrand $g(z) = z e^{-z^2/2}$ is an odd function (i.e., $g(-z) = -g(z)$). Because $e^{-z^2/2}$ decays extremely rapidly, the integral is absolutely convergent. The negative area from $-\infty$ to $0$ perfectly cancels the positive area from $0$ to $\infty$. Thus, $E(Z) = 0$.

**Part 2: Variance**
By [LOTUS]() and the fact that $E(Z) = 0$, $Var(Z) = E(Z^2) - (E(Z))^2 = E(Z^2)$.
$$
Var(Z) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} z^2 e^{-z^2/2} dz
$$
Since $z^2 e^{-z^2/2}$ is an even function, we can simplify this to:
$$
Var(Z) = \frac{2}{\sqrt{2\pi}} \int_{0}^{\infty} z^2 e^{-z^2/2} dz
$$
We use [integration by parts]() with $u = z$ and $dv = z e^{-z^2/2} dz$. Then $du = dz$ and $v = -e^{-z^2/2}$.
$$
\begin{aligned}
Var(Z) &= \frac{2}{\sqrt{2\pi}} \left( \left[ -z e^{-z^2/2} \right]_{0}^{\infty} + \int_{0}^{\infty} e^{-z^2/2} dz \right)
\end{aligned}
$$
The boundary term evaluates to $0$ because exponential decay dominates polynomial growth as $z \to \infty$. The remaining integral is exactly half of the total area under $e^{-z^2/2}$, which we proved earlier is $\sqrt{2\pi}/2$.
$$
\begin{aligned}
Var(Z) &= \frac{2}{\sqrt{2\pi}} \left( 0 + \frac{\sqrt{2\pi}}{2} \right) \\
&= 1
\end{aligned}
$$
$\blacksquare$
:::

## 4\. The General Normal Distribution

With the standard Normal distribution established, we can construct the general Normal distribution using a simple linear shift and scaling.

**Definition (General Normal Distribution):**
If $Z \sim \mathcal{N}(0, 1)$, then for any real numbers $\mu$ and $\sigma$ (with $\sigma > 0$), the random variable $X$ defined by:
$$
X = \mu + \sigma Z
$$
follows the **Normal distribution** with mean $\mu$ and variance $\sigma^2$. We write this as $X \sim \mathcal{N}(\mu, \sigma^2)$.

It is straightforward to verify the mean and variance using the linearity of expectation and properties of variance:
$$
E(X) = E(\mu + \sigma Z) = \mu + \sigma E(Z) = \mu + 0 = \mu
$$
$$
Var(X) = Var(\mu + \sigma Z) = Var(\sigma Z) = \sigma^2 Var(Z) = \sigma^2(1) = \sigma^2
$$

::: {.callout-warning}
## Scaling by Standard Deviation
Notice that to obtain a variance of $\sigma^2$, we must multiply $Z$ by the standard deviation $\sigma$, NOT the variance. If we multiplied by $\sigma^2$, the resulting variable would have a variance of $\sigma^4$, throwing off the units of the distribution.
:::

### 4.1\. Standardization

The process of reversing the location-scale transformation is known as **standardization**. If we have a general Normal random variable $X \sim \mathcal{N}(\mu, \sigma^2)$, we can center it and scale it back to a standard Normal random variable:
$$
\frac{X - \mu}{\sigma} \sim \mathcal{N}(0, 1)
$$
Standardization is practically useful because it allows us to compute probabilities for any Normal distribution using only the standard Normal CDF $\Phi$.

## 5\. PDF and CDF of the General Normal

**Theorem (Normal CDF and PDF):** Let $X \sim \mathcal{N}(\mu, \sigma^2)$. The CDF of $X$ is:
$$
F(x) = \Phi\left(\frac{x - \mu}{\sigma}\right)
$$
And the PDF of $X$ is:
$$
f(x) = \frac{1}{\sigma} \varphi\left(\frac{x - \mu}{\sigma}\right) = \frac{1}{\sqrt{2\pi}\sigma} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
$$

::: {.callout-tip collapse="true"}
## Proof of General CDF and PDF
**Step 1: The CDF**
We start with the definition of the CDF and standardize the random variable inside the probability event:
$$
\begin{aligned}
F(x) &= P(X \le x) \\
&= P\left(\frac{X - \mu}{\sigma} \le \frac{x - \mu}{\sigma}\right)
\end{aligned}
$$
Because $\frac{X - \mu}{\sigma}$ is exactly the standard Normal $Z$, this becomes:
$$
F(x) = P\left(Z \le \frac{x - \mu}{\sigma}\right) = \Phi\left(\frac{x - \mu}{\sigma}\right)
$$

**Step 2: The PDF**
To find the PDF $f(x)$, we differentiate the CDF $F(x)$ with respect to $x$, applying the chain rule:
$$
\begin{aligned}
f(x) &= \frac{d}{dx} \Phi\left(\frac{x - \mu}{\sigma}\right) \\
&= \Phi'\left(\frac{x - \mu}{\sigma}\right) \cdot \frac{d}{dx} \left(\frac{x - \mu}{\sigma}\right) \\
&= \varphi\left(\frac{x - \mu}{\sigma}\right) \cdot \frac{1}{\sigma}
\end{aligned}
$$
Substituting the definition of $\varphi$ gives the full explicit expression for the PDF. $\blacksquare$
:::

## 6\. The 68-95-99.7% Rule

A useful heuristic for dealing with Normal distributions is the **68-95-99.7% rule**, which provides quick approximations for the probability mass contained within one, two, and three standard deviations of the mean.

**Theorem (Empirical Rule):** If $X \sim \mathcal{N}(\mu, \sigma^2)$, then:

1.  $P(|X - \mu| < \sigma) \approx 0.68$
2.  $P(|X - \mu| < 2\sigma) \approx 0.95$
3.  $P(|X - \mu| < 3\sigma) \approx 0.997$

Standardizing these ranges translates the rule directly to the standard Normal distribution:

* $P(|Z| < 1) \approx 0.68$
* $P(|Z| < 2) \approx 0.95$
* $P(|Z| < 3) \approx 0.997$

### 6.1\. Example: Calculating Probabilities
Suppose $X \sim \mathcal{N}(-1, 4)$. We want to find the approximate and exact probability $P(|X| < 3)$.

**Step 1:** Translate the event.
$|X| < 3$ is equivalent to $-3 < X < 3$.

**Step 2:** Standardize the event bounds using $\mu = -1$ and $\sigma = 2$ (since variance is $4$).
$$
P(-3 < X < 3) = P\left(\frac{-3 - (-1)}{2} < \frac{X - (-1)}{2} < \frac{3 - (-1)}{2}\right) = P(-1 < Z < 2)
$$

**Step 3:** Exact calculation using $\Phi$.
$$
P(-1 < Z < 2) = \Phi(2) - \Phi(-1) \approx 0.8186
$$

**Step 4:** Approximation using the 68-95-99.7% rule.
The interval from $-1$ to $2$ spans the right half of the $2$-standard deviation bracket and the left half of the $1$-standard deviation bracket. By symmetry:
$$
\begin{aligned}
P(-1 < Z < 2) &= P(-1 < Z < 1) + P(1 < Z < 2) \\
&\approx 0.68 + \frac{P(-2 < Z < 2) - P(-1 < Z < 1)}{2} \\
&\approx 0.68 + \frac{0.95 - 0.68}{2} \\
&= 0.68 + \frac{0.27}{2} \\
&= 0.815
\end{aligned}
$$
The approximation ($0.815$) is extremely close to the true numerical value ($0.8186$).

## 7\. Transformations: The Folded Normal Distribution

We can derive interesting distributions by transforming standard Normal variables. One such distribution is the Folded Normal, which models the absolute magnitude of a standard Normal variable.

**Example (Folded Normal):**
Let $Z \sim \mathcal{N}(0, 1)$ and define $Y = |Z|$. Even though the absolute value function has a non-differentiable "corner" at $0$, $Y$ has a perfectly valid continuous distribution.

**Part (a): Find $E(Y)$**
By [LOTUS](), we compute the expectation using the PDF of $Z$:
$$
\begin{aligned}
E(Y) &= E(|Z|) = \int_{-\infty}^{\infty} |z| \frac{1}{\sqrt{2\pi}} e^{-z^2/2} dz
\end{aligned}
$$
Because the integrand is an even function, we can evaluate it from $0$ to $\infty$ and multiply by $2$:
$$
\begin{aligned}
E(Y) &= 2 \int_{0}^{\infty} z \frac{1}{\sqrt{2\pi}} e^{-z^2/2} dz \\
&= \frac{2}{\sqrt{2\pi}} \left[ -e^{-z^2/2} \right]_{0}^{\infty} \\
&= \sqrt{\frac{2}{\pi}}
\end{aligned}
$$

**Part (b): Find $Var(Y)$**
Notice that $Y^2 = |Z|^2 = Z^2$. Therefore, $E(Y^2) = E(Z^2) = Var(Z) = 1$.
Using the variance shortcut formula:
$$
\begin{aligned}
Var(Y) &= E(Y^2) - (E(Y))^2 \\
&= 1 - \left( \sqrt{\frac{2}{\pi}} \right)^2 \\
&= 1 - \frac{2}{\pi}
\end{aligned}
$$

**Part (c): Find the CDF and PDF of $Y$**
The random variable $Y$ cannot be negative, so for $y \le 0$, $F_Y(y) = 0$.
For $y > 0$:
$$
\begin{aligned}
F_Y(y) &= P(Y \le y) \\
&= P(|Z| \le y) \\
&= P(-y \le Z \le y) \\
&= \Phi(y) - \Phi(-y)
\end{aligned}
$$
Using the tail symmetry property $\Phi(-y) = 1 - \Phi(y)$:
$$
\begin{aligned}
F_Y(y) &= \Phi(y) - (1 - \Phi(y)) \\
&= 2\Phi(y) - 1
\end{aligned}
$$
To find the PDF $f_Y(y)$, differentiate the CDF for $y > 0$:
$$
f_Y(y) = \frac{d}{dy} (2\Phi(y) - 1) = 2\varphi(y)
$$
(And $f_Y(y) = 0$ elsewhere).

::: {.callout-note}
## Intuition: Probability Folding
The PDF $f_Y(y) = 2\varphi(y)$ makes logical sense: taking the absolute value "folds" all the probability mass from the negative half of the real line onto the positive half. The density at any positive point $y$ receives its original contribution $\varphi(y)$ plus the mirrored contribution from $\varphi(-y)$. Thus, the density precisely doubles on the positive side.
:::
