# Continous Expectation, LOTUS, and the Survival Function

For continuous [random variables](random-variables.ipynb), the definition of Expected Value is perfectly analogous to the discrete definition: we simply replace the sum with an integral and the PMF with the PDF.

**Definition:** The expected value of a continuous random variable $X$ with PDF $f(x)$ is:
$$E(X) = \int_{-\infty}^{\infty} x f(x) dx$$

## 1. Law of the Unconscious Statistician (LOTUS)

When evaluating the expected value of a transformed random variable $g(X)$ (such as $X^2$ or $e^X$), it is a common mathematical trap to assume $E[g(X)] = g(E[X])$. This is fundamentally false for any non-linear function $g$.

To compute $E[g(X)]$, we **do not** need to derive the complex, transformed probability distribution of the new random variable $Y = g(X)$. We can compute the expectation directly using the original PMF or PDF of $X$. This profoundly useful shortcut is known as the **Law of the Unconscious Statistician (LOTUS)**.

**Theorem (Discrete LOTUS):** Let $X$ be a discrete random variable with PMF $P(X=x)$, and let $g(x)$ be a real-valued function. Then:
$$E[g(X)] = \sum_{x} g(x) P(X=x)$$

**Theorem (Continuous LOTUS):** Let $X$ be a continuous random variable with PDF $f(x)$, and let $g(x)$ be a real-valued function. Then:
$$E[g(X)] = \int_{-\infty}^{\infty} g(x) f(x) dx$$
