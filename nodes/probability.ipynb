# Probability

## 1\. The Probability Space

We have already seen the [naive definition of probability](probability-naive-definition.ipynb) and its limitations. To solve these limitations, modern mathematics uses an axiomatic version of probability.

A **probability space** consists of a [sample-space](sample-space.ipynb) $S$, a collection of valid [events](event.ipynb) (a $\sigma$-algebra), and a probability function $P$. The function $P$ takes an event $A \subseteq S$ as input and returns $P(A)$, a real number between $0$ and $1$, as output.

## 2\. Kolmogorov Axioms for Probability

For $P$ to be a mathematically valid probability measure, it must strictly satisfy the following three axioms:

**1. Non-Negativity**
The probability of any event is a non-negative real number:
$$P(A) \geq 0$$

**2. Normalization (Unit Measure)**
The probability that *some* outcome in the sample space occurs is exactly 1:
$$P(S) = 1$$

**3. Countable Additivity**
If $A_1, A_2, \dots$ is a countable sequence of disjoint events (meaning $A_i \cap A_j = \emptyset$ for $i \neq j$), then the probability of their union is the sum of their individual probabilities:
$$P \left( \bigcup_{j=1}^{\infty} A_j \right) = \sum_{j=1}^{\infty} P(A_j)$$

From these three axioms, all other [properties of probability](probability-properties.ipynb) can be rigorously proven.
