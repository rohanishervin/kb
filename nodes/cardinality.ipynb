# Cardinality and Sizes of Infinity

## 1. Defining "Same Size"
In mathematics, we define two sets $A$ and $B$ to have the **equal cardinality** (the same size) if and only if there exists a [**bijection**](def-functions-mappings.ipynb) between them.
$$|A| = |B| \iff \text{there is a bijection from } A \text{ to } B$$
This creates an [equivalence relation](pre-set-theory.ipynb), denoted $A \sim B$.

## 2. Hierarchy of Sizes
Using bijections, we can classify sets into different sizes:

* **Finite:** $S = \emptyset$ or $S \sim \{1, 2, \dots, n\}$ for some $n \in \mathbb{N}$.
* **Infinite:** Not finite.
* **Denumerable:** $S \sim \mathbb{N}$ (It can be put into an infinite, indexed list).
* **Countable:** Either finite or denumerable.
* **Uncountable:** An infinite set that is *not* countable (cannot be put into a list).

**Sub-Theorems on Denumerability:**

* Each infinite set contains a denumerable subset.
* An infinite subset of a denumerable set is also denumerable.

## 3. Cantor's Diagonalization Argument

**Theorem:** $\mathbb{R}$ is uncountable. (By extension, any interval $(a, b)$ or $[a, b]$ is uncountable).

**Proof:**

We will prove this by contradiction. Assume the real numbers in the interval $(0, 1)$ are denumerable. This means we can list them out in a sequence: $r_1, r_2, r_3, \dots$ such that every real number in $(0, 1)$ appears exactly once on this list.
We can write out their decimal expansions:
$r_1 = 0.d_{11}d_{12}d_{13}\dots$
$r_2 = 0.d_{21}d_{22}d_{23}\dots$
$r_3 = 0.d_{31}d_{32}d_{33}\dots$

We now construct a new real number $x = 0.x_1x_2x_3\dots$ by looking at the *diagonal* digits of our list ($d_{nn}$). 
We define the $n$-th digit of $x$ such that it is different from $d_{nn}$. For example:

* If $d_{nn} = 1$, let $x_n = 2$.
* If $d_{nn} \neq 1$, let $x_n = 1$.

Because $x$ is an infinite decimal, $x \in (0, 1)$. However, $x$ cannot be anywhere on our list! 

* It cannot be $r_1$ because its 1st digit is different.
* It cannot be $r_2$ because its 2nd digit is different.
* It cannot be $r_n$ because its $n$-th digit is different.

This contradicts our assumption that our list contained *every* real number in $(0, 1)$. Thus, $(0, 1)$ is uncountable. Since $(0, 1) \subset \mathbb{R}$, $\mathbb{R}$ must also be uncountable. $\square$

## 4. Schröder-Bernstein Theorem

Comparing the sizes of sets is often difficult because finding a perfect bijection is hard. The Schröder-Bernstein Theorem provides a powerful shortcut.

**Theorem:** For any sets $A$ and $B$, if $|A| \leq |B|$ and $|B| \leq |A|$, then $|A| = |B|$.

**Practical Meaning:** If we can find an injective (one-to-one) function from $A$ to $B$, and another injective function from $B$ to $A$, this guarantees that a perfect bijection exists between them, and therefore they have the exact same cardinality.
