# Principle of Inclusion-Exclusion (PIE)

## 1\. Concept and Intuition

The **Principle of Inclusion-Exclusion (PIE)** provides a systematic, rigorous way to calculate the size (or [probability](probability.ipynb)) of a union of multiple overlapping sets. It corrects for overcounting when events are not mutually exclusive.

To build intuition, consider the simplest cases before the general formula:

  * **For 2 Events ($A, B$):** If we simply add $P(A) + P(B)$, we have counted the overlapping region $P(A \cap B)$ twice. We must *exclude* it once to correct the sum.
    $$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

  * **For 3 Events ($A, B, C$):**
    We *include* the individual probabilities, *exclude* the pairwise overlaps, but doing so subtracts the dead-center region ($A \cap B \cap C$) entirely\! We must *include* it back one final time.
    $$P(A \cup B \cup C) = P(A) + P(B) + P(C) - P(A \cap B) - P(A \cap C) - P(B \cap C) + P(A \cap B \cap C)$$

-----

## 2\. Formal General Theorem

For a collection of $n$ events $A_1, A_2, \dots, A_n$, the exact probability of their union is calculated by alternating addition and subtraction of the intersections of increasing size:

$$P\left(\bigcup_{i=1}^n A_i\right) = \sum_{i=1}^n P(A_i) - \sum_{i < j} P(A_i \cap A_j) + \sum_{i < j < k} P(A_i \cap A_j \cap A_k) - \dots + (-1)^{n-1} P\left(\bigcap_{i=1}^n A_i\right)$$

*(Note: The number of terms of size $k$ in this expansion is exactly determined by the [binomial coefficient](binomial-coefficient.ipynb), $\binom{n}{k}$. For example, there are $\binom{n}{2}$ pairwise intersections to subtract).*

-----

## 3\. Combinatorial Proof

How do we know for absolutely certain that this alternating plus-minus pattern counts every element exactly once? We can prove it using the [binomial theorem](binomial-theorem.ipynb).

**Proof:**

Consider an arbitrary outcome $\omega$ that belongs to exactly $k$ of the $n$ events (where $k \geq 1$). We need to determine how many times $\omega$ is counted in the massive PIE sum.

1.  In the first sum (individual events), $\omega$ is counted exactly $\binom{k}{1} = k$ times.
2.  In the second sum (pairwise intersections), $\omega$ is subtracted exactly $\binom{k}{2}$ times.
3.  In the third sum (triple intersections), $\omega$ is added exactly $\binom{k}{3}$ times.
4.  This continues up to size $k$. (It is not counted in any intersections larger than $k$).

The total net number of times $\omega$ is counted is the series:
$$\text{Total Count} = \binom{k}{1} - \binom{k}{2} + \binom{k}{3} - \dots + (-1)^{k-1}\binom{k}{k}$$

To evaluate this, recall the Binomial Expansion of $(1 - 1)^k$:
$$(1 - 1)^k = \binom{k}{0} - \binom{k}{1} + \binom{k}{2} - \binom{k}{3} + \dots + (-1)^k\binom{k}{k}$$

Because $(1 - 1)^k = 0^k = 0$, we can set the equation to $0$ and isolate $\binom{k}{0}$:
$$\binom{k}{0} = \binom{k}{1} - \binom{k}{2} + \binom{k}{3} - \dots + (-1)^{k-1}\binom{k}{k}$$

Since $\binom{k}{0} = 1$, we perfectly conclude:
$$\text{Total Count} = 1$$

Every element, regardless of how many sets it belongs to, is mathematically guaranteed to be counted exactly one time. $\square$

## 4\. Classic Application Example

**Question:** If you draw a single card from a standard 52-card deck, what is the probability that it is either a Spade or a Face Card (Jack, Queen, King)?

**Solution:**
Let $S$ be the event of drawing a Spade. Let $F$ be the event of drawing a Face Card.
We want $P(S \cup F)$. Because a card can be both a Spade AND a Face Card, these events are not mutually exclusive. We must use PIE.

  * $P(S) = \frac{13}{52}$ (There are 13 spades).
  * $P(F) = \frac{12}{52}$ (There are 3 face cards per suit $\times$ 4 suits).
  * $P(S \cap F) = \frac{3}{52}$ (The Jack, Queen, and King of Spades).

Apply the theorem:
$$P(S \cup F) = P(S) + P(F) - P(S \cap F)$$
$$P(S \cup F) = \frac{13}{52} + \frac{12}{52} - \frac{3}{52} = \frac{22}{52} \approx 42.3\%$$
