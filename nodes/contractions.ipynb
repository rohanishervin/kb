# Contractions

Fixed-point theorems are central to many applications in analysis, providing the foundational machinery for proving results like the general implicit function theorem and the existence of solutions to ordinary differential equations. Before diving into the theorems, we must establish the core terminology.

## 1\. Fixed Points

**Definition:**
Let $M$ be a set and $f: M \to M$ be a function. A point $p \in M$ is called a **fixed point** of $f$ if it is mapped to itself under $f$. That is, $p$ is a fixed point if and only if:
$$
f(p) = p
$$

## 2\. Contraction Mappings

**Definition:**
Let $(M, d)$ be a [metric space](metric-spaces.ipynb). A mapping $f: M \to M$ is called a **contraction** (or contraction mapping) of $M$ if there exists a real constant $k \in [0, 1)$ such that for all $x, y \in M$, the following inequality holds:
$$
d(f(x), f(y)) \le k d(x, y)
$$

The constant $k$ is often referred to as the Lipschitz constant of the contraction.

::: {.callout-warning}
## Strict Inequality vs. Contraction
It is a common pitfall to assume that a function satisfying strictly $d(f(x), f(y)) < d(x, y)$ for $x \neq y$ is a contraction. This is **false**. For $f$ to be a contraction, the ratio of the distances must be bounded away from $1$ by some uniform constant $k < 1$. 
:::

::: {.callout-note}
## Intuition: What does a contraction do?
Geometrically, a contraction map takes the entire metric space and "shrinks" it. If you apply a contraction mapping repeatedly, any two points in the space will be forced closer and closer together, eventually squishing the space toward a single locus.
:::
