# Multiplication Rule (Fundamental Principle of Counting)

## 1\. General Definition

The **Multiplication Rule** states that when a process (or random [experiment](experiment.ipynb)) is divided into multiple sequential stages, the total number of combined outcomes is the product of the number of choices available at each stage.

Specifically, if a process consists of $k$ stages, where:

  * Stage $1$ can occur in $n_1$ distinct ways,
  * Stage $2$ can occur in $n_2$ distinct ways (regardless of the specific choice made in Stage 1),
  * $\dots$
  * Stage $k$ can occur in $n_k$ distinct ways,

Then the total number of possible sequential combinations for the entire process is:

$$\text{Total Outcomes} = n_1 \times n_2 \times \dots \times n_k = \prod_{i=1}^{k} n_i$$

-----

## 2\. Set-Theoretic Foundation

Formally, this rule describes the cardinality (size) of the **Cartesian product** of finite sets.

If $A$ and $B$ are sets representing the possible outcomes of Stage 1 and Stage 2, their Cartesian product $A \times B$ is the set of all possible ordered pairs $(a, b)$. The multiplication rule mathematically guarantees that the size of this combined set is the product of their individual sizes:

$$|A \times B| = |A| \cdot |B|$$

For $k$ sets, this generalizes perfectly to $|S_1 \times S_2 \times \dots \times S_k| = |S_1| \cdot |S_2| \cdots |S_k|$.

-----

## 3\. Practical Example

**Question:** A restaurant offers a set menu where you must choose exactly 1 appetizer (from 3 options), 1 main course (from 4 options), and 1 dessert (from 2 options). How many unique 3-course meals can be ordered?

**Solution:**
Because each choice is made sequentially, and selecting a specific appetizer does not restrict your options for the main course, these are independent counting stages. We simply multiply the number of choices available at each stage:

$$\text{Total Unique Meals} = 3 \times 4 \times 2 = 24$$

There are exactly $24$ distinct meal combinations.
