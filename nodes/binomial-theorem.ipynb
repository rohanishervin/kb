# The Binomial Theorem

## 1\. Theorem Statement

The **Binomial Theorem** provides a direct algebraic method to expand expressions of the form $(x + y)^n$ into a sum involving terms of the form $ax^b y^c$.

For any integer $n \geq 0$ and any real numbers $x$ and $y$, the theorem states:

$$(x + y)^n = \sum_{k=0}^{n} \binom{n}{k} x^{n-k} y^k$$

Where the term $\binom{n}{k}$ is the standard [binomial coefficient](binomial-coefficient.ipynb), algebraically defined as:

$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$

*(Note: Because of the symmetry property of binomial coefficients, $\binom{n}{k} = \binom{n}{n-k}$, the theorem can also be written with the exponents swapped: $x^k y^{n-k}$. Both forms are mathematically identical).*

-----

## 2\. Combinatorial Proof

**Proof:**

Consider writing out the full expansion of the product:
$$(x + y)^n = \underbrace{(x + y)(x + y)\dots(x + y)}_{n \text{ times}}$$

When multiplying this out using the distributive property, every single term in the final unsimplified sum is the result of choosing exactly one variable (either $x$ or $y$) from *each* of the $n$ parentheses.

Because we are making exactly $n$ choices, every term will have a combined degree of $n$. Specifically, if we choose $y$ exactly $k$ times, we must have chosen $x$ the remaining $(n-k)$ times. This produces the algebraic term $x^{n-k}y^k$.

How many ways can we form this specific term? We have $n$ total parentheses, and we need to choose exactly $k$ of them to supply the $y$. Because the order in which we select the parentheses does not matter, this is a standard combination.

The number of ways to choose $k$ parentheses from a set of $n$ is exactly $\binom{n}{k}$. Therefore, the term $x^{n-k}y^k$ will appear exactly $\binom{n}{k}$ times in the unsimplified expansion.

Summing this logic across all possible values of $k$ (from $0$ up to $n$) yields the complete theorem:
$$(x + y)^n = \sum_{k=0}^{n} \binom{n}{k} x^{n-k} y^k$$
$\square$

-----

## 3\. Important Corollaries for Probability

By cleverly substituting values for $x$ and $y$ into the Binomial Theorem, we can instantly prove several foundational identities that are heavily used in probability theory.

**Corollary 1: The Sum of All Combinations**
If a set has $n$ elements, the total number of all possible subsets (the size of the power set) is $2^n$.
$$\sum_{k=0}^{n} \binom{n}{k} = 2^n$$
*Proof:* Let $x=1$ and $y=1$ in the Binomial Theorem. The left side becomes $(1+1)^n = 2^n$. The right side becomes $\sum \binom{n}{k} (1)^{n-k} (1)^k = \sum \binom{n}{k}$.

**Corollary 2: The Alternating Sum**
The number of even-sized subsets of a set is exactly equal to the number of odd-sized subsets.
$$\sum_{k=0}^{n} (-1)^k \binom{n}{k} = 0$$
*Proof:* Let $x=1$ and $y=-1$. The left side becomes $(1 - 1)^n = 0^n = 0$. The right side becomes $\sum \binom{n}{k} (1)^{n-k} (-1)^k$.

## 4\. Classic Example

**Question:** Expand the polynomial $(2a - 3b)^3$.

**Solution:**
Apply the Binomial Theorem with $n=3$, $x = 2a$, and $y = -3b$:

$$(2a - 3b)^3 = \sum_{k=0}^{3} \binom{3}{k} (2a)^{3-k} (-3b)^k$$

Calculate the terms one by one for $k=0, 1, 2, 3$:

  * **k=0:** $\binom{3}{0}(2a)^3(-3b)^0 = (1)(8a^3)(1) = 8a^3$
  * **k=1:** $\binom{3}{1}(2a)^2(-3b)^1 = (3)(4a^2)(-3b) = -36a^2b$
  * **k=2:** $\binom{3}{2}(2a)^1(-3b)^2 = (3)(2a)(9b^2) = 54ab^2$
  * **k=3:** $\binom{3}{3}(2a)^0(-3b)^3 = (1)(1)(-27b^3) = -27b^3$

Summing these terms yields the final expansion:
$$(2a - 3b)^3 = 8a^3 - 36a^2b + 54ab^2 - 27b^3$$
