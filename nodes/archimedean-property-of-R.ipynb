# Archimedean Property of $\mathbb{R}$

if $a,b \in \mathbb{R}$ and $a,b>0$ then:
$\exists n \in \mathbb{N}: na>b$.

**Proof** by contradiction suppose $na\leq b$. we can create $A \subset \mathbb{R}$ such that:


$$ A=\{na|n \in \mathbb{N}\}$$

$a \in A$ so it's nonempty. and by assumption $b$ is an upper bound. so we can use [AoC](axiom-of-completeness.ipynb):

$$ s=sup(A) $$

now we knew $s-a < s$ so there exist an element in $A$ that is bigger than $s-a$. so,

$$
\exists m \in \mathbb{N}: s-a<ma \implies s<(1+m)a = m'a, m' \in \mathbb{N}
$$

and it contradicts the fact that $s$ is an upperbound.

## Theorem

The following statements are equivalent.
 - *If $a$ and $b$ are positive real numbers, then there exists a positive integer $n$ such that $na > b.$*

 - *The set $\{1, 2, 3, \ldots\}$ of positive integers is unbounded above.*

 - *For each real number $x$, there exists an integer $n$ such that $n < x < n+1.$*

 - *For each positive real number $x$, there exists a positive integer $n$ such that $nx > x \geq (n-1)x.$*
